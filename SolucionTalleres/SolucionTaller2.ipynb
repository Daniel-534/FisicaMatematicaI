# Solución Problema Propuesto 1 - Taller 2

La **Delta de Dirac** permite tratar distribuciones concentradas en puntos, líneas o superficies como si fueran funciones de densidad definidas en todo el espacio volumétrico. Para que la densidad volumétrica $\rho(\vec{r})$ sea correcta, su integral sobre todo el espacio debe devolver la masa total de la distribución.

### 1. Varilla con densidad lineal de masa $\lambda$ situada en el eje $y$

En un sistema de coordenadas cartesianas $(x, y, z)$, una varilla situada exactamente sobre el eje $y$ implica que la masa solo existe donde $x = 0$ y $z = 0$. La masa está "extendida" a lo largo de la coordenada $y$ pero "colapsada" en las dimensiones $x$ y $z$.

La densidad volumétrica se expresa como:
$$\rho(x, y, z) = \lambda(y) \delta(x) \delta(z)$$

**Explicación:**
*   Las funciones $\delta(x)$ y $\delta(z)$ aseguran que la densidad sea nula en cualquier punto que no pertenezca al eje $y$.
*   Al integrar $\rho$ sobre un volumen, las integrales $\int \delta(x) dx$ y $\int \delta(z) dz$ resultan en 1, dejando únicamente la integral de la densidad lineal $\int \lambda(y) dy$ para obtener la masa total.

---

### 2. Cascarón esférico de radio $R$ con densidad de masa $\sigma$

Para esta distribución, el sistema de coordenadas más natural es el esférico $(r, \theta, \phi)$. En un cascarón o superficie esférica, la masa se concentra exclusivamente en una distancia radial fija $r = R$.

La densidad volumétrica se expresa como:
$$\rho(r, \theta, \phi) = \sigma(\theta, \phi) \delta(r - R)$$

**Explicación:**
*   La función $\delta(r - R)$ restringe la presencia de masa únicamente a la superficie de radio $R$.
*   Si la densidad superficial $\sigma$ es uniforme, la masa total es $M = \sigma (4\pi R^2)$. Al integrar en volumen esférico, el diferencial $dV = r^2 \sin\theta dr d\theta d\phi$ interactúa con la delta para evaluar la expresión en $r=R$.

---

### 3. Anillo de radio $R$ con distribución lineal de masa $\lambda$ en un plano

Considerando que el anillo se encuentra en el plano $xy$ (donde $z = 0$), utilizamos coordenadas cilíndricas $(\rho, \phi, z)$. La masa está concentrada en una circunferencia de radio constante $\rho = R$ y a una altura constante $z = 0$.

La densidad volumétrica se expresa como:
$$\rho(\rho, \phi, z) = \lambda(\phi) \delta(\rho - R) \delta(z)$$

**Explicación:**
*   $\delta(z)$ sitúa el anillo en el plano indicado.
*   $\delta(\rho - R)$ confina la masa al radio específico.
*   En este sistema, el diferencial de volumen es $dV = \rho d\rho d\phi dz$. Al integrar, la propiedad selectiva de las deltas extrae los valores en $R$ y $0$.

---

### 4. Disco de radio $R$ con distribución superficial de masa $\sigma$ en $z = z_0$

Un disco es una distribución superficial situada en un plano, pero limitada radialmente. Usando coordenadas cilíndricas $(\rho, \phi, z)$, el disco se encuentra en la elevación $z = z_0$ y se extiende desde el centro hasta $\rho = R$.

La densidad volumétrica se expresa como:
$$\rho(\rho, \phi, z) = \sigma(\rho, \phi) \delta(z - z_0) \cdot H(R - \rho)$$

Donde $H$ es la función escalón de Heaviside para limitar el radio. De manera más simplificada, si asumimos que $\sigma$ ya está definida como cero para $\rho > R$:
$$\rho(\rho, \phi, z) = \sigma(\rho, \phi) \delta(z - z_0)$$

**Explicación:**
*   La función $\delta(z - z_0)$ "aplasta" la distribución volumétrica contra el plano horizontal situado en $z_0$.
*   A diferencia del anillo, no hay una delta en la coordenada radial $\rho$, ya que la masa está distribuida continuamente sobre el área del disco (superficie) y no solo en su borde.

A continuación se presenta la solución completa y detallada al **problema propuesto 2 del Taller 2**, en el cual se solicita demostrar la representación integral de la **Delta de Dirac** en una dimensión y su posterior generalización a tres dimensiones.

# Solución Problema Propuesto 2 - Taller 2

El objetivo es demostrar que la distribución Delta de Dirac $\delta(x - x')$ puede expresarse mediante la integral de Fourier de una onda plana. Esta representación es fundamental en física matemática, especialmente en el análisis de señales y la mecánica cuántica.

## 1. Representación en una dimensión (1D)

### A. Definición mediante el límite de una secuencia
Para demostrar la identidad, partimos de una secuencia de funciones que aproximan a la delta. Consideremos la secuencia definida por la función **sinc**:
$$\delta_n(x - x') = \frac{\sin[n(x - x')]}{\pi(x - x')}$$

Esta función tiene un pico en $x = x'$ que se vuelve progresivamente más alto y estrecho a medida que $n$ aumenta, manteniendo siempre un área unitaria.

### B. Relación con la integral de Fourier
Utilizando la identidad de la integral de una exponencial compleja en un intervalo finito:
$$\int_{-n}^{n} e^{ik(x - x')} dk = \left[ \frac{e^{ik(x - x')}}{i(x - x')} \right]_{-n}^{n} = \frac{e^{in(x - x')} - e^{-in(x - x')}}{i(x - x')}$$
Aplicando la fórmula de Euler ($e^{i\theta} - e^{-i\theta} = 2i\sin\theta$):
$$\int_{-n}^{n} e^{ik(x - x')} dk = \frac{2i \sin[n(x - x')]}{i(x - x')} = \frac{2 \sin[n(x - x')]}{(x - x')}$$

Si multiplicamos esta expresión por $1/2\pi$, obtenemos exactamente la secuencia $\delta_n$:
$$\delta_n(x - x') = \frac{1}{2\pi} \int_{-n}^{n} e^{ik(x - x')} dk$$

### C. Paso al límite (Distribución)
Al tomar el límite cuando $n \to \infty$, el intervalo de integración se extiende a todo el eje real. Por lo tanto, en el sentido de las distribuciones, la Delta de Dirac se representa como:
$$\mathbf{\delta(x - x') = \frac{1}{2\pi} \int_{-\infty}^{\infty} e^{ik(x - x')} dk}$$

Esta expresión tiene significado físico cuando se encuentra bajo el signo de una integral, donde actúa "filtrando" el valor de una función de prueba $f(x)$ en el punto $x'$.

---

## 2. Generalización a tres dimensiones (3D)

### A. Definición en Coordenadas Cartesianas
La Delta de Dirac tridimensional $\delta(\vec{r} - \vec{r}')$ se define como el producto de tres deltas unidimensionales, una para cada eje coordenado:
$$\delta(\vec{r} - \vec{r}') = \delta(x - x') \delta(y - y') \delta(z - z')$$

### B. Sustitución de la representación integral
Sustituimos la forma integral obtenida en el paso anterior para cada componente espacial:
$$\delta(\vec{r} - \vec{r}') = \left[ \frac{1}{2\pi} \int_{-\infty}^{\infty} e^{ik_x(x - x')} dk_x \right] \left[ \frac{1}{2\pi} \int_{-\infty}^{\infty} e^{ik_y(y - y')} dk_y \right] \left[ \frac{1}{2\pi} \int_{-\infty}^{\infty} e^{ik_z(z - z')} dk_z \right]$$

### C. Combinación de términos
Agrupamos las constantes y combinamos las integrales en una sola integral triple sobre el espacio del vector de onda $\vec{k}$:
$$\delta(\vec{r} - \vec{r}') = \frac{1}{(2\pi)^3} \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} e^{i[k_x(x - x') + k_y(y - y') + k_z(z - z')]} dk_x \, dk_y \, dk_z$$

Utilizando la notación del producto escalar $\vec{k} \cdot (\vec{r} - \vec{r}') = k_x(x - x') + k_y(y - y') + k_z(z - z')$ y el elemento de volumen en el espacio recíproco $d^3k = dk_x \, dk_y \, dk_z$, llegamos al resultado final:

$$\mathbf{\delta(\vec{r} - \vec{r}') = \frac{1}{(2\pi)^3} \int_{\text{todo } k} e^{i\vec{k} \cdot (\vec{r} - \vec{r}')} d^3k}$$

### Conclusión
Esta representación 3D muestra que la Delta de Dirac es la **superposición uniforme de todas las ondas planas posibles**. Es una herramienta vital para resolver ecuaciones diferenciales inhomogéneas como la **ecuación de Poisson** o la **ecuación de ondas** mediante transformadas de Fourier.
