```python
import sympy as sp
```

```python
rho, z, t, T_0 = sp.symbols('rho z t T_0')
alpha, chi_01 = sp.symbols('alpha \chi_{01}')
R, L = sp.symbols("R L")

A_ml = 4*T_0/(sp.pi * chi_01 * sp.besselj(1,chi_01))
T = A_ml * sp.besselj(0,chi_01*rho/R)*sp.sin(sp.pi*z/L)*sp.exp(-alpha**2 * ((chi_01/R)**2 + (sp.pi/L)**2)*t)

T_eval1 = T.subs({R:1, L:sp.pi, rho: chi_01/2})
T_eval2 = T_eval1.subs({chi_01: 2.4048, t: 1, alpha: 1.16e-4, z: 10})

resultado = T_eval2.evalf()
```
