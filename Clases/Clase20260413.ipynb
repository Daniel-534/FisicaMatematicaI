Considere una barra delgada aislada semi-infinita a lo largo del eje x con extremo izquierdo en $x=0$.

Sea $T(x, t)$ la temperatura de la barra. La ecuación de difusión de calor para la barra es $$\partial_x^2 T(x, t) = \frac{1}{\kappa} - \partial_t T(x, t)$$

$\kappa$: Constante de conductividad térmica.

Las condiciones de frontera son:

* $T(x\to\infty,t)$ es finita

* $\partial_x T\bigg|_{x=0} = 0$

* $T(x, 0) = f(x)$

Hallar $T(x, y)$

¿Cuál de estas condiciones me asegura una función de cuadrado integrable?

# Solución

$T(x,t) = A(x)B(t)$


Tengo tres casos: $\alpha^2, -\beta^2, 0$

Par ahorra tiempo en cálculos, podemos analizar el problema y nos damos cuenta que nos sirve el caso de $-\beta^2$

$\frac{1}{A}\frac{d^2A}{dx^2} = -\beta^2 \wedge \frac{1}{\kappa}\frac{1}{B}\frac{dB}{dx} = -\beta^2$

Para completar el desarrollo de la solución para la barra semi-infinita utilizando el método de **separación de variables**, seguiremos los pasos analíticos basados en las fuentes para resolver las ecuaciones diferenciales ordinarias (EDO) resultantes y aplicar las condiciones de frontera.

### 1. Resolución de las Ecuaciones Diferenciales Ordinarias (EDO)
A partir de la constante de separación $-\beta^2$, tenemos el siguiente sistema de EDO:

1.  **Para el espacio ($x$):**
    $$\frac{d^2A}{dx^2} + \beta^2 A(x) = 0$$
    La solución general para esta ecuación es armónica:
    $$A(x) = C_1 \cos(\beta x) + C_2 \sin(\beta x)$$

2.  **Para el tiempo ($t$):**
    $$\frac{dB}{dt} + \beta^2 \kappa B(t) = 0$$
    La solución es un decaimiento exponencial:
    $$B(t) = C_3 e^{-\beta^2 \kappa t}$$

Al combinar ambas, obtenemos una solución particular para un valor dado de $\beta$:
$$T_\beta(x, t) = [C_1 \cos(\beta x) + C_2 \sin(\beta x)] e^{-\beta^2 \kappa t}$$

### 2. Aplicación de las Condiciones de Frontera (CF)
*   **Condición de aislamiento en $x = 0$:**
    Se nos da que $\partial_x T \big|_{x=0} = 0$. Calculamos la derivada de la solución respecto a $x$:
    $$\partial_x T = \beta [-C_1 \sin(\beta x) + C_2 \cos(\beta x)] e^{-\beta^2 \kappa t}$$
    Evaluando en $x=0$:
    $$\beta [0 + C_2] e^{-\beta^2 \kappa t} = 0 \implies \mathbf{C_2 = 0}$$
    Esto nos deja con $A(x) = C_1 \cos(\beta x)$.

*   **Condición de finitud al infinito ($x \to \infty$):**
    Como la solución contiene $\cos(\beta x)$, la temperatura permanece acotada para cualquier valor real de $\beta$, satisfaciendo la condición de que $T$ es finita.

### 3. Superposición e Integral de Fourier
Dado que el dominio es **semi-infinito** ($0 \le x < \infty$), el parámetro $\beta$ no está restringido a valores discretos, sino que forma un **espectro continuo**. Por el principio de superposición, la solución general es una integral:
$$T(x, t) = \int_{0}^{\infty} C(\beta) \cos(\beta x) e^{-\beta^2 \kappa t} d\beta$$

### 4. Condición Inicial (CI) y Transformada de Fourier
Aplicamos la condición inicial $T(x, 0) = f(x)$:
$$f(x) = \int_{0}^{\infty} C(\beta) \cos(\beta x) d\beta$$

Esta expresión identifica a $C(\beta)$ como la **Transformada Coseno de Fourier** de la función $f(x)$. El coeficiente se determina mediante la transformada inversa:
$$C(\beta) = \frac{2}{\pi} \int_{0}^{\infty} f(x') \cos(\beta x') dx'$$

### Solución Final
Sustituyendo el coeficiente en la integral, la temperatura de la barra en cualquier punto y tiempo es:
$$T(x, t) = \frac{2}{\pi} \int_{0}^{\infty} \left[ \int_{0}^{\infty} f(x') \cos(\beta x') dx' \right] \cos(\beta x) e^{-\beta^2 \kappa t} d\beta$$

Esta solución satisface que el flujo de calor es nulo en el extremo izquierdo (debido a la base de cosenos) y que la distribución de calor se difunde y suaviza con el tiempo debido al factor exponencial negativo.

$$B(t) = \tilde{c}_1 e^{-\kappa \beta^2 t}$$

$$A(x) = \tilde{c}_2 e^{i\beta x} + \tilde{c}_3 e^{-i\beta x}$$



Las expresiones proporcionadas corresponden a las soluciones de las ecuaciones diferenciales ordinarias (EDO) que surgen al aplicar el **método de separación de variables** a la ecuación de difusión de calor.

### ¿Quiénes son estas constantes?
Las constantes $\tilde{c}_1, \tilde{c}_2$ y $\tilde{c}_3$ son **constantes de integración arbitrarias**. Su valor específico se determina mediante las **condiciones de frontera** (espaciales) y las **condiciones iniciales** (temporales) impuestas por el problema físico.
*   **$\tilde{c}_1$**: Representa la amplitud inicial de la parte temporal.
*   **$\tilde{c}_2$ y $\tilde{c}_3$**: Son coeficientes que determinan la forma y amplitud de la distribución espacial de la temperatura.

---

### Desarrollo paso a paso

Partiendo de la suposición de una solución separable $T(x, t) = A(x)B(t)$, el desarrollo utilizando las formas exponenciales solicitadas es el siguiente:

**1. Parte Temporal: $B(t)$**
La ecuación separada para el tiempo es $\frac{1}{\kappa B} \frac{dB}{dt} = -\beta^2$. Al integrar esta ecuación de primer orden, obtenemos:
$$\frac{dB}{B} = -\kappa \beta^2 dt \implies \ln B = -\kappa \beta^2 t + \ln \tilde{c}_1$$
Esto nos da el **decaimiento exponencial** de la temperatura en el tiempo:
$$B(t) = \tilde{c}_1 e^{-\kappa \beta^2 t} \quad$$

**2. Parte Espacial: $A(x)$**
*(Nota: En su consulta escribió $A(t)$, pero de acuerdo a la variable $x$ en el exponente y la teoría de separación, esta es la función espacial $A(x)$)*. 
La EDO para el espacio es $A''(x) + \beta^2 A(x) = 0$. La solución general para esta ecuación armónica de segundo orden con coeficientes constantes puede escribirse en términos de **exponenciales complejas**:
$$A(x) = \tilde{c}_2 e^{i\beta x} + \tilde{c}_3 e^{-i\beta x} \quad$$

**3. Aplicación de las Condiciones de Frontera (Barra semi-infinita)**
Para la barra aislada en $x=0$, aplicamos la condición de flujo nulo $\partial_x A(x) \big|_{x=0} = 0$:
*   Derivamos $A(x)$: $A'(x) = i\beta \tilde{c}_2 e^{i\beta x} - i\beta \tilde{c}_3 e^{-i\beta x}$.
*   Evaluamos en $x=0$: $i\beta(\tilde{c}_2 - \tilde{c}_3) = 0$.
*   Para una solución no trivial ($\beta \neq 0$), esto implica que **$\tilde{c}_2 = \tilde{c}_3$**.

Al ser iguales las constantes, podemos usar la **identidad de Euler** ($e^{i\theta} + e^{-i\theta} = 2\cos\theta$) para simplificar la parte espacial:
$$A(x) = \tilde{c}_2 (e^{i\beta x} + e^{-i\beta x}) = 2\tilde{c}_2 \cos(\beta x) \quad$$

**4. Superposición General**
Dado que el dominio es **semi-infinito** ($x \ge 0$), el parámetro $\beta$ no es discreto sino que forma un **espectro continuo**. Combinando $A(x)$ y $B(t)$ y absorbiendo las constantes constantes en una función de peso $C(\beta)$, la solución se escribe como una **integral de Fourier**:

$$T(x, t) = \int_{0}^{\infty} C(\beta) \cos(\beta x) e^{-\kappa \beta^2 t} d\beta \quad$$

Donde los coeficientes $C(\beta)$ se hallan finalmente a través de la **Transformada Coseno de Fourier** de la condición inicial $T(x, 0) = f(x)$.

Para desarrollar explícitamente el "salto" matemático entre la representación integral de una función y la determinación de sus coeficientes (la transformada inversa), se utiliza la **propiedad de ortogonalidad de las bases continuas** y el comportamiento de la **función delta de Dirac**, tal como se describe en las fuentes.

Aquí tienes el desarrollo paso a paso:

### 1. Multiplicación por la función base e integración
Partimos de la expresión de $f(x)$ como una superposición continua de cosenos:
$$f(x) = \int_{0}^{\infty} C(\beta) \cos(\beta x) d\beta$$

Para aislar el coeficiente $C(\beta)$, multiplicamos ambos lados de la igualdad por $\cos(\beta' x)$ (donde $\beta'$ es un valor específico del parámetro) e **integramos respecto a $x$** en el dominio del problema ($0$ a $\infty$):
$$\int_{0}^{\infty} f(x) \cos(\beta' x) dx = \int_{0}^{\infty} \left[ \int_{0}^{\infty} C(\beta) \cos(\beta x) d\beta \right] \cos(\beta' x) dx$$

### 2. Intercambio del orden de integración
De acuerdo con la teoría de bases continuas, podemos intercambiar el orden de las integrales en el lado derecho:
$$\int_{0}^{\infty} f(x) \cos(\beta' x) dx = \int_{0}^{\infty} C(\beta) \left[ \int_{0}^{\infty} \cos(\beta x) \cos(\beta' x) dx \right] d\beta$$

### 3. Identificación de la Delta de Dirac (Ortogonalidad)
La integral interna representa la **condición de ortogonalidad** de la base de cosenos en el intervalo semi-infinito. Basándonos en las representaciones integrales de la delta descritas en las fuentes:
$$\int_{0}^{\infty} \cos(\beta x) \cos(\beta' x) dx = \frac{\pi}{2} \delta(\beta - \beta')$$

Sustituyendo esta identidad en nuestra ecuación:
$$\int_{0}^{\infty} f(x) \cos(\beta' x) dx = \int_{0}^{\infty} C(\beta) \left[ \frac{\pi}{2} \delta(\beta - \beta') \right] d\beta$$

### 4. Aplicación de la propiedad selectiva de la Delta
La función delta de Dirac tiene la propiedad de "seleccionar" el valor de la función en el punto donde su argumento se anula ($\beta = \beta'$):
$$\int_{0}^{\infty} C(\beta) \delta(\beta - \beta') d\beta = C(\beta')$$

Por lo tanto, la expresión se simplifica a:
$$\int_{0}^{\infty} f(x) \cos(\beta' x) dx = \frac{\pi}{2} C(\beta')$$

### 5. Resultado final
Despejando $C(\beta')$ y renombrando las variables de forma convencional ($x \to x'$ y $\beta' \to \beta$), obtenemos la expresión final para el coeficiente:
**$$C(\beta) = \frac{2}{\pi} \int_{0}^{\infty} f(x') \cos(\beta x') dx'$$**

Este procedimiento demuestra que $C(\beta)$ es la amplitud de cada modo de frecuencia espacial $\beta$ necesaria para reconstruir la función original $f(x)$.

Para desarrollar explícitamente la expresión $$\int_{0}^{\infty} \cos(\beta x) \cos(\beta' x) dx = \frac{\pi}{2} \delta(\beta - \beta')$$, podemos seguir dos caminos analíticos basados en las fuentes proporcionadas: uno a través de la **teoría de bases continuas** y otro utilizando la **representación integral de la delta de Dirac**.

### 1. Desarrollo mediante la teoría de bases continuas
Las fuentes establecen que un conjunto de funciones $\{\phi(k, x)\}$ es **ortonormal** en un intervalo $(a, b)$ si satisface la condición:
$$\int_{a}^{b} \phi^*(k, x) \phi(k', x) dx = \delta(k - k')$$.

De acuerdo con el texto de Sepúlveda, el conjunto de funciones:
$$\phi(\beta, x) = \sqrt{\frac{2}{\pi}} \cos(\beta x)$$
constituye una **base ortonormal continua** definida en el dominio semi-infinito $0 \le x < \infty$. Al aplicar la definición de ortonormalidad a esta base específica:

1.  **Sustitución en la integral de ortonormalidad:**
    $$\int_{0}^{\infty} \left( \sqrt{\frac{2}{\pi}} \cos(\beta x) \right) \left( \sqrt{\frac{2}{\pi}} \cos(\beta' x) \right) dx = \delta(\beta - \beta')$$

2.  **Extracción de constantes:**
    $$\frac{2}{\pi} \int_{0}^{\infty} \cos(\beta x) \cos(\beta' x) dx = \delta(\beta - \beta')$$

3.  **Resultado final:** Multiplicando ambos lados por $\frac{\pi}{2}$, obtenemos la expresión deseada:
    $$\int_{0}^{\infty} \cos(\beta x) \cos(\beta' x) dx = \frac{\pi}{2} \delta(\beta - \beta')$$.

---

### 2. Desarrollo mediante la representación integral de la Delta
Podemos llegar al mismo resultado partiendo de la definición fundamental de la función delta de Dirac como una integral de Fourier:
$$\delta(\beta - \beta') = \frac{1}{2\pi} \int_{-\infty}^{\infty} e^{ix(\beta - \beta')} dx$$.

1.  **Uso de la identidad de Euler:** Expandimos el exponencial complejo como $e^{i\theta} = \cos\theta + i\sin\theta$:
    $$\delta(\beta - \beta') = \frac{1}{2\pi} \int_{-\infty}^{\infty} \left[ \cos(x(\beta - \beta')) + i \sin(x(\beta - \beta')) \right] dx$$.

2.  **Análisis de paridad:** La función seno es impar, por lo que su integral sobre un dominio simétrico $(-\infty, \infty)$ es nula. Esto reduce la expresión a:
    $$\delta(\beta - \beta') = \frac{1}{2\pi} \int_{-\infty}^{\infty} \cos(x(\beta - \beta')) dx$$

3.  **Identidad trigonométrica de la diferencia:** Usamos $\cos(A-B) = \cos A \cos B + \sin A \sin B$:
    $$2\pi \delta(\beta - \beta') = \int_{-\infty}^{\infty} \left[ \cos(\beta x) \cos(\beta' x) + \sin(\beta x) \sin(\beta' x) \right] dx$$.

4.  **Separación de términos y simetría:** Debido a la ortogonalidad de la base trigonométrica continua en el intervalo infinito, las partes de coseno y seno contribuyen independientemente a la delta. Para la parte del coseno:
    $$\int_{-\infty}^{\infty} \cos(\beta x) \cos(\beta' x) dx = \pi \delta(\beta - \beta')$$

5.  **Reducción del intervalo:** Como el integrando $\cos(\beta x) \cos(\beta' x)$ es una **función par** de $x$ (el producto de dos funciones pares es par), la integral sobre el dominio completo es el doble de la integral sobre el semi-dominio:
    $$2 \int_{0}^{\infty} \cos(\beta x) \cos(\beta' x) dx = \pi \delta(\beta - \beta')$$

Al dividir por 2, se recupera exactamente la expresión:
$$\mathbf{\int_{0}^{\infty} \cos(\beta x) \cos(\beta' x) dx = \frac{\pi}{2} \delta(\beta - \beta')}$$

La relación entre el problema de difusión de calor en la barra semi-infinita y el **Lema de Riemann-Lebesgue** es fundamental para garantizar que la solución matemática sea físicamente aceptable, específicamente en lo que respecta al comportamiento de la temperatura a grandes distancias.

De acuerdo con las fuentes, la conexión se establece de la siguiente manera:

### 1. Garantía de la condición de frontera al infinito
En el problema de la barra semi-infinita (o en dominios infinitos en general), se impone la condición física de que la temperatura debe desvanecerse o permanecer finita cuando $x \to \infty$. El Lema de Riemann-Lebesgue es la herramienta matemática que **garantiza que el potencial (o la temperatura) tienda a cero** en esos límites.

### 2. Definición matemática del Lema
El lema establece que si una función $f(\alpha)$ es absolutamente integrable (es decir, $\int |f(\alpha)| d\alpha$ tiene un valor finito), entonces el valor de su integral de Fourier tiende a cero cuando la variable espacial crece indefinidamente:
$$\lim_{x \to \pm\infty} \int_a^b e^{i\alpha x} f(\alpha) d\alpha = 0$$

### 3. Aplicación a la solución del problema
En el desarrollo de la difusión de calor, la solución se expresó como una integral de Fourier (o integral de cosenos):
$$T(x, t) = \int_{0}^{\infty} \underbrace{C(\beta) e^{-\kappa \beta^2 t}}_{f(\beta)} \cos(\beta x) d\beta$$

*   **Interferencia destructiva:** Cuando $x$ se vuelve muy grande, la función $\cos(\beta x)$ oscila con una frecuencia extremadamente alta. 
*   **Anulación del flujo:** El Lema de Riemann-Lebesgue asegura que estas oscilaciones rápidas provoquen una cancelación mutua (interferencia destructiva) dentro de la integral, resultando en que la suma total (la temperatura $T$) converja a cero conforme nos alejamos del origen.

En resumen, el lema justifica matemáticamente que la superposición continua de modos normales (frecuencias $\beta$) necesaria para representar la condición inicial $f(x)$ no "arrastre" calor de forma infinita hacia el final de la barra, cumpliendo así con la **condición de finitud en el infinito** impuesta originalmente.

Un ejemplo fundamental en mecánica cuántica (QM) donde se aplica el **lema de Riemann-Lebesgue** es en la descripción de una **partícula libre** mediante un paquete de ondas $\Psi(\vec{r}, t)$, lo cual permite garantizar que la partícula esté localizada y que su probabilidad de presencia se anule a distancias infinitas.

De acuerdo con las fuentes, la relación se desarrolla de la siguiente manera:

### 1. Representación de la función de onda como integral de Fourier
En mecánica cuántica, una partícula que no está confinada (como un electrón libre) no posee un momento lineal único, sino que su estado se describe como una superposición continua de modos normales (ondas planas). La función de onda en el espacio de las coordenadas $\Psi(\vec{r}, t)$ es la **transformada de Fourier** de la función de onda en el espacio de los momentos $\Phi(\vec{k})$:

$$\Psi(\vec{r}, t) = \frac{1}{(2\pi)^{3/2}} \int_{-\infty}^{\infty} \Phi(\vec{k}) e^{i(\vec{k} \cdot \vec{r} - \omega t)} d^3k$$

Donde:
*   **$\Phi(\vec{k})$** es la amplitud de probabilidad para cada vector de onda $\vec{k}$.
*   **$e^{i(\vec{k} \cdot \vec{r} - \omega t)}$** representa la fase de cada modo individual.

### 2. Aplicación del Lema de Riemann-Lebesgue
El lema establece que si una función $f(\alpha)$ es absolutamente integrable ($\int |f(\alpha)| d\alpha$ es finito), entonces el valor de su integral de Fourier tiende a cero cuando la variable espacial crece indefinidamente:

$$\lim_{r \to \infty} \int e^{i\vec{k} \cdot \vec{r}} \Phi(\vec{k}, t) d^3k = 0$$

### 3. Significado Físico en Mecánica Cuántica
En este contexto, el lema garantiza dos aspectos cruciales para la validez física del modelo:

*   **Localización y Normalización:** Para que la función de onda sea físicamente aceptable, debe ser de cuadrado integrable ($L^2$), lo que usualmente implica que $\Phi(\vec{k})$ es una función bien comportada. El lema asegura matemáticamente que **$\Psi(\vec{r}, t)$ converja a cero cuando $r \to \infty$**. Sin esta garantía, no podríamos asegurar que la probabilidad total de encontrar la partícula en todo el espacio sea igual a 1 (normalización).
*   **Interferencia Destructiva:** Físicamente, el lema describe cómo, a grandes distancias del centro del paquete de ondas, las diferentes frecuencias $\vec{k}$ oscilan tan rápidamente que se cancelan entre sí (interferencia destructiva). Esto evita que la "amplitud de probabilidad" se arrastre de forma infinita por el espacio.

### 4. Ejemplo en Dispersión (Scattering)
En problemas de dispersión, como el estudio de una partícula chocando contra un potencial $V(r)$, se utiliza a menudo la **aproximación de Born**. La función de onda resultante se expresa como una integral que involucra la transformada de Fourier del potencial. El lema de Riemann-Lebesgue asegura que la onda dispersada disminuya su influencia conforme nos alejamos del centro de dispersión, permitiendo que la solución satisfaga las condiciones de frontera al infinito (donde solo debe quedar la onda incidente o una onda esférica que decae).

En resumen, el lema es la herramienta que justifica que la dualidad onda-partícula no resulte en una "onda infinita", sino en una **partícula localizada** cuya presencia en el infinito es nula.
