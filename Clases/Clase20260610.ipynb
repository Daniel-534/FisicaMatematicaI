# Resumen de lo que llevamos hasta el momento.

Los **polinomios asociados de Legendre**, denotados como $P_l^m(x)$, son una extensión de los polinomios de Legendre ordinarios y surgen fundamentalmente al resolver problemas físicos con **simetría esférica** que no poseen simetría azimutal completa.

A continuación se presenta un resumen de sus aspectos más importantes:

### 1. Ecuación Diferencial
Surgen como soluciones a la **ecuación asociada de Legendre**, la cual aparece al separar variables en la ecuación de Laplace en coordenadas esféricas para la parte polar ($\theta$):
$$(1-x^2)\frac{d^2y}{dx^2} - 2x\frac{dy}{dx} + \left[ l(l+1) - \frac{m^2}{1-x^2} \right]y = 0$$
Donde $x = \cos \theta$. Para que las soluciones sean finitas en los polos ($x = \pm 1$), los índices deben cumplir que **$l$ es un entero no negativo** y **$m$ es un entero** tal que $-l \leq m \leq l$.

### 2. Definición y Relación con $P_l(x)$
Para valores de $m \geq 0$, se definen a partir de la $m$-ésima derivada del polinomio de Legendre ordinario $P_l(x)$:
$$P_l^m(x) = (-1)^m (1-x^2)^{m/2} \frac{d^m}{dx^m} P_l(x)$$
*(Nota: El factor $(-1)^m$ es la fase de Condon-Shortley, utilizada para estandarizar resultados en mecánica cuántica).*

### 3. Fórmula de Rodrigues
Permite generar cualquier polinomio asociado de forma directa mediante derivación sucesiva:
$$P_l^m(x) = \frac{(-1)^m}{2^l l!} (1-x^2)^{m/2} \frac{d^{l+m}}{dx^{l+m}} (x^2-1)^l$$

### 4. Propiedades Fundamentales
*   **Ortogonalidad:** Para un valor fijo de $m$, los polinomios de diferente grado $l$ son ortogonales en el intervalo $[-1, 1]$:
    $$\int_{-1}^{1} P_l^m(x) P_k^m(x) dx = \frac{2}{2l+1} \frac{(l+m)!}{(l-m)!} \delta_{lk}$$
*   **Paridad:** La simetría respecto al origen depende de la suma de sus índices:
    $$P_l^m(-x) = (-1)^{l+m} P_l^m(x)$$
*   **Relación para $m$ negativo:** Los términos con índice inferior negativo se relacionan con los positivos mediante:
    $$P_l^{-m}(x) = (-1)^m \frac{(l-m)!}{(l+m)!} P_l^m(x)$$

### 5. Aplicaciones Principales
*   **Armónicos Esféricos:** Son la base de las funciones $Y_l^m(\theta, \phi)$, que describen la distribución angular en problemas de **potencial central**.
*   **Mecánica Cuántica:** Se utilizan para hallar las funciones de onda del **átomo de hidrógeno** (orbitales atómicos) y para describir el **momento angular**.
*   **Electrostática:** Permiten expandir potenciales creados por distribuciones de carga asimétricas (como anillos o dipolos fuera del eje).

# Ejemplo: Halle $P_2^1(x)$ con $x=\cos(\theta)$

Para hallar el polinomio asociado de Legendre **$P_2^1(x)$** partimos de su definición mediante la **fórmula de Rodrigues** generalizada, la cual se expresa como $P_l^m(x) = \frac{(-1)^m}{2^l l!} (1 - x^2)^{m/2} \frac{d^{l+m}}{dx^{l+m}} (x^2 - 1)^l$. Al identificar los índices correspondientes a nuestro caso, tenemos **$l = 2$** y **$m = 1$**, lo que nos permite escribir la expresión específica $P_2^1(x) = \frac{(-1)^1}{2^2 2!} (1 - x^2)^{1/2} \frac{d^{2+1}}{dx^{2+1}} (x^2 - 1)^2$. Comenzamos el desarrollo evaluando el factor constante inicial, que resulta ser **$-1/8$**, y expandiendo el binomio interno para obtener el polinomio **$x^4 - 2x^2 + 1$**. El proceso continúa aplicando la derivada de orden $l+m=3$ a dicho polinomio de forma sucesiva: la primera derivada es $4x^3 - 4x$, la segunda derivada es $12x^2 - 4$ y, finalmente, la **tercera derivada es $24x$**. Al reunir estos elementos, la función se reduce a $P_2^1(x) = -\frac{1}{8} \sqrt{1 - x^2} (24x)$, lo cual se simplifica anaĺıticamente a **$-3x \sqrt{1 - x^2}$**. Para concluir el desarrollo empleando la sustitución **$x = \cos \theta$**, recordamos que el término radical $\sqrt{1 - x^2}$ es equivalente a $\sqrt{1 - \cos^2 \theta}$, que por identidad trigonométrica fundamental es igual a **$\sin \theta$**. De este modo, al sustituir tanto la variable $x$ como la raíz, determinamos que la forma final del polinomio solicitado es **$P_2^1(\cos \theta) = -3 \cos \theta \sin \theta$**.

# Ejemplo: Aplicar $G^-$ a $P_2^1$ para hallar $P_2^0$

Para aplicar el **operador escalera de descenso** ($G^-$) al polinomio asociado de Legendre **$P_2^1(x)$** con el fin de obtener **$P_2^0(x)$**, partimos de la propiedad diferencial de los operadores escalera presentada en las fuentes:

$$(l + m)(l - m + 1) \sqrt{1 - x^2} P_l^{m-1}(x) = \left[ mx - (1 - x^2) \frac{d}{dx} \right] P_l^m(x)$$

Identificando los índices para nuestro caso, tenemos **$l = 2$** y **$m = 1$**. Al sustituirlos en la expresión anterior, el lado izquierdo de la ecuación se convierte en $(2 + 1)(2 - 1 + 1) \sqrt{1 - x^2} P_2^0(x)$, lo que se simplifica a **$6\sqrt{1 - x^2} P_2^0(x)$**. Por otro lado, el operador aplicado al lado derecho toma la forma **$[x - (1 - x^2) \frac{d}{dx}] P_2^1(x)$**. 

Utilizando la expresión de **$P_2^1(x) = 3x\sqrt{1 - x^2}$** (definida en la convención del texto de Sepúlveda para ser consistente con el operador dado), procedemos a realizar las operaciones indicadas:

1.  **Multiplicación por $x$:** El primer término es $x(3x\sqrt{1 - x^2}) = \mathbf{3x^2\sqrt{1 - x^2}}$.
2.  **Derivación de $P_2^1(x)$:** La derivada respecto a $x$ de $3x(1 - x^2)^{1/2}$ es $3(1 - x^2)^{1/2} + 3x \cdot \frac{1}{2}(1 - x^2)^{-1/2}(-2x)$, lo que resulta en **$3\sqrt{1 - x^2} - \frac{3x^2}{\sqrt{1 - x^2}}$**.
3.  **Aplicación del factor $-(1 - x^2)$:** Al multiplicar la derivada anterior por $-(1 - x^2)$, obtenemos **$-3(1 - x^2)^{3/2} + 3x^2\sqrt{1 - x^2}$**.

Al sumar ambos resultados para completar el lado derecho de la ecuación original, tenemos:
$$3x^2\sqrt{1 - x^2} - 3(1 - x^2)\sqrt{1 - x^2} + 3x^2\sqrt{1 - x^2}$$
Factorizando el término común $\sqrt{1 - x^2}$, la expresión se reduce a $\sqrt{1 - x^2} [3x^2 - 3(1 - x^2) + 3x^2]$, que se simplifica a **$3\sqrt{1 - x^2}(3x^2 - 1)$**.

Finalmente, igualamos ambos lados de la ecuación:
$$6\sqrt{1 - x^2} P_2^0(x) = 3\sqrt{1 - x^2}(3x^2 - 1)$$
Al dividir por $6\sqrt{1 - x^2}$, hallamos que **$P_2^0(x) = \frac{1}{2}(3x^2 - 1)$**, lo cual coincide exactamente con la definición del polinomio de Legendre ordinario **$P_2(x)$**.

# Armónicos esféricos $Y_\ell^m$

Es la base biortonormal formada a partir de las bases ortonormales:

* $\{\varphi_m(x)\} = \left\{\frac{1}{\sqrt{2\pi}}e^{i m x}\right\}$
* $\{\varphi_{\ell m}(x)\} = \left\{\sqrt{\frac{2\ell+1}{2}}\sqrt{\frac{(\ell-m)!}{(\ell+m)!}}P_{\ell}^m(x)\right\}$


$$Y_\ell^m(\theta, \phi) = \sqrt{\frac{2\ell+1}{4\pi} \frac{(\ell-m)!}{(\ell+m)!}} P_\ell^m(\cos \theta) e^{im\phi}$$

**Ortonormalidad:** Estas funciones son ortonormales sobre la superficie de una esfera unitaria, cumpliendo que $\int_{0}^{2\pi} \int_{0}^{\pi} [Y_{\ell_1}^{m_1}(\theta, \phi)]^* Y_{\ell_2}^{m_2}(\theta, \phi) \sin \theta d\theta d\phi = \delta_{\ell_1\ell_2} \delta_{m_1m_2}$.

Físicamente, los armónicos esféricos son las **autofunciones del operador de momento angular** al cuadrado ($L^2$) con autovalores $\ell(\ell+1)$ y de su componente en $z$ ($L_z$) con autovalores $m$. Constituyen una base completa, lo que permite expandir cualquier función definida sobre la superficie de una esfera (serie de Laplace).

La solución general a la **ecuación de Laplace** ($\nabla^2\phi = 0$) en coordenadas esféricas, expresada en términos de los **armónicos esféricos** $Y_\ell^m(\theta, \varphi)$, se escribe de la siguiente manera:

$$\phi(r, \theta, \varphi) = \sum_{\ell=0}^\infty \sum_{m=-\ell}^\ell \left( A_{\ell m} r^\ell + \frac{B_{\ell m}}{r^{\ell+1}} \right) Y_\ell^m(\theta, \varphi)$$

# Ejemplo: Hallar $Y_0^0, Y_2^1$

1.  **Definición del producto interno:** En un espacio de Hilbert, el producto interno de una función consigo misma representa su norma al cuadrado. Para los armónicos esféricos, la relación entre la notación de Dirac y la representación integral es:
    $$\langle Y_0^0 | Y_0^0 \rangle = \int_{\Omega} [Y_0^0(\theta, \phi)]^* Y_0^0(\theta, \phi) d\Omega$$

2.  **Sustitución del valor funcional:** Utilizando el valor hallado anteriormente, $Y_0^0 = \frac{1}{\sqrt{4\pi}}$:
    $$\langle Y_0^0 | Y_0^0 \rangle = \int_{0}^{2\pi} \int_{0}^{\pi} \left( \frac{1}{\sqrt{4\pi}} \right)^* \left( \frac{1}{\sqrt{4\pi}} \right) \sin \theta d\theta d\phi$$

3.  **Evaluación de la integral de normalización:**
    $$\langle Y_0^0 | Y_0^0 \rangle = \frac{1}{4\pi} \int_{0}^{2\pi} d\phi \int_{0}^{\pi} \sin \theta d\theta$$
    Como la integral sobre todo el ángulo sólido $\int d\Omega$ es igual a $4\pi$:
    $$\langle Y_0^0 | Y_0^0 \rangle = \frac{1}{4\pi} (4\pi) = \mathbf{1}$$

De manera idéntica, para **$Y_2^1$**, se escribiría:
$$\langle Y_2^1 | Y_2^1 \rangle = \int_{\Omega} |Y_2^1|^2 d\Omega = \mathbf{1}$$

Este formalismo demuestra que los armónicos esféricos constituyen una **base ortonormal** completa, donde de forma general se cumple que:
$$\langle Y_\ell^m | Y_{\ell'}^{m'} \rangle = \delta_{\ell\ell'} \delta_{mm'}$$

# Ejemplo: Ejercicio de arfken

### **Enunciado del Problema (Ejemplo 15.5.1 de Arfken)**
Determine el potencial electrostático dentro de una región esférica de radio $r_0$ libre de cargas, donde el potencial en la superficie de la frontera esférica se especifica como una función arbitraria $V(r_0, \theta, \phi)$ de las coordenadas angulares $\theta$ y $\phi$.

### **Solución del Problema**

Para resolver este problema, partimos de la solución general de la **ecuación de Laplace** ($\nabla^2\phi = 0$) en coordenadas esféricas, la cual se expresa como una superposición de productos de funciones radiales y armónicos esféricos:
$$\phi(r, \theta, \phi) = \sum_{\ell=0}^\infty \sum_{m=-\ell}^\ell \left( A_{\ell m} r^\ell + \frac{B_{\ell m}}{r^{\ell+1}} \right) Y_\ell^m(\theta, \phi)$$.

El primer paso consiste en aplicar las condiciones físicas de la región de interés. Dado que buscamos el potencial en el **interior** de la esfera ($r \leq r_0$) y este debe permanecer finito en el origen ($r=0$), debemos descartar los términos con potencias negativas de $r$, ya que divergen cuando $r \to 0$. Esto implica que el coeficiente **$B_{\ell m}$ debe ser igual a cero** para todos los valores de $\ell$ y $m$. En consecuencia, la expresión para el potencial se reduce a:
$$\phi(r, \theta, \phi) = \sum_{\ell=0}^\infty \sum_{m=-\ell}^\ell A_{\ell m} r^\ell Y_\ell^m(\theta, \phi)$$.

A continuación, imponemos la **condición de frontera** en la superficie de la esfera ($r = r_0$), donde el potencial debe coincidir con la función dada $V(r_0, \theta, \phi)$. Esto nos lleva a la ecuación:
$$V(r_0, \theta, \phi) = \sum_{\ell=0}^\infty \sum_{m=-\ell}^\ell A_{\ell m} r_0^\ell Y_\ell^m(\theta, \phi)$$.

Esta expresión se identifica como una **expansión de Laplace** de la función superficial en la base de armónicos esféricos. Para determinar los coeficientes desconocidos $A_{\ell m}$, utilizamos la **propiedad de ortonormalidad** de los armónicos esféricos sobre la superficie de una esfera unitaria:
$$\int_0^{2\pi} \int_0^\pi [Y_{\ell'}^{m'}(\theta, \phi)]^* Y_\ell^m(\theta, \phi) \sin\theta d\theta d\phi = \delta_{\ell \ell'} \delta_{m m'}$$.

Multiplicando ambos lados de la condición de frontera por el complejo conjugado $[Y_\ell^m(\theta, \phi)]^* \sin\theta$ e integrando sobre todos los ángulos, aislamos el término del coeficiente deseado, obteniendo:
$$A_{\ell m} r_0^\ell = \int_0^{2\pi} \int_0^\pi V(r_0, \theta, \phi) [Y_\ell^m(\theta, \phi)]^* \sin\theta d\theta d\phi$$
De donde se concluye que los coeficientes $A_{\ell m}$ están definidos por la integral:
$$A_{\ell m} = \frac{1}{r_0^\ell} \int_0^{2\pi} \int_0^\pi V(r_0, \theta, \phi) [Y_\ell^m(\theta, \phi)]^* \sin\theta d\theta d\phi$$.

Finalmente, al sustituir estos coeficientes en nuestra serie original, determinamos que el potencial en cualquier punto del interior de la esfera se puede escribir de forma compacta como:
$$\phi(r, \theta, \phi) = \sum_{\ell=0}^\infty \sum_{m=-\ell}^\ell c_{\ell m} \left( \frac{r}{r_0} \right)^\ell Y_\ell^m(\theta, \phi)$$
Donde $c_{\ell m}$ representa los coeficientes de la expansión de Laplace del potencial especificado en la frontera. Este resultado garantiza una solución única y continua que satisface tanto la ecuación diferencial en el volumen como las condiciones impuestas en la superficie.

# Ejemplo

### **Enunciado del Problema**
Determine el potencial electrostático $\phi(r, \theta, \varphi)$ en la región comprendida entre dos cáscaras esféricas concéntricas de radios **$a$** y **$b$** ($a < r < b$). La superficie interior ($r=a$) se mantiene a un potencial fijo dado por una función arbitraria **$V_a(\theta, \varphi)$**, y la superficie exterior ($r=b$) se mantiene a un potencial **$V_b(\theta, \varphi)$**.

### **Solución y Desarrollo Analítico**

**1. Planteamiento de la Solución General**
Dado que la región de interés no contiene cargas eléctricas, el potencial satisface la **ecuación de Laplace** ($\nabla^2\phi = 0$). En coordenadas esféricas, la solución más general que es finita en sus variables angulares es la expansión en armónicos esféricos $Y_\ell^m(\theta, \varphi)$:
$$\phi(r, \theta, \varphi) = \sum_{\ell=0}^\infty \sum_{m=-\ell}^\ell \left( A_{\ell m} r^\ell + \frac{B_{\ell m}}{r^{\ell+1}} \right) Y_\ell^m(\theta, \varphi)$$
En este problema, a diferencia de los casos donde se estudia solo el interior o el exterior de una esfera, **debemos conservar ambos términos radiales**. Los términos $r^\ell$ son necesarios para satisfacer la condición en la frontera interior, y los términos $r^{-(\ell+1)}$ son necesarios para la frontera exterior, ya que la región no incluye ni el origen ($r=0$) ni el infinito ($r \to \infty$).

**2. Aplicación de las Condiciones de Frontera**
Para determinar los coeficientes $A_{\ell m}$ y $B_{\ell m}$, expandimos los potenciales conocidos en las fronteras utilizando la **expansión de Laplace**:
*   En **$r = a$**: $V_a(\theta, \varphi) = \sum_{\ell, m} \alpha_{\ell m} Y_\ell^m(\theta, \varphi)$
*   En **$r = b$**: $V_b(\theta, \varphi) = \sum_{\ell, m} \beta_{\ell m} Y_\ell^m(\theta, \varphi)$

Donde los coeficientes de superficie $\alpha_{\ell m}$ y $\beta_{\ell m}$ se hallan mediante la propiedad de **ortonormalidad** de los armónicos esféricos:
$$\alpha_{\ell m} = \int_0^{2\pi} \int_0^\pi V_a(\theta, \varphi) [Y_\ell^m(\theta, \varphi)]^* \sin\theta d\theta d\varphi$$

**3. Sistema de Ecuaciones Lineales**
Al igualar término a término la expansión general con los potenciales en $r=a$ y $r=b$, obtenemos para cada par $(\ell, m)$ un sistema de dos ecuaciones con dos incógnitas:
1.  $A_{\ell m} a^\ell + B_{\ell m} a^{-(\ell+1)} = \alpha_{\ell m}$
2.  $A_{\ell m} b^\ell + B_{\ell m} b^{-(\ell+1)} = \beta_{\ell m}$

**4. Resolución de los Coeficientes**
Multiplicando la primera ecuación por $a^{\ell+1}$ y la segunda por $b^{\ell+1}$, y resolviendo algebraicamente, se obtienen las expresiones para los coeficientes:
$$A_{\ell m} = \frac{\beta_{\ell m} b^{\ell+1} - \alpha_{\ell m} a^{\ell+1}}{b^{2\ell+1} - a^{2\ell+1}}$$
$$B_{\ell m} = \frac{\alpha_{\ell m} a^{\ell+1} b^{2\ell+1} - \beta_{\ell m} b^{\ell+1} a^{2\ell+1}}{b^{2\ell+1} - a^{2\ell+1}}$$

**5. Resultado Final**
Sustituyendo estos valores en la serie original, obtenemos una expresión única para el potencial en cualquier punto del espacio anular que satisface exactamente la ecuación diferencial y las condiciones impuestas en ambas cáscaras. Este tipo de desarrollo es fundamental para estudiar campos en **condensadores esféricos asimétricos** o en la **polarización de medios dieléctricos** entre electrodos esféricos.

# Ejercicio: Solución completa para $Y_2^1$

A continuación presento el desarrollo **completo y paso a paso** para obtener la expresión explícita de $Y_2^1(\theta, \phi)$ y demostrar su normalización, haciendo énfasis en la aparición y evaluación del polinomio asociado de Legendre $P_\ell^m(\cos\theta)$, tal como lo solicitaste.

---

### 1. Definición general de los Armónicos Esféricos
La fórmula general para los armónicos esféricos en física (incluyendo la fase de Condon-Shortley) es:

$$Y_\ell^m(\theta, \phi) = (-1)^m \sqrt{\frac{2\ell + 1}{4\pi} \frac{(\ell - m)!}{(\ell + m)!}} P_\ell^m(\cos\theta) e^{im\phi}$$

Para nuestro caso, tenemos **$\ell = 2$** y **$m = 1$**.

### 2. Cálculo de la constante de normalización
Sustituimos $\ell = 2$ y $m = 1$ en el factor constante:

$$N = \sqrt{\frac{2(2) + 1}{4\pi} \frac{(2 - 1)!}{(2 + 1)!}} = \sqrt{\frac{5}{4\pi} \frac{1!}{3!}} = \sqrt{\frac{5}{4\pi} \frac{1}{6}} = \sqrt{\frac{5}{24\pi}}$$

### 3. Obtención del Polinomio Asociado de Legendre $P_2^1(\cos\theta)$
Este es el paso clave que a menudo se omite. El polinomio asociado de Legendre se define como:

$$P_\ell^m(x) = (-1)^m (1 - x^2)^{m/2} \frac{d^m}{dx^m} P_\ell(x)$$

Primero, recordamos el polinomio de Legendre ordinario para $\ell = 2$:
$$P_2(x) = \frac{1}{2}(3x^2 - 1)$$

Ahora aplicamos la definición para $m = 1$ y $x = \cos\theta$:
$$P_2^1(x) = (-1)^1 (1 - x^2)^{1/2} \frac{d}{dx} \left[ \frac{1}{2}(3x^2 - 1) \right]$$

Calculamos la derivada:
$$\frac{d}{dx} \left[ \frac{1}{2}(3x^2 - 1) \right] = \frac{1}{2}(6x) = 3x$$

Sustituimos de vuelta:
$$P_2^1(x) = - (1 - x^2)^{1/2} (3x) = -3x\sqrt{1 - x^2}$$

Finalmente, hacemos el cambio de variable $x = \cos\theta$. Sabemos que $\sqrt{1 - \cos^2\theta} = \sin\theta$ (ya que $\theta \in [0, \pi]$, el seno es positivo):
$$P_2^1(\cos\theta) = -3\cos\theta \sin\theta$$

### 4. Expresión completa de $Y_2^1(\theta, \phi)$
Unimos la constante, el polinomio de Legendre y la parte azimutal:

$$Y_2^1(\theta, \phi) = \sqrt{\frac{5}{24\pi}} \cdot (-3\cos\theta \sin\theta) \cdot e^{i\phi}$$

Podemos simplificar la constante metiendo el $3$ dentro de la raíz ($3 = \sqrt{9}$):
$$Y_2^1(\theta, \phi) = -\sqrt{\frac{5 \cdot 9}{24\pi}} \sin\theta \cos\theta e^{i\phi} = -\sqrt{\frac{45}{24\pi}} \sin\theta \cos\theta e^{i\phi}$$

Simplificando la fracción $\frac{45}{24} = \frac{15}{8}$:
$$Y_2^1(\theta, \phi) = -\sqrt{\frac{15}{8\pi}} \sin\theta \cos\theta e^{i\phi}$$

---

### 5. Proceso de Normalización $\langle Y_2^1 | Y_2^1 \rangle$
Ahora evaluamos la integral de la norma al cuadrado sobre el ángulo sólido $d\Omega = \sin\theta \, d\theta \, d\phi$:

$$\langle Y_2^1 | Y_2^1 \rangle = \int_{0}^{2\pi} \int_{0}^{\pi} \left| Y_2^1(\theta, \phi) \right|^2 \sin\theta \, d\theta \, d\phi$$

Calculamos el módulo al cuadrado de la función (el signo negativo y la fase compleja $e^{i\phi}$ desaparecen al multiplicar por su complejo conjugado):
$$\left| Y_2^1(\theta, \phi) \right|^2 = \left( -\sqrt{\frac{15}{8\pi}} \sin\theta \cos\theta e^{i\phi} \right)^* \left( -\sqrt{\frac{15}{8\pi}} \sin\theta \cos\theta e^{i\phi} \right) = \frac{15}{8\pi} \sin^2\theta \cos^2\theta$$

Sustituimos en la integral:
$$\langle Y_2^1 | Y_2^1 \rangle = \int_{0}^{2\pi} \int_{0}^{\pi} \left( \frac{15}{8\pi} \sin^2\theta \cos^2\theta \right) \sin\theta \, d\theta \, d\phi$$

Sacamos la constante y separamos las integrales, ya que los límites son constantes y el integrando es separable:
$$\langle Y_2^1 | Y_2^1 \rangle = \frac{15}{8\pi} \left( \int_{0}^{2\pi} d\phi \right) \left( \int_{0}^{\pi} \sin^3\theta \cos^2\theta \, d\theta \right)$$

#### 5.1. Evaluación de la integral en $\phi$
$$\int_{0}^{2\pi} d\phi = 2\pi$$

#### 5.2. Evaluación de la integral en $\theta$
$$I_\theta = \int_{0}^{\pi} \sin^3\theta \cos^2\theta \, d\theta$$
Para resolverla, usamos el cambio de variable $u = \cos\theta$, por lo tanto $du = -\sin\theta \, d\theta$.
Los límites cambian: si $\theta = 0 \rightarrow u = 1$; si $\theta = \pi \rightarrow u = -1$.
Además, $\sin^2\theta = 1 - \cos^2\theta = 1 - u^2$.

$$I_\theta = \int_{1}^{-1} (1 - u^2) u^2 (-du) = \int_{-1}^{1} (u^2 - u^4) \, du$$

Como la función $(u^2 - u^4)$ es par, podemos integrar de $0$ a $1$ y multiplicar por $2$:
$$I_\theta = 2 \int_{0}^{1} (u^2 - u^4) \, du = 2 \left[ \frac{u^3}{3} - \frac{u^5}{5} \right]_{0}^{1}$$
$$I_\theta = 2 \left( \frac{1}{3} - \frac{1}{5} \right) = 2 \left( \frac{5 - 3}{15} \right) = 2 \left( \frac{2}{15} \right) = \frac{4}{15}$$

### 6. Resultado Final
Multiplicamos todos los componentes evaluados:

$$\langle Y_2^1 | Y_2^1 \rangle = \left( \frac{15}{8\pi} \right) \cdot (2\pi) \cdot \left( \frac{4}{15} \right)$$

Simplificamos paso a paso:
1. Los $15$ en el numerador y denominador se cancelan.
2. Los $\pi$ se cancelan.
3. Queda: $\frac{1}{8} \cdot 2 \cdot 4 = \frac{8}{8} = 1$

$$\langle Y_2^1 | Y_2^1 \rangle = \mathbf{1}$$

---
**Conclusión:** El desarrollo explícito confirma que, al incluir correctamente el factor del polinomio asociado de Legendre $P_2^1(\cos\theta) = -3\cos\theta\sin\theta$ y su respectiva constante de normalización, la integral sobre el ángulo sólido converge exactamente a $1$, validando que $Y_2^1$ está correctamente normalizado.

# Tarea: Repasar las propiedades del operador $L^2$

### 1. Propiedades Generales del Operador $\hat{L}^2$

El operador $\hat{L}^2$ representa el cuadrado de la magnitud del momento angular orbital en mecánica cuántica. Sus propiedades fundamentales son:

*   **Definición algebraica:** Se define como la suma de los cuadrados de las componentes cartesianas del momento angular:
    $$\hat{L}^2 = \hat{L}_x^2 + \hat{L}_y^2 + \hat{L}_z^2$$
*   **Representación en coordenadas esféricas:** En el espacio de posiciones, $\hat{L}^2$ actúa únicamente sobre las variables angulares $(\theta, \phi)$ y no depende de la distancia radial $r$. Su forma explícita es:
    $$\hat{L}^2 = -\hbar^2 \left[ \frac{1}{\sin\theta} \frac{\partial}{\partial\theta} \left( \sin\theta \frac{\partial}{\partial\theta} \right) + \frac{1}{\sin^2\theta} \frac{\partial^2}{\partial\phi^2} \right]$$
    *(Nota: El término entre corchetes es exactamente el operador de Laplace-Beltrami en la esfera, a menudo denotado como $\nabla^2_{\Omega}$).*
*   **Conmutación con las componentes:** $\hat{L}^2$ conmuta con cualquiera de las componentes cartesianas del momento angular:
    $$[\hat{L}^2, \hat{L}_x] = [\hat{L}^2, \hat{L}_y] = [\hat{L}^2, \hat{L}_z] = 0$$
    Esta es la propiedad más importante, ya que garantiza que $\hat{L}^2$ y $\hat{L}_z$ (por convención) pueden tener **autoestados simultáneos**.
*   **Hermiticidad:** $\hat{L}^2$ es un operador hermítico (autoadjunto). Esto garantiza dos cosas:
    1. Sus autovalores son números reales.
    2. Sus autofunciones correspondientes a autovalores diferentes son ortogonales.
*   **Definido positivo:** Dado que $\hat{L}_x, \hat{L}_y, \hat{L}_z$ son hermíticos, el valor esperado de $\hat{L}^2$ para cualquier estado $|\psi\rangle$ es la suma de normas al cuadrado:
    $$\langle \psi | \hat{L}^2 | \psi \rangle = \langle \psi | \hat{L}_x^2 | \psi \rangle + \langle \psi | \hat{L}_y^2 | \psi \rangle + \langle \psi | \hat{L}_z^2 | \psi \rangle = \|\hat{L}_x \psi\|^2 + \|\hat{L}_y \psi\|^2 + \|\hat{L}_z \psi\|^2 \ge 0$$
    Por lo tanto, todos sus autovalores deben ser mayores o iguales a cero.

---

### 2. Propiedades de $\hat{L}^2$ actuando sobre los Armónicos Esféricos $Y_\ell^m$

Los armónicos esféricos $Y_\ell^m(\theta, \phi)$ son, por definición, las autofunciones (autoestados) del operador $\hat{L}^2$. Cuando $\hat{L}^2$ actúa sobre ellos, se manifiestan las siguientes propiedades:

*   **Ecuación de Autovalores:** La acción de $\hat{L}^2$ sobre $Y_\ell^m$ devuelve la misma función multiplicada por una constante (el autovalor):
    $$\hat{L}^2 Y_\ell^m(\theta, \phi) = \hbar^2 \ell(\ell + 1) Y_\ell^m(\theta, \phi)$$
    Donde $\ell = 0, 1, 2, 3, \dots$ es el **número cuántico azimutal** (o de momento angular orbital).
*   **Cuantización de la magnitud:** El resultado anterior nos dice que la magnitud al cuadrado del momento angular no puede tomar cualquier valor, sino que está cuantizada en unidades de $\hbar^2 \ell(\ell+1)$. La magnitud del vector momento angular es $|\vec{L}| = \hbar\sqrt{\ell(\ell+1)}$.
*   **Independencia de $m$ (Degeneración):** Observa que el autovalor $\hbar^2 \ell(\ell+1)$ **no depende de $m$**. Para un valor dado de $\ell$, el número cuántico magnético $m$ puede tomar $2\ell + 1$ valores enteros ($m = -\ell, -\ell+1, \dots, \ell$). 
    *   Esto significa que existen $2\ell + 1$ funciones diferentes ($Y_\ell^{-\ell}, \dots, Y_\ell^{\ell}$) que son autofunciones de $\hat{L}^2$ con el **mismo autovalor**.
    *   A esto se le llama **degeneración $(2\ell + 1)$-ple**.
*   **Autoestados Simultáneos con $\hat{L}_z$:** Como se mencionó, $[\hat{L}^2, \hat{L}_z] = 0$. Por lo tanto, los $Y_\ell^m$ también son autofunciones de $\hat{L}_z$:
    $$\hat{L}_z Y_\ell^m(\theta, \phi) = m\hbar Y_\ell^m(\theta, \phi)$$
    El par de operadores $\{\hat{L}^2, \hat{L}_z\}$ forma un **Conjunto Completo de Observables que Conmutan (CCOC)** para las variables angulares, lo que permite etiquetar unívocamente cada estado con los dos números cuánticos $(\ell, m)$.
*   **Relación con el Laplaciano y la Energía:** En problemas con simetría esférica (como el átomo de hidrógeno o el oscilador armónico 3D), el operador energía cinética contiene al Laplaciano $\nabla^2$. En coordenadas esféricas, el Laplaciano se separa en una parte radial y una parte angular que es proporcional a $\hat{L}^2$:
    $$\nabla^2 = \frac{1}{r^2} \frac{\partial}{\partial r} \left( r^2 \frac{\partial}{\partial r} \right) - \frac{\hat{L}^2}{\hbar^2 r^2}$$
    Debido a que $\hat{L}^2 Y_\ell^m = \hbar^2 \ell(\ell+1) Y_\ell^m$, al aplicar el Laplaciano a una función de onda separable $\psi(r, \theta, \phi) = R(r)Y_\ell^m(\theta, \phi)$, la parte angular se reemplaza simplemente por la constante $\ell(\ell+1)$, dando origen al llamado **término de barrera centrífuga** $\frac{\hbar^2 \ell(\ell+1)}{2mr^2}$ en la ecuación radial efectiva.

---

### Resumen en una sola idea
El operador $\hat{L}^2$ mide el "tamaño" del momento angular. Al actuar sobre los armónicos esféricos $Y_\ell^m$, los deja invariantes en forma, escalándolos únicamente por el factor $\hbar^2 \ell(\ell+1)$, lo que demuestra que los armónicos esféricos son los estados de momento angular bien definido, degenerados $2\ell+1$ veces debido a las diferentes orientaciones posibles ($m$) que no alteran la magnitud total del momento.

# Tarea: Haga el desarrollo para calcular $L^2Y_1^1$ explícitamente

### 1. Definiciones
Operador de momento angular al cuadrado:
$$ \hat{L}^2 = -\hbar^2 \left[ \frac{1}{\sin\theta} \frac{\partial}{\partial\theta} \left( \sin\theta \frac{\partial}{\partial\theta} \right) + \frac{1}{\sin^2\theta} \frac{\partial^2}{\partial\phi^2} \right] $$

Armónico esférico para $\ell=1, m=1$:
$$ Y_1^1(\theta, \phi) = C \sin\theta e^{i\phi} \quad \text{donde} \quad C = -\sqrt{\frac{3}{8\pi}} $$

### 2. Evaluación de las derivadas parciales

**Parte azimutal ($\phi$):**
$$ \frac{\partial}{\partial\phi} \left( C \sin\theta e^{i\phi} \right) = i C \sin\theta e^{i\phi} $$
$$ \frac{\partial^2}{\partial\phi^2} \left( C \sin\theta e^{i\phi} \right) = i^2 C \sin\theta e^{i\phi} = -C \sin\theta e^{i\phi} $$

**Parte polar ($\theta$):**
Primera derivada:
$$ \frac{\partial}{\partial\theta} \left( C \sin\theta e^{i\phi} \right) = C \cos\theta e^{i\phi} $$

Multiplicación por $\sin\theta$:
$$ \sin\theta \frac{\partial Y_1^1}{\partial\theta} = C \sin\theta \cos\theta e^{i\phi} $$

Segunda derivada (regla del producto):
$$ \frac{\partial}{\partial\theta} \left( C \sin\theta \cos\theta e^{i\phi} \right) = C e^{i\phi} \left( \cos\theta \cdot \cos\theta + \sin\theta \cdot (-\sin\theta) \right) = C e^{i\phi} (\cos^2\theta - \sin^2\theta) $$

División por $\sin\theta$:
$$ \frac{1}{\sin\theta} \frac{\partial}{\partial\theta} \left( \sin\theta \frac{\partial Y_1^1}{\partial\theta} \right) = C e^{i\phi} \left( \frac{\cos^2\theta - \sin^2\theta}{\sin\theta} \right) $$

### 3. Sustitución en el operador $\hat{L}^2$

$$ \hat{L}^2 Y_1^1 = -\hbar^2 \left[ C e^{i\phi} \left( \frac{\cos^2\theta - \sin^2\theta}{\sin\theta} \right) + \frac{1}{\sin^2\theta} \left( -C \sin\theta e^{i\phi} \right) \right] $$

Factorizamos $C e^{i\phi}$:
$$ \hat{L}^2 Y_1^1 = -\hbar^2 C e^{i\phi} \left[ \frac{\cos^2\theta - \sin^2\theta}{\sin\theta} - \frac{\sin\theta}{\sin^2\theta} \right] $$

Simplificamos el segundo término del corchete:
$$ \hat{L}^2 Y_1^1 = -\hbar^2 C e^{i\phi} \left[ \frac{\cos^2\theta - \sin^2\theta}{\sin\theta} - \frac{1}{\sin\theta} \right] $$

Combinamos las fracciones:
$$ \hat{L}^2 Y_1^1 = -\hbar^2 C e^{i\phi} \left[ \frac{\cos^2\theta - \sin^2\theta - 1}{\sin\theta} \right] $$

### 4. Simplificación trigonométrica
Usamos la identidad $\cos^2\theta - 1 = -\sin^2\theta$:
$$ \cos^2\theta - \sin^2\theta - 1 = (\cos^2\theta - 1) - \sin^2\theta = -\sin^2\theta - \sin^2\theta = -2\sin^2\theta $$

Sustituimos en el corchete:
$$ \hat{L}^2 Y_1^1 = -\hbar^2 C e^{i\phi} \left[ \frac{-2\sin^2\theta}{\sin\theta} \right] $$
$$ \hat{L}^2 Y_1^1 = -\hbar^2 C e^{i\phi} \left[ -2\sin\theta \right] $$

### 5. Resultado final
Reagrupamos los términos para recuperar la forma original de $Y_1^1$:
$$ \hat{L}^2 Y_1^1 = 2\hbar^2 \left( C \sin\theta e^{i\phi} \right) $$

$$ \hat{L}^2 Y_1^1 = 2\hbar^2 Y_1^1 $$

Lo cual verifica la ecuación de autovalores $\hat{L}^2 Y_\ell^m = \hbar^2 \ell(\ell+1) Y_\ell^m$ para $\ell=1$, ya que $1(1+1) = 2$.
