**Ejemplo 2:** Disco aislante de radio R y carga +Q

Halle $\rho(\vec{r})$

Para hallar la densidad volumétrica de carga $\rho$ de un **disco aislante** de radio $R$ y carga total $+Q$ utilizando el sistema de coordenadas esféricas $(r, \varphi, \theta)$, donde $r$ es la distancia radial, $\varphi$ el ángulo azimutal y $\theta$ el ángulo polar (colatitud), procedemos con el siguiente análisis basado en la teoría de distribuciones y cálculo integral en coordenadas curvilíneas.

### 1. Definición del sistema y geometría del disco
En este sistema de coordenadas, el diferencial de volumen ($dV$) se expresa a partir de sus factores de escala como:
$$dV = r^2 \sin\theta \, dr \, d\varphi \, d\theta$$
El disco está situado en el **plano ecuatorial**, lo que corresponde matemáticamente a la condición **$\theta = \pi/2$**. Además, la carga está confinada a una región finita donde la distancia radial cumple **$r \leq R$**.

### 2. Relación entre densidades y carga total
La carga total $Q$ es la integral de la densidad volumétrica $\rho$ sobre todo el espacio:
$$Q = \int_0^{2\pi} d\varphi \int_0^\pi d\theta \int_0^R \rho(r, \varphi, \theta) \, r^2 \sin\theta \, dr$$
Para representar una distribución superficial (2D) como una densidad volumétrica (3D), debemos utilizar la **delta de Dirac** $\delta(\theta - \pi/2)$ para "colapsar" la carga en el plano deseado.

Proponemos una densidad de la forma $\rho = f(r) \delta(\theta - \pi/2)$, donde $f(r)$ es una función radial que determinará la normalización correcta respecto a la densidad superficial $\sigma = \frac{Q}{\pi R^2}$.

### 3. Aplicación de la propiedad selectiva de la delta
Sustituimos nuestra propuesta en la integral de carga total:
$$Q = \int_0^{2\pi} d\varphi \int_0^R f(r) r^2 \left( \int_0^\pi \sin\theta \delta(\theta - \pi/2) d\theta \right) dr$$
Utilizando la **propiedad selectiva** de la delta de Dirac, evaluamos la integral en $\theta$:
$$\int_0^\pi \sin\theta \delta(\theta - \pi/2) d\theta = \sin(\pi/2) = 1$$
Esto reduce la expresión a:
$$Q = 2\pi \int_0^R f(r) r^2 dr$$

### 4. Determinación del factor radial
Sabemos que, para un disco con densidad superficial $\sigma$, la carga total también se calcula integrando sobre su área ($dA = r \, dr \, d\varphi$ en el plano):
$$Q = \int_0^{2\pi} d\varphi \int_0^R \sigma r \, dr = 2\pi \sigma \int_0^R r \, dr$$
Al comparar ambas integrales de $Q$, igualamos los integrandos radiales para asegurar que la densidad volumétrica sea equivalente a la superficial:
$$f(r) r^2 = \sigma r \implies \mathbf{f(r) = \frac{\sigma}{r}}$$

### 5. Resultado Final
Sustituyendo el factor radial $f(r)$ en nuestra propuesta original, llegamos a la expresión final solicitada para la densidad volumétrica de carga del disco:

$$\mathbf{\rho = \frac{\sigma}{r} \delta(\theta - \pi/2)}$$

*(Nota: Esta expresión es válida para $r \leq R$. El factor $1/r$ es esencial para compensar el factor $r^2$ del Jacobiano esférico, permitiendo que al integrar resulte un diferencial de carga proporcional a $r \, dr$, propio de una superficie plana)*

**Ejemplo 3:** Densidad volumétrica asociada a una carga puntual.

Para hallar la densidad volumétrica de carga $\rho(\vec{r})$ asociada a una **carga puntual** $q$ situada en el origen utilizando el sistema de coordenadas esféricas $(r, \varphi, \theta)$, se debe proceder mediante la integración de la densidad sobre todo el volumen del espacio y la aplicación de las propiedades de la **Delta de Dirac**.

A continuación se detalla el procedimiento paso a paso:

### 1. Definición de la carga total
Físicamente, la carga total $q$ se define como la integral de la densidad volumétrica $\rho(\vec{r})$ sobre todo el espacio:
$$q = \int_{V} \rho(\vec{r}) \, dV$$

### 2. Elemento de volumen en coordenadas esféricas
En el sistema solicitado $(r, \varphi, \theta)$, donde $r$ es el radio, $\varphi$ el ángulo azimutal y $\theta$ el ángulo polar (colatitud), el diferencial de volumen $dV$ se construye a partir de los factores de escala del sistema:
*   $h_r = 1$
*   $h_\varphi = r \sin\theta$
*   $h_\theta = r$

Por lo tanto, el diferencial de volumen es:
$$dV = (h_r \, dr)(h_\varphi \, d\varphi)(h_\theta \, d\theta) = r^2 \sin\theta \, dr \, d\varphi \, d\theta$$

### 3. Propuesta de la densidad con Delta de Dirac
Para localizar una carga puntual en un punto específico, la densidad debe ser nula en todas partes excepto en dicho punto. Dado que la expresión solicitada incluye una localización en el plano ecuatorial ($\theta = \pi/2$) y en el origen ($r = 0$), proponemos una densidad de la forma:
$$\rho(\vec{r}) = A \frac{1}{r^2} \delta(\theta - \pi/2) \delta(r - 0)$$

Donde $A$ es una constante de normalización que debemos determinar para asegurar que la carga total sea $q$. El factor $1/r^2$ se incluye para compensar la métrica del sistema de coordenadas.

### 4. Determinación de la constante de normalización
Sustituimos la propuesta en la integral de la carga total:
$$q = \int_0^{2\pi} d\varphi \int_0^\pi d\theta \int_0^\infty \left[ A \frac{1}{r^2} \delta(\theta - \pi/2) \delta(r - 0) \right] (r^2 \sin\theta) \, dr$$

1.  **Simplificación radial:** Los términos $r^2$ se cancelan dentro de la integral.
2.  **Separación de integrales:**
    $$q = A \left( \int_0^{2\pi} d\varphi \right) \left( \int_0^\pi \sin\theta \, \delta(\theta - \pi/2) \, d\theta \right) \left( \int_0^\infty \delta(r - 0) \, dr \right)$$
3.  **Evaluación de las integrales individuales:**
    *   Integral azimutal: $\int_0^{2\pi} d\varphi = 2\pi$.
    *   Integral polar (Propiedad selectiva): $\int_0^\pi \sin\theta \, \delta(\theta - \pi/2) \, d\theta = \sin(\pi/2) = 1$.
    *   Integral radial (Normalización de la delta): $\int_0^\infty \delta(r - 0) \, dr = 1$.

Multiplicando los resultados obtenemos:
$$q = A \cdot (2\pi) \cdot (1) \cdot (1) \implies A = \frac{q}{2\pi}$$

### 5. Resultado Final
Sustituyendo el valor de $A$ en la expresión propuesta, llegamos a la densidad volumétrica de carga para la partícula puntual:

$$\mathbf{\rho(\vec{r}) = \frac{q}{2\pi r^2} \delta(\theta - \pi/2) \delta(r - 0)}$$

Esta expresión indica que la carga se encuentra concentrada en el origen ($r=0$) y, aunque en ese punto la dirección es ambigua, matemáticamente se ha forzado su representación a través del plano ecuatorial ($\pi/2$) con simetría azimutal, resultando en la carga total $q$ tras la integración volumétrica.

**Tarea:**
Hallar $\rho(\vec{r})$ para un anillo de radio R y carga +Q uniforme

**Solución:**

Para hallar la densidad volumétrica de carga **$\rho(\vec{r})$** para un **anillo** de radio $R$ y carga total $+Q$ uniforme utilizando el sistema de coordenadas esféricas **$(r, \varphi, \theta)$**, debemos emplear la distribución **delta de Dirac** para concentrar la carga en una línea circular dentro del espacio tridimensional.

### 1. Definición del sistema y geometría del anillo
En el sistema de coordenadas esféricas solicitado $(r, \varphi, \theta)$ (donde $r$ es el radio, $\varphi$ el azimut y $\theta$ la colatitud polar), el anillo se sitúa en una posición fija en dos de las tres dimensiones:
*   **Componente radial:** La carga se encuentra exactamente a una distancia **$r = R$** del origen.
*   **Componente polar:** El anillo está en el plano ecuatorial, lo que corresponde a **$\theta = \pi/2$**.
*   **Componente azimutal:** La carga se distribuye continuamente a lo largo de todo el ángulo **$\varphi$** (de $0$ a $2\pi$).

El elemento de volumen diferencial en este sistema es:
$$dV = r^2 \sin\theta \, dr \, d\varphi \, d\theta$$

### 2. Relación con la densidad lineal ($\lambda$)
Puesto que la carga $+Q$ es uniforme, la **densidad lineal de carga** $\lambda$ es la carga total dividida por la longitud de la circunferencia:
$$\lambda = \frac{Q}{2\pi R}$$

Para representar esta línea de carga como una densidad volumétrica $\rho(\vec{r})$, debemos utilizar **dos deltas de Dirac** para colapsar las dimensiones radial y polar. Proponemos la densidad de la forma:
$$\rho(r, \varphi, \theta) = A \delta(r - R) \delta(\theta - \pi/2)$$
Donde $A$ es una constante de normalización que garantiza que la integral de volumen devuelva la carga total $Q$.

### 3. Cálculo de la constante de normalización
La carga total $Q$ se obtiene integrando la densidad sobre todo el espacio:
$$Q = \iiint \rho(\vec{r}) \, dV = \int_0^{2\pi} d\varphi \int_0^\pi d\theta \int_0^\infty \left[ A \delta(r - R) \delta(\theta - \pi/2) \right] r^2 \sin\theta \, dr$$

Separamos y evaluamos las integrales individuales aplicando la **propiedad selectiva** de la delta ($\int f(x)\delta(x-x_0)dx = f(x_0)$):
1.  **Integral radial:** $\int_0^\infty r^2 \delta(r - R) \, dr = R^2$.
2.  **Integral polar:** $\int_0^\pi \sin\theta \, \delta(\theta - \pi/2) \, d\theta = \sin(\pi/2) = 1$.
3.  **Integral azimutal:** $\int_0^{2\pi} d\varphi = 2\pi$.

Sustituyendo los resultados en la ecuación de carga total:
$$Q = A \cdot (2\pi) \cdot (1) \cdot (R^2) = 2\pi R^2 A$$
Despejamos la constante $A$:
$$A = \frac{Q}{2\pi R^2}$$

### 4. Resultado Final
Sustituyendo la constante $A$ en nuestra expresión original (o utilizando la relación con $\lambda$), obtenemos la densidad volumétrica asociada al anillo:

$$\mathbf{\rho(r, \varphi, \theta) = \frac{Q}{2\pi R^2} \delta(r - R) \delta(\theta - \pi/2)}$$

De manera alternativa, para simplificar el manejo de las integrales, la expresión suele escribirse incluyendo los términos métricos en el denominador para cancelar el Jacobiano:
$$\mathbf{\rho(r, \varphi, \theta) = \frac{Q}{2\pi r^2} \delta(r - R) \delta(\theta - \pi/2)}$$

Esta fórmula es consistente con la teoría de distribuciones en esféricas: el factor **$1/r^2$** compensa el crecimiento del volumen con el radio, y la delta **$\delta(\theta - \pi/2)$** sitúa la carga en el plano ecuatorial con un área unitaria ajustada por el factor $\sin(\pi/2)$.

# Bases discretas

Similar a los vectores. Una función puede expresarse en términos (combinación) de funciones en una base.

Espacio euclideo:
$$\vec{A} = A_i \hat{e}_i\equiv \text{Proyección}\cdot\text{Base}$$

Espacio de Hilbert (no es discreto):

$$f(x) = a_n\phi_n(x)\equiv \text{Coeficiente}\cdot\text{Base}$$

La analogía entre los vectores en espacios euclidianos y las funciones en espacios de Hilbert es fundamental para la física matemática, ya que permite tratar a las funciones como **vectores de un espacio abstracto de infinitas dimensiones**.

A continuación se detalla esta relación basada en las fuentes:

### 1. El Espacio Euclidiano n-Dimensional
En un espacio euclidiano, un conjunto de $n$ vectores linealmente independientes constituye una **base**. Si la base $\{\hat{e}_i\}$ es ortonormal, cualquier vector $\vec{A}$ se expresa como:
$$\vec{A} = \sum_{i=1}^{n} A_i \hat{e}_i$$
En esta estructura, la componente $A_i$ representa la **proyección** del vector sobre el vector unitario $\hat{e}_i$, calculada mediante el producto escalar $A_i = \vec{A} \cdot \hat{e}_i$. La suma de los productos diádicos de los vectores de la base ($\sum \hat{e}_i \hat{e}_i$) constituye la **identidad** del espacio.

### 2. El Espacio de Hilbert (Espacio de Funciones)
El espacio de Hilbert extiende estas nociones a conjuntos infinitos de funciones $\{\phi_n(x)\}$ que actúan como los "ejes" del espacio. 
*   **Representación:** Cualquier función $f(x)$ que sea de cuadrado integrable puede expresarse como una combinación lineal de las funciones de la base: $f(x) = \sum a_n \phi_n(x)$.
*   **Coeficientes:** Al igual que en el espacio euclidiano, los coeficientes $a_n$ son las "componentes" del vector función y se obtienen mediante el **producto interno integral**.
*   **Ortogonalidad:** Se define mediante la integral $\int_a^b p(x) \phi_n^*(x) \phi_m(x) dx = A_n \delta_{nm}$, donde $p(x)$ es una **función de peso** que puede ser necesaria según el sistema de coordenadas o la ecuación diferencial de origen.

### 3. Comparación de Elementos Clave
| Concepto | Espacio Euclidiano | Espacio de Hilbert (Base Discreta) |
| :--- | :--- | :--- |
| **Elemento** | Vector $\vec{A}$ | Función $f(x)$ |
| **Base** | Vectores unitarios $\{\hat{e}_i\}$ | Funciones ortogonales $\{\phi_n(x)\}$ |
| **Componente** | Proyección $A_i = \vec{A} \cdot \hat{e}_i$ | Coeficiente $a_n = \int p(x) f(x) \phi_n^*(x) dx$ |
| **Dimensión** | Finita ($n$) | Generalmente infinita |

### 4. Condición de Completez
Un aspecto crucial que une ambas representaciones es que, para que una base sea útil, debe ser **completa**. En el espacio de Hilbert, esto significa que la suma de las funciones de la base debe ser capaz de representar una **delta de Dirac**:
$$\sum_n p(x') \phi_n^*(x') \phi_n(x) = \delta(x - x')$$
Esta propiedad garantiza que no falta ninguna "dirección" en el espacio y que cualquier función puede ser expandida sin pérdida de información. Ejemplos clásicos de estas bases discretas son las funciones trigonométricas de la **serie de Fourier**, los polinomios de **Legendre** y las funciones de **Bessel**.

Basado en las lecciones de Alonso Sepúlveda, el producto interno (o escalar) es una operación fundamental que permite definir la geometría de los espacios vectoriales y de funciones. A continuación, se presenta un resumen ordenado de sus propiedades básicas sin recurrir a la notación de Dirac:

### 1. Propiedades Fundamentales (Definición Axiomática)
Para cualquier pareja de elementos $A$ y $B$ en un espacio vectorial, el producto interno $(A, B)$ satisface:
*   **Positividad:** $(A, A) \geq 0$. El producto de un vector consigo mismo siempre es mayor o igual a cero, y es exactamente cero si y solo si $A = 0$.
*   **Norma:** Permite definir la "longitud" o módulo del vector como $\|A\| = \sqrt{(A, A)}$.
*   **Distributividad:** $(A, B + C) = (A, B) + (A, C)$.

### 2. Simetría y Conjugación
El comportamiento varía si el espacio es real o complejo:
*   **En espacios reales:** Existe simetría total, es decir, $(A, B) = (B, A)$.
*   **En espacios complejos:** Existe **simetría conjugada**, donde $(A, B) = (B, A)^*$. Esto garantiza que la norma sea siempre un número real.

### 3. Linealidad y Antilinealidad
En espacios complejos (como los espacios de Hilbert), el producto interno se comporta de forma distinta según el factor que se multiplique por un escalar $\alpha$:
*   **Antilinealidad en el primer factor:** $(\alpha A, B) = \alpha^* (A, B)$. El escalar sale de la operación como su complejo conjugado.
*   **Linealidad en el segundo factor:** $(A, \alpha B) = \alpha (A, B)$.

### 4. Representación Integral para Funciones
En el contexto de espacios de funciones o de Hilbert, el producto interno se define mediante una integral en un intervalo $(a, b)$:
$$(f, g) = \int_a^b f^*(x)g(x) dx$$
*   **Ortogonalidad:** Dos funciones son ortogonales si su producto interno es nulo: $(f, g) = 0$.
*   **Función de peso:** En muchos problemas físicos (como en la teoría de Sturm-Liouville), se introduce una función de peso $p(x)$ real, de modo que la ortogonalidad se define como:
$$(p\phi_n, \phi_m) = \int_a^b p(x)\phi_n^*(x)\phi_m(x) dx = A_n \delta_{nm}$$
Donde $\delta_{nm}$ es la delta de Kronecker (1 si $n=m$, 0 si $n \neq m$).

# Espacio de Hilbert

De acuerdo con el texto, un **espacio lineal** es, por definición, un **espacio de Hilbert** si cumple con las siguientes condiciones:

*   **a)** Es un **espacio de funciones de dimensión infinita**.
*   **b)** Tiene definido un **producto escalar** $(f, g)$, el cual es **lineal respecto a $g$** (es decir, $L(ag) = aLg$) y **antilineal respecto a $f$** (es decir, $L(af) = a^*Lf$), donde $a$ es una constante compleja en general.

Este espacio debe satisfacer adicionalmente las condiciones de que $(g, f) = (f, g)^*$ y que $(f, f) > 0$ para todo $f \neq 0$. Asimismo, la cantidad $||f|| = \sqrt{(f, f)}$ se define como la **norma de $f$**.

# Función de cuadrado integrable

La relación entre una **función de cuadrado integrable** y los conceptos de espacios de funciones y bases ortogonales es fundamental, ya que define el criterio de pertenencia de una función a un **espacio de Hilbert**.

A continuación se detalla esta relación según las fuentes:

### 1. Definición y Pertenencia al Espacio de Hilbert
Una función $f(x)$ se denomina de **cuadrado integrable** en un intervalo $(a, b)$ si la integral del cuadrado de su módulo es finita:
$$\int_a^b |f(x)|^2 dx \neq \infty$$
Esta condición es necesaria para que la función tenga una **norma** definida y finita dentro de un espacio de Hilbert, donde la norma se calcula como $\|f\| = \sqrt{(f, f)}$.

# Ortogonalidad de la base

La **ortogonalidad de la base** es una propiedad fundamental tanto en espacios vectoriales finitos como en espacios de funciones (espacios de Hilbert), permitiendo simplificar el cálculo de componentes y proyecciones de manera similar a como se hace con los ejes cartesianos.

A continuación se detallan sus aspectos principales según las fuentes:

### 1. Definición Matemática
Un conjunto de elementos $\{\phi_n\}$ constituye una base ortogonal si el **producto escalar** de dos elementos distintos es nulo.
*   **En espacios discretos:** Se utiliza la delta de Kronecker ($\delta_{nm}$), la cual es 1 si $n=m$ y 0 si $n \neq m$.
*   **En espacios de funciones:** La ortogonalidad se define mediante una integral en un intervalo $(a, b)$ con una **función de peso** $p(x)$ real:
    $$\int_a^b p(x) \phi_n^*(x) \phi_m(x) dx = A_n \delta_{nm}$$.
*   **Normalización:** Si el factor $A_n$ es igual a la unidad para todos los elementos, la base se denomina **ortonormal**.

### 2. Bases Continuas y la Delta de Dirac
Cuando el índice de la base es una variable continua (como en la transformada de Fourier), la condición de ortogonalidad utiliza la **delta de Dirac** en lugar de la de Kronecker:
$$\int_a^b \phi^*(k, x) \phi(k', x) dx = \delta(k - k')$$.
Este tipo de bases se asocian generalmente a dominios infinitos o semi-infinitos.

### 3. Origen en la Teoría de Sturm-Liouville
En física matemática, la mayoría de las bases ortogonales (como los polinomios de Legendre, Hermite o las funciones de Bessel) surgen como **autofunciones de operadores hermíticos** de segundo orden.
*   El teorema de Sturm-Liouville garantiza que las autofunciones correspondientes a **autovalores distintos** son automáticamente ortogonales.
*   Si existe **degeneración** (varias funciones con el mismo autovalor), estas no son necesariamente ortogonales entre sí, pero pueden ortogonalizarse mediante métodos específicos.

### 4. Construcción de Bases (Gram-Schmidt)
Si se dispone de un conjunto de funciones que son linealmente independientes pero no ortogonales, es posible construir una base ortogonal a partir de ellas utilizando el **procedimiento de Gram-Schmidt**. Este método permite generar familias de polinomios clásicos (Legendre, Hermite, Laguerre) a partir de una base simple como la de Taylor $\{x^n\}$, variando únicamente el intervalo y la función de peso.

### 5. Utilidad Física
Contar con una base ortogonal permite expresar cualquier función de cuadrado integrable como una **combinación lineal** (serie) de dichas funciones. Los coeficientes de esta expansión actúan como las "componentes" del vector función y se calculan de forma aislada gracias a que la ortogonalidad elimina las contribuciones cruzadas de los otros términos de la base.

**Ejemplo:**

Mostrar que la base $$\{\phi_n(x)\} = \left\{\sqrt{\frac{2}{L}}\sin\left(\frac{n\pi x}{L}\right)\right\}_{n=0,1,...}$$

es ortonormal en $0\leq x \leq L$

```
F[n_,x_]:=Sqrt[2/L]*Sin[n*Pi*x/L]
```

```
Integrate[F[1,x]^2,{x,0,L}]
```

```
Integrate[F[n,x]*F[m,x],{x,0,L},Assumptions->{Element[n,Naturals],Element[n,Naturals]}]
```

Para demostrar que la base de funciones $\{\phi_n(x)\} = \left\{\sqrt{\frac{2}{L}}\sin\left(\frac{n\pi x}{L}\right)\right\}$ es **ortonormal** en el intervalo $0\leq x \leq L$, debemos verificar que se cumplen dos condiciones fundamentales: la **ortogonalidad** entre funciones con índices distintos y la **normalización** (que la norma de cada función sea la unidad).

Matemáticamente, estas dos condiciones se resumen en que el **producto escalar** (o interno) de dos funciones de la base debe ser igual a la delta de Kronecker ($\delta_{nm}$):
$$( \phi_n, \phi_m ) = \int_{0}^{L} \phi_n^*(x) \phi_m(x) \, dx = \delta_{nm}$$
Dado que las funciones son reales, $\phi_n^*(x) = \phi_n(x)$.

---

### 1. Planteamiento de la Integral
Sustituimos la expresión dada para $\phi_n(x)$ y $\phi_m(x)$ en la definición del producto interno:
$$I = \int_{0}^{L} \left[ \sqrt{\frac{2}{L}} \sin\left( \frac{n\pi x}{L} \right) \right] \left[ \sqrt{\frac{2}{L}} \sin\left( \frac{m\pi x}{L} \right) \right] dx$$

Extraemos las constantes fuera de la integral:
$$I = \frac{2}{L} \int_{0}^{L} \sin\left( \frac{n\pi x}{L} \right) \sin\left( \frac{m\pi x}{L} \right) dx$$

Para resolver esta integral, utilizamos la identidad trigonométrica del producto de senos:
$$\sin(A)\sin(B) = \frac{1}{2} [\cos(A-B) - \cos(A+B)]$$

Aplicándola a nuestra integral:
$$I = \frac{2}{L} \cdot \frac{1}{2} \int_{0}^{L} \left[ \cos\left( \frac{(n-m)\pi x}{L} \right) - \cos\left( \frac{(n+m)\pi x}{L} \right) \right] dx$$
$$I = \frac{1}{L} \int_{0}^{L} \cos\left( \frac{(n-m)\pi x}{L} \right) dx - \frac{1}{L} \int_{0}^{L} \cos\left( \frac{(n+m)\pi x}{L} \right) dx$$

---

### 2. Caso de Ortogonalidad ($n \neq m$)
Analizamos el resultado cuando los índices son diferentes. Integramos ambos términos:
1.  $\int_{0}^{L} \cos\left( \frac{(n-m)\pi x}{L} \right) dx = \left[ \frac{L}{(n-m)\pi} \sin\left( \frac{(n-m)\pi x}{L} \right) \right]_0^L = 0$
2.  $\int_{0}^{L} \cos\left( \frac{(n+m)\pi x}{L} \right) dx = \left[ \frac{L}{(n+m)\pi} \sin\left( \frac{(n+m)\pi x}{L} \right) \right]_0^L = 0$

Como $n$ y $m$ son enteros, el argumento del seno evaluado en $L$ siempre es un múltiplo de $\pi$ (por ejemplo, $(n-m)\pi$), y sabemos que $\sin(k\pi) = 0$ para cualquier entero $k$.
**Resultado:** Para $n \neq m$, la integral es $0$. Esto demuestra que las funciones son **ortogonales**.

---

### 3. Caso de Normalización ($n = m$)
Cuando los índices son iguales, el primer término de la identidad trigonométrica cambia ($\cos(0) = 1$). La integral original se convierte en:
$$I = \frac{2}{L} \int_{0}^{L} \sin^2\left( \frac{n\pi x}{L} \right) dx$$

Utilizando la identidad de reducción de potencia $\sin^2(\theta) = \frac{1 - \cos(2\theta)}{2}$:
$$I = \frac{2}{L} \int_{0}^{L} \frac{1 - \cos\left( \frac{2n\pi x}{L} \right)}{2} dx = \frac{1}{L} \int_{0}^{L} \left[ 1 - \cos\left( \frac{2n\pi x}{L} \right) \right] dx$$

Evaluamos la integral:
$$I = \frac{1}{L} \left[ x - \frac{L}{2n\pi} \sin\left( \frac{2n\pi x}{L} \right) \right]_0^L$$
$$I = \frac{1}{L} \left[ (L - 0) - (0 - 0) \right] = \frac{L}{L} = 1$$

**Resultado:** Para $n = m$, la integral es $1$. Esto demuestra que cada función tiene **norma unitaria** (está normalizada).

---

### Conclusión Final
Puesto que hemos demostrado que:
$$( \phi_n, \phi_m ) = \begin{cases} 0 & \text{si } n \neq m \\ 1 & \text{si } n = m \end{cases} = \delta_{nm}$$
Se concluye fehacientemente que la base $\{\phi_n(x)\}$ es **ortonormal** en el intervalo $[0, L]$. Esta base es fundamental en la resolución de problemas como la cuerda vibrante o la partícula en una caja unidimensional.
