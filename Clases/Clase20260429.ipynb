# Tarea

Resolver la ecuación de Laplace para el sistema

Placa rectangular infinita en $\pm x$ y con altura $b$ en el eje $y$ con condiciones:

* $u(x, b) = V(x)$
* $u(x,0) = 0$
* $\lim_{x\to \pm \infty}u(x,y) = 0$

Para resolver la **ecuación de Laplace** ($\nabla^2 u = 0$) en una placa rectangular infinita en el eje $x$ ($-\infty < x < \infty$) con una altura $b$ en el eje $y$, se utiliza el método de **separación de variables**.

### 1. Separación de Variables
Se propone una solución de la forma $u(x, y) = X(x)Y(y)$. Al sustituir en la ecuación de Laplace y dividir por $u$:
$$\frac{1}{X}\frac{d^2X}{dx^2} + \frac{1}{Y}\frac{d^2Y}{dy^2} = 0 \implies \frac{1}{X}\frac{d^2X}{dx^2} = -\frac{1}{Y}\frac{d^2Y}{dy^2}$$
Dado que cada lado depende de una variable independiente distinta, ambos deben ser iguales a una constante de separación. Para este sistema infinito en $x$ y con condiciones que tienden a cero en el infinito, se elige $-\omega^2$ como constante:
1.  $\frac{d^2X}{dx^2} + \omega^2 X = 0 \implies X(x) = A(\omega)e^{i\omega x} + B(\omega)e^{-i\omega x}$
2.  $\frac{d^2Y}{dy^2} - \omega^2 Y = 0 \implies Y(y) = C(\omega)e^{\omega y} + D(\omega)e^{-\omega y}$

### 2. Aplicación de las Condiciones de Frontera
*   **Condición $u(x, 0) = 0$:** Esto implica que $Y(0) = 0$. Sustituyendo en la solución para $Y$:
    $$C + D = 0 \implies D = -C \implies Y(y) = C(e^{\omega y} - e^{-\omega y}) = 2C \sinh(\omega y)$$
    Por lo tanto, la dependencia en $y$ es proporcional a **$\sinh(\omega y)$**.
*   **Condición $\lim_{x\to \pm \infty}u(x,y) = 0$:** Como el dominio en $x$ es infinito, el parámetro $\omega$ puede adoptar cualquier valor real (es un **continuo**), por lo que la solución general se expresa como una **integral de Fourier** sobre $\omega$.
    $$u(x, y) = \int_{0}^{\infty} [A(\omega)e^{i\omega x} + B(\omega)e^{-i\omega x}] \sinh(\omega y) d\omega$$

### 3. Determinación de los coeficientes mediante $u(x, b) = V(x)$
Al evaluar en el borde superior $y=b$, se tiene:
$$V(x) = \int_{0}^{\infty} [A(\omega)e^{i\omega x} + B(\omega)e^{-i\omega x}] \sinh(\omega b) d\omega$$
Utilizando la **transformada de Fourier** para despejar los coeficientes $A(\omega)$ y $B(\omega)$, se obtiene:
$$A(\omega) = \frac{1}{2\pi \sinh(\omega b)} \int_{-\infty}^{\infty} V(x') e^{-i\omega x'} dx'$$
$$B(\omega) = \frac{1}{2\pi \sinh(\omega b)} \int_{-\infty}^{\infty} V(x') e^{i\omega x'} dx'$$

### 4. Solución Final
Sustituyendo estos coeficientes, la solución completa para el potencial en cualquier punto $(x, y)$ de la placa es la integral:
$$\mathbf{u(x, y) = \frac{1}{2\pi} \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} V(x') e^{i\omega(x-x')} \frac{\sinh(\omega y)}{\sinh(\omega b)} dx' d\omega}$$

Esta expresión garantiza que se cumplan todas las condiciones: el potencial es nulo en la base ($y=0$), coincide con $V(x)$ en la parte superior ($y=b$) y se desvanece en los extremos infinitos de la placa según el **lema de Riemann-Lebesgue**.

Justificaciones:


La razón principal para elegir la constante de separación de la forma $-\omega^2$ (con $\omega$ real y, por convención, positiva) se basa en el **comportamiento físico de la solución en el dominio infinito** y en la **naturaleza del análisis espectral** (Transformada de Fourier).


### 1. Convergencia y Acotamiento en el Infinito
El dominio en $x$ es $(-\infty, \infty)$. Si eligiéramos una constante de separación positiva ($+\lambda^2$), la ecuación para $X(x)$ sería $X'' - \lambda^2 X = 0$, cuya solución es exponencial real:
$$ X(x) = A e^{\lambda x} + B e^{-\lambda x} $$
Estas funciones crecen indefinidamente cuando $x \to \infty$ o $x \to -\infty$. Como el potencial físico $u(x,y)$ debe ser finito (y generalmente tiende a cero en el infinito, como indica el Lema de Riemann-Lebesgue), estas soluciones exponenciales no son aceptables físicamente.

Al elegir la constante negativa ($-\omega^2$), la ecuación se vuelve $X'' + \omega^2 X = 0$, cuya solución es oscilatoria:
$$ X(x) = A e^{i\omega x} + B e^{-i\omega x} $$
Estas funciones (senos, cosenos o exponenciales complejas) están **acotadas** para todo $x$ real, lo cual es consistente con la condición de frontera en el infinito.

El parámetro $\omega$ (o $k$) representa una **frecuencia espacial** (o número de onda).
*   Matemáticamente, la frecuencia es una magnitud, por lo que se define $\omega \ge 0$.
*   La dirección de propagación (hacia la izquierda o derecha) se maneja con el signo en el exponente ($e^{i\omega x}$ vs $e^{-i\omega x}$) o integrando desde $-\infty$ hasta $\infty$.

### 3. Evitar Redundancias
En la **Sección 3.3.1** del texto de Alonso, se menciona explícitamente:
> *"En principio, la ecuación de Laplace es satisfecha por esta solución para todos los valores de $\alpha$ positivos (escrita en la forma dada arriba la solución produce redundancias si $\alpha$ toma valores positivos y negativos)."*

Si permitiéramos $\omega$ negativo en la definición de la constante $-\omega^2$, obtendríamos exactamente la misma ecuación diferencial (pues $(-\omega)^2 = \omega^2$). Los términos $e^{i(-\omega)x}$ son equivalentes a $e^{-i\omega x}$, los cuales ya están contemplados en la solución general mediante los coeficientes arbitrarios $A(\omega)$ y $B(\omega)$ o mediante la integración sobre todo el eje real. Por tanto, restringir $\omega$ a valores positivos en la definición del parámetro evita contar la misma solución dos veces.

### 4. Comportamiento en la dirección finita ($y$)
La elección del signo en $x$ determina forzosamente el comportamiento en $y$.
*   Si $X(x)$ es oscilatoria (solución de $X'' + \omega^2 X = 0$), entonces la ecuación para $Y(y)$ debe ser $Y'' - \omega^2 Y = 0$.
*   La solución para $Y$ son funciones hiperbólicas ($\sinh(\omega y), \cosh(\omega y)$).
Dado que $y$ está confinada en un intervalo finito $[0, b]$, las funciones hiperbólicas son perfectamente válidas y acotadas.

**Resumen:** Se toma $\omega^2$ positivo (es decir, constante de separación $-\omega^2$) para garantizar **soluciones oscilatorias en el eje infinito** (que no divergen) y **soluciones hiperbólicas en el eje finito**, permitiendo así el uso de la **Transformada de Fourier** para satisfacer la condición de frontera $u(x,b)=V(x)$.


## ¿Por qué se pueden fusionar las dos exponenciales en una sola?

La clave está en un **cambio de variable** y en que el dominio de integración cubre **toda la recta real**. 
### 1. Partimos de la expresión original
Tras aplicar separación de variables, la solución general se escribe como una superposición sobre frecuencias positivas $\omega > 0$:
$$
u(x,y) = \int_{0}^{\infty} \Big[ A(\omega) e^{i\omega x} + B(\omega) e^{-i\omega x} \Big] Y(y;\omega) \, d\omega
$$

### 2. Separamos la integral en dos partes
$$
u(x,y) = \underbrace{\int_{0}^{\infty} A(\omega) e^{i\omega x} Y(y;\omega) \, d\omega}_{I_1} \;+\; \underbrace{\int_{0}^{\infty} B(\omega) e^{-i\omega x} Y(y;\omega) \, d\omega}_{I_2}
$$

### 3. Aplicamos un cambio de variable en la segunda integral ($I_2$)
Queremos que el exponencial de $I_2$ también sea $e^{i\omega x}$. Para ello, definimos:
$$
\omega' = -\omega \quad \Rightarrow \quad d\omega = -d\omega'
$$
**Transformación de los límites:**
- Si $\omega = 0 \;\Rightarrow\; \omega' = 0$
- Si $\omega \to +\infty \;\Rightarrow\; \omega' \to -\infty$

Sustituimos en $I_2$:
$$
I_2 = \int_{0}^{-\infty} B(-\omega') e^{i\omega' x} Y(y;-\omega') \, (-d\omega')
$$
Invertimos los límites de integración (esto elimina el signo menos del diferencial):
$$
I_2 = \int_{-\infty}^{0} B(-\omega') e^{i\omega' x} Y(y;-\omega') \, d\omega'
$$

### 4. Renombramos la variable muda
En cálculo integral, el nombre de la variable de integración es irrelevante ("variable muda"). Cambiamos $\omega' \to \omega$ para que coincida con la notación de $I_1$:
$$
I_2 = \int_{-\infty}^{0} B(-\omega) e^{i\omega x} Y(y;-\omega) \, d\omega
$$

### 5. Unificamos ambas integrales
Ahora tenemos:
$$
u(x,y) = \int_{0}^{\infty} A(\omega) e^{i\omega x} Y(y;\omega) \, d\omega \;+\; \int_{-\infty}^{0} B(-\omega) e^{i\omega x} Y(y;-\omega) \, d\omega
$$
Observa que **ambos integrandos comparten exactamente la misma estructura funcional**: $e^{i\omega x}$ multiplicado por una función que depende solo de $\omega$. Por lo tanto, podemos definir una **única función espectral** $\mathcal{C}(\omega)$ sobre todo $\mathbb{R}$:
$$
\mathcal{C}(\omega) = 
\begin{cases}
A(\omega) Y(y;\omega), & \omega > 0 \\
B(-\omega) Y(y;-\omega), & \omega < 0
\end{cases}
$$
Y reescribir todo como una sola integral:
$$
u(x,y) = \int_{-\infty}^{\infty} \mathcal{C}(\omega) e^{i\omega x} \, d\omega
$$


El anterior sistema me dice que tengo solución continua en x y discreta en y

### 4. Resultado Final (Notación de Alonso)
Al simplificar los factores constantes y reordenar la serie, se llega a la expresión final que aparece en el libro:
$$\mathbf{G(x, x') = -\frac{2L}{\pi^2} \sum_{n=1}^\infty \frac{1}{n^2} \sin\left(\frac{n\pi x}{L}\right) \sin\left(\frac{n\pi x'}{L}\right)}$$

### 5. Solución del Problema Inhomogéneo
Una vez construida $G(x, x')$, la solución completa para cualquier fuente arbitraria $g(x)$ se obtiene mediante la integral de convolución:
$$f(x) = \int_0^L G(x, x') g(x') dx'$$
Esta integral suma las contribuciones de cada "punto fuente" $x'$ ponderadas por la intensidad de la fuente $g(x')$ en ese lugar. El resultado garantiza automáticamente que se cumplan las condiciones de frontera $f(0)=f(L)=0$ debido a que cada término de la serie de la función de Green ya las satisface.

# Separación de variables en cilíndricas

En coordenadas cilíndricas $(\rho, \phi, z)$, la **ecuación de Laplace** para una función escalar $\phi$ se escribe como:
$$\frac{1}{\rho} \frac{\partial}{\partial \rho} \left( \rho \frac{\partial \phi}{\partial \rho} \right) + \frac{1}{\rho^2} \frac{\partial^2 \phi}{\partial \phi^2} + \frac{\partial^2 \phi}{\partial z^2} = 0$$

### 1. Propuesta de solución separable
Se propone una solución que sea el producto de tres funciones independientes, cada una dependiente de una sola coordenada:
$$\phi(\rho, \phi, z) = R(\rho)G(\phi)Z(z)$$

Sustituyendo esta propuesta en la ecuación de Laplace y dividiendo por $\frac{RGZ}{\rho^2}$, se obtiene la expresión separada:
$$\frac{\rho}{R} \frac{d}{d\rho} \left( \rho \frac{dR}{d\rho} \right) + \frac{1}{G} \frac{d^2G}{d\phi^2} + \frac{\rho^2}{Z} \frac{d^2Z}{dz^2} = 0$$

### 2. Separación de la parte angular ($G$)
El término central de la ecuación anterior depende únicamente de $\phi$, mientras que los otros dos mezclan $\rho$ y $z$. Para que la igualdad se mantenga, se iguala la parte angular a una constante de separación $-\nu^2$:
$$\frac{1}{G} \frac{d^2G}{d\phi^2} = -\nu^2$$
La solución para esta ecuación (cuando $\nu \neq 0$) es de tipo armónico:
$$G(\phi) = A e^{i\nu\phi} + B e^{-i\nu\phi}$$

**Condición de continuidad:** Si el sistema abarca el ángulo completo de $2\pi$, se exige que la función sea unívoca, es decir, $G(\phi) = G(\phi + 2\pi)$. Esto solo se cumple si **$\nu$ es un número entero $n$**.

### 3. Separación de las partes axial ($Z$) y radial ($R$)
Sustituyendo el resultado angular $-\nu^2 = -n^2$ en la ecuación original y reordenando, se logra separar las variables $z$ y $\rho$:
$$\frac{1}{\rho R} \frac{d}{d\rho} \left( \rho \frac{dR}{d\rho} \right) - \frac{n^2}{\rho^2} + \frac{1}{Z} \frac{d^2Z}{dz^2} = 0$$

Introduciendo una nueva constante de separación $k^2$, obtenemos la **ecuación axial**:
$$\frac{1}{Z} \frac{d^2Z}{dz^2} = k^2 \implies Z(z) = C e^{kz} + D e^{-kz}$$

Finalmente, se obtiene la **ecuación radial**:
$$\frac{1}{\rho} \frac{d}{d\rho} \left( \rho \frac{dR}{d\rho} \right) + \left( k^2 - \frac{n^2}{\rho^2} \right) R = 0$$

### 4. La ecuación de Bessel
Al realizar el cambio de variable adimensional $x = k\rho$, la ecuación radial se transforma en la famosa **Ecuación de Bessel** de orden $n$:
$$x^2 \frac{d^2R}{dx^2} + x \frac{dR}{dx} + (x^2 - n^2)R = 0$$
Las soluciones a esta ecuación son las **funciones de Bessel** de primera especie $J_n(k\rho)$ y las funciones de Bessel de segunda especie (o de Neumann) $N_n(k\rho)$.

### 5. Solución General
La solución más general se construye como una superposición (suma sobre $n$ e integral sobre $k$) de todas las soluciones particulares permitidas por las condiciones de frontera:
$$\phi = \sum_{n=0}^{\infty} \int_{0}^{\infty} [E J_n(k\rho) + F N_n(k\rho)] (A e^{in\phi} + B e^{-in\phi}) (C e^{kz} + D e^{-kz}) dk$$

**Observaciones de Alonso:**
*   En regiones que incluyen el eje de simetría ($\rho = 0$), se debe descartar la función de Neumann ($F=0$) ya que diverge en el origen.
*   Dependiendo de las condiciones de frontera en $z$ o $\phi$, las constantes de separación pueden cambiar de signo, dando lugar a funciones sinusoidales en $z$ y funciones de Bessel modificadas $I_n$ y $K_n$ en la parte radial.

# Ejercicio:

Resolver la EDP de Helmholtz

Para resolver la ecuación diferencial parcial (EDP) de **Helmholtz**, la cual tiene la forma general $(\nabla^2 + k^2) \psi = 0$, se utiliza comúnmente el método de **separación de variables**. Esta ecuación surge frecuentemente al tratar de encontrar soluciones estacionarias o armónicas para la ecuación de onda o de difusión, separando la dependencia temporal.

A continuación se detalla la resolución en los tres sistemas de coordenadas principales:

### 1. Coordenadas Cartesianas
En este sistema, la ecuación se escribe como:
$$\frac{\partial^2\psi}{\partial x^2} + \frac{\partial^2\psi}{\partial y^2} + \frac{\partial^2\psi}{\partial z^2} + k^2\psi = 0$$

*   **Separación:** Se propone $\psi(x, y, z) = X(x)Y(y)Z(z)$.
*   **Ecuaciones resultantes:** Al sustituir y dividir por $\psi$, se obtienen tres ecuaciones diferenciales ordinarias (EDO):
    1.  $X'' + l^2 X = 0$
    2.  $Y'' + m^2 Y = 0$
    3.  $Z'' + n^2 Z = 0$
*   **Condición:** Los parámetros de separación deben satisfacer $l^2 + m^2 + n^2 = k^2$.
*   **Solución General:** La solución es una combinación lineal de exponenciales complejas o funciones trigonométricas:
    $$\psi(x, y, z) = \sum_{l, m, n} A_{lmn} e^{\pm ilx} e^{\pm imy} e^{\pm inz}$$

### 2. Coordenadas Cilíndricas $(\rho, \phi, z)$
La ecuación de Helmholtz toma la forma:
$$\frac{1}{\rho} \frac{\partial}{\partial \rho} \left( \rho \frac{\partial \psi}{\partial \rho} \right) + \frac{1}{\rho^2} \frac{\partial^2 \psi}{\partial \phi^2} + \frac{\partial^2 \psi}{\partial z^2} + k^2 \psi = 0$$

*   **Separación:** Se propone $\psi(\rho, \phi, z) = R(\rho)\Phi(\phi)Z(z)$.
*   **Ecuaciones resultantes:**
    *   **Angular:** $\Phi'' + m^2 \Phi = 0 \implies \Phi(\phi) \sim e^{\pm im\phi}$. Si el sistema abarca el ángulo completo, $m$ debe ser entero.
    *   **Axial:** $Z'' + \alpha^2 Z = 0 \implies Z(z) \sim e^{\pm i\alpha z}$.
    *   **Radial:** Al definir $\gamma^2 = k^2 - \alpha^2$, se obtiene la **Ecuación de Bessel** de orden $m$:
        $$\rho^2 R'' + \rho R' + (\gamma^2 \rho^2 - m^2) R = 0$$
*   **Solución General:** En términos de las funciones de Bessel de primera ($J_m$) y segunda ($N_m$ o $Y_m$) especie:
    $$\psi = \sum_{m, \alpha} [A J_m(\gamma \rho) + B N_m(\gamma \rho)] (C e^{im\phi} + D e^{-im\phi}) (E e^{i\alpha z} + F e^{-i\alpha z})$$

### 3. Coordenadas Esféricas $(r, \theta, \phi)$
La ecuación se expresa como:
$$\nabla^2 \psi + k^2 \psi = 0$$

*   **Separación:** Se propone $\psi(r, \theta, \phi) = R(r)Y(\theta, \phi)$.
*   **Ecuaciones resultantes:**
    *   **Angular:** La parte angular satisface la ecuación de los **armónicos esféricos** ($Y_{lm}$) con la constante de separación $l(l+1)$.
    *   **Radial:** Conduce a la **ecuación de Bessel esférica**:
        $$r^2 R'' + 2r R' + [k^2 r^2 - l(l+1)] R = 0$$
*   **Solución General:** Se escribe como una superposición de modos $(l, m)$ utilizando las funciones de Bessel esféricas de primera ($j_l$) y segunda ($y_l$) especie:
    $$\psi(r, \theta, \phi) = \sum_{l=0}^\infty \sum_{m=-l}^l [a_{lm} j_l(kr) + b_{lm} y_l(kr)] Y_{lm}(\theta, \phi)$$

En espacios infinitos sin fronteras, la solución también puede expresarse mediante la **función de Green**, que para el operador de Helmholtz $(\nabla^2 + k^2)$ con condiciones de onda saliente es:
$$G(r, r') = -\frac{e^{ik|r-r'|}}{4\pi|r-r'|}$$

# Ejercicio:

Resolver la ecuación de laplace en coordenadas polares.

Para resolver la **ecuación de Laplace** ($\nabla^2 \phi = 0$) en coordenadas polares $(\rho, \phi)$, se utiliza el método de **separación de variables**, el cual permite transformar una ecuación diferencial parcial (EDP) en un conjunto de ecuaciones diferenciales ordinarias (EDO) independientes.

### 1. Definición de la ecuación
En coordenadas polares bidimensionales, el operador Laplaciano se expresa como:
$$\nabla^2 \phi(\rho, \phi) = \frac{1}{\rho} \frac{\partial}{\partial \rho} \left( \rho \frac{\partial \phi}{\partial \rho} \right) + \frac{1}{\rho^2} \frac{\partial^2 \phi}{\partial \phi^2} = 0$$.

### 2. Separación de variables
Se propone una solución que sea el producto de dos funciones independientes, cada una dependiente de una sola coordenada:
$$\phi(\rho, \phi) = R(\rho)G(\phi)$$.

Al sustituir esta propuesta en la ecuación original y dividir por $\frac{RG}{\rho^2}$, se obtiene la forma separada:
$$\frac{\rho}{R} \frac{d}{d\rho} \left( \rho \frac{dR}{d\rho} \right) + \frac{1}{G} \frac{d^2G}{d\phi^2} = 0$$.

Dado que un término depende solo de $\rho$ y el otro solo de $\phi$, ambos deben ser iguales a una constante de separación, que definiremos como **$\nu^2$**.

### 3. Solución de la parte angular ($G$)
La EDO para la parte angular es:
$$\frac{d^2G}{d\phi^2} = -\nu^2 G$$.

*   **Caso $\nu \neq 0$:** La solución es de tipo armónico: $G(\phi) = A e^{i\nu\phi} + B e^{-i\nu\phi}$. Para que la solución sea unívoca (que no cambie al sumar $2\pi$ al ángulo), se exige que **$\nu$ sea un número entero $n$** ($n = 1, 2, 3, \dots$).
*   **Caso $\nu = 0$:** La solución es una línea recta: $G(\phi) = a\phi + b$. La condición de periodicidad obliga a que $a = 0$, resultando en una **constante** $G_0 = b$.

### 4. Solución de la parte radial ($R$)
La EDO resultante para la parte radial es una **ecuación de Euler**:
$$\rho \frac{d}{d\rho} \left( \rho \frac{dR}{d\rho} \right) - n^2 R = 0$$.

*   **Para $n \neq 0$:** Las soluciones son potencias de $\rho$: $R(\rho) = C\rho^n + D\rho^{-n}$.
*   **Para $n = 0$:** La solución involucra un término logarítmico: $R(\rho) = E \ln \rho + F$.

### 5. Construcción de la solución general
La solución completa se obtiene sumando todas las soluciones particulares permitidas por las condiciones de frontera:
$$\phi(\rho, \phi) = (a\phi + b)(E \ln \rho + F) + \sum_{n=1}^\infty (C_n \rho^n + D_n \rho^{-n})(A_n e^{in\phi} + B_n e^{-in\phi})$$.

**Consideraciones físicas importantes:**
*   **Regiones que incluyen el origen ($\rho = 0$):** Para que el potencial sea finito en el centro, se deben descartar los términos $\rho^{-n}$ y $\ln \rho$, ya que divergen en ese punto. En este caso, la solución se simplifica a:
    $$\phi(\rho, \phi) = D_0 + \sum_{n=1}^\infty \rho^n (A_n \cos(n\phi) + B_n \sin(n\phi))$$.
*   **Regiones que excluyen el origen:** Si el sistema es una corona circular o el exterior de un círculo, se mantienen ambos términos de la parte radial.

Resolver la ecuación de Laplace.

Para resolver la **ecuación de Laplace** ($\nabla^2 u = 0$) en una sección circular (cuña) de radio $R$ y ángulo de apertura $\phi_0$, bajo las condiciones de frontera dadas, utilizamos el método de **separación de variables** en coordenadas polares $(\rho, \phi)$.

### 1. Definición del Problema
La ecuación de Laplace en dos dimensiones es:
$$\frac{1}{\rho} \frac{\partial}{\partial \rho} \left( \rho \frac{\partial u}{\partial \rho} \right) + \frac{1}{\rho^2} \frac{\partial^2 u}{\partial \phi^2} = 0$$

Las condiciones de frontera son:
*   **Base ($\phi = 0$):** $u(\rho, 0) = V_1$.
*   **Tramo recto ($\phi = \phi_0$):** $u(\rho, \phi_0) = V_2$.
*   **Tramo de arco ($\rho = R$):** $u(R, \phi) = V(\phi)$.
*   **Finitud en el origen:** $u(0, \phi)$ debe ser finito, lo que implica descartar términos proporcionales a $\ln \rho$ y $\rho^{-n}$.

### 2. Superposición de Soluciones
Dado que las condiciones en los tramos rectos son constantes y no nulas, es conveniente dividir la solución en dos partes: $u(\rho, \phi) = u_0(\phi) + w(\rho, \phi)$.

**A. Solución para los bordes radiales ($u_0$):**
Buscamos una solución que satisfaga solo la dependencia angular para los bordes constantes. El término $n=0$ de la ecuación de Laplace permite una solución de la forma $u_0(\phi) = a\phi + b$.
*   Aplicando $u_0(0) = V_1 \implies b = V_1$.
*   Aplicando $u_0(\phi_0) = V_2 \implies a\phi_0 + V_1 = V_2 \implies a = \frac{V_2 - V_1}{\phi_0}$.
$$u_0(\phi) = V_1 + \frac{V_2 - V_1}{\phi_0} \phi$$

**B. Solución para el tramo de arco ($w$):**
La función $w(\rho, \phi)$ debe satisfacer la ecuación de Laplace con condiciones **homogéneas** en los bordes rectos:
*   $w(\rho, 0) = 0$
*   $w(\rho, \phi_0) = 0$
*   $w(R, \phi) = V(\phi) - u_0(\phi) \equiv G(\phi)$

Para que $w$ se anule en $\phi=0$ y $\phi=\phi_0$, la parte angular debe ser armónica de la forma $\sin(\nu\phi)$ con $\nu = \frac{m\pi}{\phi_0}$ ($m=1, 2, \dots$). La parte radial correspondiente para una región que incluye el origen es $\rho^\nu$.
$$w(\rho, \phi) = \sum_{m=1}^{\infty} A_m \rho^{\frac{m\pi}{\phi_0}} \sin\left(\frac{m\pi\phi}{\phi_0}\right)$$

### 3. Determinación de Coeficientes
Evaluamos en el arco ($\rho = R$):
$$G(\phi) = \sum_{m=1}^{\infty} \left( A_m R^{\frac{m\pi}{\phi_0}} \right) \sin\left(\frac{m\pi\phi}{\phi_0}\right)$$
Utilizando la **ortogonalidad de la base de senos** en el intervalo $[0, \phi_0]$:
$$A_m = \frac{2}{\phi_0 R^{\frac{m\pi}{\phi_0}}} \int_{0}^{\phi_0} \left[ V(\phi) - \left( V_1 + \frac{V_2 - V_1}{\phi_0} \phi \right) \right] \sin\left(\frac{m\pi\phi}{\phi_0}\right) d\phi$$

### 4. Solución Final
El potencial en cualquier punto de la sección circular es:
$$\mathbf{u(\rho, \phi) = V_1 + \frac{V_2 - V_1}{\phi_0} \phi + \sum_{m=1}^{\infty} A_m \rho^{\frac{m\pi}{\phi_0}} \sin\left(\frac{m\pi\phi}{\phi_0}\right)}$$
Donde los coeficientes $A_m$ se calculan mediante la integral de la condición en el arco ajustada por la componente lineal de los bordes rectos.

Para hallar las soluciones de la **ecuación diferencial de Bessel** utilizando el **método de Frobenius**, seguiremos un procedimiento analítico paso a paso basado en las fuentes. La ecuación de Bessel surge frecuentemente al resolver las ecuaciones de Laplace o Helmholtz en coordenadas cilíndricas.

### 1. Identificación de la Ecuación y Singularidades
La ecuación de Bessel se define como:
$$x^2 y'' + xy' + (x^2 - \nu^2)y = 0$$
Donde $\nu$ es un parámetro constante. Al analizar el punto $x=0$, observamos que es una **singularidad regular**, ya que los términos $xP(x) = 1$ y $x^2Q(x) = (x^2 - \nu^2)$ son finitos cuando $x \to 0$. Esto garantiza la existencia de al menos una solución en serie de potencias generalizada.

### 2. Propuesta de Solución (Serie de Frobenius)
Asumimos una solución de la forma:
$$y(x) = \sum_{\lambda=0}^{\infty} a_\lambda x^{\lambda+s}, \quad a_0 \neq 0$$
Sus derivadas son:
*   $y'(x) = \sum_{\lambda=0}^{\infty} a_\lambda (\lambda+s) x^{\lambda+s-1}$
*   $y''(x) = \sum_{\lambda=0}^{\infty} a_\lambda (\lambda+s)(\lambda+s-1) x^{\lambda+s-2}$

Sustituyendo en la ecuación diferencial y agrupando potencias de $x$:
$$\sum_{\lambda=0}^{\infty} a_\lambda [(\lambda+s)^2 - \nu^2] x^{\lambda+s} + \sum_{\lambda=0}^{\infty} a_\lambda x^{\lambda+s+2} = 0$$

### 3. La Ecuación Indicial y Recurrencia
Para que la igualdad se cumpla, el coeficiente de cada potencia de $x$ debe ser nulo.
*   **Para $\lambda=0$ (menor potencia $x^s$):** $a_0(s^2 - \nu^2) = 0$. Como $a_0 \neq 0$, obtenemos la **ecuación indicial**:
    $$s^2 - \nu^2 = 0 \implies s = \pm\nu$$
*   **Para $\lambda=1$:** $a_1[(s+1)^2 - \nu^2] = 0$. Dado que para $s=\pm\nu$ el término entre corchetes no es cero (salvo casos muy específicos), se exige que **$a_1 = 0$**.
*   **Relación de recurrencia general:** Para potencias superiores, igualamos el coeficiente de $x^{\lambda+s+2}$ a cero:
    $$a_{\lambda+2} = \frac{-a_\lambda}{(\lambda+s+2)^2 - \nu^2} = \frac{-a_\lambda}{(\lambda+2)(2s+\lambda+2)}$$

Como $a_1=0$, todos los coeficientes de índice impar serán cero. Para los pares ($\lambda = 2p$):
$$a_{2p} = \frac{(-1)^p a_0}{2^{2p} p! (s+1)(s+2)...(s+p)}$$

### 4. Definición de $J_\nu(x)$ (Primer Género)
Utilizando la **función Gamma** para generalizar el factorial, y eligiendo una escala convencional para $a_0$ ($a_0 = \frac{1}{2^\nu \Gamma(\nu+1)}$), llegamos a la expresión de la **función de Bessel de primer género**:
$$\mathbf{J_\nu(x) = \sum_{k=0}^{\infty} \frac{(-1)^k}{k!\Gamma(\nu+k+1)} \left( \frac{x}{2} \right)^{\nu+2k}}$$
Si $\nu$ no es un entero, $J_\nu(x)$ y $J_{-\nu}(x)$ son linealmente independientes y forman la solución general.

### 5. Definición de $N_n(x)$ o $Y_n(x)$ (Segundo Género)
Cuando $\nu$ es un entero ($n$), ocurre que $J_{-n}(x) = (-1)^n J_n(x)$, por lo que dejan de ser independientes. Se hace necesario definir una segunda solución, la **función de Neumann** (o de Weber):
$$\mathbf{Y_\nu(x) = \frac{J_\nu(x)\cos(\nu\pi) - J_{-\nu}(x)}{\sin(\nu\pi)}}$$

Para evaluar esta función cuando $\nu$ tiende a un entero $n$, se aplica la regla de **L'Hôpital**. El desarrollo analítico muestra que $Y_n(x)$ contiene:
1.  Un término **logarítmico**: $\frac{2}{\pi} J_n(x) \ln\left(\frac{x}{2}\right)$.
2.  Una serie con **potencias negativas** de $x$ (si $n>0$), lo que causa que la función diverja en el origen.

La solución general completa para cualquier $\nu$ es entonces:
$$y(x) = c_1 J_\nu(x) + c_2 Y_\nu(x)$$
