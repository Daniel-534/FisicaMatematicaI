## Ejemplo 2: Clasificación de la Ecuación de Onda Unidimensional

Consideremos la ecuación de onda que describe el desplazamiento $u(x, t)$ en una dimensión espacial $x$ y el tiempo $t$:
$$\frac{\partial^2 u}{\partial x^2} - \frac{1}{v^2} \frac{\partial^2 u}{\partial t^2} = 0$$

Para clasificar esta ecuación diferencial parcial (EDP), la comparamos con la forma general de una EDP lineal de segundo orden:
$$A u_{xx} + B u_{xt} + C u_{tt} + D u_x + E u_t + F u = G(x, t)$$

**Identificación de coeficientes:**
A partir de la estructura de la ecuación de onda, identificamos los coeficientes asociados a las segundas derivadas:
*   $A = 1$
*   $B = 0$ (no hay término cruzado $u_{xt}$)
*   $C = -\frac{1}{v^2}$

**Cálculo del discriminante $\mathscr{D}$:**
Aplicamos la fórmula del discriminante utilizada en la clasificación de cónicas aplicada a las EDP:
$$\mathscr{D} = B^2 - 4AC$$
$$\mathscr{D} = (0)^2 - 4(1)\left(-\frac{1}{v^2}\right)$$
$$\mathscr{D} = \frac{4}{v^2}$$

**Análisis y Clasificación:**
Puesto que la velocidad de propagación $v$ es una cantidad real y finita, el discriminante es estrictamente positivo:
$$\mathscr{D} > 0$$

De acuerdo con los criterios analíticos establecidos en las fuentes, cuando el discriminante es mayor que cero, la ecuación se clasifica como **hiperbólica**.

**Implicaciones Físicas y Matemáticas:**
*   **Naturaleza:** Esta clasificación confirma que la ecuación modela fenómenos de propagación de ondas con una velocidad definida $v$.
*   **Condiciones de Unicidad:** Al ser una ecuación hiperbólica, requiere la especificación de **condiciones de Cauchy** para tener una solución única; esto implica conocer el estado inicial del sistema ($u$ en $t=0$) y su velocidad inicial ($\partial_t u$ en $t=0$), además de las condiciones en las fronteras espaciales.
*   **Características:** La existencia de dos características reales ($x - vt$ y $x + vt$) permite que la solución general se exprese como la superposición de dos ondas viajeras, una hacia la derecha y otra hacia la izquierda, conocida como la solución de D’Alembert.

## Condiciones de Dirichlet

$$u\big|_s = \psi(\vec{r})$$

**Definición de las Condiciones de Dirichlet**

Las condiciones de frontera de Dirichlet representan uno de los requisitos fundamentales en el estudio de las ecuaciones diferenciales parciales (EDP), donde se especifica analíticamente el valor que la función incógnita $u$ debe tomar en todos los puntos de la superficie o contorno $S$ que limita la región de interés $V$. Matemáticamente, esta condición se expresa de la forma $u\big|_S = \psi(\vec{r})$, donde $\psi(\vec{r})$ es una función prescrita en la frontera. Estas condiciones pueden clasificarse como homogéneas si el valor en la frontera es idénticamente cero, es decir $u|_S = 0$, o inhomogéneas si se asignan valores específicos no nulos, como el potencial eléctrico en un conductor o una temperatura fija en un borde. En el análisis de espacios de Hilbert, las condiciones de Dirichlet aseguran que el operador diferencial se comporte de manera hermítica, facilitando la expansión de soluciones mediante autofunciones ortogonales.

**Ejemplo: Ecuación de Difusión con Condiciones de Dirichlet**

La ecuación de difusión, que describe procesos como la conducción del calor en sólidos o la dispersión de sustancias químicas, se define analíticamente como $\kappa \nabla^2 u = \frac{\partial u}{\partial t}$, donde $\kappa$ representa la difusividad del medio. En este marco, las condiciones de Dirichlet se utilizan para modelar sistemas donde el entorno controla estrictamente el estado en los límites espaciales.

Considere el caso de una varilla metálica delgada y uniforme de longitud $L$ alineada con el eje $x$. Para determinar la distribución de temperatura $u(x, t)$, se pueden imponer las siguientes condiciones de Dirichlet homogéneas en sus extremos:
*   $u(0, t) = 0$
*   $u(L, t) = 0$

Estas condiciones físicas significan que ambos extremos de la varilla se mantienen en contacto con un baño térmico a una temperatura constante de cero grados para todo tiempo $t \geq 0$.

Desde el punto de vista del teorema de unicidad, las ecuaciones parabólicas como la de difusión requieren información adicional para garantizar una solución única. Específicamente, el problema debe ser "cerrado" espacialmente mediante las condiciones de Dirichlet en la frontera $S$ y "abierto" temporalmente, lo que exige proporcionar una condición inicial que describa el estado del sistema en el instante $t = 0$, expresada como $u(x, 0) = f(x)$. Mediante el método de separación de variables, la solución general para este sistema se construye como una superposición de modos normales de la forma $u(x, t) = \sum_{n=1}^{\infty} B_n \sin(n\pi x / L) \exp(-\kappa n^2 \pi^2 t / L^2)$, donde la base de senos garantiza que se cumplan los valores de Dirichlet en los extremos $x=0$ y $x=L$.


## Condiciones de Newmann

$$\partial_n u \big|_s = g(\vec{r})$$

## Condiciones de Neumann

Las **condiciones de frontera de Neumann** son un requisito esencial en el estudio de las ecuaciones diferenciales parciales (EDP), en el cual se prescribe analíticamente el valor de la **derivada normal** de la función incógnita $u$ en cada punto de la superficie o contorno $S$ que limita una región $V$.

Matemáticamente, si $\mathbf{\hat{n}}$ es el vector unitario normal a la superficie (generalmente orientado hacia afuera del volumen), la condición de Neumann se expresa como:
$$\frac{\partial u}{\partial n} \equiv \nabla u \cdot \mathbf{\hat{n}} = g(\vec{r}) \quad \text{para cada } \vec{r} \in S$$
donde $g(\vec{r})$ es una función dada que representa la tasa de cambio de $u$ en la dirección normal a la frontera.

**Propiedades Analíticas**

*   **Unicidad de la Solución:** Para ecuaciones de tipo elíptico, como las de Laplace ($\nabla^2 u = 0$) o Poisson ($\nabla^2 u = \rho$), la especificación de condiciones de Neumann determina la solución de manera única **excepto por una constante aditiva**. Esto se debe a que si dos funciones satisfacen la misma tasa de cambio en la frontera, su diferencia tiene una derivada normal nula, lo que implica que dicha diferencia es una constante en todo el volumen.
*   **Requisito de Consistencia:** En problemas de Poisson, las condiciones de Neumann no pueden ser arbitrarias. Para que exista una solución, debe cumplirse el teorema de Gauss o condición de consistencia: la integral de la fuente sobre el volumen debe ser igual al flujo total a través de la frontera: $\int_V \rho \, dV = \int_S g \, dS$.
*   **Homogeneidad:** Una condición de Neumann se denomina **homogénea** cuando $g(\vec{r}) = 0$. Físicamente, esto suele interpretarse como la ausencia de flujo (de calor, fluido o carga) a través de la superficie.

**Ejemplo: Conducción de Calor en una Varilla Aislada**

Considere la ecuación de difusión (o del calor) para una varilla delgada de longitud $L$ alineada con el eje $x$, cuya distribución de temperatura es $u(x, t)$:
$$\kappa \frac{\partial^2 u}{\partial x^2} = \frac{\partial u}{\partial t}$$

Si los extremos de la varilla en $x = 0$ y $x = L$ están **térmicamente aislados**, no puede haber flujo de calor hacia el exterior. Según la ley de Fourier, el flujo es proporcional al gradiente de temperatura ($J = -k \partial_x u$). Por lo tanto, un aislamiento perfecto se traduce en condiciones de Neumann homogéneas en ambos extremos:
*   $\frac{\partial u}{\partial x} \big|_{x=0} = 0$
*   $\frac{\partial u}{\partial x} \big|_{x=L} = 0$

Bajo estas condiciones, si la temperatura inicial es $u(x, 0) = f(x)$, la solución mediante el método de separación de variables dará lugar a una serie de cosenos de la forma $u(x, t) = \frac{a_0}{2} + \sum_{n=1}^{\infty} a_n \cos(\frac{n\pi x}{L}) e^{-\kappa (\frac{n\pi}{L})^2 t}$. Aquí, el término constante $a_0/2$ representa la temperatura promedio final que alcanzará la varilla tras un tiempo largo, ilustrando cómo la condición de Neumann permite que el sistema mantenga una "energía base" (la constante aditiva) que depende del estado inicial.

## Condiciones Mixtas
$$p u\big|_s + q \partial_n u \big|_s = h(\vec{r})$$

Las **condiciones de frontera mixtas**, también conocidas como condiciones de tercera clase o de Robin, especifican una combinación lineal del valor de la función y de su derivada normal en cada punto de la superficie de contorno,. Matemáticamente, para una región $V$ limitada por una superficie $S$, esta condición se expresa de la forma $p u\big|_S + q \partial_n u \big|_S = h(\vec{r})$, donde $p$ y $q$ son coeficientes (generalmente reales) y $h(\vec{r})$ es una función prescrita en la frontera,,.

### Propiedades Analíticas

*   **Homogeneidad:** Una condición mixta se denomina homogénea cuando $h(\vec{r}) = 0$, lo que implica que la función y su derivada mantienen una proporción constante en la frontera,. Este tipo de condiciones es esencial en la teoría de Sturm-Liouville para asegurar que el operador diferencial sea hermítico y que sus autovalores sean reales,.
*   **Unicidad de la solución:** Para ecuaciones elípticas, como la de Poisson ($\nabla^2 u = \rho$), el teorema de unicidad establece que la solución es única si los coeficientes de la combinación lineal satisfacen la restricción de ser no negativos ($p, q \geq 0$),. Si estas condiciones se aplican a la ecuación de difusión, se garantiza que el perfil de temperatura o concentración sea único para todo tiempo $t \geq 0$,.
*   **Significado Físico:** A diferencia de las condiciones de Dirichlet (valor fijo) o Neumann (flujo fijo), las condiciones mixtas modelan procesos donde el estado en la frontera depende de la interacción dinámica con el entorno, como ocurre en la transferencia de calor por convección o radiación,.

### Ejemplo: Enfriamiento de Newton en la Ecuación de Difusión

Un ejemplo fundamental de estas condiciones aparece en el estudio de la conducción del calor cuando una superficie radia o intercambia energía con el medio ambiente. Considere la ecuación de difusión para la temperatura $T(\vec{r}, t)$ en un sólido:
$$\kappa \nabla^2 T = \frac{\partial T}{\partial t}$$,.

Si la superficie del sólido no está aislada ni se mantiene a una temperatura fija, sino que intercambia calor con un fluido circundante a temperatura constante $T_{ext}$, se aplica la **Ley de Newton de radiación de calor**. Según esta ley, el flujo de calor a través de la superficie es proporcional a la diferencia de temperatura entre el sólido y el ambiente:
$$k \frac{\partial T}{\partial n} = -H(T - T_{ext})$$.

Al reorganizar los términos para que coincidan con la forma general de la condición mixta, obtenemos:
$$H T\big|_S + k \partial_n T\big|_S = H T_{ext}$$.

En esta expresión:
*   $H$ representa la emisividad o coeficiente de transferencia de calor de la superficie.
*   $k$ es la conductividad térmica del material,.
*   La función $h(\vec{r}) = H T_{ext}$ representa la influencia térmica externa prescrita.

Esta condición asegura que el gradiente de temperatura en el borde del material se ajuste automáticamente según la diferencia de energía con el exterior, permitiendo modelar procesos realistas de enfriamiento o calentamiento,.

## Condiciones de Cauchy

Las **condiciones de frontera de Cauchy** representan un requisito analítico en el que se especifican simultáneamente tanto el valor de la función incógnita $u$ como el valor de su derivada normal $\partial u / \partial n$ en todos los puntos de la frontera o superficie $S$ del dominio en estudio. 

Este tipo de condiciones es particularmente relevante cuando una de las variables independientes es el tiempo ($t$), funcionando entonces como **condiciones iniciales** que definen el estado de un sistema en un instante de referencia (usualmente $t = 0$).

### Propiedades y Clasificación

*   **Ecuaciones Hiperbólicas:** Las condiciones de Cauchy son el emparejamiento adecuado para las **ecuaciones diferenciales parciales hiperbólicas**, como la ecuación de onda. 
*   **Fronteras Abiertas:** Se aplican típicamente sobre superficies "abiertas". Si se intentaran imponer condiciones de Cauchy sobre una superficie cerrada para ecuaciones elípticas (como la de Laplace), el problema resultaría sobre-especificado, lo que generalmente conduce a inconsistencias matemáticas o soluciones inestables.
*   **Unicidad:** Proporcionar tanto la posición como la velocidad inicial en un problema dinámico permite que la teoría de ecuaciones diferenciales garantice una solución única para la evolución del sistema.

### Ejemplo: La Ecuación de Onda Unidimensional

El ejemplo clásico del uso de condiciones de Cauchy es el análisis de una cuerda vibrante de longitud infinita. El desplazamiento transversal $u(x, t)$ satisface la ecuación diferencial parcial:

$$\frac{\partial^2 u}{\partial x^2} - \frac{1}{v^2} \frac{\partial^2 u}{\partial t^2} = 0$$

Para obtener una solución particular que describa el movimiento de la cuerda, es necesario especificar su estado en el tiempo $t = 0$ mediante las siguientes condiciones de Cauchy:

1.  **Desplazamiento inicial:** $u(x, 0) = \phi(x)$, que define la forma de la cuerda al inicio.
2.  **Velocidad inicial:** $\left. \frac{\partial u}{\partial t} \right|_{t=0} = \psi(x)$, que define cómo se mueve cada punto de la cuerda en ese instante.

Estas condiciones permiten derivar la **solución de D'Alembert**, la cual muestra que el desplazamiento resultante es la superposición de dos ondas de forma constante que viajan en direcciones opuestas con velocidad $v$:

$$u(x, t) = \frac{1}{2} [\phi(x - vt) + \phi(x + vt)] + \frac{1}{2v} \int_{x-vt}^{x+vt} \psi(q) \, dq$$

Analíticamente, este resultado confirma que el valor de la función en un punto $(x, t)$ depende únicamente de los datos iniciales contenidos dentro del "intervalo de influencia" $[x-vt, x+vt]$, lo cual es una manifestación directa de la causalidad en los sistemas físicos descritos por ecuaciones hiperbólicas.

# EDP Homogénea

Dado $A u_{xx} + B u_{xt} + C u_{tt} + D u_x + E u_t + F u = 0$ entonces la solución tiene forma $u=ke^{mx+ny}$

Una ecuación diferencial parcial (EDP) se define como homogénea cuando la variable dependiente $u$ y sus derivadas aparecen con la misma potencia en todos los términos de la expresión. Analíticamente, esto implica que la ecuación no posee un término fuente independiente de $u$, por lo que la expresión se iguala a cero.

**La Solución Exponencial y la Ecuación Característica**

Para el caso de ecuaciones diferenciales parciales lineales de segundo orden con coeficientes constantes, de la forma:
$$A u_{xx} + B u_{xy} + C u_{yy} + D u_x + E u_y + F u = 0$$
es posible proponer soluciones de carácter exponencial similares a las utilizadas en las ecuaciones ordinarias. Se asume que la solución tiene la forma analítica $u(x, y) = k e^{mx+ny}$, donde $k$ es una constante y $m, n$ son parámetros a determinar.

Al realizar las derivadas parciales de esta propuesta, cada operación introduce el parámetro correspondiente como factor:
*   $u_x = m u$.
*   $u_{xx} = m^2 u$.
*   $u_y = n u$.
*   $u_{yy} = n^2 u$.
*   $u_{xy} = mn u$.

Al sustituir estas expresiones en la EDP original y factorizar el término común $k e^{mx+ny}$, la relación diferencial se transforma en una relación puramente algebraica conocida como **ecuación característica** o auxiliar:
$$A m^2 + B mn + C n^2 + D m + E n + F = 0$$.

Puesto que la función exponencial es siempre distinta de cero para valores finitos, la validez de la solución depende exclusivamente de que los parámetros $m$ y $n$ satisfagan esta ecuación cuadrática. Esto implica que existe una familia de pares $(m, n)$ que definen soluciones particulares para el sistema.

**Principio de Superposición**

Dado que la EDP es lineal y homogénea, el principio de superposición garantiza que cualquier combinación lineal de estas soluciones exponenciales particulares es también una solución válida del problema. En consecuencia, la solución general puede expresarse como una suma discreta o una integral continua sobre los parámetros permitidos por la ecuación característica.

Un ejemplo típico de este análisis es la ecuación de onda unidimensional, $u_{xx} - \frac{1}{v^2} u_{tt} = 0$. Al aplicar la propuesta exponencial, la ecuación característica resultante es $m^2 - \frac{n^2}{v^2} = 0$, lo que conduce a la relación $n = \pm mv$. Esto genera soluciones de la forma $u(x, t) = c_1 e^{m(x+vt)} + c_2 e^{m(x-vt)}$, las cuales representan físicamente ondas viajeras si se utilizan parámetros imaginarios.

## Ejemplo 3
Hallar u(x, y) para la edp:

Para hallar la solución de la ecuación diferencial parcial (EDP) homogénea de segundo orden con coeficientes constantes $u_{xx} - u_{yy} - 2u_x + u = 0$, utilizaremos el método de proponer una solución exponencial de la forma $u(x, y) = e^{mx+ny}$.

**1. Cálculo de las derivadas parciales**

Partiendo de la propuesta $u = e^{mx+ny}$, calculamos las derivadas necesarias para sustituir en la ecuación original:
*   Derivada primera respecto a $x$: $u_x = m e^{mx+ny} = mu$.
*   Derivada segunda respecto a $x$: $u_{xx} = m^2 e^{mx+ny} = m^2 u$.
*   Derivada segunda respecto a $y$: $u_{yy} = n^2 e^{mx+ny} = n^2 u$.

**2. Obtención de la ecuación característica**

Sustituimos estas expresiones en la EDP original:
$$(m^2 u) - (n^2 u) - 2(mu) + u = 0$$

Factorizando el término común $u$:
$$(m^2 - n^2 - 2m + 1)u = 0$$

Dado que la función exponencial $u = e^{mx+ny}$ es siempre distinta de cero para valores finitos de las variables, la igualdad se cumple únicamente si el paréntesis es igual a cero. Esto nos da la **ecuación característica** o algebraica:
$$m^2 - 2m + 1 - n^2 = 0$$

**3. Resolución de la ecuación característica**

Observamos que los primeros tres términos del polinomio forman un trinomio cuadrado perfecto:
$$(m - 1)^2 - n^2 = 0$$

Para encontrar la relación entre los parámetros $m$ y $n$, despejamos $n$:
$$(m - 1)^2 = n^2$$
$$n = \pm(m - 1)$$

Esto indica que para cualquier valor de $m$, existen dos valores de $n$ que satisfacen la ecuación: $n_1 = m - 1$ y $n_2 = -(m - 1)$.

**4. Solución General**

De acuerdo con el análisis de Alonso Sepúlveda, la solución general se construye como una combinación lineal de las soluciones particulares obtenidas para cada valor de $n$:
$$u(x, y) = c_1 e^{mx + (m-1)y} + c_2 e^{mx - (m-1)y}$$

Esta expresión puede simplificarse factorizando el término exponencial que depende únicamente de $x$:
$$u(x, y) = e^{mx} \left[ c_1 e^{(m-1)y} + c_2 e^{-(m-1)y} \right]$$

Si m es discreto:
$$u(x, y) = \sum_m e^{mx} \left[ c_1 e^{(m-1)y} + c_2 e^{-(m-1)y} \right]$$

Si m es continuo:
$$u(x, y) = \int e^{mx} \left[ c_1 e^{(m-1)y} + c_2 e^{-(m-1)y} \right] dm$$


**Análisis adicional: Clasificación de la ecuación**

Para identificar la naturaleza de esta EDP, identificamos sus coeficientes según la forma general $A u_{xx} + B u_{xy} + C u_{yy} + \dots = 0$:
*   $A = 1$
*   $B = 0$
*   $C = -1$

Calculamos el discriminante $\mathscr{D}$:
$$\mathscr{D} = B^2 - 4AC = (0)^2 - 4(1)(-1) = 4$$

Como el discriminante es estrictamente positivo ($\mathscr{D} > 0$), la ecuación se clasifica como **hiperbólica**. Esto es consistente con la estructura de la solución obtenida, la cual permite describir fenómenos de propagación o transporte dependiendo de las condiciones de contorno que se apliquen al sistema.

## Ejemplo 4:

Aplicar el método de separación de variables a la EDP
$$u_{xx}-u_y=0$$

El **método de separación de variables** es una técnica fundamental para resolver ecuaciones diferenciales parciales (EDP) **lineales y homogéneas**. Su objetivo es descomponer una EDP de $n$ variables en un conjunto de $n$ ecuaciones diferenciales ordinarias (EDO), donde cada una depende de una sola variable independiente.

A continuación, se aplica el método paso a paso a la ecuación proporcionada: $u_{xx} - u_y = 0$.

### 1. Proponer la forma de la solución
Se asume que la solución puede escribirse como el producto de dos funciones independientes:
$$u(x, y) = A(x)B(y)$$.

### 2. Sustitución en la EDP
Calculamos las derivadas parciales necesarias:
*   $u_{xx} = \frac{d^2A}{dx^2} B = A''(x)B(y)$.
*   $u_y = A \frac{dB}{dy} = A(x)B'(y)$.

Sustituyendo en la ecuación original $u_{xx} - u_y = 0$:
$$A''(x)B(y) - A(x)B'(y) = 0$$.

### 3. Separación de variables
Dividimos toda la ecuación por $u = A(x)B(y)$ para separar los términos:
$$\frac{A''(x)}{A(x)} = \frac{B'(y)}{B(y)}$$.

Dado que el lado izquierdo depende solo de $x$ y el derecho solo de $y$, ambos deben ser iguales a una **constante de separación**. Dependiendo del signo de esta constante, se obtienen diferentes familias de soluciones:

#### Caso A: Constante positiva ($\alpha^2$)
Si igualamos ambos lados a $\alpha^2$:
1.  $\frac{1}{A} \frac{d^2A}{dx^2} = \alpha^2 \implies A(x) = c_1 e^{\alpha x} + c_2 e^{-\alpha x}$.
2.  $\frac{1}{B} \frac{dB}{dy} = \alpha^2 \implies B(y) = e^{\alpha^2 y}$.

**Solución particular:** $u_1(x, y) = (c_1 e^{\alpha x} + c_2 e^{-\alpha x}) e^{\alpha^2 y}$.

#### Caso B: Constante negativa ($-\beta^2$)
Esta elección es común en problemas físicos (como la difusión) para asegurar que la solución decaiga o sea periódica:
1.  $\frac{1}{A} \frac{d^2A}{dx^2} = -\beta^2 \implies A(x) = d_1 \cos(\beta x) + d_2 \sin(\beta x)$.
2.  $\frac{1}{B} \frac{dB}{dy} = -\beta^2 \implies B(y) = e^{-\beta^2 y}$.

**Solución particular:** $u_2(x, y) = (d_1 \cos(\beta x) + d_2 \sin(\beta x)) e^{-\beta^2 y}$.

#### Caso C: Constante nula ($0$)
Si la constante es cero:
1.  $A''(x) = 0 \implies A(x) = ax + b$.
2.  $B'(y) = 0 \implies B(y) = \text{constante}$.

**Solución particular:** $u_3(x, y) = ax + b$.

### 4. Solución General
Gracias al **principio de superposición**, la solución general es la combinación lineal de todas las soluciones particulares posibles. Dependiendo de si el dominio es finito o infinito, esta combinación se expresa como una sumatoria o una integral:
*   **Dominio Finito:** $u(x, y) = \sum_n u_n(x, y)$.
*   **Dominio Infinito:** $u(x, y) = \int [c(\alpha)e^{\alpha x} + d(\alpha)e^{-\alpha x}] e^{\alpha^2 y} d\alpha + \dots$.
