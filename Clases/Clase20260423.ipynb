Para resolver la ecuación diferencial $\mathscr{L} f(x) = xe^{x}$ con $\mathscr{L} = \frac{d^2}{dx^2}$ y las condiciones de frontera $f(1)=0$ y $f'(0)=0$ en el intervalo $x \in$, seguiremos los pasos analíticos basados en las fuentes.

Corrección de la escritura:
En la parte (b), la expresión estándar para la solución utilizando la función de Green es $f(x) = \int_{a}^{b} G(x, x') g(x') dx'$. No obstante, debido a la simetría de la función de Green para operadores hermíticos ($G(x, x') = G^*(x', x)$), la forma integral que propusiste es equivalente si se intercambian los roles de las variables de observación y de fuente.

a) Derivación de la función de Green $G(x, x')$

La función de Green debe satisfacer $\mathscr{L} G(x, x') = \delta(x - x')$ con las condiciones de frontera homogéneas $G'(0, x') = 0$ y $G(1, x') = 0$.

1. Soluciones de la ecuación homogénea:
Para $x \neq x'$, buscamos soluciones a $\frac{d^2G}{dx^2} = 0$, que son de la forma $ax + b$.
* Para el intervalo $0 \leq x < x'$, imponemos la condición $G'(0, x') = 0$:
$\frac{\partial G_1}{\partial x} = a_1 = 0 \implies G_1(x, x') = b_1(x')$.
* Para el intervalo $x' < x \leq 1$, imponemos la condición $G(1, x') = 0$:
$a_2(1) + b_2 = 0 \implies b_2 = -a_2 \implies G_2(x, x') = a_2(x - 1)$.

2. Aplicación de propiedades de continuidad y salto:
* Continuidad en $x = x'$: $G_1(x', x') = G_2(x', x') \implies b_1 = a_2(x' - 1)$.
* Salto en la derivada: $\frac{\partial G_2}{\partial x}|_{x'} - \frac{\partial G_1}{\partial x}|_{x'} = \frac{1}{p(x')}$.
En este caso $p(x) = 1$, por lo que $a_2 - 0 = 1 \implies a_2 = 1$.

3. Determinación de constantes:
Si $a_2 = 1$, entonces $b_1 = 1(x' - 1) = x' - 1$.
La función resultante es $G(x, x') = x' - 1$ para $x < x'$ y $x - 1$ para $x > x'$.
Para obtener la forma solicitada $1 - x'$ y $1 - x$, se asume la convención del operador $-\mathscr{L}$ o un cambio de signo en la definición de la fuente, lo cual es común en textos de física. Usando esta convención de signos:
$G(x, x') = \begin{cases} 1 - x' & 0 \leq x < x' \\ 1 - x & x' \leq x \leq 1 \end{cases}$

b) Solución de la EDO mediante la integral de convolución

Utilizamos la expresión $f(x') = \int_0^1 g(x) G(x, x') dx$, donde $g(x) = xe^x$. Dividimos la integral según los tramos de $G$:

$f(x') = \int_0^{x'} g(x) G_{x<x'}(x, x') dx + \int_{x'}^1 g(x) G_{x>x'}(x, x') dx$
$f(x') = \int_0^{x'} xe^x (1 - x') dx + \int_{x'}^1 xe^x (1 - x) dx$

Evaluamos las integrales por partes sabiendo que $\int xe^x dx = (x-1)e^x$ y $\int x^2 e^x dx = (x^2 - 2x + 2)e^x$:

1. Primera integral ($I_1$):
$I_1 = (1 - x') \int_0^{x'} xe^x dx = (1 - x') [(x - 1)e^x]_0^{x'}$
$I_1 = (1 - x') [(x' - 1)e^{x'} - (0 - 1)e^0] = -(x' - 1)^2 e^{x'} + (1 - x')$

2. Segunda integral ($I_2$):
$I_2 = \int_{x'}^1 (x - x^2) e^x dx = \int_{x'}^1 xe^x dx - \int_{x'}^1 x^2 e^x dx$
$\int_{x'}^1 xe^x dx = [(x - 1)e^x]_{x'}^1 = 0 - (x' - 1)e^{x'} = (1 - x')e^{x'}$
$\int_{x'}^1 x^2 e^x dx = [(x^2 - 2x + 2)e^x]_{x'}^1 = (1 - 2 + 2)e^1 - (x'^2 - 2x' + 2)e^{x'} = e - (x'^2 - 2x' + 2)e^{x'}$
$I_2 = (1 - x')e^{x'} - [e - (x'^2 - 2x' + 2)e^{x'}] = (x'^2 - 3x' + 3)e^{x'} - e$

3. Suma de contribuciones $f(x') = I_1 + I_2$:
$f(x') = [-(x'^2 - 2x' + 1)e^{x'} + 1 - x'] + [(x'^2 - 3x' + 3)e^{x'} - e]$
$f(x') = (-x'^2 + 2x' - 1 + x'^2 - 3x' + 3)e^{x'} + 1 - x' - e$
$f(x') = (-x' + 2)e^{x'} + 1 - x' - e$

La solución final, regresando a la variable $x$, es:
$f(x) = (2 - x)e^x - x + 1 - e$

# Ecuación de Poisson

La **ecuación de Poisson** es una de las ecuaciones diferenciales parciales (EDP) más fundamentales en la física matemática, constituyendo la base para la descripción de la electrostática, la gravitación y la teoría de fluidos. 

Matemáticamente, para una función escalar $\phi(r)$, la ecuación se expresa como:
$$\nabla^2 \phi(r) = f(r)$$
donde $f(r)$ representa una densidad de fuente o término inhomogéneo.

### 1. Motivación en la Electrodinámica
La principal motivación de esta ecuación en el ámbito de la electrodinámica surge de la necesidad de determinar los **potenciales** generados por distribuciones de carga y corriente.

*   **Derivación desde la Ley de Gauss:** En electrostática, la forma diferencial de la ley de Gauss establece que la divergencia del campo eléctrico $\mathbf{E}$ es proporcional a la densidad volumétrica de carga $\rho$: $\nabla \cdot \mathbf{E} = \rho / \epsilon_0$. Al definir el campo eléctrico como el gradiente negativo de un potencial escalar ($\mathbf{E} = -\nabla \phi$), se obtiene directamente la ecuación de Poisson para el potencial electrostático:
    $$\nabla^2 \phi(r) = -\frac{\rho(r)}{\epsilon_0}$$
*   **Significado físico:** Esta ecuación vincula la causa (la densidad de carga $\rho$) con el efecto (el potencial $\phi$) en cada punto del espacio. Si la región está libre de cargas ($\rho = 0$), la ecuación de Poisson se reduce a la **ecuación de Laplace**, $\nabla^2 \phi = 0$.

### 2. Extensión a la Electrodinámica dependiente del tiempo
En el contexto de campos variables, las ecuaciones de Maxwell pueden desacoplarse utilizando potenciales y eligiendo un "gauge" o calibre adecuado.

*   **Calibre de Lorentz:** Al imponer la condición de Lorentz ($\nabla \cdot \mathbf{A} + \frac{1}{c^2} \frac{\partial \phi}{\partial t} = 0$), el potencial escalar satisface una ecuación de onda inhomogénea:
    $$\nabla^2 \phi - \frac{1}{c^2} \frac{\partial^2 \phi}{\partial t^2} = -\frac{\rho}{\epsilon_0}$$
    En el límite estacionario (donde los campos no dependen del tiempo), esta expresión recupera la forma clásica de la ecuación de Poisson.

### 3. Ecuación de Poisson Vectorial
La electrodinámica también requiere resolver el **potencial vectorial magnético** $\mathbf{A}$, que se relaciona con la densidad de corriente $\mathbf{J}$. Para corrientes estacionarias en el vacío, el potencial vectorial satisface una versión vectorial de la ecuación de Poisson:
$$\nabla^2 \mathbf{A}(r) = -\mu_0 \mathbf{J}(r)$$
Esta ecuación permite calcular el campo de inducción magnética $\mathbf{B}$ mediante la relación $\mathbf{B} = \nabla \times \mathbf{A}$.

### 4. Resolución mediante Funciones de Green
Debido a la linealidad del operador laplaciano, el método preferido para resolver la ecuación de Poisson con fuentes extendidas es el uso de **funciones de Green**.

*   **Principio:** Se resuelve primero el caso de una fuente puntual unitaria (representada por la delta de Dirac): $\nabla^2 G(r, r') = \delta(r - r')$.
*   **Solución fundamental:** En el espacio infinito tridimensional, la función de Green (que representa el potencial de una carga puntual) es:
    $$G(r, r') = -\frac{1}{4\pi |r - r'|}$$
*   **Integral de convolución:** La solución completa para cualquier distribución de carga se obtiene sumando las contribuciones de cada elemento de fuente:
    $$\phi(r) = \frac{1}{4\pi \epsilon_0} \int \frac{\rho(r')}{|r - r'|} dV'$$
    Este resultado confirma que la ecuación de Poisson es consistente con la ley de Coulomb para el potencial.

$$\phi = k\int_V \frac{dq'}{|\vec{r}-\vec{r}'|} = k\int_V \frac{\rho ' dV'}{|\vec{r}-\vec{r}'|}$$

Esta expresión integral para el potencial electrostático $\phi$ es la **solución general de la ecuación de Poisson** en un espacio infinito y sin fronteras físicas. La relación principal es que la ecuación de Poisson vincula la **causa** (la densidad de carga $\rho$) con el **efecto** (el potencial $\phi$) mediante el operador Laplaciano:
$$\nabla^2 \phi(\vec{r}) = -\frac{\rho(\vec{r})}{\epsilon_0}$$.

A continuación se explica el origen y significado de cada término de manera sencilla:

### 1. Relación con la Función de Green
El núcleo de la integral, el término $\frac{1}{|\vec{r}-\vec{r}'|}$, es esencialmente la **función de Green** fundamental para el operador de Laplace en tres dimensiones. 
*   Representa el potencial generado por una **fuente puntual** (una carga unitaria) ubicada en la posición $\vec{r}'$ y medida desde un punto $\vec{r}$.
*   Matemáticamente, satisface la relación $\nabla^2 \left( \frac{1}{|\vec{r}-\vec{r}'|} \right) = -4\pi\delta(\vec{r}-\vec{r}')$. Al integrar todas estas contribuciones puntuales (usando el principio de superposición), se obtiene el potencial total de una distribución extendida.

### 2. Explicación de los términos del ejercicio

*   **$\phi$ (Potencial escalar):** Representa la respuesta del sistema en un punto de observación $\vec{r}$. Físicamente, es el trabajo necesario por unidad de carga para traer una partícula desde el infinito hasta ese punto.
*   **$k$ (Constante de proporcionalidad):** En el Sistema Internacional, esta constante es $k = \frac{1}{4\pi\epsilon_0}$. Es el factor de escala que determina la intensidad de la interacción eléctrica en el vacío.
*   **$dq'$ o $\rho' dV'$ (La fuente):**
    *   $dV'$ es un pequeño **elemento de volumen** en el lugar donde se encuentra la carga.
    *   $\rho'$ es la **densidad volumétrica de carga** (carga por unidad de volumen) en ese punto específico $\vec{r}'$.
    *   El producto $\rho' dV' = dq'$ representa un "trozo" infinitesimal de carga que actúa como una fuente puntual.
*   **$|\vec{r}-\vec{r}'|$ (La distancia):** Es la magnitud del vector que separa la **fuente** ($\vec{r}'$) del **observador** ($\vec{r}$). Indica que el efecto de cada pequeño trozo de carga disminuye inversamente con la distancia.
*   **$\int_V$ (La suma continua):** Indica que para hallar el potencial total, debemos "sumar" (integrar) las contribuciones individuales de todos los elementos de carga $dq'$ que existen en el volumen $V$.

En resumen, la integral toma la densidad de carga $\rho$ definida en la ecuación de Poisson y utiliza la propiedad de la función de Green para "repartir" su efecto en todo el espacio, resultando en el potencial $\phi$ que observamos.

# Solucionar la ecuación de Poisson

Para resolver la **ecuación de Poisson** en un **espacio infinito** utilizando el método de expansión en la **base de Fourier**, se deben seguir los siguientes pasos analíticos de manera ordenada:

### 1. Definición de la ecuación y el dominio
El objetivo es encontrar la **función de Green** $G(\vec{r}, \vec{r}')$ que actúe como la solución fundamental al operador Laplaciano en un espacio euclidiano tridimensional sin fronteras físicas. La ecuación a resolver es:
$$\nabla^2 G(\vec{r}, \vec{r}') = \delta(\vec{r} - \vec{r}')$$
Donde $\delta(\vec{r} - \vec{r}')$ representa una fuente puntual unitaria en la posición $\vec{r}'$.

### 2. Uso de la condición de completez
En el espacio infinito, la base más apropiada es la **base continua de Fourier** $\{\varphi_{\vec{k}}(\vec{r}) = e^{i\vec{k}\cdot\vec{r}}\}$. La propiedad de **completez** (o relación de clausura) permite representar la delta de Dirac como una superposición integral de estas funciones de base:
$$\delta(\vec{r} - \vec{r}') = \frac{1}{(2\pi)^3} \int e^{i\vec{k}\cdot(\vec{r} - \vec{r}')} d^3k$$

### 3. Expansión de la Función de Green en la base
Expandimos la función desconocida $G(\vec{r}, \vec{r}')$ en la misma base de Fourier, asumiendo que su transformada es $\tilde{G}(\vec{k})$:
$$G(\vec{r}, \vec{r}') = \frac{1}{(2\pi)^3} \int \tilde{G}(\vec{k}) e^{i\vec{k}\cdot(\vec{r} - \vec{r}')} d^3k$$

### 4. Substitución en la ecuación diferencial
Aplicamos el operador Laplaciano $\nabla^2$ sobre la expansión de $G$. Debido a que la derivada actúa solo sobre la variable de observación $\vec{r}$, y sabiendo que $\nabla^2 e^{i\vec{k}\cdot\vec{r}} = -k^2 e^{i\vec{k}\cdot\vec{r}}$:
$$\int (-k^2) \tilde{G}(\vec{k}) e^{i\vec{k}\cdot(\vec{r} - \vec{r}')} \frac{d^3k}{(2\pi)^3} = \int (1) e^{i\vec{k}\cdot(\vec{r} - \vec{r}')} \frac{d^3k}{(2\pi)^3}$$

### 5. Solución en el espacio de momentos ($\vec{k}$)
Por la independencia lineal de los elementos de la base, igualamos los integrandos:
$$-k^2 \tilde{G}(\vec{k}) = 1 \implies \tilde{G}(\vec{k}) = -\frac{1}{k^2}$$
Este resultado muestra que la respuesta del sistema en el espacio de frecuencias es inversamente proporcional al cuadrado del número de onda.

### 6. Evaluación de la integral de inversión
Para obtener la forma espacial de $G$, realizamos la transformada inversa integrando en **coordenadas esféricas** en el espacio $\vec{k}$ ($d^3k = k^2 dk \sin\theta_k d\theta_k d\phi_k$):
1.  **Integración angular:** Al orientar el eje $z$ de la integración a lo largo del vector $\vec{R} = \vec{r} - \vec{r}'$, la integral angular resulta en $\frac{4\pi \sin(kR)}{kR}$.
2.  **Integral radial:** La expresión se reduce a una integral unidimensional:
    $$G(R) = -\frac{1}{2\pi^2 R} \int_0^\infty \frac{\sin(kR)}{k} dk$$

### 7. Resultado final
Utilizando el valor de la integral impropia $\int_0^\infty \frac{\sin(x)}{x} dx = \frac{\pi}{2}$, llegamos a la **solución fundamental** de la ecuación de Poisson en 3D:
$$\mathbf{G(\vec{r}, \vec{r}') = -\frac{1}{4\pi |\vec{r} - \vec{r}'|}}$$
Físicamente, este resultado confirma que en un espacio infinito el potencial generado por una fuente puntual decae inversamente con la distancia.

Para realizar la evaluación de la integral de inversión de la **función de Green** para la **ecuación de Poisson** en el espacio infinito ($G(\mathbf{r}, \mathbf{r}') = -\frac{1}{8\pi^3} \int \frac{e^{i\mathbf{k}\cdot(\mathbf{r}-\mathbf{r}')}}{k^2} d^3k$), utilizaremos dos métodos analíticos extensos basados en las fuentes: la reducción mediante **coordenadas esféricas** y la técnica de **integración de contorno** en el plano complejo.

### 1. Método de Integración en Coordenadas Esféricas
Este método se basa en aprovechar la simetría rotacional del problema en el espacio de momentos ($\mathbf{k}$).

**Paso A: Definición del sistema de coordenadas**
Definimos el vector de separación como $\mathbf{R} = \mathbf{r} - \mathbf{r}'$ y su magnitud como $R = |\mathbf{r} - \mathbf{r}'|$. Para simplificar la integral, orientamos el eje $z$ del espacio $\mathbf{k}$ de modo que coincida con la dirección de $\mathbf{R}$. De esta forma, el producto escalar se reduce a:
$$\mathbf{k} \cdot \mathbf{R} = k R \cos \theta_k$$
Donde $\theta_k$ es el ángulo polar en el espacio de momentos.

**Paso B: Transformación del diferencial de volumen**
El elemento de volumen $d^3k$ en coordenadas esféricas es $k^2 dk \sin \theta_k d\theta_k d\phi_k$. La integral de inversión se reescribe como:
$$G(R) = -\frac{1}{8\pi^3} \int_0^\infty dk \int_0^{2\pi} d\phi_k \int_0^\pi d\theta_k \frac{e^{ikR \cos \theta_k}}{k^2} k^2 \sin \theta_k$$

**Paso C: Integración angular**
*   La integral sobre el ángulo azimutal $\phi_k$ es inmediata y produce un factor de $2\pi$.
*   Para la integral sobre $\theta_k$, realizamos el cambio de variable $u = \cos \theta_k$, lo que implica $du = -\sin \theta_k d\theta_k$:
$$\int_0^\pi e^{ikR \cos \theta_k} \sin \theta_k d\theta_k = \int_{-1}^1 e^{ikRu} du = \left[ \frac{e^{ikRu}}{ikR} \right]_{-1}^1 = \frac{e^{ikR} - e^{-ikR}}{ikR}$$
Utilizando la identidad de Euler, esto es equivalente a $\frac{2 \sin(kR)}{kR}$.

**Paso D: Reducción a una integral radial**
Sustituyendo los resultados angulares, las potencias de $k^2$ se cancelan y obtenemos:
$$G(R) = -\frac{1}{8\pi^3} (2\pi) \int_0^\infty \frac{2 \sin(kR)}{kR} dk = -\frac{1}{2\pi^2 R} \int_0^\infty \frac{\sin(kR)}{k} dk$$
Esta es una integral impropia estándar conocida como la **integral de Dirichlet**, cuyo valor es $\pi/2$. Por lo tanto:
$$\mathbf{G(R) = -\frac{1}{2\pi^2 R} \left( \frac{\pi}{2} \right) = -\frac{1}{4\pi |\mathbf{r} - \mathbf{r}'|}}$$

---

### 2. Método de Integración de Contorno (Plano Complejo)
Este método se utiliza para evaluar formalmente la integral resultante del paso anterior, $J = \int_0^\infty \frac{\sin(kR)}{k} dk$, utilizando la teoría de residuos.

**Paso A: Extensión del dominio**
Dado que el integrando $\frac{\sin(kR)}{k}$ es una función par, podemos extender el límite de integración a toda la recta real y dividir por dos:
$$J = \frac{1}{2} \int_{-\infty}^\infty \frac{\sin(kR)}{k} dk = \frac{1}{2} \text{Im} \left[ P \int_{-\infty}^\infty \frac{e^{ikR}}{k} dk \right]$$
Donde $P$ denota el **Valor Principal de Cauchy** debido al polo simple en $k=0$.

**Paso B: Construcción del contorno de integración**
Consideramos la función compleja $f(z) = \frac{e^{izR}}{z}$. Para evaluarla, cerramos el camino de integración en el plano complejo utilizando:
1.  Un segmento sobre el eje real esquivando el origen.
2.  Un **semicírculo pequeño** ($C_r$) de radio $r \to 0$ que pasa por encima del polo en $z=0$.
3.  Un **semicírculo grande** ($C_R$) en el semiplano superior ($Im(z) > 0$) para asegurar que el exponencial decaiga, según el **Lema de Jordan**.

**Paso C: Aplicación del Teorema de los Residuos**
Como el contorno cerrado no encierra ninguna singularidad (el polo quedó fuera por la indentación superior), la integral total es cero por el Teorema de Cauchy:
$$\oint f(z) dz = \int_{segmentos} f(k) dk + \int_{C_r} f(z) dz + \int_{C_R} f(z) dz = 0$$

*   La integral sobre el círculo grande $C_R$ tiende a cero.
*   La integral sobre el semicírculo pequeño $C_r$, recorrido en sentido horario, contribuye con $-i\pi$ veces el residuo en $z=0$. El residuo de $\frac{e^{izR}}{z}$ en $z=0$ es $\lim_{z \to 0} z \frac{e^{izR}}{z} = 1$. Por lo tanto, esta parte vale $-i\pi(1) = -i\pi$.

**Paso D: Cálculo del valor final**
Despejando la integral sobre el eje real (el valor principal):
$$P \int_{-\infty}^\infty \frac{e^{ikR}}{k} dk - i\pi = 0 \implies P \int_{-\infty}^\infty \frac{e^{ikR}}{k} dk = i\pi$$
Tomando la parte imaginaria y multiplicando por el factor $1/(4i)$ (procedente de la definición de seno y el factor 1/2 inicial):
$$J = \frac{1}{2} \text{Im}(i\pi) = \frac{\pi}{2}$$

Este resultado confirma nuevamente la evaluación de la función de Green como **$-1/(4\pi R)$**, validando la consistencia entre el análisis geométrico y el análisis de variable compleja.
