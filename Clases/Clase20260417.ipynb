# PARTE TEÓRICA

## 1. El Problema de Sturm-Liouville

### 1.1 Forma estándar

Un **problema de Sturm-Liouville (S-L)** en el intervalo $[a, b]$ consiste en encontrar los valores $\lambda$ (autovalores) para los cuales la ecuación diferencial

$$\boxed{\frac{d}{dx}\!\left[p(x)\,\frac{dy}{dx}\right] + q(x)\,y + \lambda\,\rho(x)\,y = 0,\qquad x\in[a,b]}$$

tiene soluciones no triviales $y(x)$ (autofunciones) que satisfacen condiciones de frontera homogéneas. Las funciones $p,q,\rho$ son reales y $\rho(x)>0$ es la **función de peso**.

### 1.2 El operador de S-L

Definimos el operador diferencial
$$\mathscr{L}\,y \;=\; -\frac{1}{\rho(x)}\left\{\frac{d}{dx}\!\left[p(x)\,\frac{dy}{dx}\right] + q(x)\,y\right\}$$

de manera que el problema de S-L se reescribe como el **problema de autovalores**
$$\mathscr{L}\,y_n = \lambda_n\,y_n.$$

### 1.3 Condiciones de frontera

Las condiciones de frontera homogéneas más comunes son:

| Tipo | Condición en $x=a$ | Condición en $x=b$ |
|------|--------------------|--------------------|
| **Dirichlet** | $y(a)=0$ | $y(b)=0$ |
| **Neumann** | $y'(a)=0$ | $y'(b)=0$ |
| **Mixta** | $y(a)=0$ | $y'(b)=0$ (o viceversa) |
| **Robin / 3ª especie** | $\alpha_1 y(a)+\beta_1 y'(a)=0$ | $\alpha_2 y(b)+\beta_2 y'(b)=0$ |

### 1.4 Propiedades del operador hermítico

Si $p(a)=p(b)=0$ **o** si se satisfacen condiciones de frontera homogéneas apropiadas, el operador $\mathscr{L}$ es **hermítico** (autoadjunto) con respecto al producto interno ponderado
$$(f,g)_\rho = \int_a^b f^*(x)\,g(x)\,\rho(x)\,dx.$$

Como consecuencia:
1. **Autovalores reales:** $\lambda_n \in \mathbb{R}$.
2. **Ortogonalidad:** $\displaystyle\int_a^b y_m(x)\,y_n(x)\,\rho(x)\,dx = 0$ para $m\neq n$.
3. **Completitud:** Cualquier función $f\in L^2_\rho[a,b]$ se puede expandir como
   $$f(x) = \sum_{n=1}^{\infty} c_n\,y_n(x),\qquad c_n = \frac{(y_n, f)_\rho}{(y_n,y_n)_\rho}.$$

### 1.5 Identidad de Lagrange-Green (identidad bilineal)

Para dos funciones $u, v$ que satisfacen condiciones de frontera que hacen hermítico al operador:
$$\int_a^b \left[v\,\mathscr{L}u - u\,\mathscr{L}v\right]\rho\,dx = \Big[p(x)\left(v\,u' - u\,v'\right)\Big]_a^b = 0.$$

Esta identidad es el **fundamento algebraico** de la ortogonalidad.

### 1.6 Conversión de una EDO de 2° orden a forma S-L

Dada la EDO general
$$a_2(x)\,y'' + a_1(x)\,y' + a_0(x)\,y + \lambda\,\rho_0(x)\,y = 0,$$

el **factor integrante** que la lleva a forma S-L es
$$\mu(x) = \frac{1}{a_2(x)}\exp\!\left(\int \frac{a_1(x)}{a_2(x)}\,dx\right)$$

de modo que
$$p(x)=a_2(x)\,\mu(x),\quad q(x)=a_0(x)\,\mu(x),\quad \rho(x)=\rho_0(x)\,\mu(x).$$

**Tabla resumen de ecuaciones clásicas:**

| Ecuación | $p(x)$ | $q(x)$ | $\rho(x)$ | Autofunciones |
|----------|--------|--------|-----------|---------------|
| Legendres en $[-1,1]$ | $1-x^2$ | $0$ | $1$ | $P_n(x)$ |
| Bessel orden $\nu$ en $[0,R]$ | $x$ | $-\nu^2/x$ | $x$ | $J_\nu(k_n x)$ |
| Hermite en $(-\infty,\infty)$ | $e^{-x^2}$ | $0$ | $e^{-x^2}$ | $H_n(x)$ |
| Trigonométrica en $[0,L]$ | $1$ | $0$ | $1$ | $\sin,\cos$ |

# Problema 1 — Cuerda Vibrante (Condiciones de Dirichlet)

## Enunciado físico

Una cuerda elástica de longitud $L$ está fija en ambos extremos. La ecuación de onda 1D para el desplazamiento transversal $u(x,t)$ es
$$\frac{\partial^2 u}{\partial t^2} = c^2\,\frac{\partial^2 u}{\partial x^2}, \qquad 0 < x < L,\; t > 0,$$
con condiciones de frontera de **Dirichlet** $u(0,t) = u(L,t) = 0$ (extremos fijos).

## Separación de variables

Proponiendo $u(x,t)=X(x)\,T(t)$, obtenemos
$$\frac{T''}{c^2 T} = \frac{X''}{X} = -\lambda.$$

El **problema de S-L para $X(x)$** es
$$X'' + \lambda\,X = 0, \qquad X(0)=0,\; X(L)=0.$$

Aquí $p(x)=1$, $q(x)=0$, $\rho(x)=1$.

## Solución analítica

Los **autovalores** son $\lambda_n = \left(\dfrac{n\pi}{L}\right)^2$, $n=1,2,3,\ldots$

Las **autofunciones** son $X_n(x) = \sin\!\left(\dfrac{n\pi x}{L}\right)$.

La solución general es la **serie de Fourier en senos**:
$$u(x,t) = \sum_{n=1}^{\infty}\left[A_n\cos\!\left(\frac{n\pi c}{L}t\right)+B_n\sin\!\left(\frac{n\pi c}{L}t\right)\right]\sin\!\left(\frac{n\pi x}{L}\right).$$

```
(* === Problema 1 — Cuerda vibrante: autofunciones y autovalores === *)

(* Parámetros *)
L = 1;  (* longitud de la cuerda *)
c = 1;  (* velocidad de onda *)
nMax = 6;

(* Autovalores y autofunciones *)
autovaloresCuerda = Table[(n Pi / L)^2, {n, 1, nMax}];
autofuncionesCuerda[n_][x_] := Sin[n Pi x / L];

Print["Autovalores \[Lambda]_n = (n\[Pi]/L)^2 para n=1,...,", nMax, ":"]
Print[autovaloresCuerda // N // TableForm]
```

```
(* === Gráfica 1a: Primeros modos de vibración de la cuerda === *)
modosCuerda = Table[
  Plot[autofuncionesCuerda[n][x], {x, 0, L},
    PlotStyle -> {colores[[Mod[n-1, 6]+1]], Thickness[0.003]},
    PlotLabel -> "Modo n=" <> ToString[n] <> 
                 ",  \[Lambda]=" <> ToString[N[(n Pi/L)^2, 4]],
    AxesLabel -> {"x", Row[{Subscript["X",n],"(x)"}]},
    GridLines -> Automatic, GridLinesStyle -> LightGray,
    Frame -> True, PlotRange -> {-1.2, 1.2}],
  {n, 1, 6}
];

GraphicsGrid[Partition[modosCuerda, 3],
  PlotLabel -> Style["Autofunciones de la cuerda vibrante (Dirichlet)", Bold, 14],
  ImageSize -> 800]
```

```
(* === Gráfica 1b: Todos los modos en un solo plot === *)
Plot[
  Evaluate[Table[autofuncionesCuerda[n][x], {n, 1, 5}]],
  {x, 0, L},
  PlotStyle -> (Directive[#, Thickness[0.003]] & /@ colores),
  PlotLegends -> Table["n=" <> ToString[n], {n, 1, 5}],
  AxesLabel -> {"x", "X(x)"},
  PlotLabel -> Style["Modos normales — Cuerda vibrante (Dirichlet)", Bold],
  GridLines -> Automatic, GridLinesStyle -> LightGray,
  Frame -> True, ImageSize -> 500
]
```

```
(* === Gráfica 1c: Solución u(x,t) animada para condición inicial triangular === *)
(* Condición inicial: u(x,0) = f(x), ut(x,0)=0 — perfil triangular *)
fCuerda[x_] := Piecewise[{{2x/L, 0 <= x <= L/2}, {2(1-x/L), L/2 < x <= L}}];

(* Coeficientes de Fourier *)
An[n_] := (2/L) NIntegrate[fCuerda[x] Sin[n Pi x/L], {x, 0, L}];
coefsCuerda = Table[An[n], {n, 1, 15}];

(* Solución en serie (Bn=0 porque velocidad inicial=0) *)
uCuerda[x_, t_, nTerms_] := Sum[
  coefsCuerda[[n]] Cos[n Pi c/L t] Sin[n Pi x/L],
  {n, 1, nTerms}
];

(* Comparación de la condición inicial con su serie de Fourier *)
Plot[
  {fCuerda[x], uCuerda[x, 0, 5], uCuerda[x, 0, 15]},
  {x, 0, L},
  PlotStyle -> {Directive[Black, Dashed, Thickness[0.004]],
                Directive[colores[[2]], Thickness[0.003]],
                Directive[colores[[1]], Thickness[0.003]]},
  PlotLegends -> {"f(x) exacta", "Serie N=5", "Serie N=15"},
  AxesLabel -> {"x", "u(x,0)"},
  PlotLabel -> Style["Condición inicial triangular y su serie de Fourier", Bold],
  Frame -> True, GridLines -> Automatic, GridLinesStyle -> LightGray,
  ImageSize -> 500
]
```

```
(* === Gráfica 1d: Evolución temporal de la cuerda === *)
tSnaps = {0, 0.1, 0.25, 0.5};
Plot[
  Evaluate[Table[uCuerda[x, t, 15], {t, tSnaps}]],
  {x, 0, L},
  PlotStyle -> (Directive[#, Thickness[0.003]] & /@ colores[[1;;4]]),
  PlotLegends -> ("t=" <> ToString[#] & /@ tSnaps),
  AxesLabel -> {"x", "u(x,t)"},
  PlotLabel -> Style["Evolución temporal de la cuerda vibrante", Bold],
  Frame -> True, GridLines -> Automatic, GridLinesStyle -> LightGray,
  PlotRange -> {-1, 1}, ImageSize -> 500
]
```

# Problema 2 — Partícula en Caja Cuántica Infinita (Condiciones de Dirichlet)

## Enunciado físico

La ecuación de Schrödinger independiente del tiempo para una partícula de masa $m$ confinada en $[0,L]$ con paredes infinitas es
$$-\frac{\hbar^2}{2m}\,\psi''(x) = E\,\psi(x), \qquad \psi(0)=\psi(L)=0.$$

## Forma S-L

Identificamos $p(x)=1$, $q(x)=0$, $\rho(x)=1$, $\lambda = \dfrac{2mE}{\hbar^2}$. El problema es idéntico en estructura al de la cuerda, pero la interpretación física es diferente.

## Solución

$$E_n = \frac{n^2\pi^2\hbar^2}{2mL^2}, \qquad \psi_n(x) = \sqrt{\frac{2}{L}}\,\sin\!\left(\frac{n\pi x}{L}\right).$$

Las funciones están **normalizadas**: $\int_0^L |\psi_n|^2\,dx = 1$.

```
(* === Problema 2 — Partícula en caja cuántica === *)

(* Unidades: hbar=1, 2m=1, L=1 (unidades atómicas simplificadas) *)
hbar = 1; masa = 1/2; LQ = 1;

(* Niveles de energía (en unidades de E1) *)
E1 = Pi^2 hbar^2 / (2 masa LQ^2);
energiasQ = Table[n^2 E1, {n, 1, 6}];
Print["Niveles de energía En/E1: ", Table[n^2, {n,1,6}]]

(* Funciones de onda normalizadas *)
psiQ[n_][x_] := Sqrt[2/LQ] Sin[n Pi x/LQ];

(* Verificación de normalización *)
normas = Table[NIntegrate[psiQ[n][x]^2, {x, 0, LQ}], {n, 1, 4}];
Print["Normas (deben ser 1): ", normas]
```

```
(* === Gráfica 2a: Funciones de onda y densidades de probabilidad === *)
nMostrar = 4;

plotFuncionesOnda = Plot[
  Evaluate[Table[psiQ[n][x] + n^2, {n, 1, nMostrar}]],
  {x, 0, LQ},
  PlotStyle -> (Directive[#, Thickness[0.003]] & /@ colores[[1;;nMostrar]]),
  PlotLegends -> Table["\[Psi]" <> ToString[n] <> "(x)", {n, 1, nMostrar}],
  AxesLabel -> {"x/L", "\[Psi]_n(x) + E_n/E_1"},
  PlotLabel -> Style["Funciones de onda (desplazadas al nivel de energía)", Bold],
  Epilog -> {Dashed, Table[
    Line[{{0, n^2}, {LQ, n^2}}], {n, 1, nMostrar}]},
  Frame -> True, GridLines -> None, ImageSize -> 480
];

plotDensidades = Plot[
  Evaluate[Table[psiQ[n][x]^2, {n, 1, nMostrar}]],
  {x, 0, LQ},
  Filling -> Axis,
  FillingStyle -> (Directive[Opacity[0.25], #] & /@ colores[[1;;nMostrar]]),
  PlotStyle -> (Directive[#, Thickness[0.003]] & /@ colores[[1;;nMostrar]]),
  PlotLegends -> Table["|\[Psi]" <> ToString[n] <> "|\[Squared]", {n, 1, nMostrar}],
  AxesLabel -> {"x/L", "|\[Psi]_n(x)|\[Squared]"},
  PlotLabel -> Style["Densidades de probabilidad", Bold],
  Frame -> True, GridLines -> Automatic, GridLinesStyle -> LightGray,
  ImageSize -> 480
];

GraphicsRow[{plotFuncionesOnda, plotDensidades}, ImageSize -> 980]
```

# Problema 3 — Ecuación de Calor con Extremos a Temperatura Cero (Dirichlet)

## Enunciado

Una barra de longitud $L=1$ m tiene sus extremos mantenidos a temperatura $T=0$. La distribución de temperatura inicial es $T(x,0)=f(x)$. La ecuación de calor es
$$\frac{\partial T}{\partial t} = \kappa\,\frac{\partial^2 T}{\partial x^2}, \qquad T(0,t)=T(L,t)=0.$$

## Forma S-L

Tras separación de variables, el problema espacial es:
$$X'' + \lambda X = 0, \qquad X(0)=X(L)=0 \quad \text{(Dirichlet)}.$$

## Solución

$$T(x,t) = \sum_{n=1}^{\infty} B_n\,e^{-\kappa (n\pi/L)^2 t}\,\sin\!\left(\frac{n\pi x}{L}\right),$$
$$B_n = \frac{2}{L}\int_0^L f(x)\,\sin\!\left(\frac{n\pi x}{L}\right)\,dx.$$

```
(* === Problema 3 — Ecuación de calor con Dirichlet === *)

kappa = 0.1;  (* difusividad térmica *)
LH = 1;

(* Condición inicial: perfil gaussiano centrado *)
fCalor[x_] := Exp[-20 (x - LH/2)^2];

(* Coeficientes de Fourier en senos *)
Bn[n_] := (2/LH) NIntegrate[fCalor[x] Sin[n Pi x/LH], {x, 0, LH}];
nTermsHeat = 30;
coefsCalor = Table[Bn[n], {n, 1, nTermsHeat}];

(* Temperatura en serie *)
TCalor[x_, t_] := Sum[
  coefsCalor[[n]] Exp[-kappa (n Pi/LH)^2 t] Sin[n Pi x/LH],
  {n, 1, nTermsHeat}
];

Print["Configuración lista. Primeros 5 coeficientes Bn:", Round[coefsCalor[[1;;5]], 0.0001]]
```

```
(* === Gráfica 3a: Perfil de temperatura en distintos tiempos (Dirichlet) === *)
tiemposCalor = {0, 0.5, 1, 2, 5, 10};

Plot[
  Evaluate[Table[TCalor[x, t], {t, tiemposCalor}]],
  {x, 0, LH},
  PlotStyle -> (Directive[#, Thickness[0.003]] & /@ colores),
  PlotLegends -> ("t=" <> ToString[#] & /@ tiemposCalor),
  AxesLabel -> {"x (m)", "T(x,t) (°C)"},
  PlotLabel -> Style["Difusión de calor — Extremos fijos T=0 (Dirichlet)", Bold],
  Frame -> True, GridLines -> Automatic, GridLinesStyle -> LightGray,
  PlotRange -> {-0.05, 1.1},
  ImageSize -> 520
]
```

```
(* === Gráfica 3b: Mapa de calor 2D — T(x,t) === *)
DensityPlot[
  TCalor[x, t],
  {x, 0, LH}, {t, 0, 15},
  ColorFunction -> "TemperatureMap",
  FrameLabel -> {{"t (s)", None}, {"x (m)", "T(x,t)"}},
  PlotLabel -> Style["Mapa de temperatura T(x,t) — Dirichlet", Bold],
  PlotPoints -> 80,
  ImageSize -> 520
]
```

# Problema 4 — Barra Adiabática (Condiciones de Neumann)

## Enunciado

Una barra de longitud $L$ tiene sus extremos **perfectamente aislados** (flujo de calor cero en los extremos). La ecuación de calor es
$$\frac{\partial T}{\partial t} = \kappa\,\frac{\partial^2 T}{\partial x^2}, \qquad \frac{\partial T}{\partial x}\bigg|_{x=0} = \frac{\partial T}{\partial x}\bigg|_{x=L} = 0 \quad \text{(Neumann)}.$$

## Forma S-L

El problema S-L espacial:
$$X'' + \lambda X = 0, \qquad X'(0)=X'(L)=0.$$

## Solución

Para $\lambda_0=0$: $X_0(x)=1$ (modo constante, $T$ se equilibra al promedio).

Para $\lambda_n = (n\pi/L)^2$, $n=1,2,\ldots$: $X_n(x)=\cos\!(n\pi x/L)$.

$$T(x,t) = \frac{A_0}{2} + \sum_{n=1}^{\infty} A_n\,e^{-\kappa(n\pi/L)^2 t}\,\cos\!\left(\frac{n\pi x}{L}\right),$$
$$A_0 = \frac{2}{L}\int_0^L f(x)\,dx, \qquad A_n = \frac{2}{L}\int_0^L f(x)\cos\!\left(\frac{n\pi x}{L}\right)\,dx.$$

**Nota física importante:** La temperatura converge al promedio inicial $\bar{T}=A_0/2$, no a cero. Esto refleja la conservación de la energía térmica en un sistema aislado.

```
(* === Problema 4 — Barra adiabática (Neumann) === *)

LN = 1; kappaN = 0.1;

(* Condición inicial: perfil asimétrico *)
fAd[x_] := x (1 - x);

(* Valor medio (A0/2) *)
A0ad = (2/LN) NIntegrate[fAd[x], {x, 0, LN}];
Tpromedio = A0ad/2;

(* Coeficientes en cosenos *)
Anan[n_] := (2/LN) NIntegrate[fAd[x] Cos[n Pi x/LN], {x, 0, LN}];
nTermsAd = 20;
coefsAd = Table[Anan[n], {n, 1, nTermsAd}];

(* Solución en serie *)
TAd[x_, t_] := Tpromedio + Sum[
  coefsAd[[n]] Exp[-kappaN (n Pi/LN)^2 t] Cos[n Pi x/LN],
  {n, 1, nTermsAd}
];

Print["Temperatura de equilibrio (promedio inicial): ", N[Tpromedio, 5]]
```

```
(* === Gráfica 4a: Temperatura en la barra adiabática === *)
tiemposAd = {0, 1, 2, 5, 10, 30};

Plot[
  Evaluate[Append[Table[TAd[x, t], {t, tiemposAd}], Tpromedio]],
  {x, 0, LN},
  PlotStyle -> Append[
    (Directive[#, Thickness[0.003]] & /@ colores),
    Directive[Black, Dashed, Thickness[0.003]]
  ],
  PlotLegends -> Append[
    "t=" <> ToString[#] & /@ tiemposAd,
    "T promedio"
  ],
  AxesLabel -> {"x (m)", "T(x,t)"},
  PlotLabel -> Style["Barra adiabática — Neumann (extremos aislados)", Bold],
  Frame -> True, GridLines -> Automatic, GridLinesStyle -> LightGray,
  ImageSize -> 520
]
```

```
(* === Gráfica 4b: Autofunciones de Neumann (cosenos) === *)
Plot[
  Evaluate[Table[Cos[n Pi x/LN], {n, 0, 4}]],
  {x, 0, LN},
  PlotStyle -> (Directive[#, Thickness[0.003]] & /@ colores),
  PlotLegends -> Table["n=" <> ToString[n], {n, 0, 4}],
  AxesLabel -> {"x", "X_n(x) = cos(n\[Pi]x/L)"},
  PlotLabel -> Style["Autofunciones — Condición de Neumann", Bold],
  Frame -> True, GridLines -> Automatic, GridLinesStyle -> LightGray,
  Epilog -> {
    Text[Style["X'(0)=0, X'(L)=0", Italic, 11], {0.5, -1.15}]
  },
  PlotRange -> {-1.3, 1.3}, ImageSize -> 500
]
```

# Problema 5 — Modos Acústicos en Tubo Cerrado (Neumann)

## Enunciado físico

En un tubo acústico cerrado en ambos extremos ($x=0$ y $x=L$), la **presión** satisface la ecuación de onda
$$\frac{\partial^2 p}{\partial t^2} = v_s^2\,\frac{\partial^2 p}{\partial x^2},$$
con condiciones de **Neumann** (velocidad del fluido nula en los extremos):
$$\frac{\partial p}{\partial x}\bigg|_{x=0} = \frac{\partial p}{\partial x}\bigg|_{x=L} = 0.$$

## Frecuencias naturales (modos normales)

$$f_n = \frac{n\,v_s}{2L}, \qquad n = 0, 1, 2, \ldots$$

El modo $n=0$ es la **presión uniforme estática** (no hay oscilación). Los modos $n\geq 1$ son los **armónicos**. La diferencia con el tubo abierto (Dirichlet) es que aquí aparece el modo $n=0$ y los perfiles son cosenos.

```
(* === Problema 5 — Modos acústicos en tubo cerrado (Neumann) === *)

vs = 343;   (* velocidad del sonido en aire, m/s *)
LTubo = 1; (* longitud del tubo en metros *)

(* Frecuencias naturales *)
frecuencias = Table[n vs / (2 LTubo), {n, 0, 6}];
Print["Frecuencias naturales f_n (Hz) para L=1m: ", frecuencias]

(* Perfiles de presión (autofunciones) *)
pModo[n_][x_] := Cos[n Pi x / LTubo];

(* Gráfica de los primeros 5 modos de presión *)
Plot[
  Evaluate[Table[pModo[n][x], {n, 0, 4}]],
  {x, 0, LTubo},
  PlotStyle -> (Directive[#, Thickness[0.003]] & /@ colores),
  PlotLegends -> Table[
    "n=" <> ToString[n] <> ", f=" <> ToString[N[frecuencias[[n+1]],4]] <> " Hz",
    {n, 0, 4}
  ],
  AxesLabel -> {"x (m)", "p_n(x)"},
  PlotLabel -> Style["Modos de presión — Tubo cerrado (Neumann)", Bold],
  Epilog -> {
    Gray, Dashed,
    Line[{{0, 0}, {LTubo, 0}}]
  },
  Frame -> True, GridLines -> Automatic, GridLinesStyle -> LightGray,
  PlotRange -> {-1.3, 1.5}, ImageSize -> 530
]
```

```
(* === Gráfica 5b: Comparación Dirichlet vs Neumann (tubo abierto vs cerrado) === *)
GraphicsGrid[
  {{
    Plot[Evaluate[Table[Sin[n Pi x/LTubo], {n, 1, 4}]],
      {x, 0, LTubo},
      PlotStyle -> (Directive[#, Thickness[0.003]] & /@ colores[[1;;4]]),
      PlotLabel -> Style["Tubo abierto (Dirichlet)\nPresión nula en extremos", Bold, 10],
      AxesLabel -> {"x", "p_n(x)"},
      Frame -> True, GridLines -> Automatic, GridLinesStyle -> LightGray,
      PlotLegends -> Table["n=" <> ToString[n], {n, 1, 4}]
    ],
    Plot[Evaluate[Table[Cos[n Pi x/LTubo], {n, 0, 3}]],
      {x, 0, LTubo},
      PlotStyle -> (Directive[#, Thickness[0.003]] & /@ colores[[1;;4]]),
      PlotLabel -> Style["Tubo cerrado (Neumann)\nVelocidad nula en extremos", Bold, 10],
      AxesLabel -> {"x", "p_n(x)"},
      Frame -> True, GridLines -> Automatic, GridLinesStyle -> LightGray,
      PlotLegends -> Table["n=" <> ToString[n], {n, 0, 3}]
    ]
  }},
  ImageSize -> 860
]
```

# Problema 6 — Barra con Extremo Fijo y Extremo Adiabático (Condición Mixta)

## Enunciado

Una barra de longitud $L$ tiene:
- Extremo izquierdo ($x=0$): **temperatura fija** $T=0$ → condición de **Dirichlet**.
- Extremo derecho ($x=L$): **aislado** (flujo de calor cero) → condición de **Neumann**.

$$\frac{\partial T}{\partial t} = \kappa\,\frac{\partial^2 T}{\partial x^2}, \qquad T(0,t)=0,\quad T_x(L,t)=0.$$

## Forma S-L y solución

El problema de S-L espacial es:
$$X'' + \lambda X = 0, \qquad X(0)=0,\; X'(L)=0.$$

Las condiciones mixtas obligan a:
$$X(x) = A\sin(\sqrt{\lambda}\,x),$$
$$X'(L)=0 \;\Rightarrow\; A\sqrt{\lambda}\cos(\sqrt{\lambda}\,L)=0 \;\Rightarrow\; \sqrt{\lambda_n}\,L = \frac{(2n-1)\pi}{2}.$$

Así:
$$\lambda_n = \left(\frac{(2n-1)\pi}{2L}\right)^2,\qquad X_n(x) = \sin\!\left(\frac{(2n-1)\pi x}{2L}\right), \qquad n=1,2,3,\ldots$$

Observar que los autovalores son **distintos** a los de Dirichlet puro y Neumann puro.

```
(* === Problema 6 — Condición mixta Dirichlet/Neumann === *)

LM = 1; kappaM = 0.1;

(* Autovalores: lambda_n = ((2n-1)pi/(2L))^2 *)
sqrtLambdaMix[n_] := (2n - 1) Pi / (2 LM);
lambdaMix[n_] := sqrtLambdaMix[n]^2;

(* Autofunciones: sin((2n-1)pi x/(2L)) *)
XMix[n_][x_] := Sin[sqrtLambdaMix[n] x];

(* Comparar con lambda_n de Dirichlet puro *)
Grid[
  Prepend[
    Table[{n,
      N[lambdaMix[n], 5],
      N[(n Pi/LM)^2, 5]},
      {n, 1, 5}
    ],
    {"n", "\[Lambda]_n mixto", "\[Lambda]_n Dirichlet"}
  ],
  Frame -> All, Background -> {{LightBlue}, {LightYellow, None}}
]
```

```
(* === Gráfica 6a: Autofunciones mixtas vs Dirichlet === *)
GraphicsRow[
  {
    Plot[
      Evaluate[Table[XMix[n][x], {n, 1, 5}]],
      {x, 0, LM},
      PlotStyle -> (Directive[#, Thickness[0.003]] & /@ colores),
      PlotLegends -> Table["n=" <> ToString[n], {n, 1, 5}],
      AxesLabel -> {"x", "X_n(x)"},
      PlotLabel -> Style["Autofunciones Mixtas\nX(0)=0, X'(L)=0", Bold, 10],
      Epilog -> {Blue, PointSize[0.02],
        Point[{LM, #}] & /@ Table[0, {5}]},
      Frame -> True, GridLines -> Automatic, GridLinesStyle -> LightGray,
      PlotRange -> {-1.3, 1.3}, ImageSize -> 400
    ],
    Plot[
      Evaluate[Table[Sin[n Pi x/LM], {n, 1, 5}]],
      {x, 0, LM},
      PlotStyle -> (Directive[#, Thickness[0.003]] & /@ colores),
      PlotLegends -> Table["n=" <> ToString[n], {n, 1, 5}],
      AxesLabel -> {"x", "X_n(x)"},
      PlotLabel -> Style["Autofunciones Dirichlet\nX(0)=0, X(L)=0", Bold, 10],
      Frame -> True, GridLines -> Automatic, GridLinesStyle -> LightGray,
      PlotRange -> {-1.3, 1.3}, ImageSize -> 400
    ]
  },
  ImageSize -> 850
]
```

```
(* === Problema 6 cont.: Solución de la ecuación de calor con BC mixta === *)

(* Condición inicial: perfil lineal f(x)=x *)
fMix[x_] := x;

(* Normas de las autofunciones *)
normaMix[n_] := NIntegrate[XMix[n][x]^2, {x, 0, LM}]; (* = L/2 *)

(* Coeficientes de la serie *)
BnMix[n_] := NIntegrate[fMix[x] XMix[n][x], {x, 0, LM}] / normaMix[n];
nTermsMix = 20;
coefsMix = Table[BnMix[n], {n, 1, nTermsMix}];

(* Solución *)
TMix[x_, t_] := Sum[
  coefsMix[[n]] Exp[-kappaM lambdaMix[n] t] XMix[n][x],
  {n, 1, nTermsMix}
];

(* Gráfica de la evolución *)
tiemposMix = {0, 0.5, 1, 3, 8, 20};
Plot[
  Evaluate[Table[TMix[x, t], {t, tiemposMix}]],
  {x, 0, LM},
  PlotStyle -> (Directive[#, Thickness[0.003]] & /@ colores),
  PlotLegends -> ("t=" <> ToString[#] & /@ tiemposMix),
  AxesLabel -> {"x (m)", "T(x,t)"},
  PlotLabel -> Style["Calor — BC Mixta: T(0)=0, T'(L)=0", Bold],
  Frame -> True, GridLines -> Automatic, GridLinesStyle -> LightGray,
  PlotRange -> {-0.05, 1.05}, ImageSize -> 520
]
```

# Problema 7 — Enfriamiento por Radiación (Condición de Robin / Mixta General)

## Enunciado físico

Una barra de longitud $L$ tiene el extremo izquierdo aislado ($T_x(0,t)=0$) y el extremo derecho se enfría por **convección/radiación** de Newton:
$$\kappa\,T_x(L,t) = -h\,T(L,t),$$
donde $h>0$ es el coeficiente de transferencia de calor. Reescribiendo:
$$T_x(L,t) + \frac{h}{\kappa}\,T(L,t) = 0 \quad (\text{condición de Robin}).$$

El sistema completo:
$$T_t = \kappa\,T_{xx}, \qquad T_x(0)=0,\; T_x(L)+\beta T(L)=0, \quad \beta = h/\kappa.$$

## Forma S-L y autovalores

La parte espacial es:
$$X'' + \lambda X = 0, \qquad X'(0)=0,\; X'(L)+\beta X(L)=0.$$

Con $X(x)=\cos(\mu x)$ ($\mu=\sqrt{\lambda}$), la condición en $x=L$ da la **ecuación trascendente**:
$$\mu\,\tan(\mu L) = \beta.$$

Los autovalores $\mu_n$ son las raíces positivas de esta ecuación (se obtienen numéricamente).

```
(* === Problema 7 — Condición de Robin: ecuación trascendente === *)

LR = 1;
beta = 2.5; (* h/kappa *)
kappaR = 0.1;

(* Visualizar la ecuación trascendente mu*tan(mu*L)=beta *)
Plot[
  {mu Tan[mu LR], beta},
  {mu, 0.001, 5 Pi},
  PlotStyle -> {
    Directive[colores[[1]], Thickness[0.003]],
    Directive[Red, Dashed, Thickness[0.003]]
  },
  PlotLegends -> {"\[Mu] tan(\[Mu]L)", "\[Beta]=" <> ToString[beta]},
  AxesLabel -> {"\[Mu]", ""},
  PlotRange -> {-10, 15},
  Exclusions -> {Pi/2, 3Pi/2, 5Pi/2, 7Pi/2, 9Pi/2},
  PlotLabel -> Style["Ecuación trascendente: \[Mu]tan(\[Mu]L)=\[Beta]", Bold],
  Frame -> True, GridLines -> Automatic, GridLinesStyle -> LightGray,
  ImageSize -> 520
]
```

```
(* === Cálculo numérico de los autovalores de Robin === *)

(* Encontrar raíces de mu*tan(mu*L) - beta = 0 *)
nRaices = 6;
muRobin = Table[
  mu /. FindRoot[
    mu Tan[mu LR] - beta,
    {mu, (2k-1) Pi/(2 LR) + 0.3} (* semilla cercana al k-ésimo cero *)
  ],
  {k, 1, nRaices}
];

lambdaRobin = muRobin^2;

Grid[
  Prepend[
    Table[{n, N[muRobin[[n]], 6], N[lambdaRobin[[n]], 6]}, {n, 1, nRaices}],
    {"n", "\[Mu]_n", "\[Lambda]_n = \[Mu]_n\[Squared]"}
  ],
  Frame -> All,
  Background -> {{LightBlue}, {LightYellow, None}}
]
```

```
(* === Autofunciones de Robin y solución de calor === *)

XRobin[n_][x_] := Cos[muRobin[[n]] x];

(* Normas *)
normaRobin[n_] := NIntegrate[XRobin[n][x]^2, {x, 0, LR}];

(* Condición inicial: f(x) = 1 (barra a temperatura uniforme 1) *)
fRobin[x_] := 1;

(* Coeficientes *)
CnRobin[n_] := NIntegrate[fRobin[x] XRobin[n][x], {x, 0, LR}] / normaRobin[n];
coefsRobin = Table[CnRobin[n], {n, 1, nRaices}];

(* Solución *)
TRobin[x_, t_] := Sum[
  coefsRobin[[n]] Exp[-kappaR lambdaRobin[[n]] t] XRobin[n][x],
  {n, 1, nRaices}
];

(* Gráfica de la evolución *)
tiemposRobin = {0, 1, 3, 7, 15, 40};
Plot[
  Evaluate[Table[TRobin[x, t], {t, tiemposRobin}]],
  {x, 0, LR},
  PlotStyle -> (Directive[#, Thickness[0.003]] & /@ colores),
  PlotLegends -> ("t=" <> ToString[#] & /@ tiemposRobin),
  AxesLabel -> {"x (m)", "T(x,t)"},
  PlotLabel -> Style["Enfriamiento por convección — Robin\nT'(0)=0, T'(L)+\[Beta]T(L)=0", Bold],
  Frame -> True, GridLines -> Automatic, GridLinesStyle -> LightGray,
  PlotRange -> {-0.05, 1.1}, ImageSize -> 520
]
```

```
(* === Gráfica Robin: comparar distintos coeficientes beta === *)
betas = {0.5, 2, 5, 20};

(* Para cada beta, hallar mu1 y graficar la autofunción fundamental *)
mu1Beta = Table[
  mu /. FindRoot[mu Tan[mu LR] - b, {mu, 0.5}],
  {b, betas}
];

Plot[
  Evaluate[Table[Cos[mu1Beta[[i]] x], {i, 1, 4}]],
  {x, 0, LR},
  PlotStyle -> (Directive[#, Thickness[0.003]] & /@ colores[[1;;4]]),
  PlotLegends -> ("\[Beta]=" <> ToString[#] & /@ betas),
  AxesLabel -> {"x", "X_1(x)"},
  PlotLabel -> Style["Primera autofunción de Robin para distintos \[Beta]", Bold],
  Epilog -> {
    Gray, PointSize[0.015],
    MapThread[Point[{LR, Cos[#1 LR]}] &, {mu1Beta}]
  },
  Frame -> True, GridLines -> Automatic, GridLinesStyle -> LightGray,
  ImageSize -> 520
]
```

# Problema 8 — Verificación: Ortogonalidad y Completitud

## Ortogonalidad

Para las autofunciones de un problema S-L hermítico:
$$\int_a^b X_m(x)\,X_n(x)\,\rho(x)\,dx = \|X_n\|^2\,\delta_{mn}.$$

## Convergencia de la serie

La **norma del error** de truncamiento decrece al agregar más términos:
$$E_N = \left\|f - \sum_{n=1}^{N}c_n X_n\right\|_\rho.$$

```
(* === Verificación de ortogonalidad — Dirichlet === *)

(* Matriz de productos internos <X_m, X_n> para senos en [0,1] *)
nOrto = 5;
matrizOrto = Table[
  NIntegrate[Sin[m Pi x] Sin[n Pi x], {x, 0, 1}],
  {m, 1, nOrto}, {n, 1, nOrto}
];

Print["Matriz de productos internos (m,n = 1..5):"];
MatrixForm[Round[matrizOrto, 10^-10]]
```

```
(* === Visualización de la ortogonalidad como mapa de calor === *)
nOrtoVis = 8;
matrizOrtoVis = Table[
  Abs[NIntegrate[Sin[m Pi x] Sin[n Pi x], {x, 0, 1}]],
  {m, 1, nOrtoVis}, {n, 1, nOrtoVis}
];

MatrixPlot[
  matrizOrtoVis,
  ColorFunction -> "SunsetColors",
  FrameTicks -> {
    {Range[nOrtoVis], None},
    {Range[nOrtoVis], None}
  },
  PlotLabel -> Style["| <X_m, X_n> | — Ortogonalidad (Dirichlet, senos)", Bold],
  ColorRules -> {0 -> White},
  ImageSize -> 400
]
```

```
(* === Verificación de ortogonalidad — Neumann (cosenos) === *)
nOrtoN = 5;
matrizOrtoN = Table[
  NIntegrate[
    Cos[m Pi x] Cos[n Pi x],
    {x, 0, 1}
  ],
  {m, 0, nOrtoN-1}, {n, 0, nOrtoN-1}
];
Print["Ortogonalidad cosenos (Neumann):"];
MatrixForm[Round[matrizOrtoN, 10^-10]]
```

```
(* === Convergencia de la serie de Fourier para una función discontinua === *)

(* Onda cuadrada (discontinua) — Dirichlet en [0,1] *)
fCuadrada[x_] := Piecewise[{{1, 0 < x < 1/2}, {-1, 1/2 <= x < 1}}];

(* Coeficientes *)
BnCuad[n_] := 2 NIntegrate[fCuadrada[x] Sin[n Pi x], {x, 0, 1}];
nMaxCuad = 40;
coefsCuad = Table[BnCuad[n], {n, 1, nMaxCuad}];

fSerie[x_, N_] := Sum[coefsCuad[[n]] Sin[n Pi x], {n, 1, N}];

(* Comparación de aproximaciones con N=1,3,7,19 términos *)
Plot[
  Evaluate[Prepend[
    Table[fSerie[x, N], {N, {3, 7, 19, 39}}],
    fCuadrada[x]
  ]],
  {x, 0.001, 0.999},
  PlotStyle -> Prepend[
    (Directive[#, Thickness[0.003]] & /@ colores[[1;;4]]),
    Directive[Black, Dashed, Thickness[0.004]]
  ],
  PlotLegends -> Prepend[
    Table["N=" <> ToString[N], {N, {3, 7, 19, 39}}],
    "Exacta"
  ],
  AxesLabel -> {"x", "f(x)"},
  PlotLabel -> Style["Fenómeno de Gibbs — Convergencia de la serie de Fourier (Dirichlet)", Bold],
  Frame -> True, GridLines -> Automatic, GridLinesStyle -> LightGray,
  PlotRange -> {-1.5, 1.5}, ImageSize -> 560
]
```

```
(* === Error de truncamiento en función del número de términos === *)
errores = Table[
  Sqrt[NIntegrate[(fCuadrada[x] - fSerie[x, N])^2, {x, 0.01, 0.99}]],
  {N, 1, 30}
];

ListLogPlot[
  errores,
  PlotStyle -> {Directive[colores[[1]], PointSize[0.02]]},
  Joined -> True,
  AxesLabel -> {"Número de términos N", "||Error||_2"},
  PlotLabel -> Style["Convergencia en norma L² de la serie de Fourier (Dirichlet)", Bold],
  GridLines -> Automatic, GridLinesStyle -> LightGray,
  Frame -> True, ImageSize -> 520
]
```

# Problema 9 — Oscilador Armónico Cuántico (S-L con Hermite)

## Enunciado físico

La ecuación de Schrödinger para el oscilador armónico cuántico 1D en coordenadas adimensionales es:
$$-\frac{d^2\psi}{d\xi^2} + \xi^2\psi = \varepsilon\,\psi, \qquad \xi\in(-\infty,+\infty),\quad \psi(\pm\infty)=0.$$

## Forma S-L

Reescribiendo con $p(\xi)=e^{-\xi^2}$, $\rho(\xi)=e^{-\xi^2}$, las autofunciones son los **polinomios de Hermite** normalizados:
$$\psi_n(\xi) = \frac{1}{\sqrt{2^n n! \sqrt{\pi}}}\,e^{-\xi^2/2}\,H_n(\xi), \qquad \varepsilon_n = 2n+1.$$

La condición de frontera es de tipo **natural** (decaimiento en el infinito), equivalente a condición de Dirichlet en $\pm\infty$.

```
(* === Problema 9 — Oscilador Armónico Cuántico === *)

(* Funciones de onda normalizadas del OAC *)
psiOAC[n_][xi_] := 
  (1/Sqrt[2^n n! Sqrt[Pi]]) Exp[-xi^2/2] HermiteH[n, xi];

(* Verificar normalización *)
normOAC = Table[
  NIntegrate[psiOAC[n][xi]^2, {xi, -Infinity, Infinity}],
  {n, 0, 4}
];
Print["Normas (deben ser 1): ", Round[normOAC, 10^-8]]

(* Verificar ortogonalidad <psi0, psi2> *)
orto02 = NIntegrate[psiOAC[0][xi] psiOAC[2][xi], {xi, -Infinity, Infinity}];
Print["<psi_0, psi_2> = ", orto02, " (debe ser 0)"]
```

```
(* === Gráfica 9a: Funciones de onda del OAC === *)
nOAC = 5;
energiasOAC = Table[2n + 1, {n, 0, nOAC - 1}];

Plot[
  Evaluate[Table[
    psiOAC[n][xi] + energiasOAC[[n+1]],
    {n, 0, nOAC-1}
  ]],
  {xi, -5, 5},
  PlotStyle -> (Directive[#, Thickness[0.003]] & /@ colores),
  PlotLegends -> Table[
    "n=" <> ToString[n] <> ", \[Epsilon]=" <> ToString[2n+1],
    {n, 0, nOAC-1}
  ],
  Epilog -> {
    Gray, Dashed,
    Table[Line[{{-5, e}, {5, e}}], {e, energiasOAC}],
    Thick, Black,
    (* potencial armónico *)
    Plot[xi^2, {xi, -3.5, 3.5}, PlotStyle -> {Thick, Black}][[1, 1]]
  },
  AxesLabel -> {"\[Xi]", "\[Psi]_n(\[Xi]) + \[Epsilon]_n"},
  PlotLabel -> Style["OAC — Funciones de onda sobre el pozo parabólico", Bold],
  Frame -> True, GridLines -> None,
  PlotRange -> {-1, 11}, ImageSize -> 520
]
```

```
(* === Gráfica 9b: Densidades de probabilidad del OAC === *)
Plot[
  Evaluate[Table[psiOAC[n][xi]^2, {n, 0, 4}]],
  {xi, -5, 5},
  Filling -> Axis,
  FillingStyle -> (Directive[Opacity[0.2], #] & /@ colores),
  PlotStyle -> (Directive[#, Thickness[0.003]] & /@ colores),
  PlotLegends -> Table["|\[Psi]" <> ToString[n] <> "|\[Squared]", {n, 0, 4}],
  AxesLabel -> {"\[Xi]", "|\[Psi]_n(\[Xi])|\[Squared]"},
  PlotLabel -> Style["OAC — Densidades de probabilidad", Bold],
  Frame -> True, GridLines -> Automatic, GridLinesStyle -> LightGray,
  ImageSize -> 520
]
```

# Resumen General: Tabla de Problemas S-L

| Problema | Ecuación | $p(x)$ | BC | Autovalores $\lambda_n$ | Autofunciones |
|----------|----------|---------|------|-------------------------|---------------|
| Cuerda/Caja cuántica | $y''+\lambda y=0$ | $1$ | **Dirichlet** | $(n\pi/L)^2$ | $\sin(n\pi x/L)$ |
| Barra adiabática / Tubo cerrado | $y''+\lambda y=0$ | $1$ | **Neumann** | $(n\pi/L)^2$, $n\geq 0$ | $\cos(n\pi x/L)$ |
| Barra mixta | $y''+\lambda y=0$ | $1$ | **Mixta** | $\left(\frac{(2n-1)\pi}{2L}\right)^2$ | $\sin\!\left(\frac{(2n-1)\pi x}{2L}\right)$ |
| Robin (convección) | $y''+\lambda y=0$ | $1$ | **Robin** | raíces de $\mu\tan(\mu L)=\beta$ | $\cos(\mu_n x)$ |
| OAC (Hermite) | Schrödinger arm. | $e^{-\xi^2}$ | **Natural** ($\to 0$) | $2n+1$ | $e^{-\xi^2/2}H_n(\xi)$ |

---
## Reflexión final

- Las **condiciones de Dirichlet** (función cero en los bordes) dan autofunciones en **senos** y aparecen naturalmente en cuerdas, tubos abiertos, cajas cuánticas.
- Las **condiciones de Neumann** (derivada cero) dan autofunciones en **cosenos** (incluido el modo $n=0$) y corresponden a extremos aislados o cerrados.
- Las **condiciones mixtas** producen frecuencias propias con los factores $(2n-1)/2$ en lugar de $n$.
- Las **condiciones de Robin** generan autovalores que se calculan numéricamente a partir de una ecuación trascendente, y modelan fenómenos de interfaz (convección, radiación).

En todos los casos la **ortogonalidad** de las autofunciones garantiza que podemos expandir cualquier condición inicial en una serie, y la **completitud** asegura que esa serie converge al valor exacto (en norma $L^2$).

```
(* === Panel resumen: autofunciones de los 4 tipos de BC === *)
L0 = 1;

(* Dirichlet *)
pD = Plot[Evaluate[Table[Sin[n Pi x/L0], {n,1,3}]], {x,0,L0},
  PlotStyle->(Directive[#,Thickness[0.003]]&/@colores[[1;;3]]),
  PlotLabel->Style["Dirichlet\ny(0)=y(L)=0",Bold,10],
  AxesLabel->{"x","X_n"},
  PlotLegends->Table["n="<>ToString[n],{n,1,3}],
  Frame->True, GridLines->Automatic, GridLinesStyle->LightGray
];

(* Neumann *)
pN = Plot[Evaluate[Table[Cos[n Pi x/L0], {n,0,2}]], {x,0,L0},
  PlotStyle->(Directive[#,Thickness[0.003]]&/@colores[[1;;3]]),
  PlotLabel->Style["Neumann\ny'(0)=y'(L)=0",Bold,10],
  AxesLabel->{"x","X_n"},
  PlotLegends->Table["n="<>ToString[n],{n,0,2}],
  Frame->True, GridLines->Automatic, GridLinesStyle->LightGray
];

(* Mixta *)
pM = Plot[Evaluate[Table[Sin[(2n-1) Pi x/(2 L0)], {n,1,3}]], {x,0,L0},
  PlotStyle->(Directive[#,Thickness[0.003]]&/@colores[[1;;3]]),
  PlotLabel->Style["Mixta\ny(0)=0, y'(L)=0",Bold,10],
  AxesLabel->{"x","X_n"},
  PlotLegends->Table["n="<>ToString[n],{n,1,3}],
  Frame->True, GridLines->Automatic, GridLinesStyle->LightGray
];

(* Robin con beta=2.5 *)
pR = Plot[Evaluate[Table[Cos[muRobin[[n]] x], {n,1,3}]], {x,0,L0},
  PlotStyle->(Directive[#,Thickness[0.003]]&/@colores[[1;;3]]),
  PlotLabel->Style["Robin (\[Beta]=2.5)\ny'(0)=0, y'(L)+\[Beta]y(L)=0",Bold,10],
  AxesLabel->{"x","X_n"},
  PlotLegends->Table["n="<>ToString[n],{n,1,3}],
  Frame->True, GridLines->Automatic, GridLinesStyle->LightGray
];

GraphicsGrid[
  {{pD, pN}, {pM, pR}},
  PlotLabel -> Style["Autofunciones según tipo de condición de frontera", Bold, 14],
  ImageSize -> 900
]
```
