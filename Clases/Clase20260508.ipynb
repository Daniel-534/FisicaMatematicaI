# Ecuación de Bessel

La **ecuación de Bessel** es una de las ecuaciones diferenciales ordinarias (EDO) más importantes de la física matemática, ya que surge al resolver las ecuaciones de Laplace, Helmholtz o la de ondas en sistemas con **simetría cilíndrica**. Su solución mediante el **método de Frobenius** permite obtener funciones que describen fenómenos oscilatorios no periódicos, como la vibración de una membrana circular o la difracción de la luz.

A continuación se explica la ecuación y el procedimiento analítico para resolverla paso a paso:

### 1. Definición de la Ecuación de Bessel
La forma estándar de la ecuación de Bessel de orden $\nu$ es:
$$\mathbf{x^2 \frac{d^2y}{dx^2} + x \frac{dy}{dx} + (x^2 - \nu^2)y = 0}$$

Al analizar el punto $x=0$, se observa que es una **singularidad regular**, lo que garantiza que podemos encontrar al menos una solución utilizando una serie de potencias generalizada (Serie de Frobenius).

### 2. Propuesta de Solución (Método de Frobenius)
Se asume una solución de la forma:
$$y(x) = \sum_{\lambda=0}^{\infty} a_\lambda x^{\lambda+s}, \quad a_0 \neq 0$$

Derivando término a término y sustituyendo en la EDO original, se obtiene una expresión que agrupa las potencias de $x$:
$$\sum_{\lambda=0}^{\infty} a_\lambda [(\lambda+s)^2 - \nu^2] x^{\lambda+s} + \sum_{\lambda=0}^{\infty} a_\lambda x^{\lambda+s+2} = 0$$

### 3. La Ecuación Indicial y sus Raíces
Para que la ecuación se cumpla, el coeficiente de la potencia más baja de $x$ (cuando $\lambda=0$) debe ser cero. Esto nos da la **ecuación indicial**:
$$a_0 (s^2 - \nu^2) = 0 \implies \mathbf{s = \pm\nu}$$

El método de Frobenius asegura que la raíz mayor ($s = \nu$) siempre proporcionará una solución válida en serie.

### 4. Relación de Recurrencia
Al igualar a cero los coeficientes de las potencias superiores, se determina que los coeficientes de índice impar deben ser cero ($a_1 = a_3 = \dots = 0$). Para los índices pares ($\lambda = 2k$), se obtiene la **relación de recurrencia**:
$$a_{k+2} = \frac{-a_k}{(k+s+2)^2 - \nu^2}$$

### 5. Función de Bessel de Primer Género ($J_\nu$)
Sustituyendo $s = \nu$ y aplicando la recurrencia sucesivamente, los coeficientes se expresan en términos de la **función Gamma** ($\Gamma$), que generaliza el concepto de factorial. Esto define la **función de Bessel de primer género**:
$$\mathbf{J_\nu(x) = \sum_{k=0}^{\infty} \frac{(-1)^k}{k! \Gamma(\nu + k + 1)} \left( \frac{x}{2} \right)^{\nu+2k}}$$

### 6. La Segunda Solución
La naturaleza de la segunda solución independiente depende del valor de $\nu$:

*   **Si $\nu$ no es un entero:** La segunda solución es simplemente $J_{-\nu}(x)$, obtenida usando la raíz $s = -\nu$.
*   **Si $\nu$ es un entero ($n$):** Ocurre que $J_{-n}(x) = (-1)^n J_n(x)$, por lo que dejan de ser linealmente independientes. En este caso, se requiere definir la **función de Neumann** ($N_n$ o $Y_n$), la cual incluye un término logarítmico y diverge en el origen ($x \to 0$).

La solución general completa se escribe como una combinación lineal de ambas:
$$\mathbf{y(x) = A J_\nu(x) + B Y_\nu(x)}$$

La **ecuación indicial** es una expresión algebraica fundamental en el **método de Frobenius**, utilizada para hallar soluciones en serie de potencias de ecuaciones diferenciales ordinarias (EDO) de segundo orden en las cercanías de una **singularidad regular**. Esta ecuación determina el exponente $s$ (o $\sigma$) del término de menor potencia en la serie propuesta para la solución.

### 1. Definición y obtención
Dada una ecuación diferencial de la forma:
$$x^2 \frac{d^2y}{dx^2} + x P(x) \frac{dy}{dx} + Q(x) y = 0$$
donde $P(x)$ y $Q(x)$ pueden expresarse como series de potencias ($P(x) = \sum p_m x^m$ y $Q(x) = \sum q_m x^m$), se propone una solución de la forma $y(x) = \sum_{\lambda=0}^{\infty} a_\lambda x^{\lambda+s}$, con $a_0 \neq 0$.

Al sustituir esta serie en la EDO, el coeficiente de la potencia más baja de $x$ ($x^s$) debe anularse individualmente, lo que genera la **ecuación indicial**:
$$\mathbf{s(s - 1) + s p_0 + q_0 = 0}$$
donde $p_0$ y $q_0$ son los primeros términos de las expansiones de $P(x)$ y $Q(x)$ respectivamente.

### 2. Importancia de las raíces
La naturaleza de las dos raíces de la ecuación indicial ($s_1$ y $s_2$) define el tipo de soluciones que se pueden obtener:

*   **Raíces que no difieren por un entero:** El método de Frobenius proporciona dos soluciones linealmente independientes en forma de serie.
*   **Raíces iguales ($s_1 = s_2$):** Solo se obtiene una solución directamente por series; la segunda solución independiente suele requerir un término logarítmico y debe hallarse por otros métodos (como el método de la derivada o reducción de orden).
*   **Raíces que difieren por un entero:** La raíz mayor siempre proporciona una solución válida. La raíz menor puede o no proporcionar una segunda solución independiente; si falla, la segunda solución también incluirá un término logarítmico.

### 3. Ejemplos en funciones especiales
Las fuentes detallan las ecuaciones indiciales para las EDO más comunes en física:

*   **Ecuación de Bessel:** $x^2 y'' + x y' + (x^2 - \nu^2)y = 0$. Su ecuación indicial es $s^2 - \nu^2 = 0$, con raíces **$s = \pm \nu$**.
*   **Ecuación de Legendre:** Al expandir alrededor del origen (punto ordinario), la ecuación indicial es $s(s - 1) = 0$, con raíces **$s = 0$ y $s = 1$**.
*   **Ecuación de Hermite:** Presenta las mismas raíces que la de Legendre, **$s = 0$ y $s = 1$**.
*   **Ecuación de Laguerre:** $x y'' + (1 - x) y' + n y = 0$. Posee una raíz doble **$s = 0$**, lo que garantiza una sola solución en serie simple (un polinomio si $n$ es entero).
*   **Efecto Stark (Coordenadas parabólicas):** La ecuación indicial es $s^2 - m^2/4 = 0$, con raíces **$s = \pm m/2$**.

Para llegar a la expresión de **$a_{2p}$** en el libro de **Alonso Sepúlveda**, se sigue el procedimiento del **método de Frobenius** aplicado a la **ecuación de Bessel**. A continuación se detalla el desarrollo analítico paso a paso:

### 1. Definición y Propuesta de Serie
Se parte de la ecuación de Bessel en su forma adimensional:
$$x^2\ddot{y}(x) + x\dot{y}(x) + (x^2 - \nu^2)y(x) = 0$$
Se propone una solución en serie de potencias (Serie de Frobenius):
$$y(x) = \sum_{\lambda=0}^{\infty} a_\lambda x^{\lambda+s}, \quad a_0 \neq 0$$

### 2. Sustitución y Relación de Recurrencia
Al sustituir la serie y sus derivadas en la ecuación diferencial, se agrupan las potencias de $x$ obteniendo:
$$\sum_{\lambda=0}^{\infty} a_\lambda [(\lambda+s)^2 - \nu^2]x^{\lambda+s} + \sum_{\lambda=0}^{\infty} a_\lambda x^{\lambda+s+2} = 0$$
Para que la ecuación sea válida, el coeficiente de cada potencia de $x$ debe ser cero:
*   **Ecuación indicial ($\lambda=0$):** $a_0[s^2 - \nu^2] = 0$. Como $a_0 \neq 0$, se tiene que **$s = \pm\nu$**.
*   **Coeficiente de $\lambda=1$:** $a_1[(s+1)^2 - \nu^2] = 0$. Para $s = \nu$, esto implica que **$a_1 = 0$**.
*   **Recurrencia general ($\lambda \geq 0$):** Al igualar potencias superiores, se obtiene la relación entre términos alternos:
    $$a_{\lambda+2} = \frac{-a_\lambda}{(\lambda+s+2)^2 - \nu^2}$$

### 3. Evaluación de los Coeficientes Pares
Tomando la raíz **$s = \nu$**, la relación de recurrencia se simplifica a:
$$a_{\lambda+2} = \frac{-a_\lambda}{(\lambda+2)(\lambda+2\nu+2)}$$
Dado que $a_1 = 0$, todos los coeficientes de índice impar se anulan. Para los índices pares ($\lambda = 0, 2, 4, \dots$), calculamos los primeros términos para identificar el patrón:
*   **Para $\lambda=0$:** $a_2 = \frac{-a_0}{2(2\nu+2)} = \frac{-a_0}{2^2(1)(\nu+1)}$
*   **Para $\lambda=2$:** $a_4 = \frac{-a_2}{4(2\nu+4)} = \frac{(-1)^2 a_0}{2^4(2 \cdot 1)(\nu+2)(\nu+1)}$
*   **Para $\lambda=4$:** $a_6 = \frac{-a_4}{6(2\nu+6)} = \frac{(-1)^3 a_0}{2^6(3 \cdot 2 \cdot 1)(\nu+3)(\nu+2)(\nu+1)}$

### 4. Generalización con la Función Gamma
Para escribir el producto $(\nu+p)(\nu+p-1)\dots(\nu+1)$ de forma compacta, Alonso utiliza la **función Gamma** ($\Gamma$), la cual satisface $\Gamma(x+1) = x\Gamma(x)$. Esto permite expresar el denominador como:
$$(\nu+p)(\nu+p-1)\dots(\nu+1) = \frac{\Gamma(\nu+p+1)}{\Gamma(\nu+1)}$$
Al sustituir esta propiedad y el factorial $p! = 3 \cdot 2 \cdot 1 \dots p$ en la secuencia obtenida, se llega a la expresión general para el coeficiente de orden **$2p$**:

$$\mathbf{a_{2p} = \frac{(-1)^p a_0 \Gamma(\nu + 1)}{2^{2p} p! \Gamma(\nu + p + 1)}}$$

Esta expresión es la que finalmente permite definir la **función de Bessel de primer género** $J_\nu(x)$ al elegir una normalización específica para $a_0$.

A partir de la expresión para los coeficientes **$a_{2p}$** obtenida mediante el método de Frobenius, se puede reconstruir la solución **$y(x)$** de la ecuación de Bessel. El procedimiento analítico para llegar a la definición de la función de Bessel de primer género, **$J_\nu(x)$**, es el siguiente:

### 1. Reconstrucción de la serie para $s = \nu$
Partiendo de la propuesta de serie $y(x) = \sum_{\lambda=0}^{\infty} a_\lambda x^{\lambda+s}$ y utilizando la raíz de la ecuación indicial **$s = \nu$**, la solución se expresa únicamente en términos de potencias pares debido a que los coeficientes de índice impar se anulan ($a_1 = a_3 = \dots = 0$). Sustituyendo el coeficiente general **$a_{2p}$** derivado anteriormente:

$$a_{2p} = \frac{(-1)^p a_0 \Gamma(\nu + 1)}{2^{2p} p! \Gamma(\nu + p + 1)}$$

Sustituimos esto en la sumatoria original:
$$y(x) = \sum_{p=0}^{\infty} a_{2p} x^{\nu+2p} = a_0 \Gamma(\nu + 1) \sum_{p=0}^{\infty} \frac{(-1)^p x^{\nu+2p}}{2^{2p} p! \Gamma(\nu + p + 1)}$$

### 2. Definición de $J_\nu(x)$
Para estandarizar la función, Alonso Sepúlveda realiza un ajuste algebraico multiplicando y dividiendo por **$2^\nu$**, lo que permite agrupar los términos de la siguiente manera:

$$y(x) = a_0 \Gamma(\nu + 1) 2^\nu \sum_{p=0}^{\infty} \frac{(-1)^p}{p! \Gamma(\nu + p + 1)} \left( \frac{x}{2} \right)^{\nu+2k}$$

La sumatoria resultante define la **función de Bessel de primer género y orden $\nu$**, denotada como **$J_\nu(x)$**:
$$\mathbf{J_\nu(x) = \sum_{k=0}^{\infty} \frac{(-1)^k}{k! \Gamma(\nu + k + 1)} \left( \frac{x}{2} \right)^{\nu+2k}}$$

Por lo tanto, la primera solución particular hallada por Frobenius es proporcional a esta función: **$y(x) \propto J_\nu(x)$**.

### 3. La solución general para $y(x)$
Dependiendo de la naturaleza del parámetro **$\nu$**, la expresión final para la solución general del sistema varía:

*   **Si $\nu$ no es un número entero:** Las raíces de la ecuación indicial ($s = \nu$ y $s = -\nu$) proporcionan dos soluciones linealmente independientes, $J_\nu(x)$ y $J_{-\nu}(x)$. La expresión completa para $y(x)$ es:
    $$\mathbf{y(x) = A J_\nu(x) + B J_{-\nu}(x)}$$
*   **Si $\nu$ es un entero ($n$):** Las funciones $J_n$ y $J_{-n}$ dejan de ser independientes debido a la relación **$J_{-n}(x) = (-1)^n J_n(x)$**. En este caso, se introduce la **función de Neumann** ($Y_n$ o $N_n$) como la segunda solución independiente. La solución general se escribe entonces como:
    $$\mathbf{y(x) = c_1 J_n(x) + c_2 Y_n(x)}$$

Esta forma final garantiza que se cubran todas las soluciones posibles de la ecuación de Bessel, siendo **$J_\nu(x)$** la componente regular (finita) en el origen.

Las funciones de Bessel de primer género de orden cero $J_0(x)$ y de orden uno $J_1(x)$ están estrechamente vinculadas a través de diversas relaciones de derivación, recurrencia e integración. A continuación se presentan las conexiones principales fundamentadas en las fuentes:

### 1. Relación de Derivación Primaria
La relación más directa entre ambas funciones es que la derivada de la función de orden cero es el negativo de la función de orden uno:
$$\mathbf{\frac{d}{dx} J_0(x) = -J_1(x)}$$

### 2. Relación de Derivación del Producto $x J_1(x)$
Inversamente, la derivada del producto de la variable $x$ por la función de orden uno devuelve la función de orden cero multiplicada por $x$:
$$\mathbf{\frac{d}{dx} [x J_1(x)] = x J_0(x)}$$
Esta identidad es fundamental para evaluar integrales que involucran a $J_0$ en problemas de física, como el cálculo del potencial en cilindros.

### 3. Aplicación de las Fórmulas de Recurrencia
Las funciones de Bessel satisfacen fórmulas de recurrencia generales. Al aplicar estas fórmulas para el caso $n=1$, se obtienen relaciones adicionales entre $J_0$ y $J_1$ (involucrando a veces a $J_2$):

*   **Suma de órdenes:** $J_0(x) + J_2(x) = \frac{2}{x} J_1(x)$
*   **Diferencia de órdenes (Derivada de $J_1$):** $J_0(x) - J_2(x) = 2 J'_1(x)$
*   **Identidad de paso:** $J_1(x) = \frac{x}{2}(J_0(x) + J_2(x))$ (derivada de la primera de esta lista).

### 4. Relación Integral
Derivado de la relación de derivación del punto 2, se establece que la integral de $x J_0$ está definida por $x J_1$:
$$\mathbf{\int x J_0(x) dx = x J_1(x)}$$
Asimismo, en el intervalo unitario, si $\alpha$ es una raíz de $J_0$, se cumple que:
$$\int_0^1 x J_0(\alpha x) dx = \frac{J_1(\alpha)}{\alpha}$$

### 5. Alternancia de Ceros
Existe una relación cualitativa importante respecto a sus raíces (puntos donde la función se anula): entre cualquier par de ceros consecutivos de $J_0(x)$, existe exactamente un cero de $J_1(x)$. Esta propiedad de **entrelazamiento de ceros** es característica de las funciones de Bessel de órdenes consecutivos.

### 6. Representación Integral de $J_1$ a partir de $J_0$
Mediante la diferenciación de la representación integral de $J_0(x) = \frac{1}{\pi} \int_0^\pi \cos(x \sin \theta) d\theta$, se puede obtener $J_1(x)$ como:
$$J_1(x) = \frac{1}{\pi} \int_0^\pi \sin(x \sin \theta) \sin \theta d\theta$$

Basado en las fuentes, la relación entre las funciones de Bessel y las derivadas respecto al orden (índice) surge principalmente al buscar una segunda solución linealmente independiente cuando el orden $\nu$ es un número entero.

### 1. Forma general de $J_\nu(x)$
La **función de Bessel de primer género** y orden $\nu$ se define analíticamente mediante su expansión en serie de potencias (derivada del método de Frobenius):
$$\mathbf{J_\nu(x) = \sum_{p=0}^{\infty} \frac{(-1)^p}{p! \Gamma(\nu + p + 1)} \left( \frac{x}{2} \right)^{\nu+2p}}$$
Esta función es la solución regular en el origen para cualquier valor de $\nu$.

### 2. Forma de $N_\nu$ (o $Y_\nu$) en términos de derivadas del orden
Cuando $\nu$ tiende a un número entero $n$, las funciones $J_n$ y $J_{-n}$ se vuelven linealmente dependientes ($J_{-n} = (-1)^n J_n$). Para encontrar la segunda solución independiente, la **función de Neumann** (o de Bessel de segundo género), se utiliza el límite de la definición de Weber:
$$N_\nu(x) = \frac{J_\nu(x) \cos(\nu\pi) - J_{-\nu}(x)}{\sin(\nu\pi)}$$
Al aplicar la regla de **L'Hôpital** para el caso de orden entero ($n$), la función de Neumann se expresa explícitamente en términos de las **derivadas de las funciones de Bessel respecto a su orden** evaluadas en $n = \nu$:

$$\mathbf{N_n(x) = \frac{1}{\pi} \left[ \frac{\partial J_\nu(x)}{\partial \nu} - (-1)^n \frac{\partial J_{-\nu}(x)}{\partial \nu} \right]_{\nu=n}}$$

Esta expresión demuestra que la función de Neumann contiene la información de cómo cambia la solución respecto al parámetro del orden.

### 3. Estructura analítica de la derivada
Las fuentes señalan que la derivada de la función de Bessel respecto a su orden, $\frac{\partial J_\mu(z)}{\partial \mu} \big|_{\mu=\nu}$, tiene una estructura característica que incluye un término logarítmico:
*   $\frac{\partial J_\mu(z)}{\partial \mu} \big|_{\mu=\nu} = J_\nu(z) \ln\left(\frac{z}{2}\right) + g(\nu, z)$.
*   Donde $g(\nu, z)$ es una serie de potencias en $z$. Esta es la razón por la cual la función de Neumann $N_n(x)$ presenta una **singularidad logarítmica** en el origen ($x \to 0$), comportándose como $\frac{2}{\pi} J_n(x) \ln(x/2)$ más otros términos en serie.

Los **ceros de las funciones de Bessel de primer género**, designados comúnmente como $\alpha_{\nu n}$ o $\chi_{\nu n}$, son los valores del argumento para los cuales la función se anula, es decir, $J_\nu(\alpha_{\nu n}) = 0$,.

A continuación se detallan sus propiedades y características principales según las fuentes:

### **1. Naturaleza y Cantidad de las Raíces**
*   **Número infinito:** Cada función de Bessel $J_\nu(x)$ posee un número infinito de ceros reales,.
*   **Raíces reales:** Para valores de $\nu > -1$, todas las raíces de la ecuación $J_\nu(x) = 0$ son **reales**; no existen raíces complejas ni puramente imaginarias en este rango,.
*   **Punto de origen:** El valor $x = 0$ se considera una raíz trivial, ya que para $\nu > 0$, la función $J_\nu(0) = 0$ por definición de su serie de potencias,.
*   **Raíces comunes:** Si $\nu$ es un número real, las funciones de órdenes consecutivos (como $J_\nu$ y $J_{\nu+1}$) no comparten raíces comunes, a excepción del origen.

### **2. Propiedades Cualitativas**
*   **Entrelazamiento (Interlacing):** Los ceros de funciones de Bessel de órdenes consecutivos están entrelazados; esto significa que entre cualquier par de ceros consecutivos de $J_\nu(x)$ existe exactamente un cero de $J_{\nu+1}(x)$ y viceversa,.
*   **Comportamiento Asintótico:** Para valores grandes de $x$, la separación entre dos ceros consecutivos de una función de orden $\nu$ tiende a $\pi$ ($\chi_{\nu, n+1} - \chi_{\nu, n} \approx \pi$). En este límite, las funciones de Bessel se comportan como funciones cosenoidales desplazadas en fase.

### **4. Importancia en la Física Matemática**
*   **Condiciones de Frontera:** En problemas de Dirichlet (como una membrana circular fija en su borde), los ceros determinan los valores permitidos del parámetro de separación $k$ (autovalores), exigiendo que $ka = \alpha_{\nu n}$, donde $a$ es el radio del sistema,.
*   **Ortogonalidad:** Los ceros son fundamentales para construir bases ortogonales en intervalos finitos. Las funciones $\{J_\nu(\alpha_{\nu n} \rho / a)\}$ forman un conjunto completo y ortogonal en el intervalo $[0, a]$ con factor de peso $\rho$,.
*   **Modos Normales:** En acústica y mecánica, los ceros identifican las frecuencias naturales de oscilación y las posiciones de las líneas nodales en sistemas cilíndricos o circulares,.
