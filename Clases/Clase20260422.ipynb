* Continuidad ($G_1$ y $G_2$)
La función de Green $G(x, x')$ se define usualmente por tramos para cubrir las dos regiones del dominio separadas por el punto de aplicación de la fuente puntual $x'$. Se denota como $G_1(x, x')$ para el intervalo $x < x'$ y como $G_2(x, x')$ para el intervalo $x > x'$. Una propiedad fundamental es que la función debe ser continua en todo el dominio, incluyendo el punto $x = x'$. Matemáticamente, esto se expresa como el requisito de que los valores de ambos tramos coincidan en el punto de la fuente:
$$G_1(x', x') = G_2(x', x')$$.

* Discontinuidad de la derivada de Green
A diferencia de la función misma, la primera derivada de la función de Green con respecto a la variable de observación $x$ presenta un salto finito o discontinuidad en el punto $x = x'$. Este comportamiento es necesario para que, al derivar nuevamente, el operador diferencial $\mathscr{L}$ genere la singularidad de la delta de Dirac $\delta(x - x')$. Para una ecuación de segundo orden, el salto en la derivada se define por la relación:
$$\frac{dG_2(x, x')}{dx} \bigg|_{x=x'} - \frac{dG_1(x, x')}{dx} \bigg|_{x=x'} = \frac{1}{p(x')}$$.
Donde $p(x')$ es el coeficiente que acompaña a la segunda derivada cuando el operador $\mathscr{L}$ está escrito en su forma autoadjunta de Sturm-Liouville.

* Satisfacción de la ecuación diferencial
La función de Green es la solución fundamental de la ecuación diferencial inhomogénea cuando la fuente es una distribución puntual de magnitud unitaria:
$$\mathscr{L}(\vec{r}) G(\vec{r}, \vec{r}') = \delta(\vec{r} - \vec{r}')$$.

* Cumplimiento de condiciones de frontera
$G(x, x')$ debe satisfacer las mismas condiciones de frontera homogéneas (como Dirichlet o Neumann) que se imponen a la solución general del problema original. Por ejemplo, si la solución debe anularse en los extremos del intervalo, la función de Green también debe hacerlo.

* Simetría o reciprocidad
Cuando el operador diferencial $\mathscr{L}$ es hermítico bajo las condiciones de frontera dadas, la función de Green posee la propiedad de simetría:
$$G(\vec{r}, \vec{r}') = G^*(\vec{r}', \vec{r})$$.
Físicamente, esto implica que la respuesta en $\vec{r}$ debida a una fuente en $\vec{r}'$ es idéntica (o el complejo conjugado) a la respuesta en $\vec{r}'$ si la fuente estuviera en $\vec{r}$.

# Ejercicio

Para resolver la ecuación diferencial ordinaria (EDO) inhomogénea planteada dentro del marco de **Sturm-Liouville** y las **Funciones de Green**, seguiremos un procedimiento analítico detallado. El problema consiste en hallar $f(x)$ tal que:
$$\frac{d^2}{dx^2} f(x) = -x^2 + 2e^{-x}$$
Sujeto a las condiciones de frontera: **$f(0) = 0$** (Dirichlet) y **$f'(1) = 0$** (Neumann).

---

### Paso 1: Identificación del Operador de Sturm-Liouville
El operador diferencial es $\mathscr{L} = \frac{d^2}{dx^2}$. Según las fuentes, este es un operador autoadjunto con función de peso $p(x) = 1$ y $q(x) = 0$. La ecuación se puede escribir en la forma general de Sturm-Liouville:
$$\frac{d}{dx}\left[ p(x) \frac{df}{dx} \right] + q(x)f = g(x) \implies \frac{d}{dx}\left[ 1 \cdot f' \right] = -x^2 + 2e^{-x}$$
Donde la fuente inhomogénea es **$g(x) = -x^2 + 2e^{-x}$**.

---

### Paso 2: Solución de la Ecuación Homogénea Asociada
Para construir la función de Green o usar el método de superposición, necesitamos las soluciones de la ecuación homogénea $\mathscr{L}f_c = 0$:
$$\frac{d^2 f_c}{dx^2} = 0 \implies f_c(x) = C_1 x + C_2$$
Buscamos dos soluciones particulares, $y_1$ y $y_2$, que satisfagan individualmente cada condición de frontera:
1.  **Para $x=0$:** $y_1(0) = 0 \implies C_2 = 0$. Elegimos **$y_1(x) = x$**.
2.  **Para $x=1$:** $y_2'(1) = 0 \implies C_1 = 0$. Elegimos **$y_2(x) = 1$**.

---

### Paso 3: Construcción de la Función de Green $G(x, z)$
La función de Green representa la respuesta del sistema a una fuente puntual $\delta(x-z)$. Se define por tramos para satisfacer las condiciones en los extremos y la continuidad en el punto de impulso:
$$G(x, z) = \begin{cases} A y_1(x) y_2(z) & 0 \le x < z \\ A y_2(x) y_1(z) & z < x \le 1 \end{cases} = \begin{cases} A \cdot x \cdot 1 & x < z \\ A \cdot 1 \cdot z & x > z \end{cases}$$
Para hallar la constante $A$, aplicamos la condición de **discontinuidad de la derivada**:
$$\frac{dG}{dx}\bigg|_{x=z^+} - \frac{dG}{dx}\bigg|_{x=z^-} = \frac{1}{p(z)} \implies \frac{d}{dx}(Az)\bigg|_{z^+} - \frac{d}{dx}(Ax)\bigg|_{z^-} = 1$$
$$0 - A = 1 \implies \mathbf{A = -1}$$
Así, la función de Green es:
$$\mathbf{G(x, z) = \begin{cases} -x & x < z \\ -z & x > z \end{cases}}$$

---

### Paso 4: Cálculo de la Solución Final mediante Integración
La solución general se obtiene integrando el producto de la función de Green por el término fuente $g(z)$ en todo el dominio:
$$f(x) = \int_0^1 G(x, z) g(z) dz = \int_0^x G_{x>z}(x, z) g(z) dz + \int_x^1 G_{x<z}(x, z) g(z) dz$$
Sustituyendo los valores:
$$f(x) = \int_0^x (-z) (-z^2 + 2e^{-z}) dz + \int_x^1 (-x) (-z^2 + 2e^{-z}) dz$$

**Evaluación de la primera integral:**
$$\int_0^x (z^3 - 2ze^{-z}) dz = \left[ \frac{z^4}{4} + 2(z+1)e^{-z} \right]_0^x = \mathbf{\frac{x^4}{4} + 2(x+1)e^{-x} - 2}$$

**Evaluación de la segunda integral:**
$$-x \int_x^1 (-z^2 + 2e^{-z}) dz = -x \left[ -\frac{z^3}{3} - 2e^{-z} \right]_x^1 = -x \left[ \left(-\frac{1}{3} - \frac{2}{e}\right) - \left(-\frac{x^3}{3} - 2e^{-x}\right) \right]$$
$$= \mathbf{\frac{x}{3} + \frac{2x}{e} - \frac{x^4}{3} - 2xe^{-x}}$$

---

### Paso 5: Resultado Final (Consolidación)
Sumamos ambos resultados para obtener la función $f(x)$ completa:
$$f(x) = \left( \frac{x^4}{4} - \frac{x^4}{3} \right) + (2xe^{-x} - 2xe^{-x}) + 2e^{-x} + \frac{x}{3} + \frac{2x}{e} - 2$$
$$\mathbf{f(x) = -\frac{x^4}{12} + 2e^{-x} + \left( \frac{1}{3} + \frac{2}{e} \right)x - 2}$$

**Verificación:**
*   **$f(0)$:** $0 + 2e^0 + 0 - 2 = 2 - 2 = 0$. (Cumple Dirichlet).
*   **$f'(x) = -\frac{x^3}{3} - 2e^{-x} + \frac{1}{3} + \frac{2}{e}$.**
*   **$f'(1)$:** $-\frac{1}{3} - 2e^{-1} + \frac{1}{3} + \frac{2}{e} = 0$. (Cumple Neumann).
*   **$f''(x) = -x^2 + 2e^{-x}$.** (Satisface la EDO original).

Para el operador $\mathscr{L} = d^2/dx^2$, la solución a la ecuación homogénea $\mathscr{L} G(x, x') = 0$ (válida para el dominio donde $x \neq x'$) es efectivamente una línea recta de la forma $G(x, x') = a(x') + b(x')x$.

En la explicación anterior se omitió este paso intermedio para emplear un método de construcción más directo presente en las fuentes, que simplifica el cálculo de las constantes. A continuación se explica por qué no se detalló esa forma y cómo se relaciona con el procedimiento seguido:

1. El enfoque de las cuatro constantes
Si se utiliza la forma general $G(x, x') = a(x') + b(x')x$ por tramos, se debe plantear:
*   $G_1(x, x') = a_1(x') + b_1(x')x$ para la región $x < x'$.
*   $G_2(x, x') = a_2(x') + b_2(x')x$ para la región $x > x'$.
Para resolver este sistema, se requiere determinar cuatro funciones desconocidas ($a_1, b_1, a_2, b_2$) imponiendo las condiciones de frontera en los extremos, la continuidad en el punto de la fuente y el salto de la derivada.

2. El enfoque de las soluciones particulares ($y_1$ e $y_2$)
El procedimiento que se mostró utiliza una técnica más eficiente que consiste en identificar soluciones de la ecuación homogénea que satisfagan individualmente una de las condiciones de frontera:
*   Para la frontera en $x = 0$, se busca una solución de $y'' = 0$ tal que $y(0) = 0$. La función resultante es $y_1(x) = x$. Note que esta es una recta $a + bx$ donde $a = 0$ y $b = 1$.
*   Para la frontera en $x = 1$, se busca una solución de $y'' = 0$ tal que $y'(1) = 0$. La función resultante es $y_2(x) = 1$. Note que esta es una recta $a + bx$ donde $a = 1$ y $b = 0$.

Al usar estas funciones, la función de Green se construye como un producto simétrico $A y_1(x_<) y_2(x_>)$. Esto garantiza de forma automática el cumplimiento de las condiciones de frontera y la continuidad en $x = x'$, reduciendo el problema a encontrar una sola constante $A$ mediante la condición de salto en la derivada.

3. Consistencia entre ambos métodos
Si se hubiera aplicado el método de las constantes $a(x')$ y $b(x')$ a las condiciones del ejercicio ($f(0) = 0$ y $f'(1) = 0$), el resultado habría sido idéntico:
*   $G_1(0, x') = a_1(x') + b_1(x')(0) = 0 \implies a_1(x') = 0$, quedando $G_1 = b_1(x')x$.
*   $G_2'(1, x') = b_2(x') = 0$, quedando $G_2 = a_2(x')$.
Esto conduce exactamente a las funciones $y_1(x) = x$ e $y_2(x) = 1$ multiplicadas por factores dependientes de $x'$.

En conclusión, no se mostró la forma general $a(x') + b(x')x$ para evitar la resolución de un sistema de ecuaciones más extenso, optando por la construcción basada en soluciones que ya incorporan la geometría del problema y las condiciones de contorno de Sturm-Liouville.

Para demostrar que la funcion de Green satisface la ecuacion $\mathscr{L}(\vec{r}) G(\vec{r}, \vec{r}') = \delta(\vec{r} - \vec{r}')$ en el problema planteado, trabajaremos con el caso unidimensional donde el vector de posicion $\vec{r}$ se reduce a la variable $x$ y el operador diferencial es $\mathscr{L} = \frac{d^2}{dx^2}$.

La funcion de Green construida para las condiciones $f(0) = 0$ y $f'(1) = 0$ es:
$G(x, x') = \begin{cases} -x, & x < x' \\ -x', & x > x' \end{cases}$

A continuacion se muestra paso a paso que esta funcion cumple con la ecuacion de la delta de Dirac:

1. Evaluacion para puntos fuera de la fuente ($x \neq x'$)
De acuerdo a la definicion de la funcion de Green, esta debe ser una solucion de la ecuacion homogenea asociada en todas partes excepto en el punto donde se aplica el impulso.
*   Para la region $x < x'$:
    $\frac{d}{dx}(-x) = -1$
    $\frac{d^2}{dx^2}(-x) = 0$
*   Para la region $x > x'$:
    $\frac{d}{dx}(-x') = 0$
    $\frac{d^2}{dx^2}(-x') = 0$
Esto confirma que $\mathscr{L} G(x, x') = 0$ para todo $x \neq x'$.

2. Analisis del salto en la derivada
Para que la segunda derivada produzca una delta de Dirac, la primera derivada debe presentar una discontinuidad de valor $1/p(x')$ en el punto $x = x'$. En este problema, el coeficiente de la segunda derivada es $p(x) = 1$.
Evaluamos el limite del salto:
$\Delta \frac{dG}{dx} = \frac{dG_2}{dx} \big|_{x=x'} - \frac{dG_1}{dx} \big|_{x=x'}$
Sustituyendo los valores hallados en el paso anterior:
$\Delta \frac{dG}{dx} = 0 - (-1) = 1$
Este resultado coincide exactamente con el requisito teorico $1/p(x') = 1/1 = 1$.

3. Aplicacion de la integral sobre la singularidad
Al integrar el operador $\mathscr{L}$ sobre un entorno infinitesimal $[x'-\epsilon, x'+\epsilon]$ que rodea al punto critico, obtenemos:
$\int_{x'-\epsilon}^{x'+\epsilon} \frac{d^2G}{dx^2} dx = \left[ \frac{dG}{dx} \right]_{x'-\epsilon}^{x'+\epsilon} = 1$
Puesto que la integral del resultado del operador es igual a 1 en un intervalo que tiende a cero, y el resultado es nulo fuera de ese punto, la expresion matematica que describe este comportamiento es la delta de Dirac:
$\frac{d^2G(x, x')}{dx^2} = \delta(x - x')$

Por lo tanto, queda demostrado que para el operador diferencial $\mathscr{L}$ dado, la funcion de Green satisface la identidad $\mathscr{L}(x) G(x, x') = \delta(x - x')$.

Para llegar a la expresión de la **función de Green** $G(x, x') = \begin{cases} -x, & x < x' \\ -x', & x > x' \end{cases}$ para el operador $\mathscr{L} = \frac{d^2}{dx^2}$ con las condiciones de frontera $f(0)=0$ y $f'(1)=0$, se sigue un procedimiento analítico basado en las propiedades fundamentales de estas funciones.

Aquí tienes el desarrollo paso a paso:

### 1. Solución de la ecuación homogénea por tramos
Para $x \neq x'$, la función de Green debe satisfacer la ecuación homogénea $\frac{d^2G}{dx^2} = 0$. La solución general para una segunda derivada nula es una línea recta de la forma $ax + b$. Definimos la función por tramos:
*   Para $x < x'$: $G_1(x, x') = a_1(x')x + b_1(x')$
*   Para $x > x'$: $G_2(x, x') = a_2(x')x + b_2(x')$

### 2. Aplicación de las condiciones de frontera
Debemos imponer las condiciones de frontera del problema original sobre la variable de observación $x$:
*   **En $x = 0$:** La condición $f(0)=0$ implica que $G_1(0, x') = 0$.
    *   $a_1(x')(0) + b_1(x') = 0 \implies \mathbf{b_1(x') = 0}$.
    *   Por lo tanto, **$G_1(x, x') = a_1(x')x$**.
*   **En $x = 1$:** La condición $f'(1)=0$ implica que la derivada respecto a $x$ de $G_2$ debe anularse en el extremo derecho.
    *   $\frac{\partial G_2}{\partial x} = a_2(x')$.
    *   Evaluando en $x=1$: $a_2(x') = 0 \implies \mathbf{a_2(x') = 0}$.
    *   Por lo tanto, **$G_2(x, x') = b_2(x')$** (una función que solo depende de la posición de la fuente $x'$).

### 3. Propiedad de continuidad ($G_1 = G_2$)
La función de Green debe ser continua en el punto donde se aplica el impulso $x = x'$. Esto significa que los valores de ambos tramos deben coincidir en ese punto:
*   $G_1(x', x') = G_2(x', x')$
*   $a_1(x')x' = b_2(x')$

### 4. Propiedad de discontinuidad de la derivada
La primera derivada de la función de Green presenta un salto finito en $x = x'$ igual a $1/p(x')$, donde $p(x)$ es el coeficiente de la segunda derivada (en este caso, $p=1$).
La condición de salto se escribe como:
$$\frac{\partial G_2}{\partial x} \bigg|_{x=x'} - \frac{\partial G_1}{\partial x} \bigg|_{x=x'} = 1$$
Sustituyendo las derivadas de nuestras funciones del paso 2:
*   $\frac{\partial G_2}{\partial x} = 0$
*   $\frac{\partial G_1}{\partial x} = a_1(x')$
*   $0 - a_1(x') = 1 \implies \mathbf{a_1(x') = -1}$

### 5. Determinación de las constantes finales
Ahora que conocemos $a_1(x') = -1$, usamos la relación de continuidad del paso 3 para hallar $b_2(x')$:
*   $b_2(x') = a_1(x')x' \implies b_2(x') = (-1)x' \implies \mathbf{b_2(x') = -x'}$

### Resultado Final
Sustituyendo los valores de las constantes en las funciones por tramos:
*   Para $x < x'$: $G_1(x, x') = (-1)x = \mathbf{-x}$
*   Para $x > x'$: $G_2(x, x') = \mathbf{-x'}$

Esto confirma la expresión deseada:
$$\mathbf{G(x, x') = \begin{cases} -x, & x < x' \\ -x', & x > x' \end{cases}}$$

3. Planteamiento de la integral de resolución

Dado que el núcleo de la integral (la función de Green) cambia de forma analítica en el punto donde la fuente coincide con la observación ($z = x$), debemos dividir la integración en dos partes:

$$f(x) = \int_{0}^{x} G_{x>z}(x, z) g(z) dz + \int_{x}^{1} G_{x<z}(x, z) g(z) dz$$

Sustituyendo las expresiones correspondientes y el término fuente $g(z) = -z^2 + 2e^{-z}$:

$f(x) = \int_{0}^{x} (-z) (-z^2 + 2e^{-z}) dz + \int_{x}^{1} (-x) (-z^2 + 2e^{-z}) dz$

4. Evaluación detallada de las integrales

Primera integral ($I_1$): Corresponde al efecto de todas las fuentes situadas a la izquierda del punto $x$.
$I_1 = \int_{0}^{x} (z^3 - 2ze^{-z}) dz$
Resolviendo término a término:
$\int_{0}^{x} z^3 dz = [\frac{z^4}{4}]_{0}^{x} = \frac{x^4}{4}$
$\int_{0}^{x} -2ze^{-z} dz = [-2(-ze^{-z} - e^{-z})]_{0}^{x} = [2e^{-z}(z + 1)]_{0}^{x} = 2e^{-x}(x + 1) - 2$
Entonces, $I_1 = \frac{x^4}{4} + 2xe^{-x} + 2e^{-x} - 2$

Segunda integral ($I_2$): Corresponde al efecto de todas las fuentes situadas a la derecha del punto $x$. En este tramo, el factor $(-x)$ es constante respecto a la variable de integración $z$.
$I_2 = -x \int_{x}^{1} (-z^2 + 2e^{-z}) dz$
$\int_{x}^{1} -z^2 dz = [-\frac{z^3}{3}]_{x}^{1} = -\frac{1}{3} + \frac{x^3}{3}$
$\int_{x}^{1} 2e^{-z} dz = [-2e^{-z}]_{x}^{1} = -2e^{-1} + 2e^{-x}$
Multiplicando por $(-x)$:
$I_2 = -x [ (-\frac{1}{3} - \frac{2}{e}) - (-\frac{x^3}{3} - 2e^{-x}) ] = \frac{x}{3} + \frac{2x}{e} - \frac{x^4}{3} - 2xe^{-x}$

5. Consolidación del resultado final

La solución completa es la suma de ambas contribuciones, $f(x) = I_1 + I_2$. Al sumar los términos, observamos que los productos que contienen $xe^{-x}$ se cancelan mutuamente:

$f(x) = (\frac{x^4}{4} - \frac{x^4}{3}) + (2xe^{-x} - 2xe^{-x}) + 2e^{-x} + (\frac{x}{3} + \frac{2x}{e}) - 2$

$f(x) = -\frac{x^4}{12} + 2e^{-x} + (\frac{1}{3} + \frac{2}{e})x - 2$

### 1. Solución en el intervalo $$ con condiciones de Dirichlet

El objetivo es hallar la **función de Green** $G(x, x')$ que satisface $\frac{d^2}{dx^2} G(x, x') = \delta(x - x')$ sujeta a $G(0, x') = G(1, x') = 0$.

#### A. Método Directo (Construcción por tramos)
Buscamos una solución a la ecuación homogénea $\frac{d^2G}{dx^2} = 0$ para $x \neq x'$, la cual tiene la forma de una línea recta $ax + b$:
1.  **Región $x < x'$:** Imponemos $G(0, x') = 0$, lo que nos da $G_1(x, x') = A(x')x$.
2.  **Región $x > x'$:** Imponemos $G(1, x') = 0$, lo que nos da $G_2(x, x') = C(x')(x - 1)$.
3.  **Continuidad:** En $x = x'$, $G_1 = G_2 \implies A(x')x' = C(x')(x' - 1)$.
4.  **Salto en la derivada:** $\frac{dG_2}{dx} \big|_{x'} - \frac{dG_1}{dx} \big|_{x'} = 1 \implies C(x') - A(x') = 1$.

Resolviendo el sistema para las constantes $A$ y $C$:
*   De (4): $C = 1 + A$.
*   Sustituyendo en (3): $Ax' = (1 + A)(x' - 1) = x' - 1 + Ax' - A$.
*   Resulta $A = x' - 1$ y $C = x'$.

Por lo tanto, la función de Green es:
$$\mathbf{G(x, x') = \begin{cases} x(x' - 1), & 0 \le x < x' \\ x'(x - 1), & x' < x \le 1 \end{cases}}$$

#### B. Método de Autofunciones
Expandimos $G(x, x')$ en términos de las autofunciones del operador $\frac{d^2}{dx^2}$ con condiciones de Dirichlet en $$:
1.  **Autofunciones:** $\phi_n(x) = \sqrt{2} \sin(n\pi x)$.
2.  **Autovalores:** $\lambda_n = -n^2\pi^2$ (ya que $\phi_n'' = -n^2\pi^2 \phi_n$).
3.  **Expansión:** $G(x, x') = \sum_{n=1}^\infty \frac{\phi_n(x) \phi_n(x')}{\lambda_n}$.

$$\mathbf{G(x, x') = -\frac{2}{\pi^2} \sum_{n=1}^\infty \frac{\sin(n\pi x) \sin(n\pi x')}{n^2}}$$

#### Equivalencia
Ambos métodos son equivalentes ya que la serie de Fourier de senos de la solución por tramos coincide exactamente con la expansión obtenida en el método B. La solución final para el potencial es:
$$\phi(x) = \int_0^1 G(x, x') \left[ -\frac{\rho(x')}{\epsilon_0} \right] dx'$$

### 2. Solución en el espacio infinito ($-\infty < x < \infty$)

Para el dominio infinito, utilizamos la **Transformada de Fourier** para resolver $\frac{d^2G}{dx^2} = \delta(x - x')$.
Aplicando la transformada, obtenemos $-k^2 \hat{G}(k) = \frac{1}{\sqrt{2\pi}} e^{-ikx'}$. La inversión formal de esta expresión (tratada como una distribución) conduce a la solución fundamental en 1D:
$$\mathbf{G(x, x') = \frac{1}{2} |x - x'|}$$
*Verificación:* La primera derivada es $\frac{1}{2} \text{sgn}(x - x')$, y su derivada produce la delta de Dirac $\delta(x - x')$.

La solución para el potencial en el espacio infinito es:
$$\phi(x) = -\frac{1}{2\epsilon_0} \int_{-\infty}^\infty |x - x'| \rho(x') dx'$$

### 3. Comparación con el problema en 3D

Al comparar el resultado de 1D con la solución de Poisson en 3D hallada anteriormente en el taller ($G(\vec{r}, \vec{r}') = -\frac{1}{4\pi |\vec{r} - \vec{r}'|}$), observamos diferencias fundamentales:

*   **Dependencia con la distancia:** En **3D**, la función de Green (potencial de una fuente puntual) decae inversamente con la distancia ($1/R$). En **1D**, la función de Green crece linealmente con la distancia ($|x - x'|$).
*   **Comportamiento Físico:** Esto ocurre porque en 3D las líneas de campo se dispersan en un área que crece como $R^2$, reduciendo la intensidad del campo. En 1D, el flujo de campo a través de una "superficie" (puntos a los lados) es constante, lo que implica un campo eléctrico uniforme a cada lado y un potencial que aumenta linealmente.
*   **Signo:** Dependiendo de la convención del operador, el signo puede variar, pero la característica esencial es que el potencial 1D es una "V" divergente en el infinito, mientras que el 3D converge a cero.
