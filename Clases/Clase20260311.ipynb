A continuación se presenta la información sobre la **función Gamma ($\Gamma$)** extraída de las fuentes, formateada para ser utilizada directamente en una celda de **Markdown**:

---

### 1. Definiciones de la Función $\Gamma(z)$

Existen dos formas primordiales de definir esta función según el texto:

*   **Definición Integral (Integral de Euler):** Esta es la representación más común, válida para la parte real de $z$ mayor a cero:
    $$\Gamma(z) \equiv \int_0^\infty e^{-t} t^{z-1} dt, \quad \text{Re}(z) > 0$$

*   **Límite Infinito (Fórmula de Euler):** Esta definición es válida para todos los valores de $z$ excepto para los enteros no positivos ($0, -1, -2, \dots$):
    $$\Gamma(z) \equiv \lim_{n \to \infty} \frac{1 \cdot 2 \cdot 3 \cdots n}{z(z+1)(z+2) \cdots (z+n)} n^z$$

---

### 2. Relación con el Factorial

La función Gamma actúa como una generalización de la función factorial para números complejos y reales. Para un número entero positivo $n$, la relación es:
$$\Gamma(n+1) = n!$$
O de forma equivalente:
$$\Gamma(n) = (n-1)!$$
Por convención, se establece que $0! = \Gamma(1) = 1$.

---

### 3. Principales Propiedades y Relaciones Funcionales

*   **Relación de Recurrencia:** Es la propiedad fundamental que permite extender la función paso a paso:
    $$\Gamma(z+1) = z\Gamma(z)$$

*   **Fórmula de Reflexión:** Conecta la función con valores de argumento negativo:
    $$\Gamma(z)\Gamma(1-z) = \frac{\pi}{\sin(\pi z)}$$

*   **Fórmula de Duplicación de Legendre:** Útil para simplificar productos de funciones Gamma con argumentos desplazados por un medio:
    $$\Gamma(z+1)\Gamma\left(z + \frac{1}{2}\right) = \frac{\sqrt{\pi}}{2^{2z}}\Gamma(2z+1)$$

*   **Valor Especial Notable:** Un resultado frecuente en física (como en la distribución de Maxwell-Boltzmann) es el valor de la función para $1/2$:
    $$\Gamma\left(\frac{1}{2}\right) = \sqrt{\pi}$$

---

### 4. Analiticidad y Singularidades

La función $\Gamma(z)$ es una función analítica en todo el plano complejo, excepto en los puntos $z = 0, -1, -2, -3, \dots$, donde presenta **polos simples**. El residuo en el polo $z = -n$ viene dado por:
$$R_n = \frac{(-1)^n}{n!}$$

Para representar el producto de factoriales $n!m!$ como una multiplicación de integrales y luego llevarlo a coordenadas polares, utilizaremos la definición integral de la función Gamma, dado que $z! = \Gamma(z+1)$.

A continuación, se presenta el procedimiento paso a paso en formato **Markdown/TeX**:

---

### 1. Representación de $n!m!$ como producto de integrales

Partimos de la representación de la función Gamma para argumentos reales positivos, empleando la forma cuadrática de la integral de Euler:
$$\Gamma(p) = 2 \int_0^\infty s^{2p-1} e^{-s^2} ds$$

Sustituyendo $p = n+1$ y $q = m+1$, obtenemos las expresiones para $n!$ y $m!$:
$$n! = \Gamma(n+1) = 2 \int_0^\infty s^{2n+1} e^{-s^2} ds$$
$$m! = \Gamma(m+1) = 2 \int_0^\infty t^{2m+1} e^{-t^2} dt$$

Al multiplicar ambas integrales, el producto se puede escribir como una integral doble en el primer cuadrante del plano $st$:
$$n!m! = 4 \int_0^\infty \int_0^\infty s^{2n+1} t^{2m+1} e^{-(s^2+t^2)} ds \, dt$$

---

### 2. Transformación a coordenadas polares (Paso a paso)

Para resolver esta integral doble, realizamos un cambio de variables a coordenadas polares:

*   **Paso A: Definir las sustituciones.**
    Hacemos $s = r \cos \theta$ y $t = r \sin \theta$.
*   **Paso B: Transformar el exponente.**
    La suma de los cuadrados en el exponente se convierte en el radio al cuadrado: $s^2 + t^2 = r^2$.
*   **Paso C: Determinar el Jacobiano.**
    El elemento diferencial de área en el plano se transforma de $ds \, dt$ a $r \, dr \, d\theta$.
*   **Paso D: Establecer los nuevos límites.**
    Como la integración original cubre el cuadrante positivo ($s, t \geq 0$), los nuevos límites para el radio son $r \in [0, \infty)$ y para el ángulo son $\theta \in [0, \pi/2]$.

Sustituyendo estos valores en la integral doble obtenemos:
$$n!m! = 4 \int_0^\infty \int_0^{\pi/2} (r \cos \theta)^{2n+1} (r \sin \theta)^{2m+1} e^{-r^2} r \, d\theta \, dr$$

---

### 3. Separación de la expresión resultante

Podemos separar la integral en una parte radial y una parte angular:
$$n!m! = \left( 2 \int_0^\infty r^{2n+2m+3} e^{-r^2} dr \right) \left( 2 \int_0^{\pi/2} \cos^{2n+1} \theta \sin^{2m+1} \theta \, d\theta \right)$$

Identificando cada componente según las fuentes:
1.  La primera integral radial corresponde a $\Gamma(n+m+2)$, que es equivalente a $(n+m+1)!$.
2.  La segunda integral angular define la **Función Beta** $B(n+1, m+1)$.

Por lo tanto, la expresión final en términos de coordenadas polares permite establecer la relación fundamental entre las funciones Gamma y Beta:
$$n!m! = (n+m+1)! \, B(n+1, m+1)$$

A continuación, se presenta el desarrollo paso a paso para definir la **función Beta** y establecer su relación con la **función Gamma**, partiendo del resultado en coordenadas polares obtenido anteriormente.

---

### 1. Definición de la Función Beta $B(p,q)$

A partir de la separación de la integral doble en componentes radiales y angulares, se define la **función Beta** en su forma trigonométrica como la parte angular de dicha expresión:

$$B(p,q) \equiv 2 \int_0^{\pi/2} \cos^{2p-1} \theta \sin^{2q-1} \theta \, d\theta, \quad \text{Re}(p), \text{Re}(q) > 0$$

Esta función es simétrica, lo que implica que $B(p,q) = B(q,p)$, ya que intercambiar los exponentes de seno y coseno no altera el valor de la integral.

---

### 2. Representación de $B(p,q)$ en términos de la Función Gamma

Para llegar a la identidad fundamental que conecta ambas funciones, retomamos el producto de dos funciones Gamma utilizando su representación integral cuadrática:

**Paso 1: Escribir el producto de las integrales.**
Utilizamos $s$ y $t$ como variables de integración para $\Gamma(p)$ y $\Gamma(q)$ respectivamente:
$$\Gamma(p)\Gamma(q) = \left( 2 \int_0^\infty s^{2p-1} e^{-s^2} ds \right) \left( 2 \int_0^\infty t^{2q-1} e^{-t^2} dt \right)$$
$$\Gamma(p)\Gamma(q) = 4 \int_0^\infty \int_0^\infty s^{2p-1} t^{2q-1} e^{-(s^2+t^2)} ds \, dt$$

**Paso 2: Transformación a coordenadas polares.**
Realizamos el cambio de variables $s = r \cos \theta$ y $t = r \sin \theta$, donde el diferencial de área es $ds \, dt = r \, dr \, d\theta$:
$$\Gamma(p)\Gamma(q) = 4 \int_0^\infty \int_0^{\pi/2} (r \cos \theta)^{2p-1} (r \sin \theta)^{2q-1} e^{-r^2} r \, dr \, d\theta$$

**Paso 3: Separación de términos radiales y angulares.**
Agrupamos las potencias de $r$ y las funciones trigonométricas de $\theta$:
$$\Gamma(p)\Gamma(q) = \left( 2 \int_0^\infty r^{2(p+q)-1} e^{-r^2} dr \right) \left( 2 \int_0^{\pi/2} \cos^{2p-1} \theta \sin^{2q-1} \theta \, d\theta \right)$$

**Paso 4: Identificación de las funciones.**
*   La primera integral entre paréntesis es, por definición, la representación cuadrática de la función Gamma con argumento $(p+q)$:
    $$\Gamma(p+q) = 2 \int_0^\infty r^{2(p+q)-1} e^{-r^2} dr$$
*   La segunda integral es la definición de la función Beta $B(p,q)$ establecida en el punto anterior.

**Paso 5: Resultado final.**
Sustituyendo estas identificaciones en la ecuación del Paso 3, obtenemos:
$$\Gamma(p)\Gamma(q) = \Gamma(p+q) \cdot B(p,q)$$

Despejando $B(p,q)$, llegamos a la representación buscada:
$$B(p,q) = \frac{\Gamma(p)\Gamma(q)}{\Gamma(p+q)}$$

La relación de este ejemplo con la teoría reside en que la **función Beta ($B$)** proporciona una herramienta sistemática para evaluar integrales de potencias de senos y cosenos sobre el intervalo $[0, \pi/2]$.

De acuerdo con las fuentes, la representación trigonométrica de la función Beta es:
$$B(p, q) = 2 \int_0^{\pi/2} \cos^{2p-1}\theta \sin^{2q-1}\theta \, d\theta, \quad \text{Re}(p), \text{Re}(q) > 0$$

A continuación, se detalla la resolución del ejemplo utilizando esta teoría:

### 1. Identificación de parámetros
Deseamos calcular $I = \int_0^{\pi/2} \sin^2\theta \cos\theta \, d\theta$. Comparando con la fórmula general, identificamos los exponentes:
*   Para el seno: $2q - 1 = 2 \implies 2q = 3 \implies \mathbf{q = \frac{3}{2}}$.
*   Para el coseno: $2p - 1 = 1 \implies 2p = 2 \implies \mathbf{p = 1}$.

Puesto que la definición de $B(p, q)$ incluye un factor 2 al principio, la integral buscada es:
$$\int_0^{\pi/2} \sin^2\theta \cos\theta \, d\theta = \frac{1}{2} B\left(1, \frac{3}{2}\right)$$

### 2. Relación con la función Gamma
Utilizamos la identidad fundamental que relaciona ambas funciones:
$$B(p, q) = \frac{\Gamma(p)\Gamma(q)}{\Gamma(p+q)}$$

Sustituyendo nuestros valores:
$$B\left(1, \frac{3}{2}\right) = \frac{\Gamma(1)\Gamma(\frac{3}{2})}{\Gamma(1 + \frac{3}{2})} = \frac{\Gamma(1)\Gamma(\frac{3}{2})}{\Gamma(\frac{5}{2})}$$

### 3. Evaluación de los términos
*   **$\Gamma(1)$**: Por definición de factorial, $0! = \Gamma(1) = 1$.
*   **$\Gamma(\frac{3}{2})$**: Utilizando la relación de recurrencia $\Gamma(z+1) = z\Gamma(z)$, tenemos $\Gamma(\frac{3}{2}) = \frac{1}{2}\Gamma(\frac{1}{2})$. Sabiendo que $\Gamma(\frac{1}{2}) = \sqrt{\pi}$:
    $$\Gamma\left(\frac{3}{2}\right) = \frac{\sqrt{\pi}}{2}$$
*   **$\Gamma(\frac{5}{2})$**: Aplicando de nuevo la recurrencia:
    $$\Gamma\left(\frac{5}{2}\right) = \frac{3}{2}\Gamma\left(\frac{3}{2}\right) = \frac{3}{2} \cdot \frac{\sqrt{\pi}}{2} = \frac{3\sqrt{\pi}}{4}$$

### 4. Cálculo final
Sustituimos estos valores en la expresión de la función Beta:
$$B\left(1, \frac{3}{2}\right) = \frac{1 \cdot \frac{\sqrt{\pi}}{2}}{\frac{3\sqrt{\pi}}{4}} = \frac{\sqrt{\pi}}{2} \cdot \frac{4}{3\sqrt{\pi}} = \frac{2}{3}$$

Finalmente, aplicamos el factor $1/2$ identificado en el paso 1:
$$I = \frac{1}{2} \cdot \frac{2}{3} = \frac{1}{3}$$

**Resultado:** La integral $\int_0^{\pi/2} \sin^2{\theta}\cos{\theta}d\theta$ es exactamente **$1/3$**.

Deseamos calcular la integral $I = \int_0^{\pi/2} \cos^4\theta \sin^4\theta \, d\theta$. Al comparar los exponentes, obtenemos:
*   Para el coseno: $2p - 1 = 4 \implies 2p = 5 \implies \mathbf{p = \frac{5}{2}}$.

# delta de dirac

* Es una generalización de la delta de kronecker
* No es estrictamente una función
* Es una distribución que tiene muchas representaciones


La **delta de Dirac** ($\delta(x)$) se puede introducir formalmente a través del límite de una secuencia de funciones, como la **curva gaussiana** $f_\alpha(x) = \frac{\alpha}{\sqrt{\pi}}e^{-\alpha^2 x^2}$. Independientemente del valor del parámetro $\alpha$, esta curva siempre tiene un **área unitaria**. A medida que $\alpha$ crece hacia el infinito, la secuencia genera curvas **progresivamente más altas y estrechas**. En el límite $\alpha \to \infty$, la distribución se convierte en un "pico" infinitamente alto y estrecho en el origen, siendo nula en cualquier otro punto. 

Aunque no es una función ordinaria sino una **distribución**, se define de manera general mediante su **propiedad selectiva**: al integrarla con una función arbitraria $f(x)$, la delta "selecciona" el valor de dicha función en el punto donde se encuentra su pico. Matemáticamente se expresa como:
$$\int_{-\infty}^{\infty} f(x) \delta(x - x_0) dx = f(x_0)$$

### Ejemplo interesante con Wolfram Language
En el entorno de **Wolfram**, la delta de Dirac se representa con el comando `DiracDelta[x]`. Un uso avanzado y fascinante de esta herramienta es la resolución de **ecuaciones diferenciales inhomogéneas** mediante el cálculo de la **función de Green**.

Por ejemplo, para modelar el efecto de un "golpe" instantáneo sobre un oscilador armónico, se puede utilizar el siguiente comando en Wolfram para encontrar su respuesta (función de Green):

```wolfram
(* Hallar la función de Green para un oscilador armónico *)

```
Este comando utiliza internamente la `DiracDelta` para representar una **fuente puntual** de fuerza. Además, se puede verificar su comportamiento en transformadas integrales; por ejemplo, la **Transformada de Fourier** de una delta de Dirac devuelve una constante, lo que físicamente indica que un pulso instantáneo contiene **todas las frecuencias** con la misma intensidad.

Propiedad selectiva de la delta de dirac.

La **delta de Dirac** ($\delta(x-x_0)$) y la **delta de Kronecker** ($\delta_{ij}$) comparten una función operativa idéntica en física matemática: actúan como un **mecanismo de selección**, diferenciándose únicamente en el dominio (discreto o continuo) en el que operan.

A continuación, se presenta la analogía basada en su **propiedad selectiva**:

### 1. Delta de Kronecker (Ámbito Discreto)
La delta de Kronecker se define para índices enteros como $\delta_{ij} = 1$ si $i=j$, y $\delta_{ij} = 0$ si $i \neq j$.
*   **Acción selectiva:** Cuando aparece dentro de una sumatoria, la delta de Kronecker **"elige" o "filtra"** un único término de entre todos los posibles.
*   **Expresión matemática:** 
    $$\sum_i a_i \delta_{ij} = a_j$$
    Aquí, la delta selecciona la componente $j$-ésima del vector $a$.

### 2. Delta de Dirac (Ámbito Continuo)
La delta de Dirac es la **extensión al continuo** de la delta de Kronecker. Aunque técnicamente es una distribución (función patológica), se define por su comportamiento bajo una integral.
*   **Acción selectiva:** Al realizar una integración, la delta de Dirac **selecciona**, de entre todos los valores posibles de una función arbitraria $f(x)$, únicamente su valor en el punto $x = x_0$.
*   **Expresión matemática:**
    $$\int_{-\infty}^{\infty} f(x) \delta(x - x_0) dx = f(x_0)$$
    Esta integral es la analogía directa de la sumatoria discreta mencionada anteriormente.

### Resumen de la analogía
| Característica | Delta de Kronecker ($\delta_{ij}$) | Delta de Dirac ($\delta(x-x_0)$) |
| :--- | :--- | :--- |
| **Dominio** | Discreto (enteros) | Continuo (números reales) |
| **Operación** | Sumatoria ($\sum$) | Integral ($\int$) |
| **Resultado** | Selecciona una **componente** ($a_j$) | Selecciona un **valor funcional** ($f(x_0)$) |

En esencia, mientras que la delta de Kronecker permite extraer un elemento de una lista numerable, la delta de Dirac permite extraer un punto específico de un espectro continuo de valores.

Basado en las fuentes proporcionadas, la **delta de Dirac** ($\delta$) no es una función ordinaria sino una **distribución** que puede representarse matemáticamente de diversas maneras, principalmente como el límite de una secuencia de funciones o mediante integrales.


# Representación mediante integral de Fourier
Esta forma es fundamental en el análisis de ondas y transformadas, y define a la delta en términos de una **integral sobre un dominio infinito** de frecuencias o números de onda:
$$\delta(x - x_0) = \frac{1}{2\pi} \int_{-\infty}^{\infty} e^{ik(x - x_0)} dk$$
Esta representación es una consecuencia de la **condición de completez** de la base exponencial de Fourier.

## Otras representaciones mencionadas
El texto también señala que la delta de Dirac puede obtenerse mediante el límite de otras funciones de área 1 que exhiben un pico definido, tales como:
*   **Límite de una función lorentziana:** $\lim_{\alpha \to \infty} \frac{\alpha}{\pi} \frac{1}{1 + \alpha^2 x^2}$.
*   **Límite de una función sinc:** $\lim_{\alpha \to \infty} \frac{\sin(\alpha x)}{\pi x}$.
*   **Condición de completez de bases ortonormales:** Se puede representar como la suma del producto de las funciones de una base ortonormal completa, como $\sum_n p(x') \phi_n^*(x') \phi_n(x) = \delta(x - x')$.


Basado en las fuentes proporcionadas, a continuación se detallan las **propiedades algebraicas y operacionales** de la delta de Dirac ($\delta$), las cuales son fundamentales en el manejo de distribuciones en física matemática:

### 1. Propiedades de producto con funciones
Estas propiedades muestran cómo la delta interactúa con funciones ordinarias, "extrayendo" o "anulando" valores:
*   **Propiedad de filtrado general:** $f(x)\delta(x - x_0) = f(x_0)\delta(x - x_0)$.
*   **Anulación por el argumento:** $x\delta(x) = 0$.
*   **Generalización del producto:** $g(x)\delta(x - x_0) = g(x_0)\delta(x - x_0)$.

### 2. Escalamiento y argumentos compuestos
Permiten simplificar expresiones donde el argumento de la delta no es simplemente $x$:
*   **Escalamiento lineal:** $\delta(ax) = \frac{1}{|a|}\delta(x)$.
*   **Delta de una función $g(x)$:** Si $g(x)$ tiene raíces simples $x_k$ (donde $g(x_k)=0$ y $g'(x_k) \neq 0$), entonces:
    $$\delta[g(x)] = \sum_k \frac{\delta(x - x_k)}{|g'(x_k)|} \text{}$$
*   **Caso particular de parábola:** $\delta(x^2 - a^2) = \frac{1}{2|a|} [\delta(x + a) + \delta(x - a)]$.

### 3. Propiedades de la derivada de la delta
Aunque la derivada de una delta ($\delta'$) es una distribución aún más singular, obedece reglas precisas bajo el signo de integración:
*   **Acción sobre una función:** $\int_{-\infty}^{\infty} f(x) \frac{d}{dx}\delta(x - x_0) dx = -f'(x_0)$.
*   **Derivadas de orden superior:** $\int_{-\infty}^{\infty} f(x) \frac{d^n}{dx^n}\delta(x - x_0) dx = (-1)^n \frac{d^nf(x)}{dx^n} \big|_{x=x_0}$.
*   **Identidades algebraicas con la derivada:**
    *   $x \frac{d}{dx}\delta(x) = -\delta(x)$.
    *   $x^2 \frac{d}{dx}\delta(x) = 0$.
    *   $x^{m+1} \frac{d^m}{dx^m}\delta(x) = 0$.
*   **Antisimetría:** $\frac{d}{dx}\delta(x - x_0) = -\frac{d}{dx}\delta(x_0 - x)$.

### 4. Propiedades integrales y composición
*   **Simetría:** $\delta(x - x_0) = \delta(x_0 - x)$ y es una cantidad real ($\delta = \delta^*$).
*   **Reproducción (Convolución):** $\int_{-\infty}^{\infty} \delta(x - x')\delta(x' - x'') dx' = \delta(x - x'')$.
*   **Composición de derivadas:** $\int_{-\infty}^{\infty} \frac{d^m}{dx^m} \delta(x - x') \frac{d^n}{dx'^n} \delta(x' - x'') dx' = \frac{d^{m+n}}{dx^{m+n}} \delta(x - x'')$.

Estas propiedades permiten simplificar ecuaciones diferenciales inhomogéneas y modelar fuentes puntuales de manera rigurosa en el marco de la **Teoría de Distribuciones**.

**NOTA**

La siguiente es una propiedad muy importante

La propiedad de la delta de Dirac con un argumento funcional, generalmente denotada como **$\delta(g(x))$**, permite descomponer esta distribución en una suma de deltas centradas en las raíces de la función $g(x)$.

De acuerdo con las fuentes, los detalles fundamentales de esta propiedad son:

### 1. Definición General
Si la función $g(x)$ tiene **raíces simples** $x_k$ (es decir, puntos donde $g(x_k) = 0$ y su derivada $g'(x_k) \neq 0$), la delta se expresa como:
$$\delta(g(x)) = \sum_k \frac{\delta(x - x_k)}{|g'(x_k)|}$$

Esta suma se extiende sobre todas las $N$ raíces de la función dentro del dominio de interés.

### 2. Propiedad bajo la Integral
Cuando esta expresión se utiliza dentro de una integral con una función de prueba $f(x)$, la propiedad selectiva actúa sobre cada una de las raíces de $g(x)$:
$$\int_{-\infty}^{\infty} f(x)\delta[g(x)] dx = \sum_k \frac{f(x_k)}{|g'(x_k)|}$$

Para que esta relación sea válida, se exige que la derivada en las raíces no sea nula ($g'(x_k) \neq 0$).

### 3. Caso Particular: Función Cuadrática
Un ejemplo común citado en las lecciones es el de una parábola del tipo $g(x) = x^2 - a^2$. En este caso, las raíces son $x = a$ y $x = -a$, y la derivada es $g'(x) = 2x$. Aplicando la propiedad, se obtiene:
$$\delta(x^2 - a^2) = \frac{1}{2|a|} [\delta(x + a) + \delta(x - a)]$$

### Resumen de Requisitos
*   **Raíces simples:** La función $g(x)$ debe cruzar el eje $x$; si la raíz es doble o la derivada es cero en ese punto, la fórmula estándar no aplica de forma directa.
*   **Factor de escala:** El denominador $|g'(x_k)|$ compensa la "velocidad" con la que la función $g(x)$ pasa por cero, asegurando que el área total de la distribución siga siendo consistente con la definición de la delta.

# Cálculo de Densidades de Carga y la Delta de Dirac

En física matemática, es fundamental representar distribuciones de carga (puntuales, lineales o superficiales) como si fueran **densidades volumétricas** ($\rho$). Para lograr esto de manera rigurosa, utilizamos la **Delta de Dirac** ($\delta$).

## 1. Definición en Coordenadas Esféricas

Para una esfera de radio $R$ con carga total $Q$, trabajamos en un sistema de coordenadas esféricas $(r, \theta, \varphi)$. El elemento de volumen diferencial ($dV$) se define a partir de los factores de escala $h_1, h_2, h_3$:

*   $h_r = 1$
*   $h_\theta = r$
*   $h_\varphi = r \sin\theta$

Así, el diferencial de volumen es:
$$dV = (h_r dr)(h_\theta d\theta)(h_\varphi d\varphi) = r^2 \sin\theta \, dr \, d\theta \, d\varphi$$

La carga elemental $dq$ se relaciona con la densidad volumétrica $\rho$ mediante:
$$dq = \rho \, dV = \rho (r^2 \sin\theta \, dr \, d\theta \, d\varphi)$$

---

## 2. El Problema de la Densidad Uniforme en Todo el Espacio

Si asumimos una densidad de carga constante $\rho_0$ en **todo** el espacio euclidiano (desde $r=0$ hasta $r \to \infty$), al calcular la carga total $Q$ obtenemos:

$$Q = \int_{\text{espacio}} \rho_0 \, dV = \int_0^{2\pi} \int_0^\pi \int_0^\infty \rho_0 \, r^2 \sin\theta \, dr \, d\theta \, d\varphi$$

Integrando las partes angulares (que dan $4\pi$ estereorradianes):
$$Q = 4\pi \rho_0 \int_0^\infty r^2 \, dr = 4\pi \rho_0 \left[ \frac{r^3}{3} \right]_0^\infty \to \infty$$

**¿Cuál es el problema aquí?**
El resultado infinito indica que no estamos limitando la carga a una región finita. Para modelar un objeto físico real, como una **corteza esférica** o una **esfera aislante**, debemos "filtrar" la densidad para que solo exista donde está la carga.

---

## 3. Solución: Uso de la Delta de Dirac para una Corteza Esférica

Si queremos representar una carga $Q$ distribuida únicamente en la **superficie** de una esfera de radio $R$ (una corteza), la densidad volumétrica $\rho$ debe ser nula en todas partes excepto en $r=R$.

Para que la integral de volumen nos devuelva la carga finita $Q$, redefinimos $\rho$ como:
$$\rho(r) = A \frac{\delta(r - R)}{r^2}$$

### Demostración y comprobación:
Calculamos la carga total $Q$ con esta nueva expresión para verificar la constante de normalización $A$:

$$Q = \iiint \left( A \frac{\delta(r - R)}{r^2} \right) r^2 \sin\theta \, dr \, d\theta \, d\varphi$$

1.  **Simplificación:** Los términos $r^2$ se cancelan.
2.  **Integración Angular:** $\int_0^{2\pi} d\varphi \int_0^\pi \sin\theta \, d\theta = 4\pi$.
3.  **Propiedad Selectiva de la Delta:** Según la definición $\int f(r)\delta(r-R)dr = f(R)$:
    $$Q = 4\pi A \int_0^\infty \delta(r - R) \, dr = 4\pi A (1) = 4\pi A$$

Por lo tanto, la constante es $A = \frac{Q}{4\pi}$.

### Resultado Final Corregido:
La densidad volumétrica de carga para una corteza esférica de radio $R$ y carga $Q$ es:
$$\rho(r) = \frac{Q}{4\pi r^2} \delta(r - R)$$

*(Nota: En tu nota aparecía $\rho_0$, que es equivalente a nuestra constante $A$. La forma correcta de escribirlo para que la unidad sea carga/distancia$^3$ es incluir el factor de área en el denominador o definir $\rho_0$ en términos de la densidad superficial $\sigma$)*.

---

## 4. Resumen de correcciones de notación

1.  **Jacobiano/Factores de escala:** Tu expresión $\frac{dq}{h_1h_2h_3 du_1du_2du_3}$ es correcta conceptualmente, pero en esféricas es preferible usar directamente $r^2\sin\theta$ para mayor claridad.
2.  **Diferencial de ángulo:** Es común usar $d\phi$ o $d\varphi$ para el ángulo azimutal, pero mantén la consistencia. El estándar en el texto de Sepúlveda es $\phi$ para el azimut y $\theta$ para la colatitud.
3.  **Límites de integración:** En el cálculo de la carga total, siempre especifica que para el ángulo polar $\theta$ el límite es $[0, \pi]$ y para el azimutal $\varphi$ es $[0, 2\pi]$.
4.  **La Delta de Dirac:** La expresión $\frac{\delta(r-R)}{r^2}$ es la forma técnica correcta porque compensa el factor $r^2$ del diferencial de volumen en esféricas, permitiendo que la carga sea una propiedad puntual o superficial.
