# Problema 3

Considere la base de polinomios $\{x^n\}$. Utilizando el método de Gram-Schmidt
encuentre los dos primeros polinomios de Hermite $H_n(x)$ con $-\infty < x < \infty$,
función de peso $w(x) = e^{-x^2}$, y normalización:

$$\langle H_n \mid H_m \rangle = 2^m\, m!\,\sqrt{\pi}\;\delta_{mn}.$$

## Solución

### Producto interno

Con la función de peso $w(x) = e^{-x^2}$, el producto interno entre dos funciones $f$ y $g$ se define como:

$$\langle f \mid g \rangle = \int_{-\infty}^{\infty} e^{-x^2}\, f(x)\, g(x)\, dx.$$

### Integrales gaussianas de referencia

Para aplicar el procedimiento se necesitan los siguientes resultados estándar:

$$\int_{-\infty}^{\infty} e^{-x^2}\, dx = \sqrt{\pi}, \qquad \int_{-\infty}^{\infty} x^2\, e^{-x^2}\, dx = \frac{\sqrt{\pi}}{2}, \qquad \int_{-\infty}^{\infty} x^4\, e^{-x^2}\, dx = \frac{3\sqrt{\pi}}{4}.$$

Toda integral de la forma $\displaystyle\int_{-\infty}^{\infty} x^{2k+1}e^{-x^2}dx = 0$, pues el integrando es una función impar.

### Método de Gram-Schmidt

Se parte de la base $\{u_n\} = \{1,\, x,\, x^2,\, \ldots\}$ y se construye una base ortogonal $\{\varphi_n\}$ mediante:

$$\varphi_n = u_n - \sum_{k=0}^{n-1} \frac{\langle \varphi_k \mid u_n \rangle}{\langle \varphi_k \mid \varphi_k \rangle}\,\varphi_k.$$

Una vez obtenidos los $\varphi_n$ ortogonales, se normalizan para satisfacer la condición $\langle H_n \mid H_m \rangle = 2^n n!\sqrt{\pi}\,\delta_{mn}$.

---

### Paso 0: Construcción de $H_0(x)$

Se toma $\varphi_0 = u_0 = 1$. Su norma cuadrada es:

$$\langle \varphi_0 \mid \varphi_0 \rangle = \int_{-\infty}^{\infty} e^{-x^2}\, dx = \sqrt{\pi}.$$

La condición de normalización exige $\langle H_0 \mid H_0 \rangle = 2^0\cdot 0!\cdot\sqrt{\pi} = \sqrt{\pi}$. Escribiendo $H_0 = c_0\,\varphi_0$:

$$\langle H_0 \mid H_0 \rangle = c_0^2\,\sqrt{\pi} = \sqrt{\pi} \quad\Longrightarrow\quad c_0 = 1.$$

$$\boxed{H_0(x) = 1}$$

---

### Paso 1: Construcción de $H_1(x)$

Se aplica Gram-Schmidt sobre $u_1 = x$:

$$\varphi_1 = u_1 - \frac{\langle \varphi_0 \mid u_1 \rangle}{\langle \varphi_0 \mid \varphi_0 \rangle}\,\varphi_0.$$

El coeficiente involucrado requiere calcular:

$$\langle \varphi_0 \mid u_1 \rangle = \int_{-\infty}^{\infty} x\, e^{-x^2}\, dx = 0,$$

pues $x\,e^{-x^2}$ es una función **impar**. Por tanto:

$$\varphi_1 = x.$$

Su norma cuadrada es:

$$\langle \varphi_1 \mid \varphi_1 \rangle = \int_{-\infty}^{\infty} x^2\, e^{-x^2}\, dx = \frac{\sqrt{\pi}}{2}.$$

La condición de normalización exige $\langle H_1 \mid H_1 \rangle = 2^1\cdot 1!\cdot\sqrt{\pi} = 2\sqrt{\pi}$. Escribiendo $H_1 = c_1\,\varphi_1$:

$$c_1^2\,\frac{\sqrt{\pi}}{2} = 2\sqrt{\pi} \quad\Longrightarrow\quad c_1^2 = 4 \quad\Longrightarrow\quad c_1 = 2.$$

$$\boxed{H_1(x) = 2x}$$

---

### Paso 2: Construcción de $H_2(x)$

Se aplica Gram-Schmidt sobre $u_2 = x^2$:

$$\varphi_2 = u_2 - \frac{\langle \varphi_1 \mid u_2 \rangle}{\langle \varphi_1 \mid \varphi_1 \rangle}\,\varphi_1 - \frac{\langle \varphi_0 \mid u_2 \rangle}{\langle \varphi_0 \mid \varphi_0 \rangle}\,\varphi_0.$$

Se calculan los dos coeficientes:

$$\langle \varphi_1 \mid u_2 \rangle = \int_{-\infty}^{\infty} x^3\, e^{-x^2}\, dx = 0 \quad\text{(integrando impar)},$$

$$\langle \varphi_0 \mid u_2 \rangle = \int_{-\infty}^{\infty} x^2\, e^{-x^2}\, dx = \frac{\sqrt{\pi}}{2}.$$

Por tanto:

$$\varphi_2 = x^2 - 0 - \frac{\sqrt{\pi}/2}{\sqrt{\pi}}\cdot 1 = x^2 - \frac{1}{2}.$$

Se calcula su norma cuadrada:

$$\langle \varphi_2 \mid \varphi_2 \rangle = \int_{-\infty}^{\infty}\!\left(x^2 - \frac{1}{2}\right)^2 e^{-x^2}\, dx = \int_{-\infty}^{\infty}\!\left(x^4 - x^2 + \frac{1}{4}\right)e^{-x^2}\, dx = \frac{3\sqrt{\pi}}{4} - \frac{\sqrt{\pi}}{2} + \frac{\sqrt{\pi}}{4} = \frac{\sqrt{\pi}}{2}.$$

La condición de normalización exige $\langle H_2 \mid H_2 \rangle = 2^2\cdot 2!\cdot\sqrt{\pi} = 8\sqrt{\pi}$. Escribiendo $H_2 = c_2\,\varphi_2$:

$$c_2^2\,\frac{\sqrt{\pi}}{2} = 8\sqrt{\pi} \quad\Longrightarrow\quad c_2^2 = 16 \quad\Longrightarrow\quad c_2 = 4.$$

$$\boxed{H_2(x) = 4\!\left(x^2 - \frac{1}{2}\right) = 4x^2 - 2}$$

---

### Verificación de la ortogonalidad

Se comprueba que las tres funciones son mutuamente ortogonales bajo el peso $e^{-x^2}$:

$$\langle H_0 \mid H_1 \rangle = \int_{-\infty}^{\infty} 2x\, e^{-x^2}\, dx = 0 \quad\text{(integrando impar)},$$

$$\langle H_0 \mid H_2 \rangle = \int_{-\infty}^{\infty} (4x^2-2)\,e^{-x^2}\, dx = 4\cdot\frac{\sqrt{\pi}}{2} - 2\sqrt{\pi} = 0,$$

$$\langle H_1 \mid H_2 \rangle = \int_{-\infty}^{\infty} 2x(4x^2-2)\,e^{-x^2}\, dx = \int_{-\infty}^{\infty}(8x^3 - 4x)\,e^{-x^2}\, dx = 0 \quad\text{(integrando impar)}.$$

---

### Resumen de resultados

Aplicando el método de Gram-Schmidt a la base $\{1,\, x,\, x^2,\,\ldots\}$ con el producto interno inducido por el peso $w(x) = e^{-x^2}$, y exigiendo la normalización $\langle H_n \mid H_m \rangle = 2^n n!\sqrt{\pi}\,\delta_{mn}$, se obtienen los polinomios de Hermite:

$$H_0(x) = 1, \qquad H_1(x) = 2x, \qquad H_2(x) = 4x^2 - 2,$$

en concordancia con la expresión general dada en Sepúlveda (ec. 8.62) y con la fórmula de Rodrigues $H_n(x) = (-1)^n e^{x^2}\dfrac{d^n}{dx^n}\!\left(e^{-x^2}\right)$.

## Bases Discretas vs. Bases Continuas

### Bases Discretas

Una base discreta $\{\varphi_n(x)\}$ está indexada por un entero $n = 1, 2, 3, \ldots$ y se define sobre un **intervalo finito** $(a, b)$. La condición de ortogonalidad tiene la forma:

$$\int_a^b \varphi_n^*(x)\varphi_m(x)\, dx = A_n \delta_{nm}$$

donde $\delta_{nm}$ es la **delta de Kronecker**. Cualquier función $f(x)$ se expande como una **suma discreta**:

$$f(x) = \sum_n C_n \varphi_n(x)$$

El ejemplo más clásico es la **base trigonométrica de Fourier** en el intervalo $(-L, L)$:

$$\left\{\frac{1}{\sqrt{2L}}e^{in\pi x/L}\right\}, \quad n \in \mathbb{Z}$$

con condición de ortogonalidad:

$$\int_{-L}^{L} e^{i(n-m)\pi x/L}\, dx = 2L\,\delta_{nm}$$

La expansión de $f(x)$ resulta en la conocida **serie de Fourier**:

$$f(x) = \sum_{n=-\infty}^{\infty} C_n\, e^{in\pi x/L}$$

---

### Bases Continuas

Una base continua $\{\varphi(k, x)\}$ está indexada por una variable **real y continua** $k$, y se define típicamente sobre un **dominio infinito o semi-infinito**. La condición de ortogonalidad involucra ahora la **delta de Dirac**:

$$\int_a^b \varphi^*(k, x)\varphi(k', x)\, dx = \delta(k - k')$$

La expansión de $f(x)$ se convierte en una **integral**:

$$f(x) = \int C(k)\,\varphi(k, x)\, dk$$

El ejemplo natural es el **límite** $L \to \infty$ de la base de Fourier discreta. Cuando $L \to \infty$, el parámetro $k = n\pi/L$ varía de manera continua, y la base pasa a ser:

$$\left\{\frac{1}{\sqrt{2\pi}}e^{ikx}\right\}, \quad k \in \mathbb{R}$$

con ortogonalidad:

$$\int_{-\infty}^{\infty} e^{i(k-k')x}\, dx = 2\pi\,\delta(k - k')$$

La expansión se convierte en la **transformada de Fourier**:

$$f(x) = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty} F(k)\, e^{ikx}\, dk$$

---

### Tabla Comparativa

| Característica | Base Discreta | Base Continua |
|---|---|---|
| Índice | Entero $n$ | Variable real $k$ |
| Dominio | Intervalo **finito** | Intervalo **infinito** |
| Ortogonalidad | $\delta_{nm}$ (Kronecker) | $\delta(k-k')$ (Dirac) |
| Expansión de $f(x)$ | Suma $\sum_n C_n \varphi_n$ | Integral $\int C(k)\varphi(k,x)dk$ |
| Coeficientes | Discretos $C_n$ | Función continua $C(k)$ |
| Ejemplo (Fourier) | Serie de Fourier | Transformada de Fourier |

---

### Conclusión Cualitativa

La transición de una base discreta a una continua no es un cambio abrupto, sino un **límite natural**: cuando el dominio de definición de la función crece sin límite, los modos permitidos se acercan cada vez más entre sí hasta llenar densamente el eje $k$, convirtiendo la suma en una integral. Esto revela una profunda coherencia estructural: tanto las series como las integrales de Fourier son manifestaciones del mismo principio de expansión en bases ortogonales, adaptado al tamaño del dominio. En la práctica, esto implica que la elección entre una u otra herramienta no es arbitraria, sino que está **dictada por la geometría del problema**: dominios finitos piden bases discretas y coeficientes contables, mientras que dominios infinitos exigen bases continuas y espectros que varían suavemente.

Partiendo de la expansión en la base continua de Fourier:

$$f(x) = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty} F(k)\, e^{ikx}\, dk$$

Multiplicamos ambos lados por $e^{-ik'x}$ e integramos sobre todo $x$:

$$\int_{-\infty}^{\infty} f(x)\, e^{-ik'x}\, dx = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty} F(k) \left[\int_{-\infty}^{\infty} e^{ikx}e^{-ik'x}\, dx\right] dk$$

El corchete interior es precisamente la condición de ortogonalidad de la base:

$$\int_{-\infty}^{\infty} e^{i(k-k')x}\, dx = 2\pi\,\delta(k - k')$$

Sustituyendo:

$$\int_{-\infty}^{\infty} f(x)\, e^{-ik'x}\, dx = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty} F(k)\cdot 2\pi\,\delta(k-k')\, dk$$

Usando la propiedad selectiva de la delta de Dirac, $\int F(k)\,\delta(k - k')\,dk = F(k')$:

$$\int_{-\infty}^{\infty} f(x)\, e^{-ik'x}\, dx = \frac{2\pi}{\sqrt{2\pi}}\, F(k') = \sqrt{2\pi}\, F(k')$$

Despejando y renombrando $k' \to k$:

$$\boxed{F(k) = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty} f(x)\, e^{-ikx}\, dx}$$

Este es exactamente el par transformado: $f(x)$ y $F(k)$ se obtienen una de la otra mediante el **mismo operador**, cambiando únicamente el signo del exponente, lo que refleja la simetría entre las variables conjugadas $x$ y $k$.

## Transformada de Fourier del Paquete Gaussiano

Aplicando la definición:

$$F(k) = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{\infty} Ae^{-\alpha^2 x^2}\, e^{-ikx}\, dx = \frac{A}{\sqrt{2\pi}}\int_{-\infty}^{\infty} e^{-\alpha^2 x^2 - ikx}\, dx$$

**Completamos el cuadrado** en el exponente:

$$-\alpha^2 x^2 - ikx = -\alpha^2\left(x^2 + \frac{ik}{\alpha^2}x\right) = -\alpha^2\left(x + \frac{ik}{2\alpha^2}\right)^2 - \frac{k^2}{4\alpha^2}$$

Sustituyendo:

$$F(k) = \frac{A}{\sqrt{2\pi}}\, e^{-k^2/4\alpha^2} \int_{-\infty}^{\infty} e^{-\alpha^2\left(x + \frac{ik}{2\alpha^2}\right)^2} dx$$

**Hacemos el cambio de variable** $u = \alpha\left(x + \dfrac{ik}{2\alpha^2}\right)$, con $du = \alpha\, dx$:

$$\int_{-\infty}^{\infty} e^{-\alpha^2\left(x + \frac{ik}{2\alpha^2}\right)^2} dx = \frac{1}{\alpha}\int_{-\infty}^{\infty} e^{-u^2}\, du = \frac{\sqrt{\pi}}{\alpha}$$

donde usamos la integral gaussiana estándar $\int_{-\infty}^{\infty} e^{-u^2}du = \sqrt{\pi}$. Entonces:

$$F(k) = \frac{A}{\sqrt{2\pi}}\cdot\frac{\sqrt{\pi}}{\alpha}\, e^{-k^2/4\alpha^2}$$

$$\boxed{F(k) = \frac{A}{\sqrt{2}\,\alpha}\, e^{-k^2/4\alpha^2}}$$

El resultado es también una **gaussiana**, centrada en $k=0$, con ancho $2\alpha$ en el espacio de frecuencias. Nótese la relación inversa entre anchos: si $\alpha$ es grande, $f(x)$ es estrecha y $F(k)$ es ancha, y viceversa. Esto es una expresión directa del **principio de incertidumbre**.

## Definición de los anchos

Los anchos $\Delta x$ y $\Delta k$ se definen como el **ancho total a la mitad del máximo** (FWHM) de cada gaussiana.

---

## Ancho $\Delta x$ de $f(x)$

El máximo de $f(x) = Ae^{-\alpha^2 x^2}$ ocurre en $x=0$ con valor $f(0) = A$. Los puntos de semimáximo satisfacen:

$$f(x_{1/2}) = \frac{A}{2} \implies Ae^{-\alpha^2 x_{1/2}^2} = \frac{A}{2}$$

$$e^{-\alpha^2 x_{1/2}^2} = \frac{1}{2} \implies \alpha^2 x_{1/2}^2 = \ln 2 \implies x_{1/2} = \pm\frac{\sqrt{\ln 2}}{\alpha}$$

Por tanto:

$$\boxed{\Delta x = 2x_{1/2} = \frac{2\sqrt{\ln 2}}{\alpha}}$$

---

## Ancho $\Delta k$ de $F(k)$

El máximo de $F(k) = \dfrac{A}{\sqrt{2}\,\alpha}e^{-k^2/4\alpha^2}$ ocurre en $k=0$. Los puntos de semimáximo satisfacen:

$$F(k_{1/2}) = \frac{1}{2}\cdot\frac{A}{\sqrt{2}\,\alpha} \implies e^{-k_{1/2}^2/4\alpha^2} = \frac{1}{2}$$

$$\frac{k_{1/2}^2}{4\alpha^2} = \ln 2 \implies k_{1/2} = \pm\, 2\alpha\sqrt{\ln 2}$$

Por tanto:

$$\boxed{\Delta k = 2k_{1/2} = 4\alpha\sqrt{\ln 2}}$$

---

## Producto $\Delta x \cdot \Delta k$

$$\Delta x \cdot \Delta k = \frac{2\sqrt{\ln 2}}{\alpha} \cdot 4\alpha\sqrt{\ln 2}$$

$$\Delta x \cdot \Delta k = 8\left(\sqrt{\ln 2}\right)^2$$

$$\boxed{\Delta x \cdot \Delta k = 8\ln 2}$$

---

Este resultado es notable: el producto es una **constante universal**, independiente de $A$ y de $\alpha$. Confirma que no existe ninguna elección de $\alpha$ que permita tener simultáneamente una señal estrecha en $x$ y estrecha en $k$. Es la expresión matemática precisa del **principio de incertidumbre de Heisenberg**, donde bajo la identificación $k = p/\hbar$ se tiene $\Delta x\, \Delta p = 8\hbar\ln 2$.

# Proceso de Gram Schmidt

El **proceso de ortogonalización de Gram-Schmidt** es un método matemático que permite construir una **base ortogonal** (o unitaria) a partir de un conjunto inicial de funciones o vectores que son linealmente independientes pero no ortogonales.

Este procedimiento es fundamental en la teoría de espacios de Hilbert, ya que permite generar los "ejes" independientes necesarios para expandir cualquier función en una serie de funciones ortogonales.

### Lógica del Procedimiento
La idea central consiste en tomar un elemento del conjunto inicial y restarle sus **proyecciones** sobre los elementos que ya han sido ortogonalizados previamente. De esta manera, el residuo resultante es perpendicular a todos los anteriores.

Si partimos de un conjunto de funciones $\{f_n(x)\}$ definido en un intervalo $(a, b)$, el proceso para obtener la nueva base $\{\phi_n(x)\}$ sigue estos pasos lógicos:

1.  **Primer elemento:** Se toma la primera función como punto de partida (y se normaliza si se busca una base ortonormal).
    $$\phi_1 = f_1$$
2.  **Segundo elemento:** Se toma $f_2$ y se le resta su componente en la dirección de $\phi_1$.
    $$\phi_2(x) = f_2(x) - \phi_1(x) \frac{(\phi_1, f_2)}{(\phi_1, \phi_1)}$$
3.  **Término general:** Para cualquier función $n$, el nuevo elemento de la base se construye restando las proyecciones sobre todos los elementos previos ($m < n$):
    $$\mathbf{\phi_n(x) = f_n(x) - \sum_{m=1}^{n-1} \phi_m(x) \frac{(\phi_m, f_n)}{(\phi_m, \phi_m)}}$$

### Aplicaciones y Resultados
*   **Construcción de Polinomios Ortogonales:** Una de las aplicaciones más conocidas es transformar la **base de Taylor** ($1, x, x^2, x^3, \dots$) en familias de polinomios ortogonales según el intervalo y la función de peso elegida:
    *   En el intervalo $[-1, 1]$ con peso 1, genera los **Polinomios de Legendre**.
    *   En $(-\infty, \infty)$ con peso $e^{-x^2}$, genera los **Polinomios de Hermite**.
    *   En $[0, \infty)$ con peso $e^{-x}$, genera los **Polinomios de Laguerre**.
*   **Vectores Físicos:** También se aplica a vectores ordinarios en espacios $n$-dimensionales para encontrar una base de vectores unitarios perpendiculares entre sí a partir de vectores oblicuos.

### Limitaciones
Aunque el proceso es muy potente para construir polinomios, los textos señalan que **no permite construir series infinitas ortogonales**, como las funciones de Bessel, que surgen de forma distinta en la resolución de ecuaciones diferenciales. Además, en cálculos computacionales con bases de gran tamaño, el proceso puede volverse numéricamente inestable si no se maneja con cuidado.

### 1. Prueba de no ortogonalidad
El producto escalar para funciones reales con peso $p(x)=1$ se define como la integral del producto de las funciones en el intervalo dado.
Calculamos $(f_1, f_3)$:
$$(f_1, f_3) = \int_{-1}^{1} (1)(x^2) dx = \left[ \frac{x^3}{3} \right]_{-1}^{1} = \frac{1}{3} - \left( -\frac{1}{3} \right) = \mathbf{\frac{2}{3}}$$
Como el resultado es diferente de cero, las funciones **no son ortogonales**.

El proceso de ortogonalización de **Gram-Schmidt** es un método sistemático para construir una **base ortogonal** $\{\phi_n(x)\}$ partiendo de un conjunto de funciones $\{f_n(x)\}$ que son linealmente independientes pero no necesariamente perpendiculares entre sí en un intervalo $(a, b)$.

A continuación se detalla el desarrollo de este proceso utilizando la definición recursiva solicitada, la cual consiste en restar de cada nueva función original sus **proyecciones** sobre los ejes de la base ya ortogonalizados:

### 1. Definición del Producto Escalar
Para llevar a cabo el proceso en un espacio de Hilbert, primero debemos definir el producto escalar entre dos funciones $f$ y $g$ en el intervalo de interés $(a, b)$, usualmente con un factor de peso $p(x)=1$:
$$(\phi_m, f_n) = \int_{a}^{b} \phi_m^*(x) f_n(x) dx$$.

### 2. Pasos del Procedimiento

Siguiendo la fórmula $\phi_n(x) = f_n(x) - \sum_{m=1}^{n-1} \phi_m(x) \frac{(\phi_m, f_n)}{(\phi_m, \phi_m)}$:

*   **Primer elemento ($n=1$):**
    Como no existen funciones previas ($m < 1$), el primer elemento de la base ortogonal es simplemente la primera función original:
    $$\mathbf{\phi_1(x) = f_1(x)}$$

*   **Segundo elemento ($n=2$):**
    Tomamos la función $f_2(x)$ y le restamos su proyección en la dirección de $\phi_1$:
    $$\mathbf{\phi_2(x) = f_2(x) - \phi_1(x) \frac{(\phi_1, f_2)}{(\phi_1, \phi_1)}}$$
    El término $\frac{(\phi_1, f_2)}{(\phi_1, \phi_1)}$ actúa como un coeficiente que escala a $\phi_1$ para extraer exactamente la componente de $f_2$ que es paralela a él.

*   **Tercer elemento ($n=3$):**
    Ahora debemos restar las proyecciones de $f_3(x)$ sobre los dos vectores ya ortogonalizados, $\phi_1$ y $\phi_2$:
    $$\mathbf{\phi_3(x) = f_3(x) - \left[ \phi_1(x) \frac{(\phi_1, f_3)}{(\phi_1, \phi_1)} + \phi_2(x) \frac{(\phi_2, f_3)}{(\phi_2, \phi_2)} \right]}$$

### 3. Ejemplo Práctico: Generación de Polinomios de Legendre
Partiendo de la **base de Taylor** $\{f_n(x)\} = \{1, x, x^2, \dots\}$ en el intervalo $[-1, 1]$ con peso 1:

1.  **$\phi_1 = f_1 = 1$**.
2.  Para $\phi_2$, calculamos la proyección de $f_2=x$ sobre $\phi_1=1$:
    $(\phi_1, f_2) = \int_{-1}^{1} (1)(x) dx = 0$.
    Entonces, **$\phi_2 = x - 0 = x$**.
3.  Para $\phi_3$, calculamos las proyecciones de $f_3=x^2$:
    *   Sobre $\phi_1$: $\frac{(\phi_1, f_3)}{(\phi_1, \phi_1)} = \frac{\int_{-1}^1 x^2 dx}{\int_{-1}^1 1 dx} = \frac{2/3}{2} = 1/3$.
    *   Sobre $\phi_2$: $\frac{(\phi_2, f_3)}{(\phi_2, \phi_2)} = \frac{\int_{-1}^1 x^3 dx}{\int_{-1}^1 x^2 dx} = \frac{0}{2/3} = 0$.
    Resultando en: **$\phi_3 = x^2 - 1/3$**.

### 4. Importancia y Aplicaciones
Este proceso garantiza que la nueva base $\{\phi_n\}$ cumpla la condición de **ortogonalidad**: $(\phi_n, \phi_m) = 0$ para $n \neq m$. Además, si se desea que la base sea **ortonormal**, cada $\phi_n$ debe dividirse por su norma $\sqrt{(\phi_n, \phi_n)}$. 

Mediante este procedimiento se generan familias de polinomios fundamentales para la física matemática, como los de **Legendre** (intervalo $[-1, 1]$), **Hermite** (intervalo $(-\infty, \infty)$ con peso $e^{-x^2}$) y **Laguerre** (intervalo $[0, \infty)$ con peso $e^{-x}$).

# Construcción de la Base Ortogonal con Función de Peso

## Datos del Problema

**Conjunto de funciones base:**

$$\{f_n(x)\} = \{1, x, x^2, x^3,..., x^n,...\}_{n\to\infty}$$

**Dominio:**

$$-\infty \leq x \leq \infty$$

**Función de peso:**

$$p(x) = e^{-x^2}$$

**Fórmula de Gram-Schmidt:**

$$\mathbf{\phi_n(x) = f_n(x) - \sum_{m=1}^{n-1} \phi_m(x) \frac{(\phi_m, f_n)}{(\phi_m, \phi_m)}}$$

---

## Definición del Producto Interno con Peso

Cuando existe una función de peso $p(x)$, el producto interno se define como:

$$(g, h) = \int_{-\infty}^{\infty} g(x) \, h(x) \, p(x) \, dx = \int_{-\infty}^{\infty} g(x) \, h(x) \, e^{-x^2} \, dx$$

---

## Paso 1: Primer elemento ($n=1$)

**Función base:** $f_1(x) = 1$

Para $n=1$, la sumatoria es vacía (va de $m=1$ a $0$), por lo tanto:

$$\phi_1(x) = f_1(x)$$

$$\phi_1(x) = 1$$

**Primer elemento de la base:**

$$\phi_1(x) = 1$$

---

## Paso 2: Segundo elemento ($n=2$)

**Función base:** $f_2(x) = x$

Para $n=2$, la sumatoria tiene un solo término ($m=1$):

$$\phi_2(x) = f_2(x) - \phi_1(x) \frac{(\phi_1, f_2)}{(\phi_1, \phi_1)}$$

$$\phi_2(x) = x - 1 \cdot \frac{(1, x)}{(1, 1)}$$

**Cálculo del producto interno $(1, x)$:**

$$(1, x) = \int_{-\infty}^{\infty} 1 \cdot x \cdot e^{-x^2} \, dx$$

El integrando $x \cdot e^{-x^2}$ es una función impar (producto de función impar por función par). La integral sobre un dominio simétrico es cero:

$$(1, x) = 0$$

**Cálculo del producto interno $(1, 1)$:**

$$(1, 1) = \int_{-\infty}^{\infty} 1 \cdot 1 \cdot e^{-x^2} \, dx = \sqrt{\pi}$$

**Sustituyendo en la fórmula:**

$$\phi_2(x) = x - 1 \cdot \frac{0}{\sqrt{\pi}}$$

$$\phi_2(x) = x$$

**Segundo elemento de la base:**

$$\phi_2(x) = x$$

---

## Paso 3: Tercer elemento ($n=3$)

**Función base:** $f_3(x) = x^2$

Para $n=3$, la sumatoria tiene dos términos ($m=1$ y $m=2$):

$$\phi_3(x) = f_3(x) - \phi_1(x) \frac{(\phi_1, f_3)}{(\phi_1, \phi_1)} - \phi_2(x) \frac{(\phi_2, f_3)}{(\phi_2, \phi_2)}$$

$$\phi_3(x) = x^2 - 1 \cdot \frac{(1, x^2)}{(1, 1)} - x \cdot \frac{(x, x^2)}{(x, x)}$$

**Cálculo del producto interno $(1, x^2)$:**

$$(1, x^2) = \int_{-\infty}^{\infty} 1 \cdot x^2 \cdot e^{-x^2} \, dx$$

Usando la propiedad de las integrales gaussianas:

$$(1, x^2) = \frac{\sqrt{\pi}}{2}$$

**Cálculo del producto interno $(1, 1)$:**

$$(1, 1) = \sqrt{\pi}$$

**Cálculo del producto interno $(x, x^2)$:**

$$(x, x^2) = \int_{-\infty}^{\infty} x \cdot x^2 \cdot e^{-x^2} \, dx = \int_{-\infty}^{\infty} x^3 \cdot e^{-x^2} \, dx$$

El integrando $x^3 \cdot e^{-x^2}$ es una función impar, por lo tanto:

$$(x, x^2) = 0$$

**Cálculo del producto interno $(x, x)$:**

$$(x, x) = \int_{-\infty}^{\infty} x \cdot x \cdot e^{-x^2} \, dx = \int_{-\infty}^{\infty} x^2 \cdot e^{-x^2} \, dx = \frac{\sqrt{\pi}}{2}$$

**Sustituyendo en la fórmula:**

$$\phi_3(x) = x^2 - 1 \cdot \frac{\frac{\sqrt{\pi}}{2}}{\sqrt{\pi}} - x \cdot \frac{0}{\frac{\sqrt{\pi}}{2}}$$

$$\phi_3(x) = x^2 - \frac{1}{2} - 0$$

$$\phi_3(x) = x^2 - \frac{1}{2}$$

**Tercer elemento de la base:**

$$\phi_3(x) = x^2 - \frac{1}{2}$$

---

## Resultado Final

Los primeros tres elementos de la base ortogonal con función de peso $p(x) = e^{-x^2}$ son:

$$\phi_1(x) = 1$$

$$\phi_2(x) = x$$

$$\phi_3(x) = x^2 - \frac{1}{2}$$

---

## Observación

Estos polinomios son proporcionales a los **Polinomios de Hermite** clásicos. La relación con los polinomios de Hermite $H_n(x)$ es:

- $\phi_1(x) \propto H_0(x) = 1$
- $\phi_2(x) \propto H_1(x) = 2x$
- $\phi_3(x) \propto H_2(x) = 4x^2 - 2$

La función de peso $p(x) = e^{-x^2}$ es precisamente la peso natural asociada a los polinomios de Hermite en el dominio $(-\infty, \infty)$.
