# Ecuaciones importantes en la física

De acuerdo con la sección 10.1 del texto, las ecuaciones diferenciales parciales (EDP) más importantes en la física son lineales y de segundo orden. A continuación se presentan las fórmulas y descripciones de las EDP fundamentales identificadas en dicha sección:

**1. La ecuación de onda**
$$\nabla^2 u = \frac{1}{c^2} \frac{\partial^2 u}{\partial t^2}$$
Esta ecuación describe el desplazamiento desde el equilibrio, $u(\mathbf{r}, t)$, de una cuerda o membrana vibrante, así como de un sólido, gas o líquido en vibración. En electromagnetismo, se utiliza para describir componentes de campos eléctricos o magnéticos en ondas electromagnéticas, donde $c$ representa la velocidad de propagación de las ondas.

**2. La ecuación de difusión**
$$\kappa \nabla^2 u = \frac{\partial u}{\partial t}$$
Se utiliza para describir la temperatura $u$ en una región conductora térmica que no contiene fuentes ni sumideros de calor. También es aplicable para modelar la concentración de un producto químico en un proceso de difusión. La constante $\kappa$ se denomina difusividad.

**3. La ecuación de Laplace**
$$\nabla^2 u = 0$$
Esta ecuación se obtiene al establecer el estado estacionario en la ecuación de difusión (cuando la derivada temporal es cero). Describe distribuciones de temperatura en equilibrio, el potencial gravitacional en regiones sin materia y el potencial electrostático en regiones libres de carga. También describe el potencial de velocidad en el flujo de fluidos incompresibles sin fuentes ni vórtices.

**4. La ecuación de Poisson**
$$\nabla^2 u = \rho(\mathbf{r})$$
Es la generalización de la ecuación de Laplace para regiones que sí contienen materia, cargas o fuentes de calor/fluido. La función $\rho(\mathbf{r})$ representa la densidad de la fuente; por ejemplo, en electrostática se asocia con la densidad de carga eléctrica y en gravitación con la densidad de masa.

**5. La ecuación de Schrödinger**
$$-\frac{\hbar^2}{2m} \nabla^2 u + V(\mathbf{r})u = i\hbar \frac{\partial u}{\partial t}$$
Esta ecuación es fundamental en la mecánica cuántica y describe la función de onda $u(\mathbf{r}, t)$ de una partícula no relativista de masa $m$ bajo la influencia de un potencial $V(\mathbf{r})$. Al igual que la ecuación de difusión, es de segundo orden en las variables espaciales y de primer orden en el tiempo.

# Forma general de una Ecuación Diferencial Parcial de segundo orden en dos variables

La forma general de una ecuación diferencial parcial lineal de segundo orden en dos variables independientes ($x, y$) se define analíticamente de la siguiente manera:

$$A u_{xx} + B u_{xy} + C u_{yy} + D u_x + E u_y + F u = G(x, y)$$




## Ejemplo 1

$$xu_{xx} + yu_{yy} = 0$$


equivalente a la ecuación de Laplace en 2D $\nabla^2 u(x,y) = 0$

Identificamos los coeficientes de las segundas derivadas según la forma general de una ecuación diferencial parcial lineal:
*   $A = x$
*   $B = 0$
*   $C = y$

Calculamos el discriminante $\mathscr{D}$:
$\mathscr{D} = B^2 - 4AC$
$\mathscr{D} = 0 - 4(x)(y)$
$\mathscr{D} = -4xy$

**Análisis del dominio:**
Como los coeficientes $A$ y $C$ son funciones de las variables independientes, la clasificación de la ecuación es válida de manera local y puede cambiar de tipo dependiendo de la región del espacio analizada:

*   **Región hiperbólica** ($\mathscr{D} > 0$): Se presenta cuando $-4xy > 0$, lo que implica que el producto $xy < 0$. Esto ocurre analíticamente en el segundo y cuarto cuadrantes del plano.
*   **Región parabólica** ($\mathscr{D} = 0$): Se presenta cuando $-4xy = 0$, lo cual sucede si $x = 0$ o $y = 0$. Esta condición se cumple sobre los ejes coordenados.
*   **Región elíptica** ($\mathscr{D} < 0$): Se presenta cuando $-4xy < 0$, lo que implica que el producto $xy > 0$. Esto ocurre en el primer y tercer cuadrantes.
