Ejemplo de Bases discretas (enumerables -> $n\in\mathbb{Z}$)

Se muestran varios ejemplos de bases, con 1, 2 y 3 componentes. Ojo con el intervalo sobre el cual se mueve $n$

**Tarea**

Mostrar que la base c1 y d1 son iguales o equivalentes

### Ejemplos de funciones ortonormales discretas (pág. 173)

**c1)** Base trigonométrica en un intervalo de longitud $2L$:
$$\{\phi_n(x)\} = \left\{ \frac{1}{\sqrt{L}} \text{sen} \left(\frac{n\pi x}{L}\right), \frac{1}{\sqrt{L}} \cos \left(\frac{n\pi x}{L}\right), \frac{1}{\sqrt{2L}} \right\} \quad \text{para } c \leq x \leq c + 2L$$
*(Donde $n = 1, 2, 3, \dots$)*

**d1)** Base exponencial en un intervalo de longitud $2L$:
$$\{\phi_n(x)\} = \left\{ \frac{1}{\sqrt{2L}} e^{in\pi x/L} \right\} \quad \text{para } c \leq x \leq c + 2L$$
*(Donde $n = -\infty \dots \infty$)*


**Solución**

La equivalencia entre la **base trigonométrica (c1)** y la **base exponencial (d1)** radica en que ambas describen el mismo espacio de funciones en el intervalo $[c, c+2L]$ y pueden transformarse una en la otra mediante las **identidades de Euler**.

De acuerdo con las fuentes, el desarrollo de esta equivalencia se desglosa de la siguiente manera:

### 1. Definiciones de las bases (pág. 173)
*   **Base c1 (Trigonométrica):**
    $$\{\phi_n(x)\} = \left\{ \frac{1}{\sqrt{L}} \operatorname{sen} \left(\frac{n\pi x}{L}\right), \frac{1}{\sqrt{L}} \cos \left(\frac{n\pi x}{L}\right), \frac{1}{\sqrt{2L}} \right\} \quad (n = 1, 2, 3, \dots) \text{}$$
*   **Base d1 (Exponencial):**
    $$\{\phi_n(x)\} = \left\{ \frac{1}{\sqrt{2L}} e^{in\pi x/L} \right\} \quad (n = -\infty \dots \infty) \text{}$$

### 2. Transformación mediante identidades de Euler
Para demostrar la equivalencia, utilizamos las relaciones fundamentales entre funciones circulares y exponenciales complejas:
$$\cos \left(\frac{n\pi x}{L}\right) = \frac{e^{in\pi x/L} + e^{-in\pi x/L}}{2}$$
$$\operatorname{sen} \left(\frac{n\pi x}{L}\right) = \frac{e^{in\pi x/L} - e^{-in\pi x/L}}{2i}$$

Al sustituir estas expresiones en la expansión de una función $f(x)$ escrita en la base trigonométrica:
$$f(x) = \frac{a_0}{2} + \sum_{n=1}^{\infty} \left[ a_n \cos\left(\frac{n\pi x}{L}\right) + b_n \operatorname{sen}\left(\frac{n\pi x}{L}\right) \right] \text{}$$

Se obtiene, tras agrupar los términos de potencia positiva y negativa:
$$f(x) = \frac{a_0}{2} + \sum_{n=1}^{\infty} \left[ \frac{a_n - ib_n}{2} e^{in\pi x/L} + \frac{a_n + ib_n}{2} e^{-in\pi x/L} \right] \text{}$$

### 3. Conexión de coeficientes y sumatoria única
Esta última expresión se puede escribir de forma compacta como una única sumatoria que recorre todos los enteros desde $-\infty$ hasta $\infty$ (la base **d1**):
$$f(x) = \sum_{n=-\infty}^{\infty} C_n e^{in\pi x/L} \text{}$$

Donde los coeficientes se relacionan de la siguiente forma:
*   $C_0 = \frac{a_0}{2}$ (corresponde al término constante $\frac{1}{\sqrt{2L}}$).
*   $C_n = \frac{1}{2}(a_n - ib_n)$ para $n > 0$.
*   $C_{-n} = \frac{1}{2}(a_n + ib_n)$ para $n < 0$.

### Conclusión
Las dos bases son equivalentes porque **cualquier función expresable como una combinación lineal de senos y cosenos (c1) puede expresarse idénticamente como una combinación de exponenciales complejas (d1)** y viceversa. La base trigonométrica utiliza dos funciones reales por cada frecuencia $n$, mientras que la base exponencial utiliza una única función compleja pero duplica el rango de $n$ para cubrir las mismas componentes del espacio de Hilbert.

**Series de Fourier**
Usualmente la teoría de series de Fourier se introduce con la base trigonométrica **c1** introducida por Fourier en 1807

**Tarea**

Muestre lo siguiente:

$$(\tilde{\varphi}_{in},\tilde{\varphi}_{jm}) = \int_{c}^{c+2L} \tilde{\varphi}_{in}^*(x)\,\tilde{\varphi}_{jm}(x) dx = \delta_{ij}\delta_{nm}$$

Para demostrar la condición de ortonormalidad de la **base trigonométrica c1**, primero debemos definir sus elementos y el producto escalar en el espacio de Hilbert correspondiente, según las fuentes proporcionadas.

### 1. Definición de la Base c1 e Intervalo
La base **c1** se define en el intervalo $[c, c + 2L]$ como el conjunto de funciones:
*   $\tilde{\varphi}_{1n}(x) = \frac{1}{\sqrt{L}} \operatorname{sen} \left( \frac{n\pi x}{L} \right)$ para $n = 1, 2, 3, \dots$
*   $\tilde{\varphi}_{2n}(x) = \frac{1}{\sqrt{L}} \cos \left( \frac{n\pi x}{L} \right)$ para $n = 1, 2, 3, \dots$
*   $\tilde{\varphi}_{30}(x) = \frac{1}{\sqrt{2L}}$ (término constante)

El producto escalar para funciones reales en este espacio es:
$$(\tilde{\varphi}_{in}, \tilde{\varphi}_{jm}) = \int_{c}^{c+2L} \tilde{\varphi}_{in}(x) \tilde{\varphi}_{jm}(x) dx$$

---

### 2. Demostración de la Ortogonalidad ($\delta_{ij}\delta_{nm}$)

Para que se cumpla la identidad, el resultado de la integral debe ser **1** solo si el tipo de función ($i=j$) y el índice armónico ($n=m$) coinciden, y **0** en cualquier otro caso.

#### A. Ortogonalidad entre senos y cosenos ($i \neq j$)
Consideremos el producto de un seno y un coseno ($i=1, j=2$):
$$\frac{1}{L} \int_{c}^{c+2L} \operatorname{sen} \left( \frac{n\pi x}{L} \right) \cos \left( \frac{m\pi x}{L} \right) dx$$
Utilizando la identidad trigonométrica $\operatorname{sen} A \cos B = \frac{1}{2} [\operatorname{sen}(A+B) + \operatorname{sen}(A-B)]$, la integral se descompone en senos cuyas frecuencias son múltiplos de $\pi/L$. Dado que el intervalo de integración es de longitud **2L** (un período completo o múltiplo del mismo), la integral de cualquier función seno sobre este intervalo es **0**. Por lo tanto:
$$(\tilde{\varphi}_{1n}, \tilde{\varphi}_{2m}) = 0$$

#### B. Ortogonalidad entre armónicos distintos ($n \neq m$)
Para dos senos de distinto índice ($i=j=1$):
$$\frac{1}{L} \int_{c}^{c+2L} \operatorname{sen} \left( \frac{n\pi x}{L} \right) \operatorname{sen} \left( \frac{m\pi x}{L} \right) dx = \frac{1}{2L} \int_{c}^{c+2L} \left[ \cos \frac{(n-m)\pi x}{L} - \cos \frac{(n+m)\pi x}{L} \right] dx$$
Si $n \neq m$, ambos términos son funciones periódicas integradas sobre un múltiplo de su período, lo que resulta en **0**. El mismo razonamiento aplica para el producto de dos cosenos.

#### C. Ortogonalidad con el término constante
El producto de cualquier seno o coseno por $\tilde{\varphi}_{30} = 1/\sqrt{2L}$ implica integrar un seno o coseno sobre el intervalo $2L$:
$$\int_{c}^{c+2L} \frac{1}{\sqrt{2L}} \frac{1}{\sqrt{L}} \cos \left( \frac{n\pi x}{L} \right) dx = 0$$
Esto confirma que el término constante es ortogonal a todos los demás elementos de la base.

---

### 3. Demostración de la Normalización ($i=j$ y $n=m$)

Cuando las funciones son idénticas, debemos obtener **1**:

*   **Para senos ($i=j=1, n=m$):**
    $$\frac{1}{L} \int_{c}^{c+2L} \operatorname{sen}^2 \left( \frac{n\pi x}{L} \right) dx = \frac{1}{L} \int_{c}^{c+2L} \frac{1 - \cos(2n\pi x/L)}{2} dx = \frac{1}{2L} [x - 0]_c^{c+2L} = \frac{2L}{2L} = 1$$
*   **Para cosenos ($i=j=2, n=m$):**
    $$\frac{1}{L} \int_{c}^{c+2L} \cos^2 \left( \frac{n\pi x}{L} \right) dx = \frac{1}{L} \int_{c}^{c+2L} \frac{1 + \cos(2n\pi x/L)}{2} dx = 1$$
*   **Para la constante ($i=j=3$):**
    $$\int_{c}^{c+2L} \left( \frac{1}{\sqrt{2L}} \right)^2 dx = \frac{1}{2L} \int_{c}^{c+2L} dx = \frac{2L}{2L} = 1$$

### Conclusión
Se ha demostrado que la base **c1** satisface:
$$(\tilde{\varphi}_{in}, \tilde{\varphi}_{jm}) = \delta_{ij} \delta_{nm}$$
Donde el factor $\frac{1}{\sqrt{L}}$ normaliza las funciones oscilatorias y $\frac{1}{\sqrt{2L}}$ normaliza el nivel de DC (constante), asegurando que cada "eje" del espacio de funciones tenga **longitud unitaria**.

Para escribir una función $f(x)$ utilizando los elementos de la base ortonormal $\tilde{\varphi}_{in}(x)$ (referente a la base trigonométrica **c1**), se debe realizar una expansión en un espacio de Hilbert, tratando a la función como un vector que se descompone en sus componentes respecto a dicha base.

De acuerdo con las fuentes, la escritura de $f(x)$ se desglosa de la siguiente manera:

### 1. La combinación lineal (Serie de expansión)
Cualquier función $f(x)$ de cuadrado integrable en el intervalo $(c, c+2L)$ puede expresarse como una **combinación lineal única** de los elementos de la base $\tilde{\varphi}_{in}(x)$:

$$f(x) = \sum_{i} \sum_{n} C_{in} \tilde{\varphi}_{in}(x)$$

En esta expresión:
*   **$n$** representa el índice armónico ($1, 2, 3...$).
*   **$i$** identifica el tipo de función dentro de la base **c1** (seno, coseno o término constante).
*   **$C_{in}$** son los coeficientes de expansión, que representan las **componentes o coordenadas** del "vector" $f(x)$ en el espacio de funciones.

### 2. Cálculo de los coeficientes ($C_{in}$)
Para hallar cada componente $C_{in}$, se utiliza la propiedad de ortogonalidad mediante el producto escalar de la función original con cada elemento de la base:

$$C_{in} = \int_{c}^{c+2L} f(x) \tilde{\varphi}_{in}^*(x) dx$$

### 3. Forma explícita (Serie de Fourier)
Al sustituir los elementos específicos de la base **c1** proporcionados en la página 173, esta escritura toma la forma clásica de la **serie de Fourier** en un intervalo de longitud $2L$:

$$f(x) = \frac{a_0}{2} + \sum_{n=1}^{\infty} \left[ a_n \cos\left(\frac{n\pi x}{L}\right) + b_n \operatorname{sen}\left(\frac{n\pi x}{L}\right) \right]$$

Donde los coeficientes se corresponden con las proyecciones sobre los ejes unitarios de la base:
*   $a_n$ proviene de la proyección sobre las funciones coseno normalizadas.
*   $b_n$ proviene de la proyección sobre las funciones seno normalizadas.
*   $a_0/2$ corresponde a la proyección sobre el término constante $1/\sqrt{2L}$.

### 4. Condición de reconstrucción total
Si la base es **completa**, la suma de estas proyecciones reconstruye la función original sin pérdida de información, lo cual matemáticamente equivale a la aplicación de la **Identidad de la Delta de Dirac** dentro de la integral:

$$f(x) = \int_{c}^{c+2L} f(x') \left[ \sum_{i,n} \tilde{\varphi}_{in}^*(x') \tilde{\varphi}_{in}(x) \right] dx'$$

**Preguntar:** 

Condicion que debe tener una función para poder expandirse como una serie de fourier

Para que una función $f(x)$ pueda expandirse como una **serie de Fourier** en un intervalo determinado, las fuentes establecen que debe cumplir con las siguientes condiciones fundamentales de regularidad y comportamiento:

1.  **Integrabilidad:** La función debe ser de **cuadrado integrable** en el intervalo $(a, b)$ de definición. Esto significa que la integral del cuadrado de su módulo debe ser finita ($\int_a^b |f(x)|^2 dx < \infty$).
2.  **Continuidad por tramos:** La función debe ser **seccionalmente continua** o tener, como máximo, un **número finito de discontinuidades** dentro de su dominio.
3.  **Suavidad por pedazos:** Para garantizar la convergencia de la serie, se requiere que la función sea **suave por pedazos** (piecewise smooth) en el intervalo dado.
4.  **Comportamiento en las discontinuidades:** Si la función es suave por pedazos, la serie de Fourier resultante convergerá al valor de $f(x)$ en cada punto donde la función sea continua. En los puntos de discontinuidad, la serie converge al **valor medio del salto**, definido matemáticamente como $\frac{1}{2}[f(x^+) + f(x^-)]$.
5.  **Derivabilidad:** En casos de convergencia absoluta y uniforme, se exige que la función sea continua en el intervalo y que posea **derivadas continuas por secciones**.

Desde la perspectiva de los **espacios de Hilbert**, estas condiciones permiten tratar a la función como un "vector" que puede ser proyectado sobre una base ortogonal completa (como la base trigonométrica o la exponencial), garantizando que la suma infinita de dichas proyecciones reconstruya la función original sin pérdida de información.

Para demostrar cómo se anulan los términos de la sumatoria al calcular el coeficiente **$a_0$** bajo las condiciones específicas de **$c = 0$** y **$L = \pi$**, utilizaremos la expansión de la función en la base trigonométrica **c1**. 

En este caso, el intervalo de integración es **$[0, 2\pi]$** y la serie de Fourier se escribe como:
$$f(x) = \frac{a_0}{2} + \sum_{n=1}^{\infty} a_n \cos(nx) + \sum_{n=1}^{\infty} b_n \operatorname{sen}(nx)$$

El procedimiento matemático es el siguiente:

### 1. Integración de la serie en el periodo
Se integra la función $f(x)$ a ambos lados de la igualdad en el dominio completo de $0$ a $2\pi$:
$$\int_{0}^{2\pi} f(x) \, dx = \int_{0}^{2\pi} \left[ \frac{a_0}{2} + \sum_{n=1}^{\infty} a_n \cos(nx) + \sum_{n=1}^{\infty} b_n \operatorname{sen}(nx) \right] dx$$

Por la linealidad de la integral, la expresión se descompone en tres términos principales:
$$\int_{0}^{2\pi} f(x) \, dx = \int_{0}^{2\pi} \frac{a_0}{2} \, dx + \sum_{n=1}^{\infty} a_n \int_{0}^{2\pi} \cos(nx) \, dx + \sum_{n=1}^{\infty} b_n \int_{0}^{2\pi} \operatorname{sen}(nx) \, dx$$

### 2. Anulación de las integrales oscilatorias
Para cualquier valor entero de $n \geq 1$, las funciones seno y coseno completan un número exacto de ciclos dentro del intervalo $2\pi$, lo que hace que su área neta sea cero.

*   **Para los términos de $a_n$ (Cosenos):**
    $$\int_{0}^{2\pi} \cos(nx) \, dx = \left[ \frac{\operatorname{sen}(nx)}{n} \right]_0^{2\pi} = \frac{\operatorname{sen}(2n\pi) - \operatorname{sen}(0)}{n} = \frac{0 - 0}{n} = \mathbf{0}$$
*   **Para los términos de $b_n$ (Senos):**
    $$\int_{0}^{2\pi} \operatorname{sen}(nx) \, dx = \left[ -\frac{\cos(nx)}{n} \right]_0^{2\pi} = \frac{-\cos(2n\pi) + \cos(0)}{n} = \frac{-1 + 1}{n} = \mathbf{0}$$

Debido a estas propiedades de **ortogonalidad** respecto al término constante, todas las contribuciones armónicas desaparecen de la integral.

### 3. Aislamiento de $a_0$
Puesto que las sumatorias se anulan por completo, la ecuación original se reduce únicamente al componente de la base que representa el nivel de DC (constante):
$$\int_{0}^{2\pi} f(x) \, dx = \int_{0}^{2\pi} \frac{a_0}{2} \, dx = \frac{a_0}{2} \cdot [x]_0^{2\pi} = \frac{a_0}{2} \cdot 2\pi = a_0 \pi$$

Finalmente, despejando el coeficiente, obtenemos la fórmula estándar para el valor promedio en este intervalo:
$$\mathbf{a_0 = \frac{1}{\pi} \int_{0}^{2\pi} f(x) \, dx}$$

$$f(x)=
\begin{cases}
1, & 0 \le x < \pi,\\
2, & \pi \le x < 2\pi
\end{cases}
$$

Para hallar el coeficiente **$a_0$** de la serie de Fourier para la función dada en el intervalo $[0, 2\pi)$, seguimos el procedimiento establecido en las fuentes para una base trigonométrica en un dominio de longitud $2\pi$ con $c=0$.

### 1. Fórmula de $a_0$
De acuerdo con el formalismo de Fourier, para una función definida en el intervalo $(c, c + 2\pi)$, el coeficiente $a_0$ se obtiene mediante la integral:
$$a_0 = \frac{1}{\pi} \int_{c}^{c+2\pi} f(x) \, dx$$.

### 2. Aplicación a la función por tramos
Dado que $f(x)$ cambia su valor en $x = \pi$, dividimos la integral en dos partes siguiendo los límites de su definición:
$$a_0 = \frac{1}{\pi} \left[ \int_{0}^{\pi} (1) \, dx + \int_{\pi}^{2\pi} (2) \, dx \right]$$

### 3. Cálculo de las integrales
Realizamos la integración para cada tramo:
*   **Primer tramo ($0 \le x < \pi$):**
    $$\int_{0}^{\pi} 1 \, dx = [x]_0^\pi = \pi$$
*   **Segundo tramo ($\pi \le x < 2\pi$):**
    $$\int_{\pi}^{2\pi} 2 \, dx = [2x]_\pi^{2\pi} = 2(2\pi) - 2(\pi) = 4\pi - 2\pi = 2\pi$$

### 4. Resultado final
Sumamos los resultados y multiplicamos por el factor $1/\pi$:
$$a_0 = \frac{1}{\pi} [\pi + 2\pi] = \frac{3\pi}{\pi} = \mathbf{3}$$

**Nota conceptual:** En la expansión de la serie de Fourier, el término constante se escribe como $a_0/2$. En este caso, $3/2 = 1.5$, lo cual coincide con el **valor promedio** de la función en el intervalo total (el promedio entre 1 y 2).

Para calcular los coeficientes **$a_n$** y **$b_n$** de la serie de Fourier para la función dada en el intervalo $[0, 2\pi)$, utilizaremos las fórmulas generales para un sistema con $c=0$ y $L=\pi$ proporcionadas en las fuentes.

### 1. Cálculo de $a_n$ (Coeficientes de los cosenos)
La fórmula para obtener los coeficientes de los términos en coseno es:
$$a_n = \frac{1}{\pi} \int_{0}^{2\pi} f(x) \cos(nx) \, dx$$

Dividimos la integral según los tramos de la función:
$$a_n = \frac{1}{\pi} \left[ \int_{0}^{\pi} (1) \cos(nx) \, dx + \int_{\pi}^{2\pi} (2) \cos(nx) \, dx \right]$$

Realizando las integraciones:
*   **Primer tramo:** $\int_{0}^{\pi} \cos(nx) \, dx = \left[ \frac{\sin(nx)}{n} \right]_0^\pi = \frac{\sin(n\pi) - \sin(0)}{n} = 0$
*   **Segundo tramo:** $\int_{\pi}^{2\pi} 2 \cos(nx) \, dx = 2 \left[ \frac{\sin(nx)}{n} \right]_\pi^{2\pi} = 2 \frac{\sin(2n\pi) - \sin(n\pi)}{n} = 0$

Dado que $\sin(k\pi) = 0$ para cualquier entero $k$, el resultado es:
$$\mathbf{a_n = 0} \text{ para todo } n \geq 1$$

---

### 2. Cálculo de $b_n$ (Coeficientes de los senos)
La fórmula para los coeficientes de los términos en seno es:
$$b_n = \frac{1}{\pi} \int_{0}^{2\pi} f(x) \sin(nx) \, dx$$

Nuevamente, dividimos la integral:
$$b_n = \frac{1}{\pi} \left[ \int_{0}^{\pi} (1) \sin(nx) \, dx + \int_{\pi}^{2\pi} (2) \sin(nx) \, dx \right]$$

Utilizamos la primitiva $\int \sin(nx) dx = -\frac{\cos(nx)}{n}$:
$$b_n = \frac{1}{\pi} \left[ \left( -\frac{\cos(nx)}{n} \right)_0^\pi + 2 \left( -\frac{\cos(nx)}{n} \right)_\pi^{2\pi} \right]$$
$$b_n = \frac{1}{n\pi} \left[ (-\cos(n\pi) + \cos(0)) + 2(-\cos(2n\pi) + \cos(n\pi)) \right]$$

Sustituyendo los valores conocidos ($\cos(0)=1$, $\cos(2n\pi)=1$ y $\cos(n\pi)=(-1)^n$):
$$b_n = \frac{1}{n\pi} \left[ (1 - (-1)^n) + 2((-1)^n - 1) \right]$$
$$b_n = \frac{1}{n\pi} \left[ 1 - (-1)^n + 2(-1)^n - 2 \right]$$
$$\mathbf{b_n = \frac{(-1)^n - 1}{n\pi}}$$

---

### 3. Análisis de $b_n$
Observamos el comportamiento de $b_n$ según la paridad de $n$:
*   **Si $n$ es par ($n = 2, 4, 6...$):** $(-1)^n = 1$, por lo tanto $b_n = \frac{1 - 1}{n\pi} = \mathbf{0}$.
*   **Si $n$ es impar ($n = 1, 3, 5...$):** $(-1)^n = -1$, por lo tanto $b_n = \frac{-1 - 1}{n\pi} = \mathbf{-\frac{2}{n\pi}}$.

### Resultado Final
La serie de Fourier para esta función es:
$$f(x) = \frac{3}{2} - \frac{2}{\pi} \sum_{n \text{ impar}}^{\infty} \frac{\sin(nx)}{n}$$

Este resultado es consistente con la expansión obtenida previamente mediante la base exponencial, confirmando que el valor promedio es $1.5$ ($a_0/2$) y que la función se reconstruye mediante una superposición de ondas senoidales de frecuencias impares.

**Ejercicio:**

$$f(x) = |x|, -\pi < x < \pi$$

Para representar la función **$f(x) = |x|$** en el intervalo $(-\pi, \pi)$ mediante una serie de Fourier, debemos identificar primero sus propiedades de simetría y luego calcular sus componentes (coeficientes) basándonos en la teoría de espacios de Hilbert.

### 1. Análisis de Simetría
La función $f(x) = |x|$ es una **función par**, ya que cumple con la condición $f(x) = f(-x)$ para todo punto en su dominio. Según las fuentes, esto simplifica significativamente la expansión:
*   En el dominio $(-\pi, \pi)$, las funciones pares tienen una expansión **solo en cosenos**.
*   Esto implica que los coeficientes de los senos son nulos: **$b_n = 0$**.

### 2. Cálculo de los Coeficientes
Utilizamos la base trigonométrica ortonormal en el intervalo $(c, c+2\pi)$ con $c = -\pi$, lo que nos da el intervalo $(-\pi, \pi)$.

#### A. Coeficiente $a_0$ (Término constante)
El coeficiente $a_0$ representa el valor promedio de la función multiplicado por un factor de escala. Para una función par:
$$a_0 = \frac{1}{\pi} \int_{-\pi}^{\pi} |x| dx = \frac{2}{\pi} \int_{0}^{\pi} x dx$$
$$a_0 = \frac{2}{\pi} \left[ \frac{x^2}{2} \right]_0^\pi = \frac{2}{\pi} \cdot \frac{\pi^2}{2} = \pi$$
En la serie, este término aparece como **$a_0/2 = \pi/2$**.

#### B. Coeficientes $a_n$ (Términos en coseno)
Para $n = 1, 2, 3, \dots$, aplicamos la fórmula de proyección sobre la base de cosenos:
$$a_n = \frac{1}{\pi} \int_{-\pi}^{\pi} |x| \cos(nx) dx = \frac{2}{\pi} \int_{0}^{\pi} x \cos(nx) dx$$
Resolviendo por integración por partes, obtenemos el resultado general para los coeficientes:
$$a_n = \frac{2}{\pi n^2} \left( \cos(n\pi) - 1 \right) = \mathbf{\frac{2}{\pi n^2} \left( (-1)^n - 1 \right)}$$

### 3. Análisis de los resultados de $a_n$
Dependiendo de si $n$ es par o impar, el comportamiento de los coeficientes varía:
*   **Si $n$ es par:** $(-1)^n = 1$, por lo tanto **$a_n = 0$**. Esto indica que las frecuencias pares no contribuyen a la formación de la función.
*   **Si $n$ es impar:** $(-1)^n = -1$, por lo tanto **$a_n = -\frac{4}{\pi n^2}$**.

### 4. Representación Final
Al reconstruir la función como una combinación lineal de los elementos de la base, obtenemos la serie de Fourier para el valor absoluto en el intervalo dado:

$$\mathbf{|x| = \frac{\pi}{2} + \frac{2}{\pi} \sum_{n=1}^{\infty} \frac{(-1)^n - 1}{n^2} \cos(nx)}$$

O, escribiendo explícitamente solo los términos impares:
$$\mathbf{|x| = \frac{\pi}{2} - \frac{4}{\pi} \left( \frac{\cos(x)}{1^2} + \frac{\cos(3x)}{3^2} + \frac{\cos(5x)}{5^2} + \dots \right)}$$

Esta serie converge al valor de la función en cada punto de continuidad. Además, debido a que $f(x)$ es continua en todo el intervalo y posee derivadas continuas por secciones, la serie converge a $|x|$ de manera **absoluta y uniforme**.

**Ejercicio:**

$$f(x) = |x|, -\pi < x < \pi$$

Para representar la función $f(x) = x$ en el intervalo $(-\pi, \pi)$ mediante una serie de Fourier, seguimos el procedimiento para funciones con simetría definida en un espacio de Hilbert.

### 1. Análisis de Simetría
La función $f(x) = x$ es una **función impar** (impar), ya que cumple con $f(-x) = -x = -f(x)$. Según las fuentes, esto simplifica el cálculo de la siguiente manera:
*   En el dominio $(-\pi, \pi)$, las funciones impares tienen una expansión que consiste **solo en términos de seno**.
*   Los coeficientes de los cosenos son nulos: **$a_0 = 0$** y **$a_n = 0$** para todo $n$.

### 2. Cálculo de los Coeficientes $b_n$
Para una función definida en $(c, c + 2\pi)$ con $c = -\pi$, el coeficiente $b_n$ se obtiene mediante la proyección de la función sobre la base de senos:
$$b_n = \frac{1}{\pi} \int_{-\pi}^{\pi} x \operatorname{sen}(nx) dx$$

Al ser $f(x)$ y $\operatorname{sen}(nx)$ ambas funciones impares, su producto es una función par, lo que permite simplificar la integral al semi-dominio:
$$b_n = \frac{2}{\pi} \int_{0}^{\pi} x \operatorname{sen}(nx) dx$$

Resolviendo la integral (por partes), las fuentes indican que el resultado es:
$$b_n = -\frac{2}{n} \cos(n\pi)$$
Dado que $\cos(n\pi) = (-1)^n$, podemos escribir el coeficiente como:
$$b_n = \frac{2}{n} (-1)^{n+1}$$

### 3. Representación Final de la Serie
Al reconstruir la función sumando sus componentes sobre la base ortogonal, obtenemos la serie de Fourier para $f(x) = x$:

$$\mathbf{x = \sum_{n=1}^{\infty} \frac{2(-1)^{n+1}}{n} \operatorname{sen}(nx)}$$

Escribiendo los primeros términos de la serie para visualizar la superposición de armónicos:
$$\mathbf{x = 2 \left( \operatorname{sen} x - \frac{1}{2} \operatorname{sen} 2x + \frac{1}{3} \operatorname{sen} 3x - \frac{1}{4} \operatorname{sen} 4x + \dots \right)}$$

### 4. Convergencia
De acuerdo con el teorema de convergencia para funciones suaves por pedazos, esta serie converge al valor de la función ($x$) en cada punto del intervalo abierto $(-\pi, \pi)$. En los puntos de discontinuidad del periodo ($x = \pi$ y $x = -\pi$), la serie convergerá al valor medio del salto, que en este caso es $\frac{1}{2}[\pi + (-\pi)] = 0$.

# Base continua (No enumerable) $\{\varphi(k, x)\}$

**Ejemplo:**

* Base de Fourier discreta: $\{\varphi_n(x)\} = \left\{\sqrt{\frac{1}{2L}}e^{in\pi x/L}\right\}\quad -L\leq x \leq L$

De acuerdo con las fuentes, la **condición de ortonormalidad en el continuo** representa una extensión del formalismo discreto (donde se usa la delta de Kronecker) hacia espacios de funciones con un índice no enumerable.

Para un conjunto de funciones $\{\phi(k, x)\}$, donde tanto la variable espacial $x$ como el índice $k$ son continuos, la condición de ortonormalidad en un intervalo $a \leq x \leq b$ se define matemáticamente como:

$$\mathbf{\int_{a}^{b} \phi^*(k, x) \phi(k', x) dx = \delta(k - k')}$$

### Puntos clave de esta condición:
*   **Sustitución de la Delta:** En el límite del continuo, la **delta de Kronecker** ($\delta_{nn'}$), que indica ortogonalidad entre índices enteros, se transforma en la **delta de Dirac** ($\delta(k - k')$).
*   **Dominios de Aplicación:** Mientras que las bases discretas se asocian típicamente a intervalos finitos, las **bases continuas** son fundamentales para realizar expansiones de funciones en **dominios infinitos o semi-infinitos**.
*   **Ejemplo Fundamental (Base de Fourier):** Para la base continua de Fourier $\{ \frac{1}{\sqrt{2\pi}} e^{ikx} \}$, la condición se expresa a través de la representación integral de la delta de Dirac:
    $$\frac{1}{2\pi} \int_{-\infty}^{\infty} e^{i(k-k')x} dx = \delta(k - k')$$

Esta propiedad garantiza que cada "eje" en el espacio de Hilbert continuo sea independiente, permitiendo que cualquier función de cuadrado integrable pueda expandirse mediante una **integral de Fourier** en lugar de una sumatoria.

La **condición de completez en el continuo** representa la extensión del formalismo de bases discretas a espacios de funciones con índices no enumerables, permitiendo la reconstrucción de una función mediante una integral en lugar de una sumatoria.

### Expresión Analítica General
Para un conjunto de funciones $\{\phi(k, x)\}$, donde $x$ es la variable espacial en el intervalo $(a, b)$ y $k$ es el índice continuo en el intervalo $(c, d)$, la condición de completez se expresa analíticamente como:

$$\mathbf{\int_{c}^{d} \phi^*(k, x') \phi(k, x) dk = \delta(x - x')}$$

### Puntos clave de la expresión:
*   **Significado Matemático:** Esta identidad indica que el conjunto de funciones es lo suficientemente "rico" como para representar cualquier función de cuadrado integrable en el dominio dado.
*   **Relación con la Delta de Dirac:** La expresión es, por definición, una **representación integral de la delta de Dirac**. Mientras que en el caso discreto la completez se expresa como una suma de productos que iguala a la delta, en el continuo se requiere la integración sobre todo el espectro del parámetro $k$.
*   **Variables Conjugadas:** En este formalismo, existe una simetría entre la condición de ortonormalidad y la de completez. Las variables $k$ y $x$ actúan como **variables conjugadas** (como la posición y el momento, o la frecuencia y el tiempo), permitiendo que ambas condiciones sean simétricas bajo el intercambio de dichas variables.
*   **Reconstrucción de Funciones:** Esta condición garantiza que una función $f(x)$ pueda expandirse mediante su **transformada** $C(k)$ de la siguiente manera:
    $$f(x) = \int_{c}^{d} C(k) \phi(k, x) dk$$

En conclusión, la completez en el continuo asegura que la superposición de los "ejes" continuos del espacio de Hilbert (las funciones de la base) es capaz de mapear y reconstruir perfectamente cualquier "vector" (función) del espacio.

**Tarea**
Demuestre la condición de completez para la base continua de fourier

Para demostrar la **condición de completez** para la base continua de Fourier, debemos probar que la superposición integral de sus funciones es capaz de reconstruir una función arbitraria $f(x)$, lo que equivale matemáticamente a una representación de la **delta de Dirac**.

A continuación se detalla el desarrollo lógico basado en las fuentes:

### 1. Definición de la base y la expansión
Para un dominio infinito $(-\infty, \infty)$, la base continua de Fourier está compuesta por el conjunto de funciones ortonormales $\{\phi(k, x)\} = \left\{ \frac{1}{\sqrt{2\pi}} e^{ikx} \right\}$. Cualquier función $f(x)$ de cuadrado integrable en este espacio de Hilbert se puede expresar como una combinación lineal (integral) de estos elementos:
$$f(x) = \int_{-\infty}^{\infty} C(k) \phi(k, x) \, dk \text{}$$

### 2. Obtención de los coeficientes (Transformada)
Para hallar los coeficientes $C(k)$, que representan las "componentes" del vector $f(x)$, multiplicamos la expansión por el complejo conjugado de la base y utilizamos la propiedad de ortonormalidad:
$$C(k) = \int_{-\infty}^{\infty} f(x') \phi^*(k, x') \, dx' \text{}$$

### 3. Sustitución y reordenamiento
Al sustituir la expresión de $C(k)$ dentro de la integral original de $f(x)$, obtenemos:
$$f(x) = \int_{-\infty}^{\infty} \left[ \int_{-\infty}^{\infty} f(x') \phi^*(k, x') \, dx' \right] \phi(k, x) \, dk \text{}$$

Cambiando el orden de las integraciones, la ecuación se reescribe como:
$$f(x) = \int_{-\infty}^{\infty} f(x') \left[ \int_{-\infty}^{\infty} \phi^*(k, x') \phi(k, x) \, dk \right] dx' \text{}$$

### 4. Identificación con la Delta de Dirac
Por definición, la función delta de Dirac $\delta(x - x')$ es la única que satisface la propiedad selectiva integral $f(x) = \int_{-\infty}^{\infty} f(x') \delta(x - x') \, dx'$. Al comparar ambas expresiones, se deduce necesariamente que el término entre corchetes es la condición de completez:
$$\int_{-\infty}^{\infty} \phi^*(k, x') \phi(k, x) \, dk = \delta(x - x') \text{}$$

### 5. Resultado final para la base de Fourier
Sustituyendo los valores específicos de la base continua de Fourier ($\frac{1}{\sqrt{2\pi}} e^{ikx}$), llegamos a la demostración final:
$$\frac{1}{2\pi} \int_{-\infty}^{\infty} e^{-ikx'} e^{ikx} \, dk = \frac{1}{2\pi} \int_{-\infty}^{\infty} e^{ik(x - x')} \, dk = \delta(x - x') \text{}$$

**Conclusión:** Esta identidad demuestra que la base es **completa**, ya que la suma de todas las proyecciones de la función sobre la base (integrando sobre todo el espectro de $k$) permite reconstruir la función original en su totalidad.

Efectivamente, **$C(k)$ es la transformada de Fourier**, aunque existe un error en el valor que mencionas para la función delta de Dirac. A continuación, aclaro ambos puntos basándome en las fuentes:

### 1. ¿Es $C(k)$ la transformada de Fourier?
**Sí.** Según las fuentes, cuando se trabaja con una base continua $\{\phi(k, x)\}$, cualquier función $f(x)$ de cuadrado integrable puede considerarse un "vector" y expandirse mediante una integral en lugar de una sumatoria. 

En esta expansión:
$$f(x) = \int_{c}^{d} C(k) \phi(k, x) dk$$
El coeficiente **$C(k)$** representa las "componentes" o coordenadas de la función en ese espacio de Hilbert continuo y se conoce específicamente como la **transformada de Fourier** cuando la base utilizada es la exponencial $\{\frac{1}{\sqrt{2\pi}} e^{ikx}\}$.

### 2. La transformada de Fourier para $f(x) = \delta(x)$
De acuerdo con el formalismo matemático de las fuentes, tu cálculo de $\sqrt{2\pi}$ no es correcto para la definición estándar utilizada en el texto. El valor real es **$1/\sqrt{2\pi}$**.

**Demostración:**
Para obtener la transformada $F(k)$ de la función delta de Dirac $\delta(x)$, aplicamos la fórmula de proyección (integral de la transformada):
$$F(k) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} f(x) e^{-ikx} dx$$

Sustituyendo $f(x) = \delta(x)$:
$$F(k) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \delta(x) e^{-ikx} dx$$

Utilizando la **propiedad selectiva** de la delta de Dirac, que establece que $\int f(x)\delta(x-x_0)dx = f(x_0)$, la integral evalúa la función exponencial en $x=0$:
$$F(k) = \frac{1}{\sqrt{2\pi}} \cdot e^{-ik(0)} = \frac{1}{\sqrt{2\pi}} \cdot 1 = \mathbf{\frac{1}{\sqrt{2\pi}}}$$

**Conclusión:**
*   **$C(k)$ (o $F(k)$)** es efectivamente la transformada de Fourier.
*   La transformada de $f(x) = \delta(x)$ es **$1/\sqrt{2\pi}$**, no $\sqrt{2\pi}$.
*   Como nota adicional, la fuentes indican que **$\delta(x)$ es la transformada de Fourier de $1/\sqrt{2\pi}$**, lo que confirma la relación inversa.
