# Peso

En el contexto de los **espacios de Hilbert** y las bases ortogonales, el **peso** (o función de peso), denotado comúnmente como $p(x)$, es un factor real que se introduce para extender la condición de ortogonalidad de un conjunto de funciones.

A continuación se explica su rol y propiedades según lo expuesto en la página 169 y siguientes del texto:

### 1. Definición en el Producto Escalar
El peso es una función real $p(x)$ que se incluye dentro de la integral que define el producto interno entre dos funciones $f(x)$ y $g(x)$ en un intervalo $[a, b]$. Matemáticamente, se dice que una base de funciones $\{\phi_n(x)\}$ es **ortogonal de peso $p(x)$** si cumple:
$$(p\phi_n, \phi_m) \equiv \int_a^b p(x)\phi_n^*(x)\phi_m(x) dx = A_n \delta_{nm}$$
Donde:
*   $A_n$ es una constante relacionada con la norma.
*   $\delta_{nm}$ es la delta de Kronecker (1 si $n=m$, 0 si $n \neq m$).

### 2. Importancia y Propiedades
*   **Garantía de Realidad:** Se exige que $p(x)$ sea una **función real** para garantizar que el producto escalar $(p\phi_n, \phi_n)$ sea un número real.
*   **Relación con la Normalización:** Si un conjunto de funciones $\{\phi_n(x)\}$ es ortogonal con un peso $p(x)$, entonces el conjunto compuesto por $\{\sqrt{p(x)}\phi_n(x)\}$ será ortogonal con peso 1.
*   **Condición de Completez:** En una base con peso, la relación de completez (que permite representar cualquier función como una serie) también debe incluir este factor:
$$\sum_n p(x')\phi_n^*(x')\phi_n(x) = \delta(x - x')$$

### 3. Conexión con las Ecuaciones Diferenciales
Aunque se introduce en el estudio de bases, el peso tiene un origen profundo en la **teoría de Sturm-Liouville**. El factor $p(x)$ es exactamente el mismo factor por el cual se debe multiplicar una ecuación diferencial lineal para convertirla en una **ecuación autoadjunta**. 

Para que los autovalores sean reales y las autofunciones sean ortogonales, se requiere generalmente que $p(x)$ **no cambie de signo** en el intervalo de integración.

**Ejemplo:** Base de los polinomios de Laguerre

$$\{L_n(x)\}_n$$

Con peso $e^{x}$

En el marco de las **bases ortogonales**, los polinomios de Laguerre $\{L_n(x)\}$ constituyen un sistema fundamental bajo las siguientes especificaciones técnicas extraídas de las fuentes:

### 1. Definición y Peso
Los polinomios de Laguerre forman una base **ortogonal** en el intervalo **$(0, \infty)$** utilizando como factor de peso la función **$p(x) = e^{-x}$**.

### 2. Condición de Ortonormalidad
A diferencia de otras bases, el conjunto $\{L_n(x)\}$ (con la normalización estándar $L_n(0)=1$) satisface una condición de **ortonormalidad**, lo que significa que su producto escalar es igual a la delta de Kronecker sin factores adicionales:
$$\int_{0}^{\infty} e^{-x} L_n(x) L_m(x) \, dx = \delta_{nm}$$.

### 3. Representación Matemática
*   **Fórmula de Rodrigues:** Permite generar cualquier polinomio de la base mediante derivación:
    $$L_n(x) = \frac{e^x}{n!} \frac{d^n}{dx^n} (x^n e^{-x})$$.
*   **Expansión en serie:** Su forma polinomial explícita es:
    $$L_n(x) = \sum_{p=0}^{n} \frac{(-1)^p (n!)^2}{(p!)^2 (n-p)!} x^p$$.

### 4. Primeros Elementos de la Base
Los vectores iniciales de este espacio de funciones son:
*   $L_0(x) = 1$
*   $L_1(x) = 1 - x$
*   $L_2(x) = 1 - 2x + \frac{x^2}{2}$
*   $L_3(x) = 1 - 3x + \frac{3x^2}{2} - \frac{x^3}{6}$.

### 5. Aplicación y Completez
Esta base es **completa**, lo que garantiza que cualquier función de cuadrado integrable en el dominio $(0, \infty)$ puede expandirse como una serie de Laguerre: $f(x) = \sum_{n=0}^{\infty} c_n L_n(x)$. Físicamente, esta estructura es esencial para describir la parte radial de las funciones de onda en el **átomo de hidrógeno**.

# Completez

La **completez** es una propiedad fundamental de las bases en un espacio de Hilbert que garantiza que el conjunto de funciones es suficiente para representar cualquier "vector" (función) del espacio sin pérdida de información.

De acuerdo con las fuentes, la definición se desglosa en los siguientes puntos:

### 1. Definición Conceptual
Un conjunto de funciones $\{\phi_n(x)\}$ es **completo** si cualquier función $f(x)$ que sea de cuadrado integrable en un intervalo $(a, b)$ puede expresarse como una **combinación lineal única** de los elementos de esa base:
$$f(x) = \sum_{n} C_n \phi_n(x)$$

### 2. Condición de Completez (Relación de Clausura)
Matemáticamente, la completez se expresa mediante una identidad que involucra a la **Delta de Dirac**. Para una base discreta con una función de peso $p(x)$, la condición es:
$$\sum_{n} p(x') \phi_n^*(x') \phi_n(x) = \delta(x - x')$$
Esta ecuación asegura que la suma de los productos de los elementos de la base actúa como la unidad dentro del espacio, permitiendo la reconstrucción de cualquier función.

### 3. Teorema de Completez (Identidad de Parseval)
Un conjunto de funciones se considera completo si y solo si, para cualquier función de cuadrado integrable $f(x)$, se cumple la **Identidad de Parseval**. Esta establece que la norma al cuadrado de la función es igual a la suma de los cuadrados de sus componentes (coeficientes de expansión):
$$\int_a^b |f(x)|^2 dx = \sum_{n} |C_n|^2$$

### 4. Completez en Bases Continuas
Para conjuntos no enumerables de funciones (como en la Transformada de Fourier), donde el índice $k$ es una variable continua, la condición de completez se define mediante una integral en lugar de una suma:
$$\int_c^d \phi^*(k, x') \phi(k, x) dk = \delta(x - x')$$

En resumen, la completez es el requisito que permite que las funciones especiales (como las de Fourier, Bessel o Legendre) funcionen como un sistema de ejes coordenados completo en espacios de infinitas dimensiones.

Para hallar los coeficientes de expansión $C_n$ (también llamados las "componentes" del vector función) dentro de un **espacio de Hilbert**, se utiliza la propiedad de **ortogonalidad** de la base.

De acuerdo con las fuentes, el procedimiento sistemático es el siguiente:

### 1. Representación de la función
Cualquier función $f(x)$ de cuadrado integrable en un intervalo $(a, b)$ puede expresarse como una combinación lineal de las funciones de una base completa $\{\phi_n(x)\}$:
$$f(x) = \sum_{n} C_n \phi_n(x)$$

### 2. Multiplicación por la función base y el peso
Para aislar un coeficiente específico $C_m$, se multiplica toda la ecuación anterior por el complejo conjugado de una función de la base $\phi_m^*(x)$ y por la **función de peso** $p(x)$ del sistema:
$$p(x) f(x) \phi_m^*(x) = \sum_{n} C_n p(x) \phi_m^*(x) \phi_n(x)$$

### 3. Integración en el intervalo
Se integra a ambos lados de la igualdad respecto a $x$ en el dominio definido $(a, b)$:
$$\int_a^b p(x) f(x) \phi_m^*(x) dx = \sum_{n} C_n \int_a^b p(x) \phi_m^*(x) \phi_n(x) dx$$

### 4. Aplicación de la condición de ortogonalidad
Por definición, en una base ortonormal, la integral del lado derecho cumple con la **Delta de Kronecker** ($\delta_{mn}$), la cual vale 1 solo si $n=m$ y 0 en cualquier otro caso. Esto hace que la sumatoria "colapse" y solo sobreviva el término donde los índices coinciden:
$$\int_a^b p(x) f(x) \phi_m^*(x) dx = C_m$$

### 5. Expresión final
Por lo tanto, la fórmula general para hallar cada coeficiente $C_n$ es la integral del producto de la función original por la base conjugada y el peso:
$$\mathbf{C_n = \int_a^b p(x') f(x') \phi_n^*(x') dx'}$$

**Nota sobre la normalización:** Si la base es solo ortogonal pero no está normalizada a la unidad, el coeficiente se divide por el factor de normalización $A_n = \int_a^b p(x)|\phi_n(x)|^2 dx$.

### 1. La fórmula de expansión
Si un conjunto de funciones $\{\phi_n(x)\}$ constituye una base **completa** y ortonormal en un intervalo $(a, b)$, cualquier función $f(x)$ de cuadrado integrable puede expresarse como:
$$f(x) = \sum_{n} C_n \phi_n(x)$$

Donde los coeficientes $C_n$ (las "componentes" del vector función) se calculan mediante la integral del producto de la función por el conjugado de la base y una función de peso $p(x')$:
$$C_n = \int_a^b p(x') f(x') \phi_n^*(x') dx'$$

Al sustituir el coeficiente en la sumatoria, obtenemos la estructura que iniciaste en tu consulta:
$$\mathbf{f(x) = \sum_{n} \left( \int_a^b p(x') f(x') \phi_n^*(x') dx' \right) \phi_n(x)}$$



$$\int_{a}^{b} f(x') \left[ \sum_n p(x') \phi_n^*(x') \phi_n(x) \right] dx' = f(x)$$

Esta igualdad se fundamenta en los siguientes puntos extraídos de las fuentes:

*   **La Relación de Clausura:** El término entre corchetes, $\sum_n p(x') \phi_n^*(x') \phi_n(x)$, representa la **propiedad de completez** de la base $\{\phi_n(x)\}$.
*   **Representación de la Delta de Dirac:** En una base completa, esta sumatoria se comporta exactamente como una **función Delta de Dirac**, $\delta(x - x')$, la cual tiene la propiedad selectiva de extraer el valor de la función en el punto $x$ cuando se integra.
*   **Componentes del "Vector" Función:** Esta estructura surge al sustituir la definición de los coeficientes de expansión $C_n$ (las componentes del vector función) dentro de la sumatoria original $f(x) = \sum C_n \phi_n(x)$.
*   **Función de Peso:** El factor $p(x')$ es la **función de peso** real necesaria para definir la ortogonalidad en el sistema específico (como en las bases de Bessel o Laguerre).


*Ejemplo:* Usar la base $\{\varphi_n(x)\} = \left\{\sqrt{\frac{1}{2L}}e^{in\pi x/L}\right\}$ para representar la funición:

$$f(x)=
\begin{cases}
1, & 0 \le x < \pi,\\
2, & \pi \le x < 2\pi
\end{cases}
$$

Para representar la función $f(x)$ en la base exponencial de Fourier proporcionada, debemos tratar a la función como un **vector dentro de un espacio de Hilbert** y calcular sus componentes o coordenadas respecto a los ejes definidos por la base $\{\varphi_n(x)\}$.

### 1. Identificación de los parámetros del sistema
La base dada es $\{\varphi_n(x)\} = \left\{\sqrt{\frac{1}{2L}}e^{in\pi x/L}\right\}$. Para este problema, la función $f(x)$ está definida en el intervalo $0 \le x < 2\pi$, por lo que la longitud total del intervalo es $2L = 2\pi$, lo que implica que **$L = \pi$**. Sustituyendo este valor, nuestra base ortonormal específica para este espacio es:
$$\varphi_n(x) = \frac{1}{\sqrt{2\pi}}e^{inx}$$
Esta base es **completa y ortonormal** en el intervalo $(0, 2\pi)$, lo que garantiza que cualquier función de cuadrado integrable en ese dominio puede representarse unívocamente como una combinación lineal de estos elementos.

### 2. Planteamiento de la expansión
La representación de la función consiste en escribir $f(x)$ como una suma infinita (serie) de los elementos de la base multiplicados por coeficientes $C_n$:
$$f(x) = \sum_{n=-\infty}^{\infty} C_n \varphi_n(x)$$.
De acuerdo con el formalismo de los espacios de funciones, los coeficientes $C_n$ se obtienen mediante el **producto escalar** de la función original con el complejo conjugado de cada elemento de la base:
$$C_n = ( \varphi_n, f ) = \int_{0}^{2\pi} f(x) \varphi_n^*(x) \, dx$$.

### 3. Cálculo de los coeficientes $C_n$
Sustituyendo la definición de nuestra base y la función por tramos, la integral se divide en dos partes:
$$C_n = \frac{1}{\sqrt{2\pi}} \left[ \int_{0}^{\pi} (1) e^{-inx} \, dx + \int_{\pi}^{2\pi} (2) e^{-inx} \, dx \right]$$

#### Caso $n = 0$:
Para el término constante (componente de frecuencia cero), la exponencial se convierte en la unidad ($e^0 = 1$):
$$C_0 = \frac{1}{\sqrt{2\pi}} \left[ \int_{0}^{\pi} 1 \, dx + \int_{\pi}^{2\pi} 2 \, dx \right] = \frac{1}{\sqrt{2\pi}} [ \pi + 2\pi ] = \frac{3\pi}{\sqrt{2\pi}} = 3\sqrt{\frac{\pi}{2}}$$.

#### Caso $n \neq 0$:
Resolvemos las integrales exponenciales utilizando la primitiva $\int e^{ax} dx = \frac{1}{a} e^{ax}$:
$$C_n = \frac{1}{\sqrt{2\pi}} \left[ \left( \frac{e^{-inx}}{-in} \right)_0^\pi + 2 \left( \frac{e^{-inx}}{-in} \right)_\pi^{2\pi} \right]$$
$$C_n = \frac{1}{\sqrt{2\pi} (-in)} \left[ (e^{-in\pi} - 1) + 2(e^{-i2n\pi} - e^{-in\pi}) \right]$$

Sabemos por las identidades de Euler que $e^{-i2n\pi} = 1$ y $e^{-in\pi} = (-1)^n$:
$$C_n = \frac{1}{-in\sqrt{2\pi}} \left[ (-1)^n - 1 + 2(1 - (-1)^n) \right]$$
$$C_n = \frac{1}{-in\sqrt{2\pi}} \left[ 1 - (-1)^n \right] = \frac{(-1)^n - 1}{in\sqrt{2\pi}}$$.

### 4. Análisis de los resultados
Observamos el comportamiento de $C_n$ según la paridad de $n$:
*   Si **$n$ es par**: $(-1)^n = 1$, por lo tanto $C_n = 0$. Esto indica que estos "ejes" del espacio de Hilbert no contribuyen a la formación de la función.
*   Si **$n$ es impar**: $(-1)^n = -1$, por lo tanto $C_n = \frac{-2}{in\sqrt{2\pi}} = \frac{2i}{n\sqrt{2\pi}}$.

### 5. Representación final de la función
Al reconstruir la función sumando todas sus proyecciones sobre la base, obtenemos:
$$f(x) = C_0 \varphi_0(x) + \sum_{n \neq 0} C_n \varphi_n(x)$$
$$f(x) = \left( 3\sqrt{\frac{\pi}{2}} \right) \frac{1}{\sqrt{2\pi}} + \sum_{n \text{ impar}} \left( \frac{-2}{in\sqrt{2\pi}} \right) \frac{1}{\sqrt{2\pi}} e^{inx}$$

Simplificando los términos constantes:
$$\mathbf{f(x) = \frac{3}{2} + \sum_{n \text{ impar}} \frac{2i}{n \cdot 2\pi} e^{inx} = \frac{3}{2} + \frac{i}{\pi} \sum_{n \text{ impar}} \frac{1}{n} e^{inx}}$$

Esta expresión final representa el "vector" función como una **superposición infinita de oscilaciones complejas**, donde el término $3/2$ es el valor promedio de la función en el intervalo y los términos sumados reconstruyen los saltos bruscos entre 1 y 2.
