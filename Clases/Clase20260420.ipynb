### 1. Definición del problema
Consideramos un sistema de longitud total $2a$ ($x \in [-a, a]$) formado por dos barras idénticas unidas en $x=0$.
*   **Ecuación de difusión:** $\partial_x^2 T(x, t) = \frac{1}{\kappa}\partial_t T(x, t)$.
*   **Condiciones de frontera (Neumann):** Extremos aislados en $\pm a$: $\partial_x T(\pm a, t) = 0$.
*   **Condición inicial:** $T(x, 0) = \begin{cases} 0 & -a \leq x < 0 \\ 100 & 0 < x \leq a \end{cases}$.

### 2. Separación de variables
Definimos la temperatura como la suma del estado estacionario (el promedio de $50^\circ\text{C}$ que alcanzará el sistema aislado) y una parte transitoria $u(x, t)$:
$$T(x, t) = u(x, t) + 50$$
Proponemos $u(x, t) = \sum_n A_n(x)B_n(t)$. Al sustituir en la ecuación de difusión obtenemos:
*   **Parte temporal:** $B_n(t) = e^{-\kappa \lambda_n^2 t}$.
*   **Parte espacial:** $A_n(x) = c_1 e^{i\lambda_n x} + c_2 e^{-i\lambda_n x}$.

### 3. Aplicación analítica de las fronteras (Forma Coseno)
Para satisfacer el aislamiento en el extremo derecho ($x=a$), imponemos $A'_n(a) = 0$:
$$i\lambda_n (c_1 e^{i\lambda_n a} - c_2 e^{-i\lambda_n a}) = 0 \implies c_2 = c_1 e^{2i\lambda_n a}$$
Sustituyendo $c_2$ en $A_n(x)$ y factorizando $e^{i\lambda_n a}$:
$$A_n(x) = c_1 e^{i\lambda_n a} (e^{i\lambda_n(x-a)} + e^{-i\lambda_n(x-a)}) = \mathbf{c_n \cos(\lambda_n(x-a))}$$

Para determinar **$\lambda_n$**, aplicamos la condición en el extremo izquierdo ($x=-a$):
$$A'_n(-a) = -c_n \lambda_n \sin(\lambda_n(-a-a)) = c_n \lambda_n \sin(2\lambda_n a) = 0$$
Para soluciones no triviales, $\sin(2\lambda_n a) = 0$, lo que implica $2\lambda_n a = n\pi$:
$$\mathbf{\lambda_n = \frac{n\pi}{2a}} \quad \text{con } n = 0, 1, 2, \dots.$$

### 4. Selección de modos por paridad de la condición inicial
La condición inicial para la parte transitoria $u(x, 0) = T(x, 0) - 50$ es:
$$u(x, 0) = \begin{cases} -50 & -a \leq x < 0 \\ 50 & 0 < x \leq a \end{cases}$$
Esta función es **impar** ($u(-x, 0) = -u(x, 0)$). Expandimos el coseno desplazado usando identidades:
$$c_n \cos\left( \frac{n\pi x}{2a} - \frac{n\pi}{2} \right) = c_n \left[ \cos\left(\frac{n\pi x}{2a}\right)\cos\left(\frac{n\pi}{2}\right) + \sin\left(\frac{n\pi x}{2a}\right)\sin\left(\frac{n\pi}{2}\right) \right]$$
*   Para que la función sea impar, el término en coseno debe anularse: $\cos(n\pi/2) = 0$.
*   Esto ocurre únicamente cuando **$n$ es un número impar** ($n = 1, 3, 5, \dots$).
*   Para estos valores, $\sin(n\pi/2) = (-1)^{(n-1)/2}$. Por tanto, los modos espaciales son proporcionales a **$\sin(\frac{n\pi x}{2a})$**.

### 5. Cálculo de los coeficientes $c_n$
Evaluamos $u(x, 0) = 50$ en el intervalo $[0, a]$ mediante la serie de Fourier de senos:
$$c_n = \frac{2}{a} \int_{0}^{a} 50 \sin\left( \frac{n\pi x}{2a} \right) dx = \frac{100}{a} \left[ -\frac{2a}{n\pi} \cos\left( \frac{n\pi x}{2a} \right) \right]_0^a$$
Como $n$ es impar, $\cos(n\pi/2) = 0$ y $\cos(0) = 1$:
$$c_n = \frac{100}{a} \left[ 0 - \left( -\frac{2a}{n\pi} \right) \right] = \mathbf{\frac{200}{n\pi}}.$$

### 6. Solución final completa
Reemplazando $n$ por $2m+1$ para asegurar que solo se sumen términos impares:
$$\mathbf{T(x, t) = 50 + \frac{200}{\pi} \sum_{m=0}^{\infty} \frac{1}{2m+1} \sin\left( \frac{(2m+1)\pi x}{2a} \right) \exp\left( -\frac{(2m+1)^2 \pi^2 \kappa t}{4a^2} \right)}.$$

**Validación:**
1.  **En $t=0$:** La serie de Fourier reconstruye el salto térmico de $0^\circ\text{C}$ a $100^\circ\text{C}$.
2.  **En $x = \pm a$:** La derivada es proporcional a $\cos(\frac{(2m+1)\pi}{2}) = 0$, cumpliendo el aislamiento.
3.  **En $t \to \infty$:** Los términos exponenciales decaen a cero, resultando en una temperatura uniforme de $50^\circ\text{C}$ en toda la barra.

Para resolver el problema **4.a** del Taller 5, que consiste en hallar los autovalores ($\lambda_n$) y autofunciones ($y_n$) de la ecuación diferencial **$x^2 y'' + 5xy' + \lambda y = 0$** sujeta a las condiciones de frontera **$y(1) = 0$** y **$y(e) = 0$** en el intervalo $[1, e]$, seguimos el procedimiento analítico detallado en las fuentes:

### 1. Resolución de la ecuación diferencial
La ecuación dada es de tipo **Euler**. Para resolverla, proponemos una solución de la forma $y = x^m$. Sustituyendo en la ecuación, obtenemos la ecuación característica:
$$m(m-1) + 5m + \lambda = 0 \implies \mathbf{m^2 + 4m + \lambda = 0} \text{}$$

Las raíces de esta ecuación son:
$$m = \frac{-4 \pm \sqrt{16 - 4\lambda}}{2} = \mathbf{-2 \pm \sqrt{4 - \lambda}} \text{}$$

### 2. Aplicación de las condiciones de frontera
La solución general toma la forma:
$$y(x) = \frac{1}{x^2} [A x^{\sqrt{4-\lambda}} + B x^{-\sqrt{4-\lambda}}] \text{}$$

1.  **Condición $y(1) = 0$:**
    $$y(1) = 1 [A(1) + B(1)] = 0 \implies \mathbf{B = -A} \text{}$$
    Sustituyendo esto, la solución se reduce a:
    $$y(x) = \frac{A}{x^2} [x^{\sqrt{4-\lambda}} - x^{-\sqrt{4-\lambda}}] \text{}$$

2.  **Condición $y(e) = 0$:**
    Para evitar la solución trivial ($y=0$), debemos analizar el término en el corchete evaluado en $x=e$:
    *   Si $4 - \lambda \ge 0$, no se encuentran soluciones no triviales que cumplan la condición en $e$.
    *   Si **$4 - \lambda < 0$**, podemos escribir $\sqrt{4-\lambda} = i\sqrt{\lambda-4}$. Usando la identidad $e^{i\theta} - e^{-i\theta} = 2i\sin\theta$ y sabiendo que $x = e^{\ln x}$, la solución se transforma en:
        $$y(x) = \frac{2iA}{x^2} \sin(\sqrt{\lambda-4} \ln x) \text{}$$
        Evaluando en $x=e$ (donde $\ln e = 1$):
        $$\sin(\sqrt{\lambda-4}) = 0 \implies \mathbf{\sqrt{\lambda - 4} = n\pi} \quad \text{con } n = 1, 2, 3, \dots \text{}$$

### 3. Resultados finales
*   **Autovalores ($\lambda_n$):** Despejando $\lambda$ de la condición del seno:
    $$\mathbf{\lambda_n = 4 + n^2\pi^2} \text{}$$
*   **Autofunciones ($y_n$):** Sustituyendo los valores permitidos de $\lambda$:
    $$\mathbf{y_n(x) = \frac{1}{x^2} \sin(n\pi \ln x)} \text{}$$

**Nota sobre la ortogonalidad:** Este problema es un caso hermítico de Sturm-Liouville con una función de peso **$p(x) = x^3$**. Las autofunciones obtenidas son ortogonales en el intervalo $[1, e]$ respecto a dicho peso. El conjunto ortonormalizado de peso 1 es **$\{\sqrt{2x} \sin(n\pi \ln x)\}$**.

En el desarrollo del problema **4.a** de las fuentes (Taller 5), el paso donde se utiliza la identidad **$x = e^{\ln x}$** ocurre al transformar la solución de una ecuación de Euler de su forma de potencias a una forma trigonométrica para aplicar las condiciones de frontera.

El proceso detallado es el siguiente:

1.  **Forma intermedia de la solución:** Tras aplicar la primera condición de frontera ($y(1)=0$), la solución se escribe en términos de potencias con exponentes complejos (para el caso $\lambda > 4$):
    $$y(x) = \frac{a}{x^2} \left( x^{i\sqrt{\lambda-4}} - x^{-i\sqrt{\lambda-4}} \right) \text{}$$

2.  **Sustitución explícita:** Para poder utilizar la identidad de Euler y convertir la expresión en una función seno, se reemplaza la base $x$ por su equivalente exponencial:
    $$x = \mathbf{e^{\ln x}} \text{}$$

3.  **Transformación de los términos:** Aplicando la sustitución anterior y sabiendo que por la segunda condición de frontera ($y(e)=0$) se cumple que $\sqrt{\lambda - 4} = n\pi$:
    *   $x^{in\pi} \rightarrow (e^{\ln x})^{in\pi} = \mathbf{e^{in\pi \ln x}}$
    *   $x^{-in\pi} \rightarrow (e^{\ln x})^{-in\pi} = \mathbf{e^{-in\pi \ln x}}$

4.  **Resultado final en senos:** Al sustituir estos términos en la ecuación original, se obtiene:
    $$y(x) = \frac{a}{x^2} \left( e^{in\pi \ln x} - e^{-in\pi \ln x} \right) = \frac{a}{x^2} 2i \sin(n\pi \ln x) \text{}$$

Este paso es fundamental porque permite expresar la autofunción en una forma que satisface naturalmente los ceros de la función en el intervalo logarítmico $[1, e]$.

Para resolver el problema **4.d** del Taller 5, que consiste en hallar los autovalores ($\lambda_n$) y autofunciones ($y_n$) de la ecuación diferencial **$(3 + x)^2 y'' + 2(3 + x)y' + \lambda y = 0$** sujeta a las condiciones de frontera **$y(-2) = 0$** y **$y(1) = 0$** en el intervalo $[-2, 1]$, seguimos este procedimiento analítico:

### 1. Transformación de la ecuación
La ecuación dada es de tipo **Euler** desplazada. Para simplificarla, realizamos el cambio de variable **$z = 3 + x$**.
*   Si $x = -2 \implies z = 1$.
*   Si $x = 1 \implies z = 4$.
*   Las derivadas se mantienen igual ya que $dz/dx = 1$.

La ecuación se transforma en:
$$z^2 \frac{d^2y}{dz^2} + 2z \frac{dy}{dz} + \lambda y = 0, \quad \text{con } y(1) = 0 \text{ y } y(4) = 0$$

### 2. Resolución de la ecuación característica
Proponemos una solución de la forma $y = z^m$. Al sustituir en la ecuación de Euler, obtenemos la ecuación indicial:
$$m(m-1) + 2m + \lambda = 0 \implies \mathbf{m^2 + m + \lambda = 0}$$

Las raíces para $m$ son:
$$m = \frac{-1 \pm \sqrt{1 - 4\lambda}}{2} = \mathbf{-\frac{1}{2} \pm \sqrt{\frac{1}{4} - \lambda}}$$

### 3. Aplicación de las condiciones de frontera
Para obtener soluciones no triviales que satisfagan dos ceros en el intervalo, requerimos que las raíces sean complejas, es decir, **$\lambda > 1/4$**. Definimos $\omega = \sqrt{\lambda - 1/4}$. La solución general es:
$$y(z) = \frac{1}{\sqrt{z}} \left[ A \cos(\omega \ln z) + B \sin(\omega \ln z) \right]$$

1.  **Condición $y(z=1) = 0$:**
    $$y(1) = \frac{1}{1} [A \cos(0) + B \sin(0)] = 0 \implies \mathbf{A = 0}$$
    La solución se reduce a $y(z) = \frac{B}{\sqrt{z}} \sin(\omega \ln z)$.

2.  **Condición $y(z=4) = 0$:**
    Para que $B \neq 0$, el argumento del seno debe ser un múltiplo de $\pi$:
    $$\omega \ln 4 = n\pi \implies \mathbf{\omega_n = \frac{n\pi}{\ln 4}} \quad \text{con } n = 1, 2, 3, \dots$$

### 4. Resultados finales
*   **Autovalores ($\lambda_n$):** Partiendo de $\omega^2 = \lambda - 1/4$ y sabiendo que $\ln 4 = 2 \ln 2$:
    $$\lambda_n - \frac{1}{4} = \left( \frac{n\pi}{2 \ln 2} \right)^2 \implies \mathbf{\lambda_n = \frac{1}{4} + \frac{n^2\pi^2}{4(\ln 2)^2}}$$
*   **Autofunciones ($y_n(x)$):** Regresando a la variable original $x$:
    $$\mathbf{y_n(x) = \frac{1}{\sqrt{3+x}} \sin\left( \frac{n\pi \ln(3+x)}{2 \ln 2} \right)}$$

**Nota sobre la ortogonalidad:** Esta ecuación es autoadjunta con una función de peso **$\rho(x) = 1$**. Las autofunciones son ortogonales en el intervalo $[-2, 1]$ respecto a este peso.

Esta expresión es la solución general para una **ecuación diferencial de Euler** de segundo orden cuando las raíces de su ecuación característica son complejas conjugadas. El desarrollo analítico para llegar a esta forma, basado en las fuentes proporcionadas, es el siguiente:

### 1. Identificación de la ecuación diferencial
La expresión dada surge de una ecuación del tipo:
$$z^2 y'' + a_1 z y' + a_2 y = 0$$
En el contexto de tus fuentes (específicamente el problema 4.d del Taller 5), esta estructura aparece tras realizar un cambio de variable ($z = 3+x$) en una ecuación de Euler desplazada. La ecuación resultante es:
$$\mathbf{z^2 \frac{d^2y}{dz^2} + 2z \frac{dy}{dz} + \lambda y = 0} \text{}$$

### 2. Ecuación característica e indicial
Siguiendo el método de Frobenius o la propuesta de solución de tipo potencia $y = z^m$:
1. Sustituimos $y, y', y''$ en la ecuación diferencial.
2. Obtenemos la ecuación característica (indicial):
   $$m(m-1) + 2m + \lambda = 0 \implies \mathbf{m^2 + m + \lambda = 0} \text{}$$
3. Resolvemos para $m$ usando la fórmula cuadrática:
   $$m = \frac{-1 \pm \sqrt{1 - 4\lambda}}{2} = \mathbf{-\frac{1}{2} \pm \sqrt{\frac{1}{4} - \lambda}} \text{}$$

### 3. Caso de raíces complejas ($\lambda > 1/4$)
Para obtener una solución oscilatoria (necesaria para cumplir condiciones de frontera con ceros), el discriminante debe ser negativo. Definimos la frecuencia angular espacial **$\omega$** como:
$$\omega = \sqrt{\lambda - \frac{1}{4}}$$
Esto nos da las raíces complejas conjugadas: $m = -1/2 \pm i\omega$.

### 4. Transformación de potencias a senos y cosenos
La solución general en términos de potencias es:
$$y(z) = c_1 z^{-1/2 + i\omega} + c_2 z^{-1/2 - i\omega}$$
Factorizando el término real:
$$y(z) = z^{-1/2} \left[ c_1 z^{i\omega} + c_2 z^{-i\omega} \right]$$

Para convertir las potencias complejas, utilizamos la identidad fundamental **$z = e^{\ln z}$**:
$$y(z) = \frac{1}{\sqrt{z}} \left[ c_1 e^{i\omega \ln z} + c_2 e^{-i\omega \ln z} \right]$$

Finalmente, aplicando la **Identidad de Euler** ($e^{i\theta} = \cos \theta + i \sin \theta$) y redefiniendo las constantes arbitrarias $A$ y $B$ como combinaciones de $c_1$ y $c_2$:
$$\mathbf{y(z) = \frac{1}{\sqrt{z}} \left[ A \cos(\omega \ln z) + B \sin(\omega \ln z) \right]}$$

**Conclusión:** Esta forma es el resultado de representar analíticamente la parte imaginaria del exponente de una función de Euler mediante funciones trigonométricas, utilizando el logaritmo natural para mapear la variable independiente $z$ al dominio de fase.
