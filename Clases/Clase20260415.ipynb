# Teoría de Sturm Loiuville

La **Teoría de Sturm-Liouville** constituye la "arquitectura común" de todas las ecuaciones diferenciales ordinarias (EDO) lineales de segundo orden y es una herramienta fundamental en física matemática. Su propósito principal es demostrar que, bajo condiciones de frontera apropiadas, estas ecuaciones exhiben un conjunto infinito de **soluciones ortogonales** que funcionan como bases en un espacio de Hilbert.

### 1. La Ecuación de Sturm-Liouville
Cualquier EDO lineal de segundo orden con coeficientes reales puede expresarse en la forma estándar de Sturm-Liouville:

$$\frac{d}{dx} \left[ p(x) \frac{dy}{dx} \right] + q(x)y + \lambda \rho(x)y = 0$$

Donde:
*   **$p(x), q(x)$ y $\rho(x)$** son funciones reales de la posición.
*   **$\rho(x)$** (o $w(x)$) se denomina **función de peso** y debe ser positiva en el intervalo de integración.
*   **$\lambda$** es el **autovalor** asociado a la solución o autofunción $y(x)$.

### 2. Propiedades de los Operadores Hermíticos
Cuando el operador de Sturm-Liouville es **hermítico** (lo cual depende tanto de la forma de la ecuación como de que las condiciones de frontera anulen ciertos términos en los extremos), se garantizan tres propiedades críticas:
1.  **Autovalores reales:** Los valores de $\lambda$ asociados a fenómenos físicos (como frecuencias naturales o energías) son siempre números reales.
2.  **Ortogonalidad:** Las autofunciones correspondientes a autovalores distintos son ortogonales entre sí con respecto a la función de peso $\rho(x)$.
3.  **Completitud:** Las autofunciones forman una **base completa**, lo que significa que cualquier función "bien comportada" puede expandirse como una serie de estas autofunciones.

### 3. Condiciones de Frontera
Para que el problema sea hermítico, se deben imponer condiciones de frontera homogéneas en los extremos del intervalo $[a, b]$, tales como:
*   **Dirichlet:** La función se anula en los extremos ($y(a) = y(b) = 0$).
*   **Neumann:** La derivada se anula en los extremos ($y'(a) = y'(b) = 0$).
*   **Mixtas:** Una combinación lineal de la función y su derivada se anula.

### 4. Ejemplos y Funciones Especiales
La mayoría de las funciones especiales utilizadas en física surgen de problemas de Sturm-Liouville al separar variables en ecuaciones como las de Laplace, Helmholtz o Schrödinger. Algunos ejemplos clásicos incluyen:
*   **Polinomios de Legendre:** Surgen en problemas con simetría esférica.
*   **Funciones de Bessel:** Aparecen en sistemas con simetría cilíndrica.
*   **Polinomios de Hermite:** Describen el oscilador armónico cuántico.
*   **Polinomios de Laguerre:** Fundamentales para describir el átomo de hidrógeno.

Sea $\mathscr{L}$ el operador diferencial 

$$\mathscr{L} f(x) = a_2 \frac{d^2}{dx^2} f + a_1\frac{d}{dx} + a_0 f$$

que aplica sobre f(x) en el dominio $a\leq x \leq b$.

En el espacio de Hilbert el producto interno entre dos funciones $f(x)$ y $g(x)$ en el intervalo $[a, b]$ se define como:

$$(f, g) = \int_a^b f^*(x) g(x) \rho(x) dx$$

donde $f^*(x)$ representa el complejo conjugado de $f(x)$ y **$\rho(x)$** (o $w(x)$) es una **función de peso** real, continua y no negativa en el dominio de integración.

Para completar el análisis del operador $\mathscr{L}$ en este espacio, se deben considerar los siguientes puntos:

*   **Propiedad de Hermiticidad:** Para que el operador $\mathscr{L}$ sea considerado **hermítico** en este espacio, el producto interno debe cumplir que $(f, \mathscr{L}g) = (\mathscr{L}f, g)$. Esto garantiza que sus autovalores sean reales y sus autofunciones ortogonales.
*   **Forma de Sturm-Liouville:** Si el operador no es autoadjunto originalmente (es decir, si $a_1 \neq a_2'$), se puede transformar a la forma estándar de Sturm-Liouville multiplicando toda la ecuación por un factor integrante o función de peso $w(x) = \frac{1}{a_2} \exp\left(\int \frac{a_1}{a_2} dx\right)$.
*   **Condiciones de Frontera:** La definición completa del producto interno para este operador requiere que las funciones $f$ y $g$ pertenezcan a un subespacio que satisfaga **condiciones de frontera homogéneas** (como Dirichlet o Neumann), asegurando que los términos de borde en la integración por partes se anulen.

Para escribir la expresión completa y aplicar la **integración por partes** al producto interno $(\mathscr{L}f, g)$, partimos de la definición del producto interno en un espacio de Hilbert sobre el intervalo $[a, b]$ y la forma del operador diferencial de segundo orden proporcionada:

### 1. Expresión completa del producto interno
La expresión se escribe como la integral del producto de la función resultante de operar sobre $f$ (tomando su complejo conjugado) por la función $g$:

**$$(\mathscr{L}f, g) = \int_a^b (\mathscr{L}f)^*(x) g(x) dx$$**

Sustituyendo el operador $\mathscr{L} f = a_2 f'' + a_1 f' + a_0 f$:
**$$(\mathscr{L}f, g) = \int_a^b [a_2^*(x) (f'')^*(x) + a_1^*(x) (f')^*(x) + a_0^*(x) f^*(x)] g(x) dx$$**

---

### 2. Aplicación de integración por partes
Para transferir las operaciones de derivación de $f^*$ a $g$, se aplica integración por partes término a término:

*   **Para el término de segundo orden ($a_2^*$):** Se integra dos veces por partes.
    $$\int_a^b g a_2^* (f'')^* dx = [g a_2^* (f')^*]_a^b - \int_a^b (g a_2^*)' (f')^* dx$$
    $$\int_a^b g a_2^* (f'')^* dx = [g a_2^* (f')^* - (g a_2^*)' f^*]_a^b + \int_a^b f^* (g a_2^*)'' dx$$

*   **Para el término de primer orden ($a_1^*$):** Se integra una vez por partes.
    $$\int_a^b g a_1^* (f')^* dx = [g a_1^* f^*]_a^b - \int_a^b f^* (g a_1^*)' dx$$

---

### 3. Resultado Final
Al agrupar los términos bajo la integral y los términos evaluados en la frontera, obtenemos:

**$$(\mathscr{L}f, g) = \int_a^b f^* [D^2(a_2^* g) - D(a_1^* g) + a_0^* g] dx + [g a_2^* (f')^* - f^* (a_2^* g)' + a_1^* g f^*]_a^b$$**

Esta expresión permite identificar dos componentes fundamentales:
1.  **El operador adjunto ($\mathscr{L}^\dagger$):** El corchete dentro de la integral define a $\mathscr{L}^\dagger g = D^2(a_2^* g) - D(a_1^* g) + a_0^* g$.
2.  **El término de frontera (Concomitante):** Se puede rescribir utilizando el Wronskiano ($W$) para mayor simplicidad:
    **$$(\mathscr{L}f, g) = (f, \mathscr{L}^\dagger g) + [-a_2^* W(f^*, g) + f^* g (a_1^* - (a_2^*)')]^b_a$$**

Para que el operador sea **hermítico**, el término entre corchetes evaluado en los límites $a$ y $b$ debe anularse, lo cual depende de las condiciones de frontera impuestas al problema.

El **operador adjunto** $\overline{\mathscr{L}}$ de un operador diferencial lineal $\mathscr{L}$ se define a través de una relación integral (producto interno) entre dos funciones arbitrarias $f(x)$ y $g(x)$ en un intervalo $[a, b]$. El objetivo de este operador es transferir todas las operaciones de derivación que actúan sobre la función $f$ hacia la función $g$ mediante integraciones por partes sucesivas.

Para un operador diferencial de **segundo orden** de la forma $\mathscr{L} = a_2 D^2 + a_1 D + a_0$, su adjunto $\overline{\mathscr{L}}$ se define como:
**$$\overline{\mathscr{L}}(\cdot) = D^2(a_2^* \cdot) - D(a_1^* \cdot) + a_0^* \cdot$$**.

Las características principales de esta definición son:

*   **Relación fundamental:** Se establece mediante la identidad $(\mathscr{L}f, g) = (f, \overline{\mathscr{L}}g) + \text{términos de frontera}$. 
*   **Términos de frontera:** El desarrollo genera un residuo evaluado en los límites del intervalo, conocido como el **concomitante bilineal**, que para el segundo orden es $[ -a_2^* W(f^*, g) + f^*g(a_1^* - \dot{a}_2^*) ]_a^b$, donde $W$ representa el Wronskiano.
*   **Generalización:** Para un operador de **orden $p$** dado por $\mathscr{L} = \sum_{k=0}^p a_k D^k$, el operador adjunto se define de forma general como:
    **$$\overline{\mathscr{L}}(\cdot) = \sum_{k=0}^p (-1)^k D^k(a_k^* \cdot)$$**.
*   **Autoadjunto:** Se dice que un operador es **autoadjunto** si coincide exactamente con su adjunto ($\mathscr{L} = \overline{\mathscr{L}}$). Si además las condiciones de frontera hacen que los términos de borde se anulen, el operador se denomina **hermítico**.

Si las funciones $a_i(x)$ son **reales**, se derivan las siguientes conclusiones fundamentales para el operador diferencial de segundo orden $\mathscr{L} = a_2 D^2 + a_1 D + a_0$ y su relación con el operador adjunto $\overline{\mathscr{L}}$:

### 1. Definición del operador adjunto
El operador adjunto $\overline{\mathscr{L}}$ se simplifica al no requerir el complejo conjugado de los coeficientes, quedando definido como:
**$$\overline{\mathscr{L}}(\cdot) = D^2(a_2 \cdot) - D(a_1 \cdot) + a_0 \cdot$$**.

Al desarrollar las derivadas por la regla del producto, este operador puede expresarse como:
$$\overline{\mathscr{L}}g = a_2 g'' + (2a'_2 - a_1)g' + (a''_2 - a'_1 + a_0)g$$.

### 2. Condición para que sea autoadjunto
Para que el operador sea **autoadjunto** ($\mathscr{L} = \overline{\mathscr{L}}$), es necesario y suficiente que se cumpla la relación:
**$$a_1 = a'_2$$**.

Si esta condición se satisface, el operador puede escribirse en la forma compacta característica de la teoría de Sturm-Liouville:
$$\mathscr{L}(\cdot) = \frac{d}{dx} \left[ a_2 \frac{d}{dx} (\cdot) \right] + a_0(\cdot)$$.

### 3. Transformación a la forma autoadjunta
Si el operador original no es autoadjunto (es decir, $a_1 \neq a'_2$), siempre es posible convertirlo en uno multiplicándolo por un **factor integrante** $p(x)$ definido como:
**$$p(x) = \frac{1}{a_2} \exp\left( \int \frac{a_1}{a_2} dx \right)$$**.

### 4. Propiedad de Hermiticidad
Bajo el supuesto de coeficientes reales, el operador será **hermítico** si, además de ser autoadjunto, las funciones sobre las que actúa satisfacen condiciones de frontera tales que el **concomitante bilineal** se anule:
$$[a_2 W(f, g)]^b_a = 0$$.
Donde $W(f, g)$ es el Wronskiano de las funciones. Esto garantiza que los **autovalores sean reales** y las **autofunciones sean ortogonales**.

# Ejemplo 1

Operador diferencial de Legendre.

$(1-x^2)f'' - 2xf' + l(l+1)f = 0$

El **operador diferencial de Legendre** es uno de los ejemplos más importantes de la teoría de Sturm-Liouville en física matemática, ya que surge naturalmente al resolver la ecuación de Laplace en coordenadas esféricas para problemas con simetría axial.

A continuación se presenta el análisis detallado del operador basado en las fuentes:

### 1. Identificación de coeficientes
Dada la ecuación diferencial:
$$(1-x^2)f''(x) - 2xf'(x) + l(l+1)f(x) = 0$$

Podemos identificar los coeficientes de la forma general $\mathscr{L} f = a_2 f'' + a_1 f' + a_0 f$ como:
*   **$a_2(x) = 1 - x^2$**
*   **$a_1(x) = -2x$**
*   **$a_0(x) = l(l+1)$** (donde $l$ es una constante, usualmente un entero)

### 2. Condición de autoadjunción
Para determinar si el operador es autoadjunto, verificamos si se cumple la relación $a'_2(x) = a_1(x)$:
$$\frac{d}{dx}(1 - x^2) = -2x$$

Como la derivada de $a_2$ es exactamente igual a $a_1$, se concluye que **el operador de Legendre es intrínsecamente autoadjunto**. No requiere de un factor integrante adicional para adquirir esta propiedad.

### 3. Forma de Sturm-Liouville
Debido a su carácter autoadjunto, la ecuación puede escribirse directamente en la forma compacta de Sturm-Liouville $\frac{d}{dx}[p(x)f'(x)] + [q(x) + \lambda \rho(x)]f(x) = 0$:

**$$\frac{d}{dx} \left[ (1 - x^2) \frac{df}{dx} \right] + l(l+1)f = 0$$**

En esta estructura:
*   **$p(x) = 1 - x^2$**.
*   **$q(x) = 0$**.
*   **$\rho(x) = 1$** (función de peso unitaria).
*   **$\lambda = l(l+1)$** (autovalor).

### 4. Dominio y Singularidades
*   **Intervalo natural:** El dominio de interés suele ser **$x \in [-1, 1]$**, donde $x$ representa el coseno del ángulo polar ($\cos \theta$).
*   **Puntos singulares:** Los extremos $x = 1$ y $x = -1$ son **puntos singulares regulares** de la ecuación. En estos puntos, la función $p(x)$ se anula ($1 - (\pm 1)^2 = 0$), lo que define el intervalo de ortogonalidad natural del operador.

### 5. Autofunciones: Polinomios de Legendre
Las soluciones que permanecen finitas y regulares en los puntos singulares ($x = \pm 1$) existen solo cuando **$l$ es un entero no negativo** ($l = 0, 1, 2, \dots$). Estas soluciones son los **Polinomios de Legendre**, denotados como $P_l(x)$.

Debido a que el operador es hermítico en este intervalo, los polinomios de Legendre cumplen con la **propiedad de ortogonalidad** con peso unitario:
$$\int_{-1}^{1} P_l(x) P_k(x) dx = \frac{2}{2l+1} \delta_{lk}$$

Cualquier función bien comportada en el intervalo $[-1, 1]$ puede expandirse como una serie de estas autofunciones (Serie de Legendre).

La relación entre el número cuántico orbital **$l$** y los niveles de energía de un átomo se deriva de la resolución de la ecuación de Schrödinger en un potencial central. A continuación se detalla esta conexión según las fuentes:

### 1. Definición y valores de $l$
El número cuántico **$l$** representa el **momento angular orbital** del electrón. Al separar las variables de la ecuación de Schrödinger en coordenadas esféricas, la parte angular genera una constante de separación de la forma **$l(l+1)$**. Para un nivel de energía dado por el número cuántico principal **$n$**, los valores permitidos de $l$ son números enteros que van desde **$0$ hasta $n-1$**.

### 2. Notación Espectroscópica
Los valores numéricos de $l$ se asocian con letras que definen los **subniveles o orbitales**:
*   **$l = 0$**: orbital **s** (*sharp*).
*   **$l = 1$**: orbital **p** (*principal*).
*   **$l = 2$**: orbital **d** (*diffuse*).
*   **$l = 3$**: orbital **f** (*fundamental*).

### 3. El papel de $l$ en la energía
*   **Átomo de Hidrógeno:** En un potencial puramente culombiano (como el del hidrógeno), la energía de un estado estacionario **depende únicamente de $n$** ($E_n = -K^2m / (2\hbar^2n^2)$). Esto significa que los subniveles con diferente $l$ dentro de un mismo $n$ (por ejemplo, $2s$ y $2p$) son **degenerados**, es decir, tienen la misma energía.
*   **Átomos Multielectrónicos:** Aunque las fuentes se centran en el modelo de un solo electrón, mencionan que en sistemas más complejos la interacción entre electrones y el **apantallamiento del núcleo** hacen que la energía dependa también de $l$. Por ejemplo, en la configuración **$1s^2 2s^2 2p^4$** (Oxígeno):
    *   **$1s$**: Nivel $n=1$, subnivel $l=0$. Es el de menor energía.
    *   **$2s$**: Nivel $n=2$, subnivel $l=0$.
    *   **$2p$**: Nivel $n=2$, subnivel $l=1$.
    *   En este caso, el orbital **$2s$ tiene menor energía que el $2p$** debido a que los electrones en $s$ "penetran" más hacia el núcleo que los de $p$.

### 4. Capacidad de los niveles
Para cada valor de $l$, existen **$2l+1$** posibles valores del número cuántico magnético $m$, lo que define la cantidad de estados independientes (orientaciones) en cada subnivel. Si se incluye el **espín** del electrón ($\pm 1/2$), el número de electrones que puede albergar cada subnivel es **$2(2l+1)$**:
*   **Subnivel s ($l=0$)**: $2(0+1) = \mathbf{2}$ electrones.
*   **Subnivel p ($l=1$)**: $2(2+1) = \mathbf{6}$ electrones.
*   **Subnivel d ($l=2$)**: $2(4+1) = \mathbf{10}$ electrones.

Esto explica por qué en tu ejemplo la configuración se escribe como **$1s^2$** (lleno con 2), **$2s^2$** (lleno con 2) y **$2p^4$** (parcialmente lleno, con un máximo de 6).

# Ejemplo 2
Ecuación de Bessel
$$x^2y''+xy'+(k^2x^2-\nu^2)y = 0$$

La **Ecuación de Bessel** es una de las ecuaciones diferenciales más importantes de la física matemática, especialmente en problemas que presentan **simetría cilíndrica**.

A continuación, se desarrolla el análisis de este ejemplo basado en las fuentes:

### 1. Origen y Contexto
La ecuación de Bessel de orden $\nu$:
$$x^2y''+xy'+(k^2x^2-\nu^2)y = 0$$
surge típicamente al aplicar el método de **separación de variables** a las ecuaciones de Laplace, Helmholtz o de onda en **coordenadas cilíndricas** (específicamente en la componente radial $\rho$).

### 2. Forma de Sturm-Liouville
El operador asociado a esta ecuación no es autoadjunto originalmente porque $a_2'(x) = (x^2)' = 2x$, lo cual es diferente de $a_1(x) = x$. Para convertirla a la **forma estándar de Sturm-Liouville**, se debe multiplicar toda la ecuación por el factor integrante $w(x) = 1/x$. La forma resultante es:

$$\mathbf{\frac{d}{dx} \left[ x \frac{dy}{dx} \right] + \left( k^2x - \frac{\nu^2}{x} \right)y = 0}$$

En esta estructura:
*   **Función de peso**: $\rho(x) = x$.
*   **Autovalor**: $\lambda = k^2$.
*   **Intervalo natural**: Generalmente $[0, b]$ o $[0, \infty)$.

### 3. Soluciones Fundamentales
Las soluciones se obtienen mediante el **método de Frobenius** alrededor del punto singular regular $x=0$:

*   **Funciones de Bessel de primera clase ($J_\nu$)**: Es la solución que permanece **finita y regular en el origen** ($x=0$) para $\nu \ge 0$. Su expansión en serie es:
    $$J_\nu(x) = \sum_{s=0}^{\infty} \frac{(-1)^s}{s!\Gamma(\nu+s+1)} \left( \frac{x}{2} \right)^{\nu+2s}$$
*   **Funciones de Bessel de segunda clase ($Y_\nu$ o Neumann)**: Es la segunda solución linealmente independiente. A diferencia de $J_\nu$, esta función **diverge (se hace infinita)** cuando $x \to 0$. Se define para que sea válida incluso cuando $\nu$ es un entero $n$.

**Nota sobre la independencia lineal**: Si $\nu$ no es un entero, $J_\nu$ y $J_{-\nu}$ son independientes. Si $\nu$ es un entero $n$, se cumple que $J_{-n}(x) = (-1)^n J_n(x)$, por lo que se requiere obligatoriamente de la función de Neumann $Y_n$ para completar la solución general.

### 4. Propiedades Críticas
*   **Comportamiento Asintótico**: Para valores grandes de $x$, las funciones de Bessel se comportan como oscilaciones amortiguadas que decaen proporcionalmente a $x^{-1/2}$.
*   **Ortogonalidad**: Bajo condiciones de frontera adecuadas (como $y(a)=0$), las funciones $J_\nu(k_i x)$ son **ortogonales con respecto al peso $x$** en el intervalo $[0, a]$. Esto permite expandir funciones arbitrarias en **series de Bessel**, de forma análoga a las series de Fourier.
*   **Variaciones**: Si el término $k^2x^2$ fuera negativo (argumento imaginario), la ecuación daría lugar a las **Funciones de Bessel Modificadas** $I_\nu$ y $K_\nu$, que tienen un comportamiento exponencial en lugar de oscilatorio.

La **Ecuación de Hermite** es un componente esencial de la física matemática, fundamental para la descripción del **oscilador armónico cuántico**. Su estructura y soluciones se analizan a continuación:

### 1. La Ecuación Diferencial
La forma estándar de la ecuación de Hermite es:
$$\ddot{H}(x) - 2x\dot{H}(x) + 2nH(x) = 0$$
donde $n$ es un parámetro que, en aplicaciones físicas, suele restringirse a valores enteros no negativos para garantizar soluciones convergentes.

### 2. Forma de Sturm-Liouville
Originalmente, el operador de Hermite no es autoadjunto porque $a'_2 = (1)' = 0$, que es diferente de $a_1 = -2x$. Para llevarla a la forma de Sturm-Liouville, se multiplica por el **factor integrante** $e^{-x^2}$:

$$\frac{d}{dx} \left[ e^{-x^2} \frac{dH}{dx} \right] + 2ne^{-x^2}H = 0$$

En esta estructura se identifican:
*   **Función de peso:** $\rho(x) = e^{-x^2}$.
*   **Intervalo natural:** $(-\infty, \infty)$, ya que en los límites infinitos el término $e^{-x^2}$ se anula, satisfaciendo la condición de Hermiticidad.
*   **Autovalor:** $\lambda = 2n$.

### 3. Soluciones: Polinomios de Hermite ($H_n$)
Al aplicar el método de **Frobenius**, se obtienen series de potencias que solo convergen en el infinito si se truncan para formar polinomios. Estos se conocen como **Polinomios de Hermite** y se pueden generar mediante la fórmula de Rodrigues:
$$H_n(x) = (-1)^n e^{x^2} \frac{d^n}{dx^n} (e^{-x^2})$$

Los primeros términos son:
*   $H_0(x) = 1$
*   $H_1(x) = 2x$
*   $H_2(x) = 4x^2 - 2$
*   $H_3(x) = 8x^3 - 12x$

### 4. Propiedades Críticas
*   **Ortogonalidad:** Debido a la naturaleza hermítica del problema en el intervalo infinito, los polinomios cumplen:
    $$\int_{-\infty}^{\infty} e^{-x^2} H_n(x) H_m(x) dx = 2^n \sqrt{\pi} n! \delta_{nm}$$
*   **Paridad:** Los polinomios tienen la misma paridad que su índice $n$: $H_n(-x) = (-1)^n H_n(x)$.

### 5. Aplicación Física: El Oscilador Armónico Cuántico
En mecánica cuántica, la ecuación de Schrödinger para un potencial parabólico se transforma en la ecuación de **Weber-Hermite**:
$$\frac{d^2\psi}{dy^2} + (\lambda - y^2)\psi = 0$$
La solución normalizada para los estados de energía es:
$$\psi_n(y) = \frac{1}{\sqrt{2^n n! \sqrt{\pi}}} e^{-y^2/2} H_n(y)$$
Esta función describe la **amplitud de probabilidad** de encontrar una partícula, donde el factor exponencial asegura que la probabilidad decaiga a cero en el infinito.

El enlace fundamental entre la **Ecuación de Hermite** y el **operador hermítico** reside en la **Teoría de Sturm-Liouville**, la cual permite transformar esta ecuación diferencial en un problema de autovalores dentro de un espacio de Hilbert, garantizando propiedades físicas esenciales como la realidad de las energías y la ortogonalidad de los estados.

A continuación se detalla esta conexión según las fuentes:

### 1. De la Ecuación Diferencial al Operador Autoadjunto
La ecuación de Hermite original, $H''(x) - 2xH'(x) + 2nH(x) = 0$, **no es autoadjunta** por sí sola porque no cumple la condición $a_1 = a_2'$ (donde $-2x \neq 0$). Sin embargo, las fuentes indican que cualquier ecuación de segundo orden puede volverse autoadjunta multiplicándola por un **factor de peso** $\rho(x)$. 
*   En el caso de Hermite, este factor es **$e^{-x^2}$**.
*   Al aplicarlo, el operador se convierte en: $\mathscr{L} = \frac{d}{dx} [e^{-x^2} \frac{d}{dx}] + 2ne^{-x^2}$, que tiene la estructura de un operador autoadjunto.

### 2. La Condición de Hermiticidad
Un operador autoadjunto se denomina **hermítico** si, además de su forma diferencial, las funciones sobre las que actúa satisfacen condiciones de frontera que anulan los "términos de borde" (el concomitante bilineal) en una integración por partes. 
*   Para la ecuación de Hermite, el dominio es el intervalo infinito $(-\infty, \infty)$.
*   Debido a que el factor de peso $e^{-x^2}$ decae extremadamente rápido hacia cero cuando $x \to \pm\infty$, el término de frontera $[e^{-x^2} W(f^*, g)]_{-\infty}^{\infty}$ se anula automáticamente para funciones bien comportadas.
*   Esto eleva al operador de Hermite a la categoría de **operador hermítico** en el espacio de funciones de cuadrado integrable ($L^2$).

### 3. Consecuencias de ser un Operador Hermítico
El hecho de que el operador de Hermite sea hermítico es lo que otorga validez matemática y física a los resultados del **oscilador armónico cuántico** descritos anteriormente:
*   **Autovalores Reales:** En mecánica cuántica, los autovalores corresponden a cantidades observables (como la energía). La hermiticidad garantiza que los niveles de energía $E_n = (n + 1/2)\hbar\omega$ sean siempre números reales.
*   **Ortogonalidad:** Las autofunciones (los Polinomios de Hermite multiplicados por la gaussiana) correspondientes a diferentes niveles de energía son **ortogonales** bajo el producto interno con peso $e^{-x^2}$. Esto permite que cada estado cuántico sea físicamente distinto e independiente de los demás.
*   **Completitud:** El conjunto de funciones de Hermite forma una **base completa** en el espacio de Hilbert, lo que significa que cualquier estado físico del oscilador puede representarse como una combinación lineal de estos estados básicos.

### 4. Isomorfismo con Matrices Hermíticas
Finalmente, existe un enlace directo entre este operador diferencial y el álgebra matricial. Las fuentes señalan que a cada operador diferencial hermítico se le puede asociar una **matriz hermítica** ($H = H^\dagger$) cuyos elementos son los productos internos de las funciones de base. Esta es la base de la equivalencia entre la mecánica ondulatoria de Schrödinger y la mecánica matricial de Heisenberg.

Para mostrar que, si un operador es **hermítico**, su acción sobre una función $f$ es equivalente a su acción sobre una función $g$ dentro de un producto interno (es decir, $(\mathscr{L}f, g) = (f, \mathscr{L}g)$), se debe recurrir a la definición matemática del **operador adjunto** y a las **condiciones de frontera** del problema.

A continuación se presenta el desarrollo formal basado en las fuentes:

### 1. Definición del Producto Interno
En un espacio de Hilbert, el producto interno entre la función resultante de aplicar el operador $\mathscr{L}$ sobre $f$ y una función $g$ se define mediante la integral:
$$(\mathscr{L}f, g) = \int_a^b (\mathscr{L}f)^*(x) g(x) dx$$.

### 2. Identificación del Operador Adjunto
Mediante el uso de **integración por partes** sucesivas, se pueden transferir las derivadas que actúan sobre $f^*$ hacia la función $g$. Este procedimiento general genera la siguiente identidad fundamental:
$$(\mathscr{L}f, g) = (f, \overline{\mathscr{L}}g) + [ \text{Términos de frontera} ]_a^b$$.
Donde **$\overline{\mathscr{L}}$** (o $\mathscr{L}^\dagger$) es el **operador adjunto** de $\mathscr{L}$.

### 3. Condiciones para la Hermiticidad
Para que el operador $\mathscr{L}$ sea estrictamente **hermítico**, deben satisfacerse simultáneamente dos requisitos:

*   **Autoadjunción:** El operador debe ser idéntico a su adjunto, es decir, **$\mathscr{L} = \overline{\mathscr{L}}$**. Esto depende de la estructura interna de los coeficientes del operador.
*   **Anulación de los términos de frontera:** Las funciones $f$ y $g$ deben pertenecer a un subespacio que satisfaga condiciones de frontera (como que las funciones o sus derivadas se anulen en los extremos $a$ y $b$) tales que el residuo de la integración por partes sea **cero**.

### 4. Conclusión: La igualdad del operador hermítico
Cuando se cumplen las condiciones anteriores, el término de frontera desaparece y el operador adjunto se reemplaza por el operador original. Por lo tanto, se demuestra que:
**$$(\mathscr{L}f, g) = (f, \mathscr{L}g)$$**.

Esta propiedad garantiza, entre otras cosas, que los **autovalores** del operador sean siempre números **reales** y que sus autofunciones sean **ortogonales**, lo cual es la base de la mecánica cuántica y la teoría de Sturm-Liouville.
