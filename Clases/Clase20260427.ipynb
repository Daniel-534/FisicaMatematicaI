# Ejercicio

Hallar $f(x)$ para la siguiente EDO:

$$\frac{d^2}{x^2}f(x) = x$$

con condiciones $f(0)=f(L=1)=0$

Para resolver la ecuación diferencial ordinaria (EDO) inhomogénea planteada dentro del marco de la física matemática, utilizaremos el **método de la función de Green**, el cual es ideal para tratar términos fuente ($g(x) = x$) bajo condiciones de frontera específicas.

El problema es:
$$\frac{d^2}{dx^2} f(x) = x, \quad \text{con } f(0) = 0 \text{ y } f(1) = 0$$

---

### 1. Identificación del Operador y Condiciones
El operador diferencial es $\mathscr{L} = \frac{d^2}{dx^2}$. Este es un operador autoadjunto con función de peso $p(x) = 1$ y $q(x) = 0$. El dominio de la solución es el intervalo $x \in$. Las condiciones dadas son de tipo **Dirichlet homogéneas** ($f=0$ en los extremos).

### 2. Construcción de la Función de Green $G(x, x')$
La función de Green representa la respuesta del sistema a una fuente puntual $\delta(x - x')$. Debe satisfacer:
$$\frac{d^2}{dx^2} G(x, x') = \delta(x - x')$$
Sujeta a $G(0, x') = 0$ y $G(1, x') = 0$.

**Paso A: Solución de la ecuación homogénea por tramos**
Fuera del punto $x'$, la ecuación es $G'' = 0$, cuya solución son líneas rectas de la forma $ax + b$.
1.  **Región $x < x'$ ($G_1$):** Aplicamos $G_1(0, x') = 0$.
    $$a_1(0) + b_1 = 0 \implies b_1 = 0 \implies \mathbf{G_1(x, x') = a_1 x}$$
2.  **Región $x > x'$ ($G_2$):** Aplicamos $G_2(1, x') = 0$.
    $$a_2(1) + b_2 = 0 \implies b_2 = -a_2 \implies \mathbf{G_2(x, x') = a_2(x - 1)}$$

**Paso B: Aplicación de propiedades de continuidad y salto**
1.  **Continuidad en $x = x'$:** $G_1(x', x') = G_2(x', x')$
    $$a_1 x' = a_2(x' - 1)$$
2.  **Discontinuidad de la derivada (salto):** $\frac{dG_2}{dx} \big|_{x'} - \frac{dG_1}{dx} \big|_{x'} = \frac{1}{p(x')}$
    Como $p(x') = 1$:
    $$a_2 - a_1 = 1 \implies a_2 = 1 + a_1$$

Sustituyendo $a_2$ en la ecuación de continuidad:
$$a_1 x' = (1 + a_1)(x' - 1) \implies a_1 x' = x' - 1 + a_1 x' - a_1 \implies \mathbf{a_1 = x' - 1}$$
Luego, $\mathbf{a_2 = x'}$.

**Paso C: Expresión de la función de Green**
$$G(x, x') = \begin{cases} x(x' - 1) & 0 \le x < x' \\ x'(x - 1) & x' < x \le 1 \end{cases}$$

### 3. Resolución de la EDO mediante integración
La solución final $f(x)$ se obtiene mediante la **integral de convolución** entre la función de Green y el término fuente $g(x') = x'$:
$$f(x) = \int_0^1 G(x, x') x' dx'$$

Debido a la definición por tramos de $G(x, x')$, dividimos la integral en el punto donde el observador $x$ coincide con la fuente $x'$:
$$f(x) = \int_0^x G_{x > x'}(x, x') x' dx' + \int_x^1 G_{x < x'}(x, x') x' dx'$$
$$f(x) = \int_0^x [x'(x - 1)] x' dx' + \int_x^1 [x(x' - 1)] x' dx'$$

**Evaluación de las integrales:**
1.  **Primera integral ($I_1$):**
    $$I_1 = (x - 1) \int_0^x (x')^2 dx' = (x - 1) \left[ \frac{(x')^3}{3} \right]_0^x = \mathbf{\frac{x^4 - x^3}{3}}$$
2.  **Segunda integral ($I_2$):**
    $$I_2 = x \int_x^1 ((x')^2 - x') dx' = x \left[ \frac{(x')^3}{3} - \frac{(x')^2}{2} \right]_x^1 = x \left[ \left( \frac{1}{3} - \frac{1}{2} \right) - \left( \frac{x^3}{3} - \frac{x^2}{2} \right) \right]$$
    $$I_2 = x \left[ -\frac{1}{6} - \frac{x^3}{3} + \frac{x^2}{2} \right] = \mathbf{-\frac{x}{6} - \frac{x^4}{3} + \frac{x^3}{2}}$$

### 4. Resultado Final
Sumamos ambas contribuciones:
$$f(x) = \left( \frac{x^4}{3} - \frac{x^4}{3} \right) + \left( \frac{x^3}{2} - \frac{x^3}{3} \right) - \frac{x}{6}$$
$$f(x) = \frac{3x^3 - 2x^3}{6} - \frac{x}{6} = \frac{x^3}{6} - \frac{x}{6}$$

Por lo tanto, la solución exacta es:
$$\mathbf{f(x) = \frac{1}{6}(x^3 - x)}$$

**Verificación rápida:**
*   $f(0) = \frac{1}{6}(0-0) = 0$. (Cumple)
*   $f(1) = \frac{1}{6}(1-1) = 0$. (Cumple)
*   $f'(x) = \frac{3x^2 - 1}{6} \implies f''(x) = \frac{6x}{6} = x$. (Satisface la EDO original).

En un ejercicio de resolución de una EDO inhomogénea mediante funciones de Green, como el de $f''(x) = x$ con $f(0) = f(1) = 0$, existen varios fundamentos teóricos que suelen aplicarse de forma mecánica pero que encierran conceptos profundos de la física matemática.

A continuación, explico los detalles teóricos que usualmente se pasan por alto:

### 1. Autoadjunticidad vs. Hermiticidad
Es común decir que el operador $\mathscr{L} = \frac{d^2}{dx^2}$ es "autoadjunto", pero hay una distinción técnica importante:
*   **Operador Autoadjunto:** Se refiere a la forma del operador. Para un operador de segundo orden $a_2(x) \frac{d^2}{dx^2} + a_1(x) \frac{d}{dx} + a_0(x)$, es autoadjunto si $a_1(x) = a_2'(x)$. En nuestro caso, $a_2=1$ y $a_1=0$, por lo que cumple la condición $0 = (1)'$.
*   **Operador Hermítico:** Un operador autoadjunto solo es **Hermítico** si además se especifican condiciones de frontera que anulen los términos de borde al integrar por partes (identidad de Green). Las condiciones de Dirichlet ($f(0)=f(1)=0$) garantizan que el término $[qW]_{a}^{b}$ (donde $W$ es el Wronskiano) sea cero, convirtiendo el problema en un **problema regular de Sturm-Liouville**.

### 2. Por qué $G(x, x')$ debe satisfacer condiciones homogéneas
En el ejercicio, impusimos $G(0, x') = 0$ y $G(1, x') = 0$. Esto no es una elección arbitraria, sino un requisito para que la solución integral sea válida.
Según la teoría, la solución general $f(x)$ se expresa como la suma de una contribución de la fuente y una de la frontera. Al exigir que $G$ cumpla las **condiciones de frontera homogéneas** del problema original, logramos que la integral de superficie en la fórmula de Green desaparezca, dejando solo la integral de convolución:
$$f(x) = \int_a^b G(x, x') g(x') dx'$$

### 3. La propiedad de Simetría (Reciprocidad)
Un detalle que a menudo se ignora es que, para operadores Hermíticos, la función de Green debe ser simétrica: **$G(x, x') = G(x', x)$**.
Si observas el resultado del ejercicio anterior:
*   $G(x, x') = x(x'-1)$ para $x < x'$
*   $G(x, x') = x'(x-1)$ para $x > x'$
Verás que si intercambias $x$ por $x'$, las expresiones se intercambian exactamente, manteniendo el valor de la función. Físicamente, esto significa que la respuesta en $x$ debida a una fuente en $x'$ es idéntica a la respuesta en $x'$ debida a una fuente en $x$.

### 4. El origen de la discontinuidad en la derivada
Al resolver, aplicamos el "salto" de la derivada: $\frac{\partial G}{\partial x}\big|_{x=x'^+} - \frac{\partial G}{\partial x}\big|_{x=x'^-} = 1$.
Teóricamente, este salto es necesario porque estamos operando con una **Delta de Dirac** $\delta(x-x')$. Para que la segunda derivada de $G$ produzca una singularidad infinita (la delta), la primera derivada debe tener un escalón o salto finito en ese punto. El valor del salto es siempre $1/p(x')$, donde $p(x')$ es el coeficiente de la derivada de mayor orden; en este ejercicio $p(x)=1$, por eso el salto es unitario.

### 5. Teorema de Unicidad
¿Por qué estamos seguros de que $f(x) = \frac{1}{6}(x^3 - x)$ es la **única** solución?
El teorema de unicidad para la ecuación de Poisson (o de Laplace) establece que si se especifican condiciones de Dirichlet en toda la frontera de una región finita, la solución es única. Esto se demuestra asumiendo dos soluciones distintas $f_1$ y $f_2$ y probando, mediante la primera identidad de Green, que su diferencia debe tener un gradiente nulo en todo el volumen, lo que implica que son idénticas.

### 6. Interpretación física del término fuente puntual
Físicamente, la función de Green $G(x, x')$ representa la **función de transferencia** o la "influencia" de un punto sobre otro. En este ejercicio, el resultado nos dice cómo un "impulso unitario" de carga o fuerza en el punto $x'$ deforma la solución en el punto $x$, respetando que los extremos estén "anclados" a cero. El potencial total es simplemente la **superposición lineal** de todas esas influencias puntuales ponderadas por el valor de la fuente real $g(x)=x$.

Para resolver la ecuación diferencial de autovalores $\frac{d^2}{dx^2} f = \lambda f$ bajo condiciones de frontera de **Dirichlet homogéneas** ($f(0) = f(1) = 0$), debemos analizar el comportamiento de la solución según el signo de $\lambda$. Este es un problema fundamental de la **teoría de Sturm-Liouville**.

### 1. Caso $\lambda = \beta^2$ (Autovalores Positivos)

Para el caso de autovalores positivos, donde definimos **$\lambda = \beta^2 > 0$**, la ecuación diferencial se expresa como $f''(x) - \beta^2 f(x) = 0$. A continuación, se desarrolla el análisis utilizando exclusivamente la forma exponencial:

1.  **Solución General:**
    La solución general para una ecuación diferencial de segundo orden con coeficientes constantes y raíces reales distintas ($\pm \beta$) es una combinación lineal de exponenciales naturales:
    $$f(x) = A e^{\beta x} + B e^{-\beta x}$$
    donde $A$ y $B$ son constantes arbitrarias determinadas por las condiciones de contorno.

2.  **Aplicación de la primera condición ($f(0) = 0$):**
    Evaluamos la función en el extremo izquierdo del dominio:
    $$f(0) = A e^{\beta(0)} + B e^{-\beta(0)} = A(1) + B(1) = 0$$
    De esta igualdad se deduce que las constantes deben ser opuestas:
    $$\mathbf{B = -A}$$
    Sustituyendo esto en la solución general, obtenemos:
    $$f(x) = A (e^{\beta x} - e^{-\beta x})$$

3.  **Aplicación de la segunda condición ($f(1) = 0$):**
    Imponemos la condición en el extremo derecho para buscar soluciones no triviales:
    $$f(1) = A (e^{\beta} - e^{-\beta}) = 0$$
    Para que exista una solución distinta de la trivial ($A \neq 0$), el término entre paréntesis debe anularse:
    $$e^{\beta} - e^{-\beta} = 0 \implies e^{\beta} = e^{-\beta}$$
    Multiplicando ambos lados por $e^{\beta}$ obtenemos:
    $$e^{2\beta} = 1$$
    Esta igualdad solo se cumple en los números reales si el exponente es cero ($2\beta = 0$), lo que implica que **$\beta = 0$**.

4.  **Conclusión:**
    Dado que nuestra premisa inicial era que $\lambda > 0$ (lo que implica $\beta \neq 0$), llegamos a una contradicción. La única forma de satisfacer las condiciones de frontera de Dirichlet con un autovalor positivo es que la amplitud $A$ sea igual a cero.
    Por lo tanto, **no existen autovalores positivos** para este problema; solo se obtiene la **solución trivial $f(x) = 0$**.

**Check de consistencia:**
Nótese que la expresión $e^{\beta x} - e^{-\beta x}$ es exactamente el doble del seno hiperbólico ($2 \sinh \beta x$). Dado que el $\sinh(z)$ solo tiene un cero en el origen para valores reales, es físicamente imposible que la función "regrese" a cero en $x=1$ sin ser la función nula, validando que el espectro de autovalores debe ser necesariamente negativo para permitir oscilaciones (senos y cosenos) que crucen el eje horizontal de nuevo.

-----



Si suponemos que $\lambda$ es una constante positiva ($\lambda > 0$), la ecuación se escribe como $f'' - \beta^2 f = 0$.
*   **Solución general:** La solución para este tipo de EDO es una combinación de exponenciales o funciones hiperbólicas: $f(x) = A \cosh(\beta x) + B \sinh(\beta x)$.
*   **Aplicación de condiciones de frontera:**
    1.  **En $x=0$:** $f(0) = A \cosh(0) + B \sinh(0) = A \cdot 1 + B \cdot 0 = 0 \implies \mathbf{A = 0}$.
    2.  **En $x=1$:** $f(1) = B \sinh(\beta) = 0$. Como $\beta \neq 0$ (ya que $\lambda > 0$), la función $\sinh(\beta)$ nunca es cero para argumentos reales distintos de cero. Por lo tanto, **$B = 0$**.
*   **Resultado:** Solo se obtiene la **solución trivial** $f(x) = 0$. No existen autovalores positivos para estas condiciones de borde.


-----




### 2. Caso $\lambda = 0$
La ecuación se reduce a $f''(x) = 0$.
*   **Solución general:** Integrando dos veces, obtenemos la ecuación de una línea recta: $f(x) = ax + b$.
*   **Aplicación de condiciones de frontera:**
    1.  **En $x=0$:** $f(0) = a(0) + b = 0 \implies \mathbf{b = 0}$.
    2.  **En $x=1$:** $f(1) = a(1) = 0 \implies \mathbf{a = 0}$.
*   **Resultado:** Nuevamente, solo se obtiene la **solución trivial**. $\lambda = 0$ no es un autovalor para este sistema.

### 3. Caso $\lambda = -\alpha^2$ (Autovalores Negativos)
Si $\lambda$ es negativo, la ecuación toma la forma de un oscilador armónico espacial: $f'' + \alpha^2 f = 0$.
*   **Solución general:** La solución se expresa en términos de funciones trigonométricas: **$f(x) = A \cos(\alpha x) + B \sin(\alpha x)$**.
*   **Aplicación de condiciones de frontera:**
    1.  **En $x=0$:** $f(0) = A \cos(0) + B \sin(0) = A = 0$. La solución se reduce a **$f(x) = B \sin(\alpha x)$**.
    2.  **En $x=1$:** $f(1) = B \sin(\alpha) = 0$.
*   **Cuantización:** Para obtener una solución no trivial ($B \neq 0$), el argumento del seno debe ser un múltiplo entero de $\pi$:
    $$\alpha_n = n\pi \quad \text{para } n = 1, 2, 3, \dots$$.
*   **Autovalores:** Sustituyendo $\alpha_n$ en $\lambda = -\alpha^2$, obtenemos el espectro discreto:
    $$\mathbf{\lambda_n = -(n\pi)^2}$$.
*   **Autofunciones:** El conjunto de soluciones (base de funciones) es:
    $$\mathbf{f_n(x) = c_n \sin(n\pi x)}$$.

---

### Check (Verificación Ordenada)
Para validar que el resultado satisface la EDO original $f'' = \lambda f$:

1.  **Tomamos la autofunción:** $f_n(x) = c_n \sin(n\pi x)$.
2.  **Primera derivada:** $\frac{d}{dx} f_n(x) = c_n (n\pi) \cos(n\pi x)$.
3.  **Segunda derivada:** $\frac{d^2}{dx^2} f_n(x) = -c_n (n\pi)^2 \sin(n\pi x)$.
4.  **Comparación con el operador:** Observamos que:
    $$\frac{d^2}{dx^2} f_n(x) = \underbrace{-(n\pi)^2}_{\lambda_n} \cdot \underbrace{c_n \sin(n\pi x)}_{f_n(x)}$$.
5.  **Conclusión:** La igualdad $f'' = \lambda f$ se cumple con **$\lambda = -(n\pi)^2$**.

**Nota sobre la nomenclatura:** En muchos textos de física se define el autovalor $\lambda$ positivo para el operador $-\frac{d^2}{dx^2}$ (el Laplaciano negativo), lo que explicaría por qué en tu solución aparece $\lambda_m = (n\pi)^2$ sin el signo negativo.

Expandir la función de Green en autofunciones del operador $\mathscr{L}$

La expansión de la **función de Green** en términos de las autofunciones de un operador diferencial $\mathscr{L}$ es un método formal poderoso que permite construir la solución fundamental de un problema inhomogéneo basándose en las propiedades de simetría y completez de la base de funciones del sistema.

A continuación, se detallan los pasos y fundamentos teóricos para realizar esta expansión:

### 1. Definición del problema de autovalores
Para un operador lineal $\mathscr{L}$ que actúa en un dominio con condiciones de frontera específicas, primero se resuelve la **ecuación de autovalores** asociada:
$$\mathscr{L} \phi_n(\vec{r}) = \lambda_n \phi_n(\vec{r})$$
Donde:
*   $\phi_n(\vec{r})$ son las **autofunciones** del operador.
*   $\lambda_n$ son los **autovalores** correspondientes.
*   Si el operador es **Hermítico** bajo las condiciones de frontera dadas, los autovalores $\lambda_n$ son reales y las autofunciones $\phi_n$ forman una base completa y ortogonal.

### 2. Representación de la Delta de Dirac
La función de Green $G(\vec{r}, \vec{r}')$ se define como la solución a la ecuación con una fuente puntual:
$$\mathscr{L} G(\vec{r}, \vec{r}') = \delta(\vec{r} - \vec{r}')$$
Utilizando la **condición de completez** (o relación de clausura) de las autofunciones, la delta de Dirac se puede expresar como una suma sobre todos los estados de la base:
$$\delta(\vec{r} - \vec{r}') = \sum_{n} \phi_n(\vec{r}) \phi_n^*(\vec{r}')$$

### 3. Procedimiento de expansión
Se propone que la función de Green puede escribirse como una combinación lineal de las mismas autofunciones, donde los coeficientes $a_n$ dependen de la posición de la fuente $\vec{r}'$:
$$G(\vec{r}, \vec{r}') = \sum_{n} a_n(\vec{r}') \phi_n(\vec{r})$$

Al sustituir esta propuesta en la ecuación definitoria de la función de Green:
$$\mathscr{L} \left( \sum_{n} a_n(\vec{r}') \phi_n(\vec{r}) \right) = \sum_{n} \phi_n(\vec{r}) \phi_n^*(\vec{r}')$$
Debido a la linealidad del operador y a la ecuación de autovalores ($\mathscr{L} \phi_n = \lambda_n \phi_n$), obtenemos:
$$\sum_{n} a_n(\vec{r}') \lambda_n \phi_n(\vec{r}) = \sum_{n} \phi_n^*(\vec{r}') \phi_n(\vec{r})$$

### 4. Determinación de los coeficientes y resultado final
Dada la **independencia lineal** de las autofunciones, podemos igualar los coeficientes término a término para cada $n$:
$$a_n(\vec{r}') \lambda_n = \phi_n^*(\vec{r}') \implies a_n(\vec{r}') = \frac{\phi_n^*(\vec{r}')}{\lambda_n}$$
Sustituyendo estos coeficientes en la serie original, se llega a la **forma bilineal** de la función de Green:
$$\mathbf{G(\vec{r}, \vec{r}') = \sum_{n} \frac{\phi_n(\vec{r}) \phi_n^*(\vec{r}')}{\lambda_n}}$$

### Observaciones importantes
*   **Simetría:** Esta expansión revela inmediatamente que $G(\vec{r}, \vec{r}') = G^*(\vec{r}', \vec{r})$, propiedad fundamental de los sistemas Hermíticos conocida como reciprocidad.
*   **Validez:** La serie es válida siempre que $\lambda_n \neq 0$. Si el operador tiene un autovalor nulo, la función de Green no existe en este sentido estricto o requiere un tratamiento especial.
*   **Bases Continuas:** En dominios infinitos donde el espectro de autovalores es continuo, la suma se convierte en una integral:
$$G(\vec{r}, \vec{r}') = \int \frac{\phi(\vec{k}, \vec{r}) \phi^*(\vec{k}, \vec{r}')}{\lambda(\vec{k})} d\vec{k}$$
Un ejemplo típico es la expansión de la función de Green para la **ecuación de Poisson** en el espacio infinito utilizando la base de Fourier ($e^{i\vec{k}\cdot\vec{r}}$).

### Problema de Alonso Sepúlveda (7.42)

El objetivo es hallar la solución $f(x)$ para el siguiente problema inhomogéneo en el intervalo $[0, L]$:
$$\mathscr{L} f(x) = g(x) \implies \frac{d^2 f(x)}{dx^2} = g(x)$$
Sujeto a las condiciones de frontera de **Dirichlet homogéneas**:
$$f(0) = 0, \quad f(L) = 0$$

### 1. Definición de la Función de Green
La función de Green $G(x, x')$ para este operador debe satisfacer la ecuación con una fuente puntual (delta de Dirac) y cumplir con las mismas condiciones de frontera:
$$\frac{d^2 G(x, x')}{dx^2} = \delta(x - x')$$
$$G(0, x') = 0, \quad G(L, x') = 0$$

### 2. Obtención de las Autofunciones del Operador
Para resolver esto mediante expansión, primero se consideran las autofunciones $\phi_n(x)$ del operador $\mathscr{L} = \frac{d^2}{dx^2}$ que se anulan en los extremos:
$$\frac{d^2 \phi_n(x)}{dx^2} = \lambda_n \phi_n(x)$$
Las soluciones que satisfacen $\phi_n(0) = \phi_n(L) = 0$ son de tipo sinusoidal:
$$\phi_n(x) = \sqrt{\frac{2}{L}} \sin\left(\frac{n\pi x}{L}\right)$$
Con sus respectivos autovalores asociados:
$$\lambda_n = -\left(\frac{n\pi}{L}\right)^2, \quad n = 1, 2, 3, \dots$$

### 3. Expansión de la Función de Green
De acuerdo con la teoría general, la función de Green se puede expresar como una combinación lineal de estas autofunciones (forma bilineal):
$$G(x, x') = \sum_{n=1}^\infty \frac{\phi_n(x) \phi_n^*(x')}{\lambda_n}$$
Sustituyendo las expresiones halladas en el paso anterior:
$$G(x, x') = \sum_{n=1}^\infty \frac{\left[ \sqrt{\frac{2}{L}} \sin\left(\frac{n\pi x}{L}\right) \right] \left[ \sqrt{\frac{2}{L}} \sin\left(\frac{n\pi x'}{L}\right) \right]}{-\left(\frac{n\pi}{L}\right)^2}$$

### 4. Resultado Final (Notación de Alonso)
Al simplificar los factores constantes y reordenar la serie, se llega a la expresión final que aparece en el libro:
$$\mathbf{G(x, x') = -\frac{2L}{\pi^2} \sum_{n=1}^\infty \frac{1}{n^2} \sin\left(\frac{n\pi x}{L}\right) \sin\left(\frac{n\pi x'}{L}\right)}$$

### 5. Solución del Problema Inhomogéneo
Una vez construida $G(x, x')$, la solución completa para cualquier fuente arbitraria $g(x)$ se obtiene mediante la integral de convolución:
$$f(x) = \int_0^L G(x, x') g(x') dx'$$
Esta integral suma las contribuciones de cada "punto fuente" $x'$ ponderadas por la intensidad de la fuente $g(x')$ en ese lugar. El resultado garantiza automáticamente que se cumplan las condiciones de frontera $f(0)=f(L)=0$ debido a que cada término de la serie de la función de Green ya las satisface.

Para desarrollar con detalle la **condición de completez** necesaria en la resolución del **problema 7.42** de Alonso Sepúlveda, es fundamental entender cómo esta propiedad permite representar la **delta de Dirac** como una superposición de las funciones de base del sistema.

### 1. Definición Teórica de la Completez
En un espacio de Hilbert, un conjunto de funciones $\{\phi_n(x)\}$ es **completo** (o satisface la **relación de clausura**) si cualquier función "bien comportada" del espacio puede expresarse como una combinación lineal única de dichas funciones. Matemáticamente, esta propiedad se manifiesta a través de la representación de la delta de Dirac:
$$\sum_{n} p(x') \phi_n^*(x') \phi_n(x) = \delta(x - x')$$
donde $p(x')$ es la función de peso asociada al operador. En el caso del operador $\mathscr{L} = \frac{d^2}{dx^2}$, la función de peso es $p(x) = 1$.

### 2. Aplicación a las Autofunciones de Seno
Para el problema de Dirichlet en el intervalo $[0, L]$, las autofunciones normalizadas obtenidas son $\phi_n(x) = \sqrt{\frac{2}{L}} \sin\left(\frac{n\pi x}{L}\right)$. Aplicando la condición de completez para esta base específica:
$$\sum_{n=1}^\infty \left[ \sqrt{\frac{2}{L}} \sin\left(\frac{n\pi x}{L}\right) \right] \left[ \sqrt{\frac{2}{L}} \sin\left(\frac{n\pi x'}{L}\right) \right] = \delta(x - x')$$
Al simplificar los factores constantes, obtenemos la **identidad fundamental** que se requiere para resolver la ecuación de la función de Green:
$$\frac{2}{L} \sum_{n=1}^\infty \sin\left(\frac{n\pi x}{L}\right) \sin\left(\frac{n\pi x'}{L}\right) = \delta(x - x')$$
Esta expresión es una representación en serie de Fourier de la delta de Dirac en el intervalo dado.

### 3. Necesidad de la Completez en la Expansión de Green
La construcción de la función de Green $G(x, x')$ mediante el método de expansión parte de la ecuación definitoria:
$$\mathscr{L}_x G(x, x') = \delta(x - x')$$
Si proponemos que $G(x, x')$ se puede expandir en la base de autofunciones como $G(x, x') = \sum_n a_n(x') \phi_n(x)$, al aplicar el operador $\mathscr{L}$ obtenemos:
$$\sum_{n} a_n(x') \mathscr{L} \phi_n(x) = \delta(x - x')$$
Sustituyendo el autovalor ($\mathscr{L} \phi_n = \lambda_n \phi_n$) y utilizando la **condición de completez** para reemplazar la delta de Dirac en el lado derecho:
$$\sum_{n} a_n(x') \lambda_n \phi_n(x) = \sum_{n} \phi_n^*(x') \phi_n(x)$$
Dada la **independencia lineal** de las autofunciones, los coeficientes de ambos lados deben ser iguales para cada $n$, lo que obliga a que:
$$a_n(x') \lambda_n = \phi_n^*(x') \implies a_n(x') = \frac{\phi_n^*(x')}{\lambda_n}$$
Sin la garantía de la **completez**, no podríamos asegurar que la suma de autofunciones sea capaz de reconstruir la singularidad puntual de la delta de Dirac, invalidando el paso fundamental de igualar los términos de la serie.

### 4. Conclusión en el Problema 7.42
En el ejercicio de Alonso, la completez de la base de senos es lo que permite pasar de la forma diferencial $\frac{d^2 G}{dx^2} = \delta(x - x')$ a la forma bilineal sumatoria. El resultado final, que incluye el factor $1/n^2$, proviene directamente de dividir cada término de la expansión de la delta (donde cada término tiene peso unitario) por su autovalor correspondiente $\lambda_n = -(n\pi/L)^2$, demostrando que la **función de Green** es la respuesta del sistema "repartida" sobre todos los modos normales de oscilación permitidos por la base completa.

Para resolver el problema planteado en el libro de Alonso Sepúlveda (página 267, ejercicio siguiente al 7.42), debemos hallar la solución a la ecuación diferencial inhomogénea:
$$\frac{d^2y(x)}{dx^2} = g(x)$$
Sujeto a las condiciones iniciales en un solo punto (**Problema de Valor Inicial**):
$$y(0) = 0, \quad y'(0) = 0$$

A diferencia del caso anterior (Dirichlet en ambos extremos), este problema se caracteriza por la **causalidad**, donde el estado del sistema en $x$ solo depende de las fuentes $g(x')$ situadas "en el pasado" o a su izquierda ($x' < x$).

### 1. Construcción de la Función de Green $G(x, x')$

La función de Green debe satisfacer $\frac{d^2}{dx^2} G(x, x') = \delta(x - x')$ con las condiciones iniciales homogéneas $G(0, x') = 0$ y $\frac{\partial G}{\partial x} \big|_{x=0} = 0$.

*   **Para $x < x'$:** La función debe cumplir las condiciones en el origen. La única solución de $G'' = 0$ que satisface $G(0) = 0$ y $G'(0) = 0$ es la solución trivial:
    $$\mathbf{G(x, x') = 0, \quad \text{para } x < x'}$$
*   **Para $x > x'$:** La función debe satisfacer la ecuación homogénea $G'' = 0$, por lo que tiene la forma $A(x - x') + B$. Debido a la continuidad en $x = x'$ (donde $G$ debe ser 0 para coincidir con el tramo izquierdo), obtenemos $B = 0$.
*   **Salto en la derivada:** La propiedad del salto unitario en $x = x'$ exige que:
    $$\frac{\partial G}{\partial x} \big|_{x=x'^+} - \frac{\partial G}{\partial x} \big|_{x=x'^-} = 1 \implies A - 0 = 1 \implies A = 1$$

Por lo tanto, la **función de Green causal** es:
$$\mathbf{G(x, x') = \begin{cases} 0, & x < x' \\ x - x', & x > x' \end{cases}}$$

### 2. Solución mediante la integral de Green

De acuerdo con la teoría general de Alonso para problemas de valor inicial, la solución $y(x)$ se expresa mediante la integral de convolución:
$$y(x) = \int_0^L G(x, x') g(x') dx'$$

Sustituyendo la forma de la función de Green, observamos que el integrando se anula para todo $x' > x$. Por lo tanto, el límite superior de la integral se reduce de $L$ a $x$:
$$\mathbf{y(x) = \int_0^x (x - x') g(x') dx'}$$

### 3. Interpretación y Verificación

Esta expresión final representa el potencial o desplazamiento acumulado debido a la fuente distribuida $g(x)$. Podemos verificar que satisface las condiciones iniciales y la EDO original:
1.  **$y(0)$**: El límite de la integral es de 0 a 0, por lo que **$y(0) = 0$**.
2.  **$y'(x)$**: Utilizando la regla de Leibniz para derivar la integral:
    $$y'(x) = (x - x)g(x) + \int_0^x \frac{\partial}{\partial x}(x - x')g(x') dx' = \int_0^x g(x') dx'$$
    Evaluando en cero, **$y'(0) = 0$**.
3.  **$y''(x)$**: Derivando $y'(x)$ obtenemos directamente **$y''(x) = g(x)$**, cumpliendo la EDO original.

Este resultado es equivalente a realizar una **integración sucesiva** de la fuente $g(x)$, un método estándar para resolver este tipo de problemas de trayectoria o respuesta temporal.

```
Integrate[x*Sin[n*Pi*x],{x,0,1}]
```
