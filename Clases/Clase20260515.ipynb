Las funciones de Bessel surgen al resolver ecuaciones diferenciales parciales (como las de Laplace, Helmholtz o Schrödinger) mediante el método de separación de variables en sistemas con simetrías específicas. Dependiendo del sistema de coordenadas utilizado, se obtienen las funciones de **Bessel cilíndricas** (u ordinarias) o las **Bessel esféricas**.

A continuación se presenta una tabla comparativa detallada basada en las fuentes:

### Tabla Comparativa: Bessel Cilíndrico vs. Esférico

| Característica | Bessel Cilíndrico (Ordinario) | Bessel Esférico |
| :--- | :--- | :--- |
| **Sistema de Coordenadas** | Cilíndrico polar $(\rho, \phi, z)$. | Esférico polar $(r, \theta, \phi)$. |
| **Ecuación Diferencial** | $x^2 y'' + x y' + (x^2 - \nu^2)y = 0$. | $x^2 y'' + 2x y' + [x^2 - n(n+1)]y = 0$. |
| **Símbolos (1.ª y 2.ª especie)** | $J_\nu(x)$ y $Y_\nu(x)$ (o $N_\nu$). | $j_n(x)$ y $y_n(x)$ (o $\eta_n$). |
| **Relación entre ambas** | Es la función fundamental de orden $\nu$. | $f_n(x) = \sqrt{\frac{\pi}{2x}} F_{n+1/2}(x)$. |
| **Forma Analítica ($n \in \mathbb{Z}$)** | Series de potencias infinitas (no elementales). | Combinaciones finitas de senos, cosenos y potencias. |
| **Comportamiento en $x \to 0$** | $J_\nu \propto x^\nu$; $Y_\nu$ diverge (log o $x^{-\nu}$). | $j_n \propto x^n$; $y_n$ diverge ($\propto x^{-n-1}$). |
| **Comportamiento en $x \to \infty$** | Oscilaciones amortiguadas $\propto x^{-1/2}$. | Oscilaciones amortiguadas $\propto x^{-1}$. |
| **Ejemplo de Orden 0** | $J_0(0) = 1$ (serie infinita). | $j_0(x) = \frac{\sin x}{x}$. |
| **Aplicación Física Típica** | Vibración de membranas circulares y guías de onda. | Partícula en una caja esférica y dispersión (scattering). |

---

### Diferencias Clave y Observaciones

1.  **Origen de la Separación:**
    *   En **coordenadas cilíndricas**, la parte radial de la ecuación de Helmholtz conduce directamente a la ecuación de Bessel ordinaria de orden $\nu$, donde $\nu$ suele ser el índice de la parte angular (azimutal).
    *   En **coordenadas esféricas**, la parte radial de la ecuación de Helmholtz genera la ecuación de Bessel esférica de orden $n$, donde el término $n(n+1)$ proviene de los autovalores del operador de momento angular $L^2$.

2.  **Naturaleza de las Soluciones:**
    *   Las **funciones cilíndricas** ($J_n$) no tienen una forma cerrada en términos de funciones elementales para argumentos generales.
    *   Las **funciones esféricas** ($j_n, y_n$) son más "simples" en el sentido de que siempre pueden expresarse como una suma finita de funciones trigonométricas multiplicadas por potencias de $x$. Por ejemplo, $j_1(x) = \frac{\sin x}{x^2} - \frac{\cos x}{x}$.

3.  **Regularidad en el Origen:**
    *   Para problemas físicos que incluyen el eje central (cilindro) o el centro (esfera), se debe descartar la segunda solución ($Y_\nu$ o $y_n$) porque **divergen en el origen**. Por lo tanto, en una cavidad cilíndrica se usa solo $J_\nu(k\rho)$ y en un pozo esférico solo $j_n(kr)$.

4.  **Ortogonalidad:**
    *   Ambas familias forman bases completas y ortogonales. La diferencia principal es el **factor de peso** en la integral de ortogonalidad: para las cilíndricas es $\rho$ (o $x$), mientras que para las esféricas es $r^2$ (o $x^2$), lo cual coincide con el elemento de volumen de sus respectivos sistemas de coordenadas.

# Operadores escalera

Los **operadores escalera** ($G_+$ y $G_-$) permiten obtener las funciones de Bessel de orden inmediatamente superior o inferior a partir de una función $J_\nu(x)$ dada. Estos se definen como:

*   **Operador para aumentar el orden ($J_{\nu+1}$):**
    $$J_{\nu+1}(x) = G_+ J_\nu(x) = \left( \frac{\nu}{x} - \frac{d}{dx} \right) J_\nu(x)$$

*   **Operador para disminuir el orden ($J_{\nu-1}$):**
    $$J_{\nu-1}(x) = G_- J_\nu(x) = \left( \frac{\nu}{x} + \frac{d}{dx} \right) J_\nu(x)$$

### Origen de las expresiones
Estas fórmulas se derivan directamente de las dos identidades de recurrencia fundamentales de las funciones de Bessel de primer género:
1.  $\frac{2\nu}{x} J_\nu(x) = J_{\nu-1}(x) + J_{\nu+1}(x)$
2.  $2 J'_\nu(x) = J_{\nu-1}(x) - J_{\nu+1}(x)$

Al sumar y restar estas dos ecuaciones, se aíslan los términos $J_{\nu-1}$ y $J_{\nu+1}$, dando lugar a los operadores diferenciales mencionados.

Para demostrar analíticamente la propiedad **$-\frac{d}{dx} J_0(x) = J_1(x)$** en pocos pasos, se utiliza la definición en serie de potencias de las funciones de Bessel de primer género obtenida mediante el método de Frobenius.

### 1. Definición de la serie para $J_0(x)$
Partimos de la expresión general de la función de Bessel de orden $\nu$ evaluada en **$\nu = 0$**:
$$J_0(x) = \sum_{s=0}^{\infty} \frac{(-1)^s}{(s!)^2} \left( \frac{x}{2} \right)^{2s} = 1 - \frac{x^2}{2^2} + \frac{x^4}{2^2 \cdot 4^2} - \dots$$

### 2. Derivación término a término
Derivamos la serie con respecto a $x$. El término constante ($s=0$) desaparece y la sumatoria comienza desde $s=1$:
$$\frac{d}{dx} J_0(x) = \sum_{s=1}^{\infty} \frac{(-1)^s}{(s!)^2} \cdot (2s) \cdot \left( \frac{x}{2} \right)^{2s-1} \cdot \frac{1}{2}$$
Simplificamos el factor $2s$ con un $s!$ del denominador, recordando que $s! = s(s-1)!$:
$$\frac{d}{dx} J_0(x) = \sum_{s=1}^{\infty} \frac{(-1)^s}{s! (s-1)!} \left( \frac{x}{2} \right)^{2s-1}$$

### 3. Cambio de índice e identificación de $J_1(x)$
Realizamos un cambio de variable en el índice haciendo **$n = s - 1$** (lo que implica $s = n + 1$):
$$\frac{d}{dx} J_0(x) = \sum_{n=0}^{\infty} \frac{(-1)^{n+1}}{n! (n+1)!} \left( \frac{x}{2} \right)^{2n+1}$$
Factorizamos el signo negativo fuera de la sumatoria:
$$\frac{d}{dx} J_0(x) = - \sum_{n=0}^{\infty} \frac{(-1)^n}{n! (n+1)!} \left( \frac{x}{2} \right)^{2n+1}$$

### 4. Conclusión
La expresión resultante dentro de la sumatoria corresponde exactamente a la definición de la serie para la función de Bessel de orden 1, **$J_1(x)$**. Por lo tanto:
$$\frac{d}{dx} J_0(x) = -J_1(x) \implies \mathbf{-\frac{d}{dx} J_0(x) = J_1(x)}$$

Para demostrar que **$J_{1/2}(x) = \sqrt{\frac{2}{\pi x}} \sin(x) = N_{-1/2}(x)$**, utilizaremos las definiciones en serie de las funciones de Bessel y las relaciones analíticas de las funciones de Neumann proporcionadas en las fuentes.

### 1. Demostración de $J_{1/2}(x) = \sqrt{\frac{2}{\pi x}} \sin(x)$
Partimos de la definición general de la **función de Bessel de primer género** de orden $\nu$ obtenida mediante el método de Frobenius,:
$$J_\nu(x) = \sum_{p=0}^{\infty} \frac{(-1)^p}{p! \Gamma(\nu + p + 1)} \left( \frac{x}{2} \right)^{\nu+2p}$$

Sustituimos el orden **$\nu = 1/2$** en la serie:
$$J_{1/2}(x) = \sum_{p=0}^{\infty} \frac{(-1)^p}{p! \Gamma(p + 3/2)} \left( \frac{x}{2} \right)^{1/2+2p}$$

Utilizando la propiedad de la **función Gamma** para argumentos semienteros, donde $\Gamma(p + 3/2) = \frac{(2p+1)! \sqrt{\pi}}{2^{2p+1} p!}$, la expresión se transforma en:
$$J_{1/2}(x) = \sqrt{\frac{x}{2}} \sum_{p=0}^{\infty} \frac{(-1)^p p! 2^{2p+1}}{p! (2p+1)! \sqrt{\pi}} \left( \frac{x^{2p}}{2^{2p}} \right)$$

Simplificando los términos constantes y extrayendo los factores comunes fuera de la sumatoria:
$$J_{1/2}(x) = \sqrt{\frac{2}{\pi x}} \sum_{p=0}^{\infty} \frac{(-1)^p x^{2p+1}}{(2p+1)!}$$

La sumatoria resultante es exactamente la **serie de Taylor para $\sin(x)$**,,. Por lo tanto:
$$\mathbf{J_{1/2}(x) = \sqrt{\frac{2}{\pi x}} \sin(x)}$$

### 2. Relación con la función de Neumann $N_{-1/2}(x)$
La solución de la ecuación de Bessel de segundo género, o **función de Neumann** ($N_\nu$ o $Y_\nu$), se define analíticamente para órdenes no enteros como,,:
$$N_\nu(x) = \frac{J_\nu(x) \cos(\nu\pi) - J_{-\nu}(x)}{\sin(\nu\pi)}$$

Evaluamos esta expresión para el orden **$\nu = -1/2$**:
$$N_{-1/2}(x) = \frac{J_{-1/2}(x) \cos(-\pi/2) - J_{1/2}(x)}{\sin(-\pi/2)}$$

Conocemos los valores de las funciones trigonométricas involucradas:
*   $\cos(-\pi/2) = 0$
*   $\sin(-\pi/2) = -1$

Al sustituir estos valores, el término que contiene a $J_{-1/2}(x)$ desaparece:
$$N_{-1/2}(x) = \frac{0 - J_{1/2}(x)}{-1} = J_{1/2}(x)$$

Como ya demostramos que $J_{1/2}(x) = \sqrt{\frac{2}{\pi x}} \sin(x)$, concluimos que:
$$\mathbf{N_{-1/2}(x) = J_{1/2}(x) = \sqrt{\frac{2}{\pi x}} \sin(x)}$$

Esta identidad confirma que para órdenes semienteros, las funciones de Bessel se reducen a funciones trigonométricas elementales,.

Para demostrar la propiedad **$J_n(x) = x^n \left( -\frac{1}{x} \frac{d}{dx} \right)^n J_0(x)$**, nos basamos en las identidades de derivación y recurrencia de las funciones de Bessel de primer género proporcionadas en las fuentes.

### 1. Demostración de la propiedad general

La demostración se realiza mediante inducción matemática, utilizando la identidad fundamental de derivación:
$$\frac{d}{dx} [x^{-\nu} J_\nu(x)] = -x^{-\nu} J_{\nu+1}(x)$$

Esta relación puede reescribirse para despejar $J_{\nu+1}(x)$ en términos de un operador diferencial actuando sobre $J_\nu$:
$$-x^\nu \frac{d}{dx} [x^{-\nu} J_\nu(x)] = J_{\nu+1}(x) \implies x^{\nu+1} \left( -\frac{1}{x} \frac{d}{dx} \right) [x^{-\nu} J_\nu(x)] = J_{\nu+1}(x)$$

**Paso de inducción:**
Definimos la hipótesis como $\left( -\frac{1}{x} \frac{d}{dx} \right)^n J_0(x) = x^{-n} J_n(x)$.

*   **Caso base ($n=1$):**
    Utilizando la propiedad $\frac{d}{dx} J_0(x) = -J_1(x)$:
    $$\left( -\frac{1}{x} \frac{d}{dx} \right) J_0(x) = -\frac{1}{x} (-J_1(x)) = x^{-1} J_1(x)$$
    Multiplicando por $x^1$, obtenemos $J_1(x) = x \left( -\frac{1}{x} \frac{d}{dx} \right) J_0(x)$, lo cual es correcto.

*   **Paso inductivo:**
    Si asumimos que se cumple para $n$, entonces para $n+1$ aplicamos el operador una vez más:
    $$\left( -\frac{1}{x} \frac{d}{dx} \right)^{n+1} J_0(x) = \left( -\frac{1}{x} \frac{d}{dx} \right) [x^{-n} J_n(x)]$$
    Aplicando la identidad fundamental con $\nu = n$:
    $$\left( -\frac{1}{x} \frac{d}{dx} \right) [x^{-n} J_n(x)] = -\frac{1}{x} [-x^{-n} J_{n+1}(x)] = x^{-(n+1)} J_{n+1}(x)$$
    Multiplicando ambos lados por $x^{n+1}$, llegamos a la expresión general:
    $$\mathbf{J_n(x) = x^n \left( -\frac{1}{x} \frac{d}{dx} \right)^n J_0(x)}$$

---

### 2. Desarrollo para $n=2$

Para desarrollar la expresión cuando **$n=2$**, aplicamos el operador diferencial sucesivamente:

$$J_2(x) = x^2 \left( -\frac{1}{x} \frac{d}{dx} \right) \left[ \left( -\frac{1}{x} \frac{d}{dx} \right) J_0(x) \right]$$

1.  **Primera aplicación del operador:**
    Como se vio anteriormente, $\left( -\frac{1}{x} \frac{d}{dx} \right) J_0(x) = \frac{J_1(x)}{x}$.

2.  **Segunda aplicación del operador:**
    $$J_2(x) = x^2 \left( -\frac{1}{x} \frac{d}{dx} \right) \left[ \frac{J_1(x)}{x} \right] = x^2 \left[ -\frac{1}{x} \frac{d}{dx} (x^{-1} J_1(x)) \right]$$
    Utilizamos la regla del producto para derivar el término entre corchetes:
    $$\frac{d}{dx} (x^{-1} J_1(x)) = -x^{-2} J_1(x) + x^{-1} J_1'(x)$$
    Sustituimos $J_1'(x)$ usando la relación $J_1'(x) = J_0(x) - \frac{1}{x} J_1(x)$ (derivada de la recurrencia):
    $$\frac{d}{dx} (x^{-1} J_1(x)) = -x^{-2} J_1(x) + x^{-1} [J_0(x) - \frac{1}{x} J_1(x)] = x^{-1} J_0(x) - 2x^{-2} J_1(x)$$

3.  **Resultado final en términos de $J_0$ y $J_1$:**
    $$J_2(x) = x^2 \left[ -\frac{1}{x} (x^{-1} J_0(x) - 2x^{-2} J_1(x)) \right] = x^2 [ -x^{-2} J_0(x) + 2x^{-3} J_1(x) ]$$
    $$\mathbf{J_2(x) = \frac{2}{x} J_1(x) - J_0(x)}$$

Este resultado coincide con la **fórmula de recurrencia estándar** para las funciones de Bessel de primer género. También puede expresarse puramente en términos de derivadas de $J_0(x)$ como:
$$\mathbf{J_2(x) = J_0''(x) - \frac{1}{x} J_0'(x)}$$

La **función generatriz** de las funciones de Bessel de primer género y orden entero, denotada como $g(x, t)$, es una herramienta analítica que encapsula a toda la familia de estas funciones ($J_n$) en una sola expresión mediante una expansión en serie de Laurent.

### **Expresión Matemática**
La expresión matemática fundamental es:
$$\mathbf{g(x, t) = e^{\frac{x}{2}(t - \frac{1}{t})} = \sum_{n=-\infty}^{\infty} J_n(x) t^n}$$

Esta serie es válida para cualquier valor de $x$ y para todo $t \neq 0$.

### **Importancia de la función generatriz**
Su importancia en la física matemática radica en que permite manipular y deducir propiedades complejas de las funciones de Bessel de forma algebraica simple:

1.  **Deducción de Relaciones de Recurrencia:** Al derivar $g(x, t)$ respecto a la variable física $x$ o al parámetro auxiliar $t$, se obtienen directamente las identidades que conectan funciones de órdenes distintos, como:
    *   $\frac{2n}{x} J_n(x) = J_{n-1}(x) + J_{n+1}(x)$.
    *   $2J'_n(x) = J_{n-1}(x) - J_{n+1}(x)$.
2.  **Expansión de Jacobi-Anger:** Permite representar ondas planas como una superposición de modos cilíndricos. Por ejemplo, al sustituir $t = e^{i\theta}$, se obtiene la identidad $\mathbf{e^{ix \sin \theta} = \sum_{n=-\infty}^{\infty} J_n(x) e^{in\theta}}$, fundamental en el estudio de la **difracción**, antenas y modulación de señales.
3.  **Establecimiento de Propiedades y Cotas:** Es el punto de partida para demostrar la paridad de las funciones ($J_{-n}(x) = (-1)^n J_n(x)$) y para establecer cotas de amplitud, como mostrar que $|J_0(x)| \leq 1$ mediante la identidad del producto de funciones generatrices.
4.  **Modelado Físico:** Facilita la resolución de problemas con **simetría cilíndrica**, como la vibración de membranas o la distribución de temperatura en cilindros, al proveer una base completa para expandir funciones iniciales.

Para visualizar cómo la **función generatriz** $g(x, t)$ produce las funciones de Bessel, realizamos una expansión en serie de potencias y agrupamos los términos según las potencias del parámetro auxiliar $t$. El procedimiento, detallado en las fuentes, es el siguiente:

### 1. Descomposición de la función generatriz
Partimos de la definición de la función generatriz de orden entero:
$$g(x, t) = e^{\frac{x}{2}(t - \frac{1}{t})} = e^{xt/2} \cdot e^{-x/2t}$$

### 2. Expansión en series de potencias
Expandimos ambos exponenciales utilizando su serie de Maclaurin estándar ($e^u = \sum \frac{u^k}{k!}$):
*   **Primer factor:** $e^{xt/2} = \sum_{r=0}^{\infty} \frac{(x/2)^r t^r}{r!}$
*   **Segundo factor:** $e^{-x/2t} = \sum_{s=0}^{\infty} \frac{(-1)^s (x/2)^s t^{-s}}{s!}$

Al multiplicar ambas series, obtenemos una **doble sumatoria** que contiene todas las combinaciones posibles de $r$ y $s$:
$$g(x, t) = \sum_{r=0}^{\infty} \sum_{s=0}^{\infty} \frac{(-1)^s (x/2)^{r+s} t^{r-s}}{r! s!}$$

### 3. Agrupamiento de términos (Cambio de índice)
Para visualizar el resultado de la serie de Bessel, debemos agrupar los términos que tienen la **misma potencia de $t$**. Definimos un nuevo índice $n = r - s$, que representa el orden de la función de Bessel. Esto implica que $r = n + s$.

Sustituyendo $r$ en la sumatoria y reorganizando para aislar $t^n$:
$$g(x, t) = \sum_{n=-\infty}^{\infty} \left[ \sum_{s=0}^{\infty} \frac{(-1)^s}{s! (n+s)!} \left( \frac{x}{2} \right)^{n+2s} \right] t^n$$

*Nota: Para valores de $n$ negativos, la suma interna comienza desde $s = -n$ para asegurar que el factorial $(n+s)!$ sea de un número no negativo.*

### 4. Identificación del resultado
Al observar el término entre corchetes, reconocemos exactamente la **definición en serie de la función de Bessel de primer género y orden $n$**, $J_n(x)$:
$$J_n(x) = \sum_{s=0}^{\infty} \frac{(-1)^s}{s! (n+s)!} \left( \frac{x}{2} \right)^{n+2s}$$

Por lo tanto, la expansión completa se visualiza como una serie de Laurent en $t$ donde cada coeficiente es una función de Bessel:
$$\mathbf{g(x, t) = \sum_{n=-\infty}^{\infty} J_n(x) t^n}$$

### Visualización conceptual de los términos:
Al expandir y agrupar, los términos se ven de la siguiente manera:
*   **Términos constantes ($t^0$):** Se obtienen cuando $r=s$, dando como resultado la serie de $J_0(x)$.
*   **Términos con $t^1$:** Se obtienen cuando $r=s+1$, resultando en la serie de $J_1(x)$.
*   **Términos con $t^{-1}$:** Se obtienen cuando $s=r+1$, resultando en $J_{-1}(x) = -J_1(x)$.

Este agrupamiento demuestra que la función generatriz es una "máquina" que contiene codificada toda la familia de funciones de Bessel de orden entero en una sola expresión compacta.

# Tarea

Usar la función g(x, t) para mostrar que

* $J_{\nu+1}(x)-J_{\nu-1}(x) = -2 d/dx J_\nu$
* $J_{\nu+1}(x)+J_{\nu-1}(x) = 2\nu/x J_\nu$

Para demostrar estas identidades de recurrencia utilizando la **función generatriz** $g(x, t)$ de las funciones de Bessel de orden entero ($J_n$), partimos de su definición en serie de Laurent:

$$g(x, t) = e^{\frac{x}{2}\left(t - \frac{1}{t}\right)} = \sum_{n=-\infty}^{\infty} J_n(x) t^n$$

### 1. Demostración de $J_{n+1}(x) - J_{n-1}(x) = -2 \frac{d}{dx} J_n(x)$

Esta relación vincula la derivada de la función con las funciones de órdenes adyacentes.

*   **Paso 1:** Diferenciamos la función generatriz respecto a la variable física **$x$**:
    $$\frac{\partial g}{\partial x} = \frac{1}{2}\left(t - \frac{1}{t}\right) e^{\frac{x}{2}\left(t - \frac{1}{t}\right)} = \frac{1}{2}\left(t - \frac{1}{t}\right) \sum_{n=-\infty}^{\infty} J_n(x) t^n$$
*   **Paso 2:** Diferenciamos la expansión en serie término a término respecto a **$x$**:
    $$\frac{\partial g}{\partial x} = \sum_{n=-\infty}^{\infty} J'_n(x) t^n$$
*   **Paso 3:** Igualamos ambas expresiones y distribuimos el factor sobre la serie:
    $$\sum_{n=-\infty}^{\infty} J'_n(x) t^n = \frac{1}{2} \sum_{n=-\infty}^{\infty} J_n(x) t^{n+1} - \frac{1}{2} \sum_{n=-\infty}^{\infty} J_n(x) t^{n-1}$$
*   **Paso 4:** Alineamos las potencias de **$t$** a la forma $t^n$ realizando cambios de índice:
    *   En el primer término de la derecha, sustituimos $n \to n-1$, obteniendo $\sum J_{n-1}(x) t^n$.
    *   En el segundo término de la derecha, sustituimos $n \to n+1$, obteniendo $\sum J_{n+1}(x) t^n$.
*   **Paso 5:** Igualamos los coeficientes de $t^n$:
    $$J'_n(x) = \frac{1}{2} [J_{n-1}(x) - J_{n+1}(x)]$$
    Multiplicando por $-2$, obtenemos la identidad solicitada:
    $$\mathbf{J_{n+1}(x) - J_{n-1}(x) = -2 \frac{d}{dx} J_n(x)}$$

---

### 2. Demostración de $J_{n+1}(x) + J_{n-1}(x) = \frac{2n}{x} J_n(x)$

Esta relación conecta los valores de la función de diferentes órdenes sin involucrar derivadas.

*   **Paso 1:** Diferenciamos la función generatriz respecto a la variable auxiliar **$t$**:
    $$\frac{\partial g}{\partial t} = \frac{x}{2}\left(1 + \frac{1}{t^2}\right) e^{\frac{x}{2}\left(t - \frac{1}{t}\right)} = \frac{x}{2}\left(1 + \frac{1}{t^2}\right) \sum_{n=-\infty}^{\infty} J_n(x) t^n$$
*   **Paso 2:** Diferenciamos la serie respecto a **$t$** término a término:
    $$\frac{\partial g}{\partial t} = \sum_{n=-\infty}^{\infty} n J_n(x) t^{n-1}$$
*   **Paso 3:** Igualamos y distribuimos los términos:
    $$\sum_{n=-\infty}^{\infty} n J_n(x) t^{n-1} = \frac{x}{2} \sum_{n=-\infty}^{\infty} J_n(x) t^n + \frac{x}{2} \sum_{n=-\infty}^{\infty} J_n(x) t^{n-2}$$
*   **Paso 4:** Ajustamos los índices para obtener la potencia común **$t^{n-1}$**:
    *   En el primer término de la derecha, sustituimos $n \to n-1$, obteniendo $\sum J_{n-1}(x) t^{n-1}$.
    *   En el segundo término de la derecha, sustituimos $n \to n+1$, obteniendo $\sum J_{n+1}(x) t^{n-1}$.
*   **Paso 5:** Igualamos los coeficientes de $t^{n-1}$:
    $$n J_n(x) = \frac{x}{2} [J_{n-1}(x) + J_{n+1}(x)]$$
    Reorganizando algebraicamente, llegamos a la segunda identidad:
    $$\mathbf{J_{n+1}(x) + J_{n-1}(x) = \frac{2n}{x} J_n(x)}$$

Estas propiedades, derivadas para órdenes enteros $n$, suelen extenderse a órdenes reales $\nu$ en la física matemática.

La **ecuación de Bessel**, expresada en términos de la coordenada radial $\rho$ (frecuente en problemas con simetría cilíndrica), se completa de la siguiente manera:

$$\mathbf{\frac{d^2J_\nu(k\rho)}{d\rho^2} + \frac{1}{\rho}\frac{dJ_\nu(k\rho)}{d\rho} + \left(k^2 - \frac{\nu^2}{\rho^2}\right)J_\nu(k\rho) = 0}$$

Esta forma surge directamente al aplicar el método de **separación de variables** a las ecuaciones de Laplace o Helmholtz en coordenadas cilíndricas.

### Detalles de la ecuación:
*   **Origen:** Proviene de la parte radial de la ecuación diferencial, que originalmente se escribe como $\frac{1}{\rho}\frac{d}{d\rho}\left(\rho\frac{dR}{d\rho}\right) + \left(k^2 - \frac{\nu^2}{\rho^2}\right)R = 0$. Al expandir la derivada del producto, se obtiene la expresión solicitada.
*   **Parámetros:**
    *   **$k$:** Es la constante de separación asociada a la dependencia en el eje $z$ o a la frecuencia en el caso de ondas.
    *   **$\nu$ (o $n$):** Representa el orden de la función de Bessel, que generalmente es un número entero cuando se requiere continuidad en el ángulo azimutal $\phi$.
*   **Forma estándar:** Si se realiza el cambio de variable $x = k\rho$, la ecuación adopta su forma más conocida:
    $$x^2 \frac{d^2y}{dx^2} + x \frac{dy}{dx} + (x^2 - \nu^2)y = 0 \quad$$

Esta ecuación es fundamental para describir fenómenos como la **vibración de membranas circulares**, la propagación de ondas en **guías de onda** y la distribución de potenciales en cilindros.

$\lambda_n, f_n \to k^2, J_\nu(k\rho)$

$\lambda_m, f_m \to k'^2, J_\nu(k'\rho)$

Para analizar el **Wronskiano** y la **condición de hermiticidad** aplicados a las funciones de Bessel en el marco de la teoría de **Sturm-Liouville (S.L.)**, partimos de la ecuación radial de Bessel identificando los elementos de tu consulta: $\lambda \to k^2$ y $f \to J_\nu(k\rho)$.

### 1. Estructura de Sturm-Liouville
La ecuación radial de Bessel se puede escribir en su **forma autoadjunta** como,,:
$$\frac{d}{d\rho} \left[ \rho \frac{dJ_\nu(k\rho)}{d\rho} \right] + \left( k^2 \rho - \frac{\nu^2}{\rho} \right) J_\nu(k\rho) = 0$$

Comparando con la forma canónica de S.L., $\frac{d}{dx}[q(x)f'] + [r(x) + \lambda p(x)]f = 0$, identificamos,:
*   **$q(\rho) = \rho$** (coeficiente de la derivada).
*   **$p(\rho) = \rho$** (factor de peso para la ortogonalidad).
*   **$r(\rho) = -\nu^2/\rho$**.
*   **$\lambda = k^2$** (autovalor).

### 2. El Wronskiano y la Identidad de Lommel
Para dos autofunciones $f_n = J_\nu(k\rho)$ y $f_m = J_\nu(k'\rho)$ con autovalores distintos ($k^2 \neq k'^2$), la integración de la ecuación diferencial en un intervalo $[a, b]$ conduce a la siguiente relación fundamental,,:
$$(k'^2 - k^2) \int_a^b \rho J_\nu(k\rho) J_\nu(k'\rho) d\rho = \left[ \rho W(J_\nu(k\rho), J_\nu(k'\rho)) \right]_a^b$$

Donde el **Wronskiano** $W$ está definido por la relación entre las funciones y sus derivadas respecto a $\rho$:
$$W(f_n, f_m) = J_\nu(k\rho) \frac{d}{d\rho}J_\nu(k'\rho) - J_\nu(k'\rho) \frac{d}{d\rho}J_\nu(k\rho)$$
$$W = J_\nu(k\rho) [k' J'_\nu(k'\rho)] - J_\nu(k'\rho) [k J'_\nu(k\rho)] \quad$$

### 3. Condición de Hermiticidad
Para que el operador diferencial asociado sea **Hermítico**, el término de frontera resultante de la integración debe anularse,,:
$$\mathbf{\left[ \rho W(J_\nu(k\rho), J_\nu(k'\rho)) \right]_a^b = 0}$$

Si esta condición se cumple, las autofunciones correspondientes a autovalores distintos son **ortogonales** con factor de peso $\rho$, es decir, $\int_a^b \rho J_\nu(k\rho) J_\nu(k'\rho) d\rho = 0$.

#### **Análisis en el intervalo típico $[0, R]$:**
*   **En el origen ($\rho = 0$):** Para $\nu > -1$, el término $\rho W$ se anula automáticamente debido a que $J_\nu(k\rho) \propto \rho^\nu$, por lo que $\rho \cdot \rho^\nu \cdot \rho^{\nu-1} \to 0$ cuando $\rho \to 0$,.
*   **En la frontera ($\rho = R$):** La hermiticidad depende de que el Wronskiano sea cero en $R$. Esto se logra imponiendo condiciones de frontera que determinan los autovalores permitidos para $k$ y $k'$,:
    *   **Dirichlet:** $J_\nu(kR) = 0$ (donde $kR = \alpha_{\nu n}$, la $n$-ésima raíz de la función),.
    *   **Neumann:** $J'_\nu(kR) = 0$ (donde $kR = \beta_{\nu n}$, la $n$-ésima raíz de la derivada),.
    *   **Mixtas (Robin):** $\alpha J_\nu(kR) + \beta J'_\nu(kR) = 0$,.

**Resumen técnico:**
Bajo estas condiciones, el problema de Bessel es un **sistema de Sturm-Liouville autoadjunto y hermítico**, lo que garantiza que:
1.  Los autovalores $\lambda_n = k^2$ son **reales**,.
2.  Las funciones de Bessel $J_\nu(k_n\rho)$ forman una **base completa y ortogonal** en el espacio de Hilbert $L^2_\rho[0, R]$, permitiendo expandir funciones arbitrarias en series de Fourier-Bessel,,.

Cuando el parámetro **$k'$ tiende a $k$** en el límite de una integral de ortogonalidad (como las de Bessel), se produce una indeterminación de tipo **$0/0$** que, al resolverse, permite hallar la **normalización** de la función.

Este procedimiento es fundamental en la teoría de Sturm-Liouville y se detalla a continuación basándose en las fuentes:

### 1. El problema de la indeterminación
Para las funciones de Bessel, la relación de ortogonalidad (identidad de Lommel) establece que:
$$\int_{a}^{b} \rho J_{\nu}(k\rho) J_{\nu}(k'\rho) d\rho = \frac{\left[ \rho W(J_{\nu}(k'\rho), J_{\nu}(k\rho)) \right]_{a}^{b}}{k^2 - k'^2}$$
Si simplemente se hace $k' = k$, el denominador se anula y el numerador (el Wronskiano) también se anula (ya que es el Wronskiano de una función consigo misma), resultando en una forma divergente o indeterminada.

### 2. Resolución mediante la Regla de l'Hôpital
Para encontrar el valor de la integral cuando los autovalores coinciden ($k=k'$), se aplica el **límite cuando $k' \to k$** utilizando la regla de l'Hôpital. Esto implica derivar el numerador y el denominador respecto a $k'$ antes de evaluar el límite.

En el caso específico de un intervalo $[0, b]$ donde $J_{\nu}(kb) = 0$, el proceso es:
*   **Numerador:** Se deriva el término del Wronskiano en la frontera respecto a $k'$.
*   **Denominador:** La derivada de $(k^2 - k'^2)$ respecto a $k'$ es $-2k'$.
*   **Resultado:** Al evaluar el límite, la indeterminación desaparece y se obtiene la **norma** de la función.

### 3. Resultado: La integral de normalización
Al aplicar este límite para las funciones de Bessel de primer género, se llega al factor de normalización estándar:
$$\int_{0}^{b} \rho [J_{\nu}(k\rho)]^2 d\rho = \mathbf{\frac{b^2}{2} [J_{\nu+1}(kb)]^2}$$
Este resultado es vital porque define la "longitud" de la función en el espacio de Hilbert y permite usar estas funciones como una base ortonormalizada.

### 4. Transición al continuo (Delta de Dirac)
En un contexto más general, cuando el espaciamiento entre autovalores tiende a cero ($k' \to k$ en un dominio infinito), la **delta de Kronecker** ($\delta_{nn'}$) se transforma en una **delta de Dirac** ($\delta(k - k')$). 
*   En bases discretas, el límite $k' \to k$ da un número finito (la norma).
*   En bases continuas, este límite representa un pico de intensidad infinita en el punto de coincidencia, lo que constituye la representación integral de la delta de Dirac.

En resumen, hacer que $k'$ tienda a $k$ es el paso matemático necesario para **pasar de la ortogonalidad (distinguir funciones) a la normalización (medir la función)**.
