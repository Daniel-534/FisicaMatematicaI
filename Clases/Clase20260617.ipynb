# Resumen

1. **Ecuación de Laguerre asociada**
$x\ddot{y} + (k + 1 - x)\dot{y} + ny = 0$.

2. **Polinomios de Laguerre asociados escritos con la formula de Rodrigues**
$L_n^k(x) = \frac{e^x x^{-k}}{n!} \frac{d^n}{dx^n} (e^{-x} x^{n+k})$.

**Tarea**

Mostrar que:

$$\psi_p^k = x^{(k+1)/2}e^{-x/2}L_p^k(x)$$

cumple la siguiente ecuación:

$$\ddot{\psi}_p^k + \psi_p^k \left[-\frac{k^2-1}{4x^2} + \frac{2p+k+1}{2x} - \frac{1}{4}\right] = 0$$



**Solución**
### Ecuación diferencial de los polinomios de Laguerre asociados

Los polinomios de Laguerre asociados $L_p^k(x)$ son soluciones de la ecuación diferencial asociada de Laguerre:

$$x \ddot{L}_p^k + (k+1-x)\dot{L}_p^k + p L_p^k = 0$$

Para simplificar la notación en las derivadas, denotaremos $L \equiv L_p^k(x)$. La ecuación anterior se puede reescribir agrupando los términos con derivadas:

$$x \ddot{L} + (k+1-x)\dot{L} = -p L \quad \text{--- (Ec. 1)}$$

### Cálculo de la primera derivada de $\psi_p^k$

Definimos nuestra función como:
$$\psi = x^{(k+1)/2} e^{-x/2} L$$

Aplicamos la regla del producto para derivar respecto a $x$ (usando el punto para denotar $d/dx$):

$$\dot{\psi} = \left[ \frac{k+1}{2} x^{(k-1)/2} \right] e^{-x/2} L + x^{(k+1)/2} \left[ -\frac{1}{2} e^{-x/2} \right] L + x^{(k+1)/2} e^{-x/2} \dot{L}$$

Factorizamos el término común $x^{(k-1)/2} e^{-x/2}$:

$$\dot{\psi} = x^{(k-1)/2} e^{-x/2} \left[ \frac{k+1}{2} L - \frac{x}{2} L + x \dot{L} \right]$$

Sacando un factor de $1/2$ en el corchete, obtenemos una expresión compacta para la primera derivada:

$$\dot{\psi} = \frac{1}{2} x^{(k-1)/2} e^{-x/2} \Big[ (k+1-x)L + 2x\dot{L} \Big] \quad \text{--- (Ec. 2)}$$

### Cálculo de la segunda derivada de $\psi_p^k$

Para hallar $\ddot{\psi}$, derivamos la **Ec. 2**. Definimos $A(x) = x^{(k-1)/2} e^{-x/2}$ y $B(x) = (k+1-x)L + 2x\dot{L}$, de modo que $\dot{\psi} = \frac{1}{2} A(x) B(x)$.

La derivada de $A(x)$ es:
$$\dot{A} = \frac{k-1}{2} x^{(k-3)/2} e^{-x/2} - \frac{1}{2} x^{(k-1)/2} e^{-x/2} = \frac{1}{2} x^{(k-3)/2} e^{-x/2} (k-1-x)$$

La derivada de $B(x)$ es:
$$\dot{B} = -L + (k+1-x)\dot{L} + 2\dot{L} + 2x\ddot{L} = -L + (k+3-x)\dot{L} + 2x\ddot{L}$$

Aplicando la regla del producto $\ddot{\psi} = \frac{1}{2}(\dot{A}B + A\dot{B})$:

$$\ddot{\psi} = \frac{1}{4} x^{(k-3)/2} e^{-x/2} (k-1-x) \Big[ (k+1-x)L + 2x\dot{L} \Big] + \frac{1}{2} x^{(k-1)/2} e^{-x/2} \Big[ -L + (k+3-x)\dot{L} + 2x\ddot{L} \Big]$$

Para simplificar, factorizamos $\frac{1}{4} x^{(k-3)/2} e^{-x/2}$ en toda la expresión (notando que $x^{(k-1)/2} = x \cdot x^{(k-3)/2}$, lo que multiplica el segundo término por $2x$):

$$\ddot{\psi} = \frac{1}{4} x^{(k-3)/2} e^{-x/2} \Big\{ (k-1-x)(k+1-x)L + 2x(k-1-x)\dot{L} + 2x \big[ -L + (k+3-x)\dot{L} + 2x\ddot{L} \big] \Big\}$$

### Agrupación de términos

Expandimos y agrupamos los términos dentro de las llaves según las funciones $L$, $\dot{L}$ y $\ddot{L}$:

*   **Términos con $L$:**
    $$(k-1-x)(k+1-x) - 2x = (k-x)^2 - 1 - 2x = x^2 - 2(k+1)x + (k^2 - 1)$$
*   **Términos con $\dot{L}$:**
    $$2x(k-1-x) + 2x(k+3-x) = 2kx - 2x - 2x^2 + 2kx + 6x - 2x^2 = 4x(k+1-x)$$
*   **Términos con $\ddot{L}$:**
    $$2x(2x\ddot{L}) = 4x^2\ddot{L}$$

Sustituyendo esto de nuevo en la expresión de $\ddot{\psi}$:

$$\ddot{\psi} = \frac{1}{4} x^{(k-3)/2} e^{-x/2} \Big\{ 4x^2\ddot{L} + 4x(k+1-x)\dot{L} + \big[ x^2 - 2(k+1)x + (k^2 - 1) \big] L \Big\}$$

### Sustitución de la ecuación de Laguerre

Observamos que los dos primeros términos dentro de las llaves son exactamente $4x$ veces la **Ec. 1**. 
Si multiplicamos la **Ec. 1** por $4x$, obtenemos:
$$4x^2\ddot{L} + 4x(k+1-x)\dot{L} = -4pxL$$

Sustituimos esta identidad en nuestra expresión para $\ddot{\psi}$:

$$\ddot{\psi} = \frac{1}{4} x^{(k-3)/2} e^{-x/2} \Big\{ -4pxL + \big[ x^2 - 2(k+1)x + (k^2 - 1) \big] L \Big\}$$

Factorizamos $L$:

$$\ddot{\psi} = \frac{1}{4} x^{(k-3)/2} e^{-x/2} L \Big\{ x^2 - (2k + 2 + 4p)x + (k^2 - 1) \Big\}$$

### Expresión final en términos de $\psi$

Recordemos que la función original es $\psi = x^{(k+1)/2} e^{-x/2} L$. 
Podemos reescribir el prefactor de nuestra ecuación como:
$$\frac{1}{4} x^{(k-3)/2} e^{-x/2} L = \frac{1}{4x^2} \left( x^{(k+1)/2} e^{-x/2} L \right) = \frac{\psi}{4x^2}$$

Sustituyendo esto:

$$\ddot{\psi} = \frac{\psi}{4x^2} \Big[ x^2 - (2p + k + 1)2x + (k^2 - 1) \Big]$$

Distribuimos el $\frac{1}{4x^2}$ en cada término del corchete:

$$\ddot{\psi} = \psi \left[ \frac{x^2}{4x^2} - \frac{2(2p + k + 1)x}{4x^2} + \frac{k^2 - 1}{4x^2} \right]$$

$$\ddot{\psi} = \psi \left[ \frac{1}{4} - \frac{2p + k + 1}{2x} + \frac{k^2 - 1}{4x^2} \right]$$

Finalmente, pasamos todos los términos al lado izquierdo de la ecuación y factorizamos un signo negativo dentro del corchete para llegar a la forma solicitada:

$$\ddot{\psi} + \psi \left[ -\frac{k^2-1}{4x^2} + \frac{2p+k+1}{2x} - \frac{1}{4} \right] = 0$$

# Átomo de Hidrógeno (H)

### El Átomo de Hidrógeno: Del Núcleo a la Nube Cuántica

**El experimento de Rutherford y el nacimiento del núcleo**
A principios del siglo XX, el modelo atómico aceptado era el de Thomson, conocido como el "pudin de pasas": se creía que el átomo era una esfera difusa de carga positiva con los electrones incrustados en ella, como pasas en un pastel. En 1911, Ernest Rutherford decidió poner a prueba este modelo utilizando un haz de partículas alfa (que son pesadas, rápidas y tienen carga positiva) disparadas contra una lámina de oro extremadamente delgada. 

La gran mayoría de las partículas atravesaban la lámina sin desviarse, lo que demostraba que el átomo es casi todo espacio vacío. Sin embargo, el hallazgo revolucionario fue que **una pequeñísima fracción de las partículas alfa rebotaba violentamente hacia atrás**. Rutherford famosamente comparó esto con disparar un obús de quince pulgadas a un trozo de papel higiénico y que el obús rebotara hacia ti. Clásicamente, esto no tenía ningún sentido bajo el modelo de Thomson; una carga difusa no podría ejercer la fuerza repulsiva suficiente para frenar y devolver una partícula alfa pesada y veloz. La única explicación lógica era que casi toda la masa del átomo y toda su carga positiva estaban concentradas en un volumen infinitesimal y ultra-denso en el centro: **había nacido el núcleo atómico**.

**Niveles de energía y la emisión de luz**
Aunque el modelo de Rutherford explicaba la estructura básica, la física clásica predecía que un electrón orbitando el núcleo debería irradiar energía continuamente en forma de ondas electromagnéticas, perdiendo velocidad y espiralando hasta colapsar contra el núcleo en una fracción de segundo. Los átomos, por tanto, no deberían ser estables. 

Para resolver esto, Niels Bohr propuso que los electrones no podían orbitar a cualquier distancia, sino que estaban restringidos a **niveles de energía específicos y cuantizados** (los famosos $n=1, 2, 3...$). Mientras el electrón se mantuviera en uno de estos niveles, no emitía energía. 

Sin embargo, cuando un electrón absorbía energía (por ejemplo, por calor o electricidad), saltaba a un nivel superior ($n$), entrando en un estado "excitado". Como este estado es inestable, el electrón tiende a caer de regreso a un nivel inferior ($n'$). Aquí entra la conservación de la energía: el electrón no puede simplemente "perder" la diferencia de energía entre el nivel alto y el bajo. En su lugar, **emite esa energía exacta al espacio en forma de un paquete discreto de luz llamado fotón**. El color (la frecuencia) de esa luz depende exactamente de la "distancia" energética entre ambos niveles. Esto explicaba por qué el gas de hidrógeno, al ser excitado, no brilla con un arcoíris continuo, sino que emite líneas de colores muy específicos y puros.

**La revolución de la Mecánica Cuántica (circa 1923)**
Aunque el modelo de Bohr funcionaba matemáticamente para el hidrógeno, era un "parche" clásico: seguía tratando al electrón como una bolita con una órbita definida. 

Alrededor de 1923-1924, la física experimentó un quiebre conceptual total. Físicos como Louis de Broglie propusieron que si la luz (una onda) podía comportarse como partícula, entonces la materia (como el electrón) también debía comportarse como una **onda**. 

Esta idea detonó el nacimiento de la **Mecánica Cuántica** moderna, desarrollada a mediados de los años 20 por Schrödinger, Heisenberg y Dirac. En este nuevo marco, la idea de "órbitas" circulares fue destruida. El electrón dejó de ser una partícula puntual girando alrededor del núcleo para convertirse en una **nube de probabilidad** (un orbital). Los números $n$ dejaron de ser distancias físicas y se convirtieron en "números cuánticos" que definían la energía, la forma y el tamaño de estas complejas nubes tridimensionales donde es *probable* encontrar al electrón. El átomo de hidrógeno pasó de ser un sistema solar en miniatura a ser el primer y más puro laboratorio de la extraña y probabilista naturaleza cuántica del universo.

# Mecánica clásica del electrón

**Momento lineal:**
$$\vec{p} = m\vec{v}$$

**Energía total (clásica):**
$$E = K + V = \frac{1}{2}mv^2 + V(\vec{r})$$

Para el átomo de hidrógeno, el potencial es coulombiano (atractivo):
$$V(r) = -\frac{ke^2}{r} = -\frac{e^2}{4\pi\varepsilon_0 r}$$

Reescribiendo la energía en términos del momento $p = mv$:
$$E = \frac{p^2}{2m} - \frac{ke^2}{r}$$

---

## Transición a Mecánica Cuántica (Postulados)

En mecánica cuántica, las observables físicas se reemplazan por **operadores**:

$$\vec{p} \longrightarrow \hat{p} = -i\hbar\vec{\nabla}$$

$$E \longrightarrow \hat{E} = i\hbar\frac{\partial}{\partial t}$$

---

## Ecuación de Schrödinger

Sustituyendo los operadores en la expresión clásica de la energía:

$$\left[\frac{(-i\hbar\vec{\nabla})^2}{2m} - \frac{ke^2}{r}\right]\psi = E\psi$$

$$\left[-\frac{\hbar^2\nabla^2}{2m} - \frac{ke^2}{r}\right]\psi = E\psi$$

Reorganizando:

$$-\frac{\hbar^2}{2m}\nabla^2\psi - \frac{ke^2}{r}\psi = E\psi$$

Definiendo el **operador Hamiltoniano** $\hat{H}$:

$$\hat{H}\psi = E\psi$$

donde:
$$\hat{H} = -\frac{\hbar^2}{2m}\nabla^2 - \frac{ke^2}{r} = -\frac{\hbar^2}{2m}\nabla^2 + V(r)$$

Esta es la **ecuación de Schrödinger independiente del tiempo** para el átomo de hidrógeno.

---

## Solución en Coordenadas Esféricas

Dado que el potencial $V(r)$ tiene simetría esférica (solo depende de $r$), es natural usar coordenadas esféricas $(r, \theta, \varphi)$.

**Separación de variables:**

Proponemos una solución de la forma:
$$\psi(r, \theta, \varphi) = R(r)Y(\theta, \varphi)$$

Para simplificar la ecuación radial, se hace el cambio:
$$R(r) = \frac{u(r)}{r}$$

Por lo tanto:
$$\psi(r, \theta, \varphi) = \frac{u(r)}{r}Y(\theta, \varphi)$$

**¿Por qué este cambio?** Al sustituir $R(r) = u(r)/r$ en el laplaciano esférico, la ecuación radial se simplifica a una forma similar a la ecuación de Schrödinger unidimensional:

$$-\frac{\hbar^2}{2m}\frac{d^2u}{dr^2} + \left[V(r) + \frac{\hbar^2\ell(\ell+1)}{2mr^2}\right]u = Eu$$

donde el término $\frac{\hbar^2\ell(\ell+1)}{2mr^2}$ es el **potencial centrífugo** que aparece al separar la parte angular.

---

La solución completa lleva a los niveles de energía cuantizados del átomo de hidrógeno:
$$E_n = -\frac{me^4}{2\hbar^2 n^2} = -\frac{13.6 \text{ eV}}{n^2}$$

Para volver de la variable radial $u(r)$ a la función de onda completa $\psi(r, \theta, \varphi)$ y simplificar, recordemos la relación entre ambas. En la separación de variables de la ecuación de Schrödinger en coordenadas esféricas, definimos:

$$ \psi(r, \theta, \varphi) = R(r) Y_\ell^m(\theta, \varphi) $$

Y para simplificar la ecuación radial, se hace el cambio $u(r) = r R(r)$, de modo que:

$$ \psi(r, \theta, \varphi) = \frac{u(r)}{r} Y_\ell^m(\theta, \varphi) $$

### 1. Volviendo a $\psi$ y simplificando

Partimos de tu ecuación radial:

$$ -\frac{\hbar^2}{2m}\frac{d^2u}{dr^2} + \left[V(r) + \frac{\hbar^2\ell(\ell+1)}{2mr^2}\right]u = Eu $$

Multiplicamos toda la ecuación por $\frac{Y_\ell^m(\theta, \varphi)}{r}$:

$$ -\frac{\hbar^2}{2m} \frac{d^2u}{dr^2} \frac{Y_\ell^m}{r} + V(r) \frac{u}{r} Y_\ell^m + \frac{\hbar^2\ell(\ell+1)}{2mr^2} \frac{u}{r} Y_\ell^m = E \frac{u}{r} Y_\ell^m $$

Reconocemos que $\frac{u}{r} Y_\ell^m = \psi$. Ahora analizamos los otros términos:

**Primer término (derivada radial):**
Como $u = r \frac{\psi}{Y_\ell^m}$, la derivada segunda respecto a $r$ es $\frac{d^2u}{dr^2} = \frac{1}{Y_\ell^m} \frac{\partial^2}{\partial r^2}(r\psi)$. Por lo tanto:
$$ \frac{d^2u}{dr^2} \frac{Y_\ell^m}{r} = \frac{1}{r} \frac{\partial^2}{\partial r^2}(r\psi) $$

**Tercer término (término centrífugo):**
Aquí es donde usamos la propiedad de los armónicos esféricos. Sabemos que $\ell(\ell+1) Y_\ell^m = \hat{L}^2 Y_\ell^m$ (donde $\hat{L}^2$ es el operador de momento angular adimensional). Entonces:
$$ \frac{\ell(\ell+1)}{r^2} \frac{u}{r} Y_\ell^m = \frac{1}{r^2} \hat{L}^2 \left( \frac{u}{r} Y_\ell^m \right) = \frac{1}{r^2} \hat{L}^2 \psi $$

Sustituyendo esto en la ecuación multiplicada:

$$ -\frac{\hbar^2}{2m} \left[ \frac{1}{r} \frac{\partial^2}{\partial r^2}(r\psi) - \frac{\hat{L}^2 \psi}{r^2} \right] + V(r)\psi = E\psi $$

El término entre corchetes es exactamente la definición del **Laplaciano en coordenadas esféricas** ($\nabla^2 \psi$). Por lo tanto, la ecuación se simplifica a la **Ecuación de Schrödinger independiente del tiempo en 3D**:

$$ -\frac{\hbar^2}{2m} \nabla^2 \psi + V(r)\psi = E\psi $$

# Desarrollo completo: De Schrödinger a la ecuación radial de Coulomb

A continuación presentamos el desarrollo sistemático que conecta la ecuación de Schrödinger con potencial coulombiano hasta la ecuación radial en la forma que permite compararla con la familia de Laguerre.

## 1. Punto de partida: Ecuación de Schrödinger estacionaria

La ecuación de Schrödinger independiente del tiempo para una partícula de masa $m$ en un potencial $V(\vec{r})$ es:

$$-\frac{\hbar^2}{2m}\nabla^2\psi(\vec{r}) + V(\vec{r})\psi(\vec{r}) = E\psi(\vec{r})$$

Para el átomo de hidrógeno, el potencial es coulombiano (atractivo):

$$V(r) = -\frac{K}{r}, \quad \text{donde } K = \frac{q^2}{4\pi\varepsilon_0}$$

Reescribiendo la ecuación:

$$\nabla^2\psi + \frac{2m}{\hbar^2}\left(E + \frac{K}{r}\right)\psi = 0$$

## 2. Separación de variables en coordenadas esféricas

Dado que el potencial $V(r)$ depende únicamente de la coordenada radial, el sistema posee simetría esférica. Es natural utilizar coordenadas esféricas $(r, \theta, \varphi)$. El laplaciano en estas coordenadas es:

$$\nabla^2 = \frac{1}{r^2}\frac{\partial}{\partial r}\left(r^2\frac{\partial}{\partial r}\right) + \frac{1}{r^2\sin\theta}\frac{\partial}{\partial\theta}\left(\sin\theta\frac{\partial}{\partial\theta}\right) + \frac{1}{r^2\sin^2\theta}\frac{\partial^2}{\partial\varphi^2}$$

Es conveniente reescribir el término radial usando la identidad:

$$\frac{1}{r^2}\frac{\partial}{\partial r}\left(r^2\frac{\partial\psi}{\partial r}\right) = \frac{1}{r}\frac{\partial^2}{\partial r^2}(r\psi)$$

Proponemos una solución separable de la forma:

$$\psi(r, \theta, \varphi) = \frac{u(r)}{r} Y(\theta, \varphi)$$

El factor $1/r$ se introduce para simplificar la ecuación radial. Sustituyendo en la ecuación de Schrödinger y dividiendo por $\psi = uY/r$:

$$\frac{1}{u}\frac{d^2u}{dr^2} + \frac{2m}{\hbar^2}\left(E + \frac{K}{r}\right) = -\frac{1}{Y}\left[\frac{1}{\sin\theta}\frac{\partial}{\partial\theta}\left(\sin\theta\frac{\partial Y}{\partial\theta}\right) + \frac{1}{\sin^2\theta}\frac{\partial^2 Y}{\partial\varphi^2}\right]$$

## 3. El operador de momento angular $\hat{L}^2$

El lado derecho de la ecuación anterior es precisamente el operador de momento angular adimensional al cuadrado, $\hat{L}^2$, actuando sobre $Y(\theta, \varphi)$.

$$\hat{L}^2 = -\left[\frac{1}{\sin\theta}\frac{\partial}{\partial\theta}\left(\sin\theta\frac{\partial}{\partial\theta}\right) + \frac{1}{\sin^2\theta}\frac{\partial^2}{\partial\varphi^2}\right]$$

Por lo tanto, la ecuación se escribe:

$$\frac{1}{u}\frac{d^2u}{dr^2} + \frac{2m}{\hbar^2}\left(E + \frac{K}{r}\right) = \frac{\hat{L}^2 Y}{Y}$$

## 4. Separación de la parte angular

Dado que el lado izquierdo depende solo de $r$ y el derecho solo de $\theta, \varphi$, ambos deben ser iguales a una constante. Escribimos esta constante como $\ell(\ell+1)$:

$$\hat{L}^2 Y(\theta, \varphi) = \ell(\ell+1) Y(\theta, \varphi)$$

Esta es la ecuación de autovalores para los armónicos esféricos. Las soluciones son los armónicos esféricos $Y_\ell^m(\theta, \varphi)$ con $\ell = 0, 1, 2, \ldots$ y $m = -\ell, \ldots, \ell$.

## 5. Ecuación radial

La parte radial queda entonces:

$$\frac{d^2u}{dr^2} + \frac{2m}{\hbar^2}\left(E + \frac{K}{r}\right)u = \ell(\ell+1)\frac{u}{r^2}$$

Reorganizando:

$$\frac{d^2u}{dr^2} + \left[-\frac{\ell(\ell+1)}{r^2} + \frac{2mK}{\hbar^2 r} + \frac{2mE}{\hbar^2}\right]u = 0$$

## 6. Cambio a la función $R(r)$

Recordemos que definimos $\psi = u(r)Y(\theta, \varphi)/r$, por lo que la parte radial de la función de onda es $R(r) = u(r)/r$. Sin embargo, la ecuación anterior ya está en términos de $u(r)$. Si queremos expresarla en términos de $R(r)$, notamos que:

$$u(r) = rR(r)$$

$$\frac{d^2u}{dr^2} = \frac{d^2}{dr^2}[rR(r)] = r\frac{d^2R}{dr^2} + 2\frac{dR}{dr}$$

Sustituyendo en la ecuación radial:

$$r\frac{d^2R}{dr^2} + 2\frac{dR}{dr} + \left[-\frac{\ell(\ell+1)}{r^2} + \frac{2mK}{\hbar^2 r} + \frac{2mE}{\hbar^2}\right]rR = 0$$

Dividiendo por $r$:

$$\frac{d^2R}{dr^2} + \frac{2}{r}\frac{dR}{dr} + \left[-\frac{\ell(\ell+1)}{r^2} + \frac{2mK}{\hbar^2 r} + \frac{2mE}{\hbar^2}\right]R = 0$$

Esta es la ecuación radial en términos de $R(r)$.

## 7. Forma final de la ecuación radial (para $u(r)$)

La ecuación para $u(r) = rR(r)$ es:

$$\frac{d^2u}{dr^2} + \left[-\frac{\ell(\ell+1)}{r^2} + \frac{2mK}{\hbar^2 r} + \frac{2mE}{\hbar^2}\right]u = 0$$

Definiendo las constantes:

$$\lambda \equiv \frac{2mK}{\hbar^2}, \quad b \equiv \frac{2mE}{\hbar^2}$$

Obtenemos

$$\boxed{\ddot{u}(r) + \left[-\frac{\ell(\ell+1)}{r^2} + \frac{\lambda}{r} + b\right]u(r) = 0}$$

donde $\ddot{u} = d^2u/dr^2$.

## 8. Adimensionalización y conexión con Laguerre

Para comparar esta ecuación con la familia de Laguerre, necesitamos una variable adimensional. Introducimos el cambio $r = \alpha x$, donde $\alpha$ tiene dimensiones de longitud y $x$ es adimensional.

Las derivadas se transforman como:

$$\frac{du}{dr} = \frac{1}{\alpha}\frac{du}{dx}, \quad \frac{d^2u}{dr^2} = \frac{1}{\alpha^2}\frac{d^2u}{dx^2}$$

Sustituyendo en la ecuación radial:

$$\frac{1}{\alpha^2}\frac{d^2u}{dx^2} + \left[-\frac{\ell(\ell+1)}{\alpha^2 x^2} + \frac{\lambda}{\alpha x} + b\right]u = 0$$

Multiplicando por $\alpha^2$:

$$\frac{d^2u}{dx^2} + \left[-\frac{\ell(\ell+1)}{x^2} + \frac{\alpha\lambda}{x} + b\alpha^2\right]u = 0$$

## 9. Comparación con la familia de Laguerre

La transformación $\psi_p^k(x) = x^{(k+1)/2}e^{-x/2}L_p^k(x)$ aplicada a la ecuación de Laguerre asociada conduce a la familia:

$$\ddot{\psi}_p^k(x) + \psi_p^k(x)\left[-\frac{k^2-1}{4x^2} + \frac{2p+k+1}{2x} - \frac{1}{4}\right] = 0$$

Comparando término a término con nuestra ecuación radial:

1. **Término en $1/x^2$:** $\ell(\ell+1) = \frac{k^2-1}{4} \implies k = 2\ell+1$

2. **Término constante:** $b\alpha^2 = -\frac{1}{4} \implies \alpha^2 = -\frac{1}{4b} = -\frac{\hbar^2}{8mE}$

3. **Término en $1/x$:** $\alpha\lambda = \frac{2p+k+1}{2} = p + \ell + 1$

De la tercera condición, sustituyendo $\lambda$ y $\alpha$:

$$\sqrt{-\frac{\hbar^2}{8mE}}\cdot\frac{2mK}{\hbar^2} = p + \ell + 1$$

$$\sqrt{-\frac{mK^2}{2\hbar^2 E}} = p + \ell + 1 \equiv n$$

donde $n$ es un entero positivo (número cuántico principal). Despejando $E$:

# Normalización de $\psi$

# Normalización de la función de onda $\psi_{n\ell m}$ del átomo de hidrógeno

La respuesta corta es: **no es conceptualmente difícil, pero sí algebraicamente tediosa**. La dificultad radica en manejar los polinomios de Laguerre asociados y las integrales que involucran exponenciales y potencias. Sin embargo, el procedimiento es sistemático y se basa en propiedades matemáticas bien establecidas.

## 1. Separación del problema de normalización

La condición de normalización para la función de onda es:

$$\int |\psi_{n\ell m}|^2 \, d^3r = 1$$

En coordenadas esféricas, el elemento de volumen es $d^3r = r^2 \sin\theta \, dr \, d\theta \, d\varphi = r^2 \, dr \, d\Omega$. Dado que la función de onda se factoriza como (Alonso Sepúlveda, Sección 8.6.3):

$$\psi_{n\ell m}(r,\theta,\varphi) = R_{n\ell}(r) \, Y_\ell^m(\theta,\varphi)$$

La integral de normalización se separa en dos partes independientes:

$$\int_0^\infty |R_{n\ell}(r)|^2 \, r^2 \, dr \times \int_{4\pi} |Y_\ell^m(\theta,\varphi)|^2 \, d\Omega = 1$$

## 2. La parte angular: trivial

La integral angular es **inmediata** porque los armónicos esféricos ya están normalizados por construcción. Como se demuestra en la Sección 8.4.5 de Alonso Sepúlveda (Ecuación 8.37):

$$\int_{4\pi} Y_{\ell'}^{m'*}(\theta,\varphi) \, Y_\ell^m(\theta,\varphi) \, d\Omega = \delta_{\ell\ell'} \, \delta_{mm'}$$

Para $\ell' = \ell$ y $m' = m$, obtenemos:

$$\int_{4\pi} |Y_\ell^m(\theta,\varphi)|^2 \, d\Omega = 1$$

Por lo tanto, **toda la dificultad se concentra en la parte radial**.

## 3. La parte radial: el verdadero desafío

La función radial del átomo de hidrógeno tiene la forma (Alonso, Sección 8.6.3, Problema al final):

$$R_{n\ell}(r) = N_{n\ell} \, e^{-r/na_0} \left(\frac{2r}{na_0}\right)^\ell L_{n-\ell-1}^{2\ell+1}\left(\frac{2r}{na_0}\right)$$

donde:
- $N_{n\ell}$ es la constante de normalización que queremos determinar
- $a_0$ es el radio de Bohr
- $L_{n-\ell-1}^{2\ell+1}$ son los polinomios de Laguerre asociados
- $n$ es el número cuántico principal, $\ell$ el número cuántico orbital

La condición de normalización radial es:

$$\int_0^\infty |R_{n\ell}(r)|^2 \, r^2 \, dr = 1$$

Sustituyendo la expresión de $R_{n\ell}$:

$$|N_{n\ell}|^2 \int_0^\infty e^{-2r/na_0} \left(\frac{2r}{na_0}\right)^{2\ell} \left[L_{n-\ell-1}^{2\ell+1}\left(\frac{2r}{na_0}\right)\right]^2 r^2 \, dr = 1$$

## 4. Cambio de variable y uso de la ortogonalidad de Laguerre

Para simplificar, introducimos la variable adimensional:

$$\rho = \frac{2r}{na_0} \quad \Rightarrow \quad r = \frac{na_0}{2}\rho, \quad dr = \frac{na_0}{2}d\rho$$

La integral se transforma en:

$$|N_{n\ell}|^2 \left(\frac{na_0}{2}\right)^3 \int_0^\infty e^{-\rho} \, \rho^{2\ell+2} \left[L_{n-\ell-1}^{2\ell+1}(\rho)\right]^2 d\rho = 1$$

Aquí es donde entra la **relación de ortogonalidad de los polinomios de Laguerre asociados** (Alonso, Sección 8.6.1, Ecuación 8.79):

$$\int_0^\infty x^k e^{-x} L_n^k(x) L_m^k(x) \, dx = \frac{(n+k)!}{n!} \, \delta_{nm}$$

En nuestro caso, identificamos:
- $k = 2\ell + 1$
- $n = m = n - \ell - 1$ (el índice del polinomio de Laguerre)
- $x = \rho$

Pero hay un detalle: nuestra integral tiene $\rho^{2\ell+2}$ mientras que la relación de ortogonalidad tiene $x^k = x^{2\ell+1}$. Necesitamos un factor adicional de $\rho$. Esto se resuelve usando la relación (Alonso, Sección 8.6.1):

$$\int_0^\infty x^{k+1} e^{-x} [L_n^k(x)]^2 \, dx = \frac{(n+k)!}{n!} (2n + k + 1)$$

Aplicando esta fórmula con nuestros parámetros:

$$\int_0^\infty \rho^{2\ell+2} e^{-\rho} \left[L_{n-\ell-1}^{2\ell+1}(\rho)\right]^2 d\rho = \frac{[(n-\ell-1)+(2\ell+1)]!}{(n-\ell-1)!} \left[2(n-\ell-1) + (2\ell+1) + 1\right]$$

Simplificando:

$$= \frac{(n+\ell)!}{(n-\ell-1)!} \cdot 2n$$

## 5. Determinación de la constante de normalización

Sustituyendo este resultado en la ecuación de normalización:

$$|N_{n\ell}|^2 \left(\frac{na_0}{2}\right)^3 \frac{(n+\ell)!}{(n-\ell-1)!} \cdot 2n = 1$$

Despejando $N_{n\ell}$:

$$N_{n\ell} = \sqrt{\left(\frac{2}{na_0}\right)^3 \frac{(n-\ell-1)!}{2n \, [(n+\ell)!]^3}}$$

Este es el factor de normalización que aparece en el Problema al final de la Sección 8.6.3 de Alonso Sepúlveda.

## 6. ¿Por qué parece difícil?

Las dificultades prácticas son:

1. **Manejo de índices**: Los polinomios de Laguerre asociados $L_{n-\ell-1}^{2\ell+1}$ tienen índices que dependen de dos números cuánticos, lo que hace fácil cometer errores.

2. **Factoriales grandes**: Para estados excitados (n grande), los factoriales $(n+\ell)!$ y $(n-\ell-1)!$ crecen rápidamente, requiriendo cuidado en los cálculos numéricos.

3. **Cambios de variable**: El paso de $r$ a $\rho = 2r/na_0$ introduce factores de escala que deben rastrearse cuidadosamente.

4. **Convenciones diferentes**: Existen diferentes convenciones para definir los polinomios de Laguerre asociados (algunos autores incluyen factores adicionales), lo que puede generar confusión al comparar resultados de diferentes fuentes.

## 7. Conclusión

Normalizar $\psi_{n\ell m}$ **no es conceptualmente difícil** porque:
- La parte angular ya está normalizada
- La parte radial se reduce a aplicar la relación de ortogonalidad de Laguerre
- El procedimiento es sistemático y repetible para cualquier $(n,\ell,m)$

**Sí es algebraicamente tedioso** porque requiere:
- Manejar expresiones con múltiples factoriales
- Realizar cambios de variable cuidadosos
- Usar identidades específicas de los polinomios de Laguerre

Una vez que se conoce el procedimiento, la normalización de cualquier estado del hidrógeno sigue los mismos pasos mecánicos. La verdadera dificultad aparece cuando se intentan calcular valores esperados de operadores más complejos (como $\langle r \rangle$, $\langle 1/r \rangle$, etc.), donde las integrales son aún más elaboradas.

### 1. Los números cuánticos espaciales (La solución de Schrödinger)

Cuando resolvemos la ecuación de Schrödinger para el átomo de hidrógeno (o cualquier potencial central $V(r)$) mediante separación de variables en coordenadas esféricas, la función de onda espacial $\psi(\vec{r})$ queda determinada por tres números cuánticos que surgen naturalmente de las condiciones de frontera y la ortogonalidad de las funciones especiales (Polinomios de Laguerre y Armónicos Esféricos):

1. **Número cuántico principal ($n$):** $n = 1, 2, 3, \dots$ Determina la energía del nivel ($E_n \propto -1/n^2$).
2. **Número cuántico orbital o azimutal ($\ell$):** $\ell = 0, 1, 2, \dots, n-1$. Determina el momento angular.
3. **Número cuántico magnético ($m$):** $m = -\ell, -\ell+1, \dots, \ell$. Determina la proyección del momento angular en el eje $z$.

Cada combinación única $(n, \ell, m)$ define un **orbital atómico** (una función de onda espacial $\psi_{n\ell m}(\vec{r})$). 

**¿Cuántos orbitales espaciales hay para un nivel $n$ dado?**
Simplemente sumamos las posibles combinaciones de $\ell$ y $m$:
$$ \text{Número de orbitales} = \sum_{\ell=0}^{n-1} (2\ell + 1) = n^2 $$
Por lo tanto, el nivel $n$ tiene una **degeneración espacial** de $n^2$ estados.

---

### 2. El cuarto número cuántico: El Espín

La ecuación de Schrödinger no relativista que resolvimos en clase describe la parte *espacial* del electrón. Sin embargo, experimentalmente (y teóricamente a través de la ecuación de Dirac), se sabe que el electrón posee un momento angular intrínseco llamado **espín** ($s = 1/2$).

Al igual que el momento angular orbital $\ell$ tiene proyecciones $m$, el espín tiene su propia proyección $m_s$. Para un electrón ($s=1/2$), solo hay dos valores posibles:
$$ m_s = +\frac{1}{2} \quad (\text{espín "arriba", } \uparrow) \quad \text{o} \quad m_s = -\frac{1}{2} \quad (\text{espín "abajo", } \downarrow) $$

Para describir el estado **completo** de un electrón en el átomo, necesitamos especificar los cuatro números cuánticos:
$$ |n, \ell, m, m_s\rangle $$

---

### 3. El Principio de Exclusión de Pauli

El principio de Pauli establece que **dos fermiones (como los electrones) idénticos no pueden ocupar el mismo estado cuántico simultáneamente**. 

En términos de nuestros números cuánticos: **No puede haber dos electrones en un átomo que tengan los cuatro números cuánticos $(n, \ell, m, m_s)$ iguales.**

Como el espín solo tiene 2 valores posibles ($+1/2$ y $-1/2$), esto significa que **cada orbital espacial $\psi_{n\ell m}$ puede albergar un máximo de 2 electrones**: uno con espín arriba y otro con espín abajo.

---

### 4. Contando los electrones: El llenado de capas

Ahora podemos calcular cuántos electrones caben en cada nivel de energía $n$. Si hay $n^2$ orbitales espaciales y cada uno acepta 2 electrones, la capacidad máxima del nivel $n$ es:
$$ \text{Capacidad máxima} = 2 \times n^2 $$

Veamos cómo funciona esto paso a paso para los primeros niveles:

#### **Caso $n = 1$ (Capa K)**
* **Orbitales espaciales:** $\ell$ solo puede ser $0$. Si $\ell=0$, $m$ solo puede ser $0$.
* Tenemos **1 solo orbital espacial**: $\psi_{100}$ (el orbital $1s$).
* **Espín:** Ese único orbital puede tener un electrón con $m_s = +1/2$ y otro con $m_s = -1/2$.
* **Total de electrones:** $2 \times (1)^2 = \mathbf{2}$ electrones.
* *Estados completos:* $|1, 0, 0, +1/2\rangle$ y $|1, 0, 0, -1/2\rangle$.

#### **Caso $n = 2$ (Capa L)**
* **Orbitales espaciales:** 
  * Si $\ell = 0 \implies m = 0$ (1 orbital: el $2s$).
  * Si $\ell = 1 \implies m = -1, 0, +1$ (3 orbitales: los $2p_x, 2p_y, 2p_z$).
* Tenemos $1 + 3 = \mathbf{4}$ orbitales espaciales.
* **Espín:** Cada uno de los 4 orbitales acepta 2 electrones.
* **Total de electrones:** $4 \times 2 = \mathbf{8}$ electrones.
* *Fórmula:* $2 \times (2)^2 = 8$.

#### **Caso $n = 3$ (Capa M)**
* **Orbitales espaciales:**
  * $\ell = 0 \implies m = 0$ (1 orbital: $3s$)
  * $\ell = 1 \implies m = -1, 0, 1$ (3 orbitales: $3p$)
  * $\ell = 2 \implies m = -2, -1, 0, 1, 2$ (5 orbitales: $3d$)
* Tenemos $1 + 3 + 5 = \mathbf{9}$ orbitales espaciales.
* **Total de electrones:** $9 \times 2 = \mathbf{18}$ electrones.
* *Fórmula:* $2 \times (3)^2 = 18$.

---

### Resumen desde la perspectiva del curso

Lo que estamos viendo es la **estructura de los espacios de Hilbert** en mecánica cuántica. 

1. Al resolver la ecuación de Sturm-Liouville para la parte radial y angular (Capítulo 8 de Alonso), generamos una base de autofunciones espaciales $\psi_{n\ell m}$ que son ortogonales y completas para la parte espacial.
2. La **degeneración** de los autovalores de energía $E_n$ es de $n^2$ (hay $n^2$ funciones espaciales distintas con la misma energía).
3. Al introducir el espín, estamos ampliando nuestro espacio de Hilbert. El espacio total es el **producto tensorial** del espacio espacial y el espacio de espín (que es un espacio de dimensión 2, $\mathbb{C}^2$).
4. La dimensión total del subespacio asociado a la energía $E_n$ es $n^2 \times 2 = 2n^2$.
5. El Principio de Pauli nos dice que, al construir el estado de un sistema de múltiples electrones (usando determinantes de Slater, aunque eso es un tema más avanzado), cada uno de estos $2n^2$ estados base puede ser ocupado por un solo electrón. 

Por eso, el llenado de la tabla periódica no es arbitrario; es una consecuencia directa y elegante de la geometría de las ecuaciones diferenciales (que dicta $n^2$) y de la naturaleza fermiónica de la materia (que multiplica por 2).

# Ejemplo: Configuración electrónica del Oxígeno ($Z=8$)

## 0. Corrección preliminar

Hay una pequeña imprecisión en el enunciado. La configuración $1s^2 2p^6$ corresponde al **Neón** ($Z=10$), no al oxígeno. El **oxígeno** tiene $Z=8$ electrones y su configuración correcta es:

$$\boxed{1s^2 \, 2s^2 \, 2p^4}$$

Vamos a construir esta configuración desde cero, usando la teoría que hemos desarrollado.

---

## 1. Punto de partida: la solución del átomo de hidrógeno

En la **Sección 8.6.3** de Alonso Sepúlveda, se resuelve la ecuación de Schrödinger para un potencial coulombiano $V(r) = -K/r$. Mediante separación de variables en coordenadas esféricas $\psi(r,\theta,\varphi) = R(r)Y(\theta,\varphi)/r$, se obtienen tres ecuaciones:

| Parte | Ecuación | Solución | Números cuánticos |
|-------|----------|----------|-------------------|
| **Azimutal** ($\varphi$) | $\Phi'' = -m^2 \Phi$ | $e^{im\varphi}$ | $m \in \mathbb{Z}$ |
| **Polar** ($\theta$) | Legendre asociada | $P_\ell^m(\cos\theta)$ | $\ell \geq \|m\|$, $\ell \in \mathbb{N}_0$ |
| **Radial** ($r$) | Laguerre asociada | $e^{-r/n a_0} L_{n-\ell-1}^{2\ell+1}(2r/na_0)$ | $n \geq \ell+1$, $n \in \mathbb{N}$ |

La función de onda completa queda etiquetada por **tres números cuánticos**:

$$\psi_{n\ell m}(r,\theta,\varphi) = R_{n\ell}(r)\, Y_\ell^m(\theta,\varphi)$$

con las restricciones:

$$n = 1, 2, 3, \ldots \qquad \ell = 0, 1, \ldots, n-1 \qquad m = -\ell, \ldots, \ell$$

---

## 2. Energía y degeneración

De la condición de convergencia de los polinomios de Laguerre (Sección 8.6.2 de Alonso), la energía resulta **cuantizada**:

$$E_n = -\frac{mK^2}{2\hbar^2 n^2} = -\frac{13.6 \text{ eV}}{n^2}$$

**Observación crucial:** la energía solo depende de $n$, no de $\ell$ ni de $m$. Esto significa que todos los estados con el mismo $n$ tienen la **misma energía** (están **degenerados**).

### Conteo de estados por nivel $n$

Para un $n$ fijo, contamos cuántas combinaciones $(\ell, m)$ existen:

$$\text{Nº de orbitales espaciales} = \sum_{\ell=0}^{n-1} (2\ell + 1) = n^2$$

| $n$ | $\ell$ permitidos | Orbitales ($m$) | Total orbitales |
|-----|-------------------|-----------------|-----------------|
| 1 | 0 | $m=0$ (1) | $1^2 = 1$ |
| 2 | 0, 1 | $m=0$; $m=-1,0,1$ | $1+3 = 4 = 2^2$ |
| 3 | 0, 1, 2 | $1+3+5$ | $9 = 3^2$ |

---

## 3. El cuarto número cuántico: el espín

La ecuación de Schrödinger no relativista solo describe la parte **espacial**. El electrón posee además un momento angular intrínseco (espín) $s=1/2$, con proyección:

$$m_s = +\frac{1}{2} \quad (\uparrow) \qquad \text{o} \qquad m_s = -\frac{1}{2} \quad (\downarrow)$$

El estado completo de un electrón se especifica con **cuatro números cuánticos**:

$$\ket{n, \ell, m, m_s}$$

---

## 4. Principio de exclusión de Pauli

> **Dos electrones en un mismo átomo no pueden tener los cuatro números cuánticos iguales.**

Como el espín solo tiene 2 valores, cada orbital espacial $\psi_{n\ell m}$ puede albergar como máximo **2 electrones** (uno con espín arriba, otro con espín abajo).

### Capacidad máxima de cada capa

$$\text{Capacidad}(n) = 2 \times n^2$$

| Capa | $n$ | Orbitales | Electrones máximos |
|------|-----|-----------|---------------------|
| K | 1 | 1 | **2** |
| L | 2 | 4 | **8** |
| M | 3 | 9 | **18** |

---

## 5. Llenado sistemático del oxígeno ($Z=8$)

El oxígeno neutro tiene **8 electrones**. Los vamos colocando en orden de energía creciente (principio de Aufbau), respetando Pauli.

### Paso 1: Capa $n=1$ (capa K)

- $\ell = 0 \Rightarrow m = 0$ (un solo orbital: $1s$)
- Capacidad: 2 electrones

$$1s^2 \quad \Rightarrow \quad \ket{1,0,0,+\tfrac{1}{2}}, \quad \ket{1,0,0,-\tfrac{1}{2}}$$

**Electrones usados:** 2. **Restantes:** 6.

### Paso 2: Capa $n=2$ (capa L)

Aquí hay dos subcapas:

**Subcapa $2s$** ($\ell=0$):
- $m=0$ (un orbital)
- Capacidad: 2 electrones

$$2s^2 \quad \Rightarrow \quad \ket{2,0,0,+\tfrac{1}{2}}, \quad \ket{2,0,0,-\tfrac{1}{2}}$$

**Electrones usados:** 4. **Restantes:** 4.

**Subcapa $2p$** ($\ell=1$):
- $m = -1, 0, +1$ (tres orbitales: $2p_{-1}, 2p_0, 2p_{+1}$, o equivalentemente $2p_x, 2p_y, 2p_z$)
- Capacidad total: $3 \times 2 = 6$ electrones
- Solo nos quedan **4 electrones** para colocar aquí.

---

## 6. Regla de Hund para los 4 electrones $2p$

Los 3 orbitales $2p$ son **degenerados** (misma energía $E_2$). ¿Cómo distribuir 4 electrones en 3 orbitales?

**Regla de Hund (máxima multiplicidad):** los electrones se distribuyen primero con espines paralelos en orbitales distintos, y solo después se aparean.

| Orbital | Electrón 1 | Electrón 2 | Electrón 3 | Electrón 4 |
|---------|-----------|-----------|-----------|-----------|
| $2p_{-1}$ | $\uparrow$ | $\downarrow$ | — | — |
| $2p_0$ | $\uparrow$ | — | — | — |
| $2p_{+1}$ | $\uparrow$ | — | — | — |

- Los 3 primeros electrones ocupan los 3 orbitales con espín $\uparrow$ (estado de **triplete**, $S=1$).
- El 4º electrón se aparea en uno de ellos (digamos $m=-1$) con espín $\downarrow$.

Resultado: **2 electrones desapareados** $\Rightarrow$ el oxígeno es **paramagnético** (esto se verifica experimentalmente).

---

## 7. Configuración final

$$\boxed{\text{O} \; (Z=8): \quad 1s^2 \, 2s^2 \, 2p^4}$$

En notación de números cuánticos, los 8 electrones ocupan los estados:

| # | $n$ | $\ell$ | $m$ | $m_s$ | Orbital |
|---|-----|--------|-----|-------|---------|
| 1 | 1 | 0 | 0 | $+1/2$ | $1s$ |
| 2 | 1 | 0 | 0 | $-1/2$ | $1s$ |
| 3 | 2 | 0 | 0 | $+1/2$ | $2s$ |
| 4 | 2 | 0 | 0 | $-1/2$ | $2s$ |
| 5 | 2 | 1 | $-1$ | $+1/2$ | $2p$ |
| 6 | 2 | 1 | $0$ | $+1/2$ | $2p$ |
| 7 | 2 | 1 | $+1$ | $+1/2$ | $2p$ |
| 8 | 2 | 1 | $-1$ | $-1/2$ | $2p$ |

## 8. Resumen conceptual

La tabla periódica emerge **naturalmente** de la estructura matemática que hemos estudiado:

1. **Ecuación de Sturm-Liouville** (Laguerre + Legendre) $\Rightarrow$ números cuánticos $n, \ell, m$.
2. **Degeneración** $n^2$ de los autovalores de energía $E_n$.
3. **Espín** (doble el espacio de Hilbert) $\Rightarrow$ capacidad $2n^2$.
4. **Principio de Pauli** $\Rightarrow$ llenado secuencial.
5. **Regla de Hund** $\Rightarrow$ distribución dentro de subcapas degeneradas.

Para el oxígeno, la capa $n=1$ está llena (2 e$^-$), la subcapa $2s$ está llena (2 e$^-$), y la subcapa $2p$ está **parcialmente llena** con 4 de 6 electrones posibles, dejando 2 electrones desapareados que explican su reactividad química y su paramagnetismo.
