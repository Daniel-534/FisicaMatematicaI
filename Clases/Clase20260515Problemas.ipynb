**Enunciado:**
Considere la **ecuaciĂłn diferencial de Bessel** de orden $\nu$:
$$x^2 \frac{d^2y}{dx^2} + x \frac{dy}{dx} + (x^2 - \nu^2)y = 0$$

Como se establece en la teorĂ­a de ecuaciones diferenciales de segundo orden, esta ecuaciĂłn posee dos soluciones linealmente independientes, denotadas como $J_\nu(x)$ (funciĂłn de Bessel de primer gĂ©nero) y $Y_\nu(x)$ (funciĂłn de Bessel de segundo gĂ©nero o funciĂłn de Neumann).

3.  **EvaluaciĂłn AnalĂ­tica:** A partir de las expansiones en serie de potencia para argumentos pequeĂ±os ($x \to 0$) o de las formas asintĂłticas para argumentos grandes ($x \to \infty$), determine el valor de la constante $C$ que garantiza la escala estĂˇndar de estas funciones.

### **Problema: DeterminaciĂłn de la Norma en el Espacio de Hilbert para Funciones de Bessel**

**Enunciado:**

Considere el operador diferencial de Bessel de orden $\nu$, definido en el intervalo $[0, a]$ como:
$$\mathcal{L} = -\left( \frac{d^2}{d\rho^2} + \frac{1}{\rho}\frac{d}{d\rho} - \frac{\nu^2}{\rho^2} \right)$$

De acuerdo con la **teorĂ­a de Sturm-Liouville**, este operador es autoadjunto respecto al **factor de peso $w(\rho) = \rho$**. Al integrar el producto de dos soluciones, $J_\nu(k\rho)$ y $J_\nu(k'\rho)$, asociadas a distintos parĂˇmetros de escala $k$ y $k'$, se obtiene la **primera identidad de Lommel**:

$$\int_{0}^{a} \rho J_{\nu}(k\rho) J_{\nu}(k'\rho) \, d\rho = \frac{a \left[ k' J_{\nu}(ka) J_{\nu+1}(k'a) - k J_{\nu}(k'a) J_{\nu+1}(ka) \right]}{k^2 - k'^2}$$

1.  **AnĂˇlisis de la IndeterminaciĂłn:** Observe que si se intenta evaluar directamente la integral para el caso en que los autovalores coinciden ($k' = k$), la expresiĂłn del lado derecho adopta la forma indeterminada **$0/0$**, debido a que tanto el numerador (basado en el Wronskiano) como el denominador se anulan simultĂˇneamente.
2.  **Tarea AnalĂ­tica:** Utilice la **regla de l'HĂ´pital** para resolver esta indeterminaciĂłn. Para ello, realice el cĂˇlculo del lĂ­mite de la expresiĂłn anterior cuando el parĂˇmetro auxiliar $k'$ tiende al valor fijo $k$:
    $$\lim_{k' \to k} \frac{a \left[ k' J_{\nu}(ka) J_{\nu+1}(k'a) - k J_{\nu}(k'a) J_{\nu+1}(ka) \right]}{k^2 - k'^2}$$
3.  **Identidades Auxiliares:** En su desarrollo, emplee las **relaciones de recurrencia** para derivadas de Bessel, especĂ­ficamente $k \frac{d}{dk} J_\nu(ka) = -a \rho J_{\nu+1}(ka) + \frac{\nu}{k} J_\nu(ka)$ (o sus equivalentes), para simplificar el resultado final.
4.  **InterpretaciĂłn:** Muestre que este lĂ­mite permite definir la **norma de la funciĂłn** (su "longitud" al cuadrado en el espacio de funciones) y determine el valor exacto de la integral de normalizaciĂłn $\int_{0}^{a} \rho [J_{\nu}(k\rho)]^2 \, d\rho$.

---

Al ejecutar el comando solicitado en Mathematica, el software aplica la derivaciĂłn respecto a $k'$ en el numerador y denominador, obteniendo el siguiente resultado simplificado en **TeXForm**:

$$\mathbf{\frac{a^2}{2} \left[ J_{\nu}(ka)^2 + J_{\nu+1}(ka)^2 - \frac{2\nu}{ka} J_{\nu}(ka) J_{\nu+1}(ka) \right]}$$

### **Contexto AnalĂ­tico segĂşn las fuentes**

*   **Importancia del LĂ­mite:** Este cĂˇlculo es el paso matemĂˇtico crĂ­tico para pasar de la **ortogonalidad** (distinguir dos funciones diferentes) a la **normalizaciĂłn** (medir la magnitud de una funciĂłn especĂ­fica).
*   **SimplificaciĂłn por Condiciones de Frontera:** Si el parĂˇmetro $k$ se elige de modo que cumpla la **condiciĂłn de Dirichlet** ($J_\nu(ka) = 0$), la soluciĂłn anterior se colapsa a la forma estĂˇndar utilizada en fĂ­sica: $\frac{a^2}{2} J_{\nu+1}(ka)^2$.
*   **Fundamento en Sturm-Liouville:** Este lĂ­mite garantiza que las funciones de Bessel puedan ser utilizadas para construir una **base ortonormal completa**, permitiendo expandir cualquier funciĂłn arbitraria en una **serie de Fourier-Bessel**.

````
Limit[(a (kp BesselJ[nu, k a] BesselJ[nu + 1, kp a] - k BesselJ[nu, kp a] BesselJ[nu + 1, k a])) / 
      (k^2 - kp^2), kp -> k] // TeXForm
````

### **Problema: Coeficientes de Fourier-Bessel para la funciĂłn monomial $x^n$**

**Enunciado:**

Considere el espacio de funciones de cuadrado integrable $L^2_x$ con factor de peso $\rho(x) = x$. En este espacio, el conjunto de funciones $\{J_n(\chi_{nl} x)\}_{l=1}^\infty$ constituye una **base completa y ortogonal**, donde $J_n$ es la funciĂłn de Bessel de primer gĂ©nero de orden $n$ (entero) y $\chi_{nl}$ representa el $l$-Ă©simo cero positivo de dicha funciĂłn ($J_n(\chi_{nl}) = 0$).

Cualquier funciĂłn $f(x)$ definida en el intervalo que sea suficientemente suave puede expresarse como la serie:
$$f(x) = \sum_{l=1}^{\infty} A_{nl} J_n(\chi_{nl} x)$$

1.  **DeterminaciĂłn AnalĂ­tica:** Utilice la relaciĂłn de ortonormalizaciĂłn:
    $$\int_{0}^{1} x [J_n(\chi_{nl} x)]^2 dx = \frac{1}{2} [J_{n+1}(\chi_{nl})]^2$$
    para demostrar que los coeficientes de la expansiĂłn estĂˇn dados por la integral:
    $$A_{nl} = \frac{2}{J_{n+1}^2(\chi_{nl})} \int_{0}^{1} x f(x) J_n(\chi_{nl} x) dx$$
2.  **Tarea de EvaluaciĂłn:** Sea la funciĂłn $f(x) = x^n$ (donde la potencia coincide con el orden de la base de Bessel). Calcule el coeficiente $A_{nl}$ realizando la integraciĂłn simbĂłlica.
3.  **Uso de Identidades:** Para resolver la integral, aplique la identidad de recurrencia diferencial:
    $$\frac{d}{dx}[x^{\mu+1} J_{\mu+1}(x)] = x^{\mu+1} J_\mu(x)$$
    y realice un cambio de variable adecuado.
4.  **VerificaciĂłn SimbĂłlica:** Implemente el cĂˇlculo en un sistema de Ăˇlgebra computacional y simplifique el resultado bajo las suposiciones de que el orden $n$ es no negativo y el Ă­ndice del modo $l$ es un entero positivo.

---

### **Resultado de la soluciĂłn en Mathematica (TeXForm)**

Al ejecutar el cĂłdigo proporcionado, el sistema evalĂşa la integral radial utilizando la propiedad de la derivada de Bessel y simplifica la constante de normalizaciĂłn, arrojando el siguiente coeficiente:

$$\mathbf{A_{nl} = \frac{2}{\chi_{nl} J_{n+1}(\chi_{nl})}}$$

### **AnĂˇlisis TĂ©cnico segĂşn las Fuentes**

*   **EvaluaciĂłn de la Integral:** Al sustituir $f(x) = x^n$, la integral se convierte en $\int_0^1 x^{n+1} J_n(\chi_{nl} x) dx$. Usando el cambio $u = \chi_{nl} x$, se integra como $\frac{1}{\chi_{nl}^{n+2}} \int u^{n+1} J_n(u) du = \frac{1}{\chi_{nl}^{n+2}} [u^{n+1} J_{n+1}(u)]$, lo que resulta en $J_{n+1}(\chi_{nl}) / \chi_{nl}$.
*   **CancelaciĂłn de TĂ©rminos:** El tĂ©rmino $J_{n+1}(\chi_{nl})$ en el numerador cancela una potencia del mismo tĂ©rmino en el factor de normalizaciĂłn $2/J_{n+1}^2(\chi_{nl})$, dejando el resultado simplificado.
*   **Importancia AnalĂ­tica:** Este resultado es fundamental en la **fĂ­sica matemĂˇtica** para resolver la ecuaciĂłn de calor en cilindros o los modos de una membrana vibrante donde la condiciĂłn inicial es una potencia de la coordenada radial.
*   **Propiedad de los Ceros:** El denominador nunca es cero porque las funciones de Bessel de Ăłrdenes consecutivos ($J_n$ y $J_{n+1}$) no comparten raĂ­ces comunes (excepto el origen), garantizando que el coeficiente $A_{nl}$ estĂ© siempre bien definido para $l \ge 1$.

````
f[x_] := x^n;
root = BesselJZero[n, l]; (* l-Ă©sima raĂ­z de Jn *)
Anl = (2 / BesselJ[n + 1, root]^2) * Integrate[x * f[x] * BesselJ[n, root * x], {x, 0, 1}] // TeXForm

FullSimplify[Anl, Assumptions -> l > 0 && n >= 0]
````

### **Problema: VerificaciĂłn de la Ortogonalidad Modal en Sistemas de Sturm-Liouville Singulares**

**Enunciado:**

Considere el espacio de funciones de cuadrado integrable $L^2_x$ dotado de un producto escalar con **factor de peso $w(x) = x$**. En este dominio, el operador diferencial de Bessel de orden $\nu = 0$ se define en su forma autoadjunta como:

Defina las dos primeras autofunciones del sistema como $\phi_1(x) = J_0(\alpha_{0,1} x)$ y $\phi_2(x) = J_0(\alpha_{0,2} x)$.
2.  **AnĂˇlisis de Hermiticidad:** Explique por quĂ© el operador de Bessel, a pesar de tener una singularidad en $x = 0$, se comporta como un operador **HermĂ­tico** en el intervalo bajo la condiciĂłn de que las soluciones sean finitas en el origen [7, 211; 10, 284].
3.  **Tarea AnalĂ­tica:** Utilice la **primera identidad de Lommel** para el Wronskiano de funciones de Bessel [10, 284; 7, 354]:
    $$(k_1^2 - k_2^2) \int_{0}^{1} x J_{0}(k_1 x) J_{0}(k_2 x) dx = \left[ x (k_2 J_{0}(k_1 x) J'_{0}(k_2 x) - k_1 J_{0}(k_2 x) J'_{0}(k_1 x)) \right]_{0}^{1}$$
    para evaluar el producto escalar de las dos primeras autofunciones identificadas en el inciso 1.
