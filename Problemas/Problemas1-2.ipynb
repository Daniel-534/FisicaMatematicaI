**Problema 2:** vector en coordenadas esféricas



```
A = {x y, -x, 3 x};

(* Relaciones esféricas *)
xs = r Sin[θ] Cos[ϕ];
ys = r Sin[θ] Sin[ϕ];
zs = r Cos[θ];

(* Campo con sustitución esférica *)
As = Simplify[A /. {x -> xs, y -> ys, z -> zs}];

(* Base esférica en cartesianas *)
er  = {Sin[θ] Cos[ϕ], Sin[θ] Sin[ϕ], Cos[θ]};
eθ  = {Cos[θ] Cos[ϕ], Cos[θ] Sin[ϕ], -Sin[θ]};
eϕ  = {-Sin[ϕ], Cos[ϕ], 0};

(* Componentes esféricas *)
Ar  = Simplify[As . er]
Aθ  = Simplify[As . eθ]
Aϕ  = Simplify[As . eϕ]
```
